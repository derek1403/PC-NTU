# Group

**Notation**

* $Z = \{\dots, -3, -2, -1, 0, 1, 2, 3, \dots\}$ of integers
  * **$Z$ (所有整數):** 這是數學最基礎的集合，相當於 Python 中沒有大小限制的 `int` 型別。
* $Z_n = \{0, 1, 2, \dots, n-1\}$
  * **$Z_n$ (模 $n$ 的同餘類):** 這是將無限的整數局限在一個長度為 $n$ 的循環中。
  * **程式思維：** 這非常類似於執行 `range(n)` 所產生的結果，裡面的元素就是任何數字除以 $n$ 之後可能出現的餘數。
* $Z_n^* = \{a \in Z_n \mid \text{gcd}(a, n) = 1\}$ 
  * **定義推導：** $\text{gcd}(a, n)$ 代表兩數的最大公因數 (Greatest Common Divisor)。當 $\text{gcd}(a, n) = 1$ 時，代表 $a$ 與 $n$ 「互質」。在模反元素的定理中，一個數字 $a$ 在模 $n$ 之下要有「乘法反元素」（也就是能找到一個 $x$ 使得 $a \times x \equiv 1 \pmod n$），充要條件就是 $a$ 與 $n$ 必須互質。
  * **程式思維：** 這就像是寫了一個 List Comprehension 過濾器：`[a for a in range(n) if math.gcd(a, n) == 1]`。以圖中的 $Z_{12}^*$ 為例，**1** 到 **11** 中與 **12** 互質的數字正是 **1, 5, 7, 11**。
  * $Z_{12}^* = \{1, 5, 7, 11\}$
  * $Z_{\text{prime number}}^* = \{1,2,3,4,5,6,\dots ,\text{prime number}-1\} = Z_{\text{prime number}}$
* $N = \{0, 1, 2, 3, \dots\}$ of non-negative integers **非零整數集合**
* $P = \{1, 2, 3, \dots\}$ of positive integers **正整數集合**
* $Q =$ The set of rational numbers **有理數集合**
* $R =$ The set of real numbers **實數集合**
* $C =$ The set of complex numbers **複數集合**
* $Q^*, R^*, C^* =$ The set of non-zero rational, real, complex numbers
  * 代表「排除 **0** 的集合」
  * 探討乘法時，**0** 不能放在分母沒有乘法反元素，將 **0** 排除後，剩下的元素才能形成一個完美的「乘法群」
* $GL_n(R) (一般線性群, General Linear Group): =$ The set of $n \times n$ invertible matrices with entries in $R$ 
  * **定義推導：** 一個矩陣要能夠反轉（求反矩陣），其行列式 (Determinant) 絕對不能為 **0**。因此 $GL_n(R)$ 就是所有行列式不為 **0** 的方陣集合。
  * **幾何推導與意義：** 在線性代數中，矩陣代表一種「空間變換」。行列式的絕對值 $|\det(A)|$，代表了這個變換將空間「放大或縮小了多少倍的面積（或體積）」
  * 建構一個 $GL_2(R)$ 的元素：**檢查元素：** $2, 0, 3$ 都是實數 ($R$)。 $M = \begin{pmatrix} 2 & 0 \\ 0 & 3 \end{pmatrix}$
* $SL_n(R) (特殊線性群, Special Linear Group): =$ The subset of $GL_n(R)$ with each matrix of determinant 1 
  * 它是 $GL_n(R)$ 的嚴格子集。它不僅要求矩陣可逆，還嚴格規定該矩陣的行列式必須精準等於 **1**。這類矩陣在幾何變換上代表「保持體積且不翻轉」的變換。
  * **例子 1：錯切變換 (Shear Mapping)** $S = \begin{pmatrix} 1 & 1 \\ 0 & 1 \end{pmatrix}$
  * **例子 2：旋轉變換 (Rotation Matrix)**  $R_\theta = \begin{pmatrix} \cos \theta & -\sin \theta \\ \sin \theta & \cos \theta \end{pmatrix}$


## **Definition** : 群的定義

A **group** $(G, *)$ is a set $G$ with an operation $*$, such that the following conditions are satisfied:

1. **封閉性 (Closure)** $a * b \in G$ for all $a, b \in G$
   * **意義：** 集合裡面的任何兩個元素互相作用後，產生的新結果必須「依然留在這個集合內」，不能跑出去。
   * **程式思維：** 這就像是 Python 裡的「型別限制」或是對物件實作 Magic Method。假設我們定義了一個類別 `MySet`，封閉性要求當你實作 `__add__(self, other)` 或是 `__mul__(self, other)` 時，它 `return` 的結果型別必須保證還是 `MySet` 的實例，而不能突然變成一個普通的 `int` 或 `list`。

2. **結合律 (Associativity)** $a * (b * c) = (a * b) * c$ for all $a, b, c \in G$
   * **意義：** 當有多個元素要連著運算時，括號刮在哪裡（先算前面還是先算後面）不影響最終結果。
   * **注意，這裡並沒有要求 $a * b = b * a$ (交換律)**，一個群不一定要滿足交換律（滿足交換律的群特別稱為阿貝爾群 Abelian Group）。
   * **減法不是群的運算**，因為 $a - (b - c) \ne (a - b) - c$

3. **單位元素 (Identity)** There is an element $e \in G$ such that $a = a * e = e * a$ for each $a \in G$
   * **意義：** 集合裡面必須存在一個特別的元素 $e$（通常稱為 identity element），它的功用就像是「透明人」，任何元素跟它作用後，都會保持原樣。
   * **注意，單位元素必定唯一** ，可以[查看證明](derek1403.github.io\Theory_Playground\_build\html\02_Concepts\Abstract_Algebra\proof_group_theory_identity_uniqueness.html)
   * **程式思維：** 如果這個運算是整數加法，這個 $e$ 就是 `0`；如果是整數乘法，$e$ 就是 `1`。如果你在處理字串串接 (String Concatenation)，這個 $e$ 就是空字串 `""`，因為 `"hello" + "" == "hello"`。
  
4. **反元素 (Inverse)** For each $a \in G$, there is an element $b \in G$ such that $a * b = b * a = e$
   * **意義：** 對於集合裡的每一個元素 $a$，你必定能找到一個對應的專屬夥伴 $b$（通常記作 $a^{-1}$），讓他們兩個作用後互相抵消，變回那個透明的單位元素 $e$。
   * **程式思維：** 這非常類似系統設計裡的「復原 (Undo)」機制。如果你執行了一個改變狀態的函數 `action_a()`，群的定義保證了你一定能找到並呼叫一個 `action_b()`，把系統狀態完美地重置回初始狀態 $e$。


在數學中，我們習慣了加減乘除，但如果我們想要把「運算」的概念推廣到更抽象的事物上（例如矩陣相乘、多項式相加，甚至魔術方塊的旋轉）我們就需要一個通用的定義，只要一個集合 (Set) $G$ 加上一個運算規則 $*$ 滿足以上4個條件，我們就可以稱它們為一個「群」。
