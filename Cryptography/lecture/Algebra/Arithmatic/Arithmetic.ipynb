## **Definition**: Floor and Ceiling

1. The **floor** $\lfloor x \rfloor$ of $x \in \boldsymbol{R}$ is the largest integer $\leq x$ ，代表不大於 $x$ 的「最大整數」
2. The **ceiling** $\lceil x \rceil$ of $x \in \boldsymbol{R}$ is the smallest integer $\geq x$ ，代表不小於 $x$ 的「最小整數」


**Example**
* $\lfloor e \rfloor = 2$, $\lceil e \rceil = 3$
* $\lfloor -3.1416 \rfloor = -4$, $\lceil -3.1416 \rceil = -3$
* $25 = 3 \times 7 + 4$ [7: divisor, 3: quotient, 4: remainder]
  * $25 \bmod 7 = 25 - \lfloor 25 / 7 \rfloor \times 7 = 25 - 3 \times 7 = 25 - 21=4$ (remainder)

## **Definition**: Modular Function (同餘函數)

$m, n \in \boldsymbol{Z}, m > 0$, define $\boldsymbol{n \bmod m} = n - \lfloor n/m \rfloor \times m$ 
* i.e., the remainder after dividing $n$ by $m$, which is $\ge 0$ and $< m$


* **Example** 58 in the base 3 representation

$$\begin{gather*}
a_0 =& 58 \bmod 3 = 1 &\qquad 19 &= \left\lfloor \frac{58}{3} \right\rfloor \\
a_1 =& 19 \bmod 3 = 1 &\qquad 6 &= \left\lfloor \frac{19}{3} \right\rfloor\\
a_2 =&  6 \bmod 3 = 0 &\qquad 2 &= \left\lfloor \frac{6}{3} \right\rfloor \\
a_3 =&  2 \bmod 3 = 2 &\qquad 0 &= \left\lfloor \frac{2}{3}  \right\rfloor \\
\end{gather*}$$

$$(2011)_3 = 2 \times 3^3 + 0 \times 3^2 + 1 \times 3^1 + 1 \times 3^0 = 58$$

### 2. 老師的觀念解析與推導

我們一樣把這個過程拆解成幾個步驟，來看看這段數學是怎麼運作的。

#### 第一步：理解同餘 (Modulo) 的本質

投影片上半部定義了 $\boldsymbol{n \bmod m} = n - \lfloor n/m \rfloor \times m$。這其實就是我們上一張投影片最後推導出來的「求餘數」公式。

**觀念類比**：
想像你是一個裝箱工人，你有 $n$ 顆蘋果，每個箱子只能裝 $m$ 顆。

* $\lfloor n/m \rfloor$ 算的是「你可以裝滿幾個完整的箱子」。
* $\lfloor n/m \rfloor \times m$ 算的是「已經被裝進箱子裡的蘋果總數」。
* 原本的總數 $n$ 減去被裝箱的數量，剩下的當然就是「裝不滿一箱、剩下來的蘋果」，也就是餘數（Remainder）。

#### 第二步：進位制轉換 (Base Conversion) 的數學推導

投影片下半部展示了如何把十進位的 $58$ 轉換成三進位的 $(2011)_3$。這個方法叫做「連續除法」（Repeated Division）。為什麼「一直取餘數」就能得到其他進位的數字？我們來推導一下。

假設一個十進位的數字 $N$ 可以被寫成 $b$ 進位（在投影片的例子中 $b=3$），那麼它在數學上可以表示為一個多項式：


$$N = a_k \cdot b^k + \dots + a_2 \cdot b^2 + a_1 \cdot b^1 + a_0 \cdot b^0$$


（這裡的 $a_k, \dots, a_0$ 就是我們要找的 $b$ 進位數字的每一位數，且 $0 \leq a_i < b$）

**尋找最低位數 $a_0$：**
如果我們對 $N$ 取 $b$ 的餘數（也就是 $N \bmod b$ 會發生什麼事？
因為前面所有包含 $b^k, \dots, b^1$ 的項，都乘上了 $b$，這代表它們都能被 $b$ 整除，餘數為 0。唯一無法被整除的，就是最後一項常數項 $a_0 \cdot b^0$（也就是 $a_0$）。


$$N \bmod b = 0 + 0 + \dots + 0 + a_0 = a_0$$


這證明了：**對數字取 $b$ 的餘數，可以直接得到 $b$ 進位表示法中的最低位數（個位數）。**
對應投影片：$58 \bmod 3 = 1$，所以個位數 $a_0$ 是 $1$。

**尋找下一位數 $a_1$：**
我們已經找到 $a_0$ 了，接下來要把目光移向 $a_1$。我們要怎麼把 $a_1$ 變成新的「個位數」呢？答案是：將原本的 $N$ 除以 $b$ 並向下取整（這就是 $\lfloor N/b \rfloor$）。


$$\lfloor \frac{N}{b} \rfloor = \lfloor a_k \cdot b^{k-1} + \dots + a_2 \cdot b^1 + a_1 + \frac{a_0}{b} \rfloor$$


因為 $\frac{a_0}{b}$ 是一個小於 1 的分數，向下取整時會被捨去（就像上一張投影片說的 Floor 概念）。結果會變成：


$$N' = \lfloor \frac{N}{b} \rfloor = a_k \cdot b^{k-1} + \dots + a_2 \cdot b^1 + a_1$$


你看！現在 $a_1$ 變成新的常數項（個位數）了。我們只要對新的 $N'$ 再次取 $\bmod b$，就能得到 $a_1$。
對應投影片：右邊先算 $\lfloor 58 / 3 \rfloor = 19$，左邊接著算 $19 \bmod 3 = 1$，得到 $a_1 = 1$。

重複這個「取餘數找當前位數 $\rightarrow$ 取商數推進到下一位」的過程，直到商數變成 $0$ 為止，就能把所有位數找出來。

#### 第三步：與 Python 程式碼的連結

這段數學推導在寫 Python 時非常直觀。你可能已經聯想到，求商數 $\lfloor n / m \rfloor$ 其實就是 Python 裡的 `//`，而求餘數就是 `%`。

更有趣的是，Python 內建了一個函數叫做 `divmod(x, y)`，它會同時執行這兩個動作，並回傳一個 tuple `(x // y, x % y)`。這完美契合了這個演算法：一次操作就能同時得到「推進到下一輪的數字」與「當前位數的答案」。
