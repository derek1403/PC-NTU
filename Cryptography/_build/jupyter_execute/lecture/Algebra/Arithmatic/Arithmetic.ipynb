## **Definition**: Floor and Ceiling

1. The **floor** $\lfloor x \rfloor$ of $x \in \boldsymbol{R}$ is the largest integer $\leq x$ ，代表不大於 $x$ 的「最大整數」
2. The **ceiling** $\lceil x \rceil$ of $x \in \boldsymbol{R}$ is the smallest integer $\geq x$ ，代表不小於 $x$ 的「最小整數」


**Example**
* $\lfloor e \rfloor = 2$, $\lceil e \rceil = 3$
* $\lfloor -3.1416 \rfloor = -4$, $\lceil -3.1416 \rceil = -3$
* $25 = 3 \times 7 + 4$ [7: divisor, 3: quotient, 4: remainder]
  * $25 \bmod 7 = 25 - \lfloor 25 / 7 \rfloor \times 7 = 25 - 3 \times 7 = 25 - 21=4$ (remainder)

## **Definition**: Modular Function (同餘函數)

$m, n \in \boldsymbol{Z}, m > 0$, define $\boldsymbol{n \bmod m} = n - \lfloor n/m \rfloor \times m$ 
* i.e., the remainder after dividing $n$ by $m$, which is $\ge 0$ and $< m$


* **Example** 58 in the base 3 representation

$$\begin{gather*}
a_0 =& 58 \bmod 3 = 1 &\qquad 19 &= \left\lfloor \frac{58}{3} \right\rfloor \\
a_1 =& 19 \bmod 3 = 1 &\qquad 6 &= \left\lfloor \frac{19}{3} \right\rfloor\\
a_2 =&  6 \bmod 3 = 0 &\qquad 2 &= \left\lfloor \frac{6}{3} \right\rfloor \\
a_3 =&  2 \bmod 3 = 2 &\qquad 0 &= \left\lfloor \frac{2}{3}  \right\rfloor \\
\end{gather*}$$

$$(2011)_3 = 2 \times 3^3 + 0 \times 3^2 + 1 \times 3^1 + 1 \times 3^0 = 58$$

可以把「體」(field)想像成一個「完美的數字世界」。在這個世界裡你可以自由地進行加、減、乘，以及**除法（除了除以 0 之外）**，而且運算完的結果絕對不會掉出這個世界的範圍外。

「可以進行除法」在抽象代數中的嚴格定義是：**每一個非零元素，都必須存在一個「乘法反元素」（Multiplicative Inverse）。** 換句話說，對於任何不為 0 的 $a$ 都要能找到一個 $b$，使得 $a \otimes b = 1$（1 是乘法的單位元素）。

現在，我們來檢查乘法凱萊表（$\otimes$）：

* 找元素 1 的反元素：$1 \otimes 1 = 1$ (存在)
* 找元素 5 的反元素：$5 \otimes 5 = 25 \bmod 6 = 1$ (存在)
* **找元素 2 的反元素**：看表格中 2 的那一個橫列：`0, 2, 4, 0, 2, 4`。**這列裡面根本沒有 1！**
這代表在 $Z_6$ 的世界裡，找不到任何數字 $b$ 可以讓 $2 \otimes b = 1$。也就是說，在這裡你無法「除以 2」。

此外，我們還觀察到一個違反直覺的現象：$2 \otimes 3 = 0$。兩個非零的數字相乘結果居然變成了 0。這兩個數字在代數中被稱為「零因子」（Zero Divisors），在一個「體」的完美世界裡面，是不允許有零因子存在的。

**結論**：因為 $Z_6$ 中並非每個非零元素都有乘法反元素（且存在零因子），所以 **$Z_6$ 不是一個體 (Field)**。

對比之前 $Z_6$ 的表格，你可以從 $Z_7$ 的乘法表中觀察到兩個非常重要的性質，這也是它成為一個「體」（Field）的關鍵：

1. **無零因子 (No Zero Divisors)**：除了 $0$ 乘上任何數等於 $0$ 之外，你找不到任何兩個非零的數字相乘會等於 $0$。
2. **每個非零元素都有乘法反元素 (Multiplicative Inverses)**：請看乘法表內部，除了第 0 行與第 0 列之外， **每一行與每一列都剛好出現一次 $1$** ，這代表每一個非零數字都能找到一個對應的數字，相乘後結果為 $1$：
* $1 \otimes 1 = 1$ (1 的反元素是 1)
* $2 \otimes 4 = 1$ (2 的反元素是 4)
* $3 \otimes 5 = 1$ (3 的反元素是 5)
* $4 \otimes 2 = 1$ (4 的反元素是 2)
* $5 \otimes 3 = 1$ (5 的反元素是 3)
* $6 \otimes 6 = 1$ (6 的反元素是 6)

### 1. 圖片文字與 LaTeX 轉換

**Greatest Common Divisor**

* Assume $a, b \in \boldsymbol{Z}$, $a \neq 0$ or $b \neq 0$
* **Definition** $d \neq 0$ is a **common divisor** of $a$ and $b$ if $d \mid a$ and $d \mid b$
* **Definition** The **greatest common divisor** $d = \gcd(a, b)$ is the largest of the common divisors


* **Example** * Divisors of $20: \pm 1, \pm 2, \pm 4, \pm 5, \pm 10, \pm 20$
* Divisors of $16: \pm 1, \pm 2, \pm 4, \pm 8, \pm 16 \Rightarrow \gcd(20, 16) = 4$


* **Proposition** $a \in \boldsymbol{P}$ (positive integer)
* $\gcd(a, a) = a$
* $\gcd(a, 0) = a$


* **Proposition** $a, b \in \boldsymbol{Z}$, $a \neq 0$ or $b \neq 0$
* $\gcd(a, b) = \gcd(|a|, |b|)$
* $\gcd(a, b) = \gcd(b, a)$

### 2. 老師的觀念解析與推導

我們把這些數學定義當作寫程式前的邏輯梳理，一步步來看。

#### 第一步：定義「因數」與「公因數」

**觀念類比**：想像你有兩根長度分別為 $20$ 公分和 $16$ 公分的木條。所謂的「因數」，就是你能找到一種標準長度的短尺，可以把木條「剛好量完（沒有剩下）」。而「公因數」就是這把短尺不但能剛好量完 $20$ 公分的木條，也能剛好量完 $16$ 公分的木條。

* 數學符號 $d \mid a$ 讀作「$d$ 整除 $a$」。這代表存在某個整數 $k$，使得 $a = k \cdot d$。
* 在 Python 的邏輯中，這等同於 `a % d == 0`。

#### 第二步：觀察特例與基本性質 (Propositions)

投影片下半部列出了幾個看起來很直觀，但在演算法中非常實用的性質。我們來推導並理解它們：

1. **$\gcd(a, a) = a$**
* **推導**：$a$ 能被 $a$ 整除（$a = 1 \cdot a$）。既然兩個數字一樣，能同時整除它們的最大整數當然就是 $a$ 自己。


2. **$\gcd(a, 0) = a$**
* **推導**：這是一個初學者常卡住的地方。為什麼 $0$ 有因數？因為對於任何非零整數 $d$，都滿足 $0 = 0 \cdot d$。也就是說，**任何非零整數都是 0 的因數**（在 Python 裡你不能寫 `0 % 0`，但數學上任何非零整數都能整除 $0$）。
* 既然 $0$ 可以被所有數字整除，那麼 $a$ 和 $0$ 的「公因數」其實就取決於 $a$ 的因數有哪些。而 $a$ 的最大因數就是 $a$，所以 $\gcd(a, 0) = a$。


3. **$\gcd(a, b) = \gcd(|a|, |b|)$**
* **推導**：如投影片 Example 所示，因數總是正負成對出現的（$\pm 1, \pm 2 \dots$）。因為我們要找的是「最大的」公因數，所以一定會挑選正數。因此，原本的數字帶不帶負號（絕對值）根本不影響結果。


4. **$\gcd(a, b) = \gcd(b, a)$**
* **推導**：這只是說明這是一個對稱的關係，誰放前面、誰放後面，共同擁有的最大因數都是同一個。



#### 第三步：從暴力解法到演算法思維

投影片的 Example 展示了「列舉法」：把所有因數列出來，然後找交集裡最大的。
