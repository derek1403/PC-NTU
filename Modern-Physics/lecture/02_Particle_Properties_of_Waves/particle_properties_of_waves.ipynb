## 2.3 光電效應 (Photoelectric Effect)

被光所釋放之電子能量和光的頻率相關。
當光的頻率夠高時會放出電子，這個現象被稱為光電
效應(photoelectric effect)，且被放出的電子稱為光電子
(photoelectron)


![](./pic/The_Photoelectric_Effect.png)


1905年，愛因斯坦將普朗克的概念延伸：光在空間中的傳播也是一個個能量包，稱為「光子」(Photon)。

* 單一光子能量：$E = h\nu$
* 電子游離能量、功函數：$IE = \phi = h\nu_0$
* **解釋光電效應 (能量守恆方程式):**
一顆電子只能「一次吞下一整顆」光子。電子吸收能量 $h\nu$ 後，一部分用來當作「過路費」克服金屬表面的束縛能（稱為功函數 Work function，記為 $\phi$ 或 $W$ 或 $\psi$），剩下的能量就變成電子飛出去的最大動能 $K_{\text{max}}$。

$$\begin{aligned}
h\nu &= K_{\text{max}} + \phi \\
K_{\text{max}} &= h\nu - \phi \\
K_{\text{max}} &= h(\nu - \nu_0)
\end{aligned}$$


光子能量的表述法: 
$$
E &= h\nu = \frac{6.626*10^-34 Js }{1.602*10^-19 J/ev}\nu = 4.136*10^-15\nu ev*s
&= h(c/\lambda) = \frac{4.136*10^-15ev*s*2.998*10^8 m/s }{\lambda} = \frac{1.240*10^-6ev*m}{\lambda}
$$

x 射線管。選定一個金屬作為靶材，加速電壓 V 越高時，電子速度越快且 x 射線波長越短

[金屬鎢和在不同電位下的x射線光譜圖示]

**Example** 以 100 kV 電壓加速電子撞擊鎢靶，求出產生的 x 射線的最短波長。

此波長對應到的頻率
$\nu = \frac{c}{\lambda} = \frac{3.00 \times 10^8 \text{ m/s}}{1.24 \times 10^{-11} \text{ m}} = 2.42 \times 10^{19} \text{ Hz}$

## 2.6 康普頓效應 Compton Effect

當 x 射線照射到物質上時，除了繞射之外，還會發生康普頓效應(Compton effect)，這是一種光子與電子之間的散射現象。當 x 射線光子與物質中的電子碰撞時，光子會將部分能量轉移給電子，導致光子的波長增加（能量減少）。這個效應提供了對光子粒子性的證據，並且是量子力學的重要實驗基礎之一。 ##重點:滿足牛頓力學的碰撞過程，光子和電子在碰撞過程中滿足動量守恆和能量守恆

[放一張康普頓效應圖示]
康普頓效應的數學描述可以通過康普頓散射公式來表達：
$$\lambda' - \lambda = \frac{h}{m_e c} (1 - \cos \phi)$$
其中 $\lambda$ 是入射光子的波長，$\lambda'$ 是散射後光子的波長，$h$ 是普朗克常數，$m_e$ 是電子的質量，$c$ 是光速，$\phi$ 是光子散射角度，$\theta$ 是電子散射角度。


### 進一步驗證光子模型

光子損失的能量 = 電子得到的能量 : $$h\nu - h\nu' = \frac{1}{2} m_e v^2$$

無質量粒子的動量和其能量相關，藉由下列公式 $$p = \frac{E}{c}$$ 可以得出光子動量的表達式 $$p = \frac{h\nu}{c}$$


* For X-axis: 初始動量 = 最終動量 (動量守恆) 與電子散射角度 $\theta$ 的關係? 與光子的關係是?

$$\frac{h\nu}{c}+0 = \frac{h\nu'}{c}\cos \phi + m_e v \cos \theta$$


* For Y-axis: 初始動量 = 最終動量 (動量守恆)

$$0 = \frac{h\nu'}{c}\sin \phi - m_e v \sin \theta$$


$$E=KE+mc^2=\sqrt{(pc)^2+(mc^2)^2}$$


先消除\theta，得到電子的速度 v 與光子散射角度 \phi 的關係，再將 v 代入動量守恆方程式中，最後得到康普頓散射公式

$$p^2c^2 = (h\nu - h\nu')^2 + 2m_e c^2 (h\nu - h\nu')$$



最後得到康普頓散射公式 $$\lambda' - \lambda = \frac{h}{m_e c} (1 - \cos \phi)$$
定義康普頓波長 $\lambda_c = \frac{h}{m_e c}$，當 $\phi = 180^\circ$ 時，$\lambda' - \lambda$ 的最大值為 $\lambda_c$。

物理意義: 光子與電子的碰撞導致光子能量減少（波長增加），同時電子獲得動能，電子被散射的角度會影響光子能量的變化(是誰決定誰?)。這個效應證明了光子具有粒子性質，並且是量子力學的重要實驗基礎之一。

>能量是否守恆? 答案是肯定的，因為光子損失的能量等於電子獲得的動能，這符合能量守恆定律。
>是光子決定電子的散射角度，還是電子決定光子的散射角度? 答案是兩者互相影響，光子和電子在碰撞過程中會同時改變彼此的動量和能量，因此散射角度是由兩者的交互作用決定的。 但是實驗觀察到的\phi分布是均勻的，這意味著光子和電子的散射角度沒有特定的偏好，這也符合量子力學中粒子行為的隨機性。是否正確? 還是說\phi分布是由電子的初始狀態決定的? 還是說\phi分布是由光子的初始狀態決定的? 還是說\phi分布是由兩者的初始狀態共同決定的? 這個問題我也不確定。


[放一張康普頓散射的實驗圖示 相對強度比上波長在不同角度\phi的關係圖示]
康普頓散射的實驗確認，根據式 (2.21)，散射角度越大，則波長的變化越大。


**Example** 以 0.1 nm 波長的 x 射線照射到靶材上，當散射角度 $\phi$ 為 90 度時，求出散射後光子的波長 $\lambda'$。求出在散射 x 射線中的極大波長，求出反彈電子的極大動能

康普頓波長 $\lambda_c = \frac{h}{m_e c} = 2.43 \times 10^{-12} \text{ m}$

波長變化 $\lambda' - \lambda = \lambda_c (1 - \cos \phi) = 2.43 \times 10^{-12} \text{ m} \times (1 - \cos 90^\circ) = 2.43 \times 10^{-12} \text{ m}$

散射後光子波長 $\lambda' = \lambda + (\lambda' - \lambda) = 0.1 \text{ nm} + 2.43 \times 10^{-12} \text{ m} = 0.10243 \text{ nm}$

反彈電子的極大動能


$$K_{\text{max}} = h\nu - h\nu' = \frac{hc}{\lambda} - \frac{hc}{\lambda'} = hc \left(\frac{1}{\lambda} - \frac{1}{\lambda'}\right)$$
$$K_{\text{max}} = (6.626 \times 10^{-34} \text{ J} \cdot \text{s})(3.00 \times 10^8 \text{ m/s}) \left(\frac{1}{0.1 \times 10^{-9} \text{ m}} - \frac{1}{0.10243 \times 10^{-9} \text{ m}}\right)$$
$$K_{\text{max}} \approx 1.94 \text{ keV}$$


如果x射線不去撞電子而是去撞原子核，會發生什麼事? 答案是會發生核散射(nuclear scattering)，這是一種光子與原子核之間的散射現象。當 x 射線光子與原子核碰撞時，光子會將部分能量轉移給原子核，導致光子的波長增加（能量減少）。這個效應提供了對光子粒子性的證據，並且是量子力學的重要實驗基礎之一。跟電子散射相比，核散射的效應通常較小，因為原子核的質量遠大於電子的質量，因此光子與原子核的碰撞會導致較小的能量轉移和波長變化。

x射線波長~10^-10 m
晶格大小~10^-10 m
核大小~10^-15 m
電子大小~10^-18 m
因此x射線的波長和晶格大小相當，可以發生繞射
x射線的波長遠大於核大小，無法發生繞射，但可以發生散射，這就是康普頓效應的來源
x射線的波長遠大於電子大小，無法發生繞射，但可以發生散射，這也是康普頓效應的來源

散射和繞射的區別，散射是指光子與物質中的粒子（如電子或原子核）發生碰撞，導致光子能量和方向的改變；繞射是指光子在遇到障礙物或通過狹縫時，波前發生彎曲和干涉現象，形成特定的圖樣。散射通常涉及能量轉移，而繞射則主要涉及波的干涉和衍射效應。

>無線電波為什麼不會和牆體發生散射?如果是因為無線電特徵尺度長於牆體，那麼x射線的特徵長於核，為什麼還是會和核發生散射? 答案是無線電波的波長遠大於牆體的特徵尺度，因此無線電波無法與牆體發生有效的散射；而x射線的波長雖然遠大於核的特徵尺度，但仍然可以與核發生散射，這是因為x射線具有足夠的能量來與核中的粒子（如質子和中子）進行相互作用，導致能量轉移和波長變化










## 2.7 配對產生 Pair Production 

在碰撞中一個光子會將其全部（光電效應）或部分（康普頓效應）能量給予電子，對一個光子而言也可能轉化成一個電子和一個正子（帶正電的電子）也是有可能的，這個過程稱為配對產生(pair production)，電磁能量會轉換為物質。



當光子具有足夠的能量時，可以在核附近產生一對粒子和反粒子，這個過程稱為配對產生(pair production)。例如，一個高能光子可以在核附近產生一對電子和正電子。這個過程需要滿足能量守恆和動量守恆的條件，並且通常需要光子的能量至少為 1.022 MeV（兩倍的電子靜止能量）才能發生。

因此可以得出$KE_e = KE_p = \sqrt{(h\nu /2)^2 + (m_e c^2)^2} - m_e c^2$


$h\nu=2\gamma m_e c^2 (v/c)\cos \theta$
因為v/c<1 ,cos theta<1,所以h\nu<2\gamma m_e c^2

## 2.8  光子與重力 Photons and Gravity

在一個重力場中降落的光子和石頭一樣會得到能量，這個能量的獲得會顯示在頻率由 v 增加至 v’ 的時候。

雖然光子沒有靜止質量，它們仍顯得像是具有重量一樣。
– 光子「質量」$m=P/v=h \nu/c^2$

光子在地球下降高度 H 之後的能量 hnu'=hnu(1+gH/c^2)


引力紅位移 (gravitational red shift)

$$\frac{\delta \nu}{\nu} =1-\frac{\nu'}{\nu} = \frac{GM}{c^2 R}$$


黑洞 (Black Hole)
– 史瓦茲希德半徑 $R_s = \frac{2GM}{c^2}$
– (factor 2 from general relativity correction)


**Example** 計算地球表面附近的引力紅位移，假設地球的質量為 $M = 5.97 \times 10^{24} \text{ kg}$，半徑為 $R = 6.371 \times 10^6 \text{ m}$，以及光速 $c = 3.00 \times 10^8 \text{ m/s}$。1960年時，下降光子的能量增加首度由龐德(Pound)和瑞布卡(Rebka)在哈佛大學觀察到。在他們的研究中， H 為 22.5m，求出頻率7.3*10^14 Hz的紅光子從 22.5 m 處降落之後，頻率變化為何。

$$
\nu'-\nu=(gH/c^2)\nu=(9.8 m/s^2 * 22.5 m / (3.00 \times 10^8 m/s)^2) \nu = 2.45 \times 10^{-15} \nu = 1.8Hz
