### 量子力學基本規則

* 波函數 $\Psi$
* 在那個位子和時間發現物體的機率 $\Psi \Psi^*$
  * $\Psi , \Psi^* \in \mathbb{C}$ 是共軛負數
  * 例如 $\Psi= A+Bi$ 則 $\Psi^*=A-Bi$
* 歸一化條件（Normalization Condition） $\int_{-\infty}^{\infty} |\Psi|^2 dV = 1$
  * $|\Psi|^2$ 是 機率密度
  * $dV$ 體積微元
  * $= 1$ 代表全空間找到粒子的總機率為 100%。
* 波函數 $\Psi$ 本身和其對空間的一階微分要連續而且有單一值
  * $\Psi$ 本身要和 $\frac{d}{dt} [\Psi]$ $\frac{d}{dx} [\Psi]$ 要連續且唯一
    * 因為要符合波動方程式 $\frac{\partial^2 y}{\partial x^2} = \frac{1}{v^2} \frac{\partial^2 y}{\partial t^2}$
    * 只要一接微分連續就可，二街微分不需要連續(一般理想的位能井都不會連續)，但要可積分
    * 波函數不連續：意味著粒子在A點出現的機率與緊鄰的B點有巨大的斷層，這會導致機率流變成無窮大（粒子瞬間移動），這違反了因果律。
    * 一階微分不連續：這代表動量在瞬間發生無窮大的改變。在現實中，任何力場（位能）都有一個緩衝的空間，所以微分項通常是平滑的。
    * 物理世界真的連續嗎?
      * 廣義相對論與時空的「平滑性」:廣義相對論認為時空是連續且可微的「流形」（Manifold）。這意味著你可以無限地分割空間，且在任何一點都能定義切平面（微分）。
        * 危機：當我們把大量的質量壓縮到極小空間時（如黑洞中心或大霹靂初期），時空曲率會趨向無窮大（奇異點）。這顯示廣義相對論這套「連續」的語言在極限下失效了，我們需要一種更底層的物理架構。
      * 量子引力中的「時空量子化」: 許多量子引力理論認為，時空並非如絲綢般連續，而是像織布一樣，由細小的「纖維」編織而成：
        * 圈量子重力（Loop Quantum Gravity, LQG）：這是最直接挑戰連續性的理論。它預言空間本身是量子化的，存在最小的體積與面積單位。在普朗克尺度下，你無法再往下分割，這就像電腦螢幕的像素，遠看連續，近看則是一個個點。
        * 弦論（String Theory）：雖然弦論在數學上仍使用連續背景，但它認為基本的物理實體不是「點」，而是「弦」。因為弦有最小長度（普朗克長度），我們無法探測比弦更小的尺度，這在實際上為空間的「連續分割」設定了一個天花板。
      * 非交換幾何（Non-commutative Geometry）:有一些觀點認為，在極小尺度下，座標 x和y 可能不滿足乘法交換律（類似量子力學中的位置與動量）。如果 xy \ne yx ，那麼「點」的概念就消失了。在這種尺度下，空間不再是點的集合，而是一種模糊的幾何結構，連續性自然無從談起。
      * 物理直覺的挑戰：全像原理
        * 全像原理（Holographic Principle） 認為，一個區域內的所有資訊可以編碼在其邊界表面上。這暗示了資訊量是有限的（離散的），而非無限的（連續的）。如果資訊是離散的，那麼承載資訊的時空背景也很難是絕對連續的。
      * 總結觀點

## 5.3 運算子 Operators

$\Psi = Ae^{-(i/\hbar)(Et - px)}$


$\frac{\partial \Psi}{\partial x} = \frac{i}{\hbar} p \Psi$

$\frac{\partial \Psi}{\partial t} = - \frac{i}{\hbar} E \Psi$


可以得到

$p \Psi = \frac{\hbar}{i} \frac{\partial}{\partial x} \Psi$

$E \Psi = i\hbar \frac{\partial}{\partial t} \Psi$


所以就定義 運算子。當運算子碰到 $\Phi$，就可以的到 運算子\hat下面的東西*$\Phi$


$\hat{p} \overset{\text{def}}{=} \frac{\hbar}{i} \frac{\partial}{\partial x}$

$\hat{E}\overset{\text{def}}{=}  i\hbar \frac{\partial}{\partial t}$

$\hat{E} \overset{\text{def}}{=} \hat{\text{KE}} + \hat{U}$

$\hat{\text{KE}} \overset{\text{def}}{=} \frac{\hat{p}^2}{2m} = \frac{1}{2m} \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right)^2 = -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2}$


組合後可以變成剛剛eqution的operator 

$$\begin{gather*}
i\hbar \frac{\partial}{\partial t} &=& -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + U\\
i\hbar \frac{\partial \Psi}{\partial t}&=&  -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U\Psi\\
\end{gather*}$$

* 總能
* 動能
* 位能

* 透過這樣的組合我們不用知道 G(x)這個物理量本身對x的關係(通常我們都不知道)，只要知道他的算子 $\hat{G}$就可以操作了


### 實際操作

$$\begin{gather*}
\langle p \rangle &=& \int_{-\infty}^{\infty} \Psi^* \hat{p} \Psi dx \\
&=&\int_{-\infty}^{\infty} \Psi^* \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right) \Psi dx \\
&=& \frac{\hbar}{i} \int_{-\infty}^{\infty} \Psi^* \frac{\partial \Psi}{\partial x} dx \\
\end{gather*}$$

$$\begin{gather*}
\langle E \rangle &=& \int_{-\infty}^{\infty} \Psi^* \hat{E} \Psi dx \\
&=& \int_{-\infty}^{\infty} \Psi^* \left( i\hbar \frac{\partial}{\partial t} \right) \Psi dx \\
&=& i\hbar \int_{-\infty}^{\infty} \Psi^* \frac{\partial \Psi}{\partial t} dx
\end{gather*}$$

$$\begin{gather*}
\langle G(x, t) \rangle = \int_{-\infty}^{\infty} \Psi^* \hat{G} \Psi dx
\end{gather*}$$




物理意義: 在相同的量子態 $\Psi $ 下，重複測量物理量 $G$ 無數次後得到的平均值。


. 為什麼一定要是 $\int \Psi^* \hat{G} \Psi \, dx$？在量子力學中，物理量（如位置、動量、能量）並不是簡單的數值，而是算符（通常包含微分）。$\Psi $ 是「被操作者」：算符 $\^{G}$ 必須作用在波函數 $\Psi $ 上才有意義。就像 $\frac{d}{dx}$ 必須寫在 $x^{2}$ 前面（即 $\frac{d}{dx}x^2$），不能寫成 $x^2 \frac{d}{dx}$。$\Psi ^{*}$ 的角色：為了讓計算結果是一個實數（物理測量值必須是實數），我們需要利用複數共軛 $\Psi ^{*}$ 與原函數 $\Psi $ 相乘（類似向量內積的概念）。

如果交換順序會怎麼樣？如果我們隨便調換位置，例如寫成 $\int \hat{G} (\Psi^* \Psi) \, dx$ 或 $\int \Psi \Psi^* \hat{G} \, dx$：數學錯誤：如果 $\^{G}$ 是動量算符 $\hat{p} = -i\hbar \frac{\partial}{\partial x}$，它會去微分它後面的所有東西。在 $\int \Psi^* \hat{G} \Psi \, dx$ 中，它只微分 $\Psi $。如果你寫成 $\int \hat{G} (\Psi^* \Psi) \, dx$，它會同時微分 $\Psi ^{*}$ 和 $\Psi $，根據微分的乘法法則（Product rule），這會多出多餘的項，算出來的結果會完全錯誤。複數問題：隨便交換順序通常會導致結果變成複數，而現實世界中你不可能量測到一個「長度是 $3 + 2i$ 公尺」的距離。

## 5.4 薛丁格方程式：穩態型式 Schrödinger’s Equation: Steady-State Form


本徵值及本徵函數


$$\Psi = Ae^{-(i/\hbar)(Et-px)} = Ae^{-(iE/\hbar)t} e^{(ip/\hbar)x} = \psi e^{-(iE/\hbar)t}$$

$$i\hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U\Psi$$

$$\frac{\partial^2 \psi}{\partial x^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$$

$$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial z^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$$


本徵的意思是 hat 放的operator 會變成數字

$\hat{G}\psi_n = G_n\psi_n$

* 這個 $G_n$就是數字(純量)
* 這在數學上的意思是：當算符 $\^{G}$ 作用在特定的函數 $\psi _{n}$ 上時，結果居然只是把這個函數縮放了 $G_{n}$ 倍，而函數本身的形狀完全沒變。這時候，我們稱 $G_{n}$ 是這個系統可能的觀測值

漢彌爾頓運算子 定義

$\hat{H} \overset{\text{def}}{=} -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + U$
* 它代表的是系統的總能量算符（動能 + 位能）。你可以把它想像成一個「能量處理機」。



薛丁格方程式
$\hat{H}\psi_n = E_n\psi_n$

薛丁格方程式 ($\hat{H}\psi_n = E_n\psi_n$)：這是最著名的應用。當你把「能量處理機」作用在波函數上，吐出來的縮放倍率 $E_{n}$ 就是該量子態下的能量等級。這也說明了為什麼能量是「一份一份」的（量子化），因為只有特定的 $\psi _{n}$ 才能滿足這個等號。

* 只有H作用在上面的形式才叫薛丁格方程式
* En能量量子化

$G = 4$

## 5.7 穿隧效應 (Tunnel Effect)


一個沒有足夠能量可以跨越電位障的粒子仍有可能穿隧電位障


近似穿隧機率

$$T = e^{-2k_2 L}$$

$$k_2 = \frac{\sqrt{2m(U - E)}}{\hbar}$$


證明

1. 分段薛丁格方程式與解這是將空間分為三層（入射區、障礙區、穿透區）的波函數解：


第一區 (x < 0) 與 第三區 (x > L)

\frac{d^2 \psi_{\text{I}}}{dx^2} + \frac{2m}{\hbar^2} E \psi_{\text{I}} = 0 \implies \psi_{\text{I}} = Ae^{ik_1 x} + Be^{-ik_1 x}

第二區 (0 < x < L, 障礙區)
\frac{d^2 \psi_{\text{II}}}{dx^2} + \frac{2m}{\hbar^2} (E - U) \psi_{\text{II}} = 0 \implies \psi_{\text{II}} = Ce^{-k_2 x} + De^{k_2 x}

定義波數
k_1 = \frac{\sqrt{2mE}}{\hbar}, \quad k_2 = \frac{\sqrt{2m(U - E)}}{\hbar}

2. 邊界條件 (Boundary Conditions)為了求出係數，必須要求波函數及其斜率在 \(x=0\) 和 \(x=L\) 處連續：

x = 0 處
A + B = C + D
ik_1 A - ik_1 B = -k_2 C + k_2 D

x = L 處
Ce^{-k_2 L} + De^{k_2 L} = Fe^{ik_1 L}
-k_2 Ce^{-k_2 L} + k_2 De^{k_2 L} = ik_1 Fe^{ik_1 L}

3. 穿隧機率 \(T\) 的精確解與近似解



精確穿隧機率
T = \left[ 1 + \frac{1}{4} \left( \frac{k_1}{k_2} + \frac{k_2}{k_1} \right)^2 \sinh^2(k_2 L) \right]^{-1} \approx \left[ \frac{16}{4 + (k_2/k_1)^2} \right] e^{-2k_2 L}

參數關係
\left( \frac{k_2}{k_1} \right)^2 = \frac{2m(U - E)/\hbar^2}{2mE/\hbar^2} = \frac{U}{E} - 1

最終簡化近似
T \approx e^{-2k_2 L}



% 薛丁格方程式 (第一區與第三區)
\frac{d^2 \psi_{\text{I}}}{dx^2} + \frac{2m}{\hbar^2} E \psi_{\text{I}} = 0

\frac{d^2 \psi_{\text{III}}}{dx^2} + \frac{2m}{\hbar^2} E \psi_{\text{III}} = 0

% 通解形式
\psi_{\text{I}} = Ae^{ik_1 x} + Be^{-ik_1 x}

\psi_{\text{III}} = Fe^{ik_1 x} + Ge^{-ik_1 x}

% 波數定義
k_1 = \frac{\sqrt{2mE}}{\hbar} = \frac{p}{\hbar} = \frac{2\pi}{\lambda}

% 第一區波函數分解 (入射與反射)
\psi_{\text{I}+} = Ae^{ik_1 x}

\psi_{\text{I}-} = Be^{-ik_1 x}

\psi_{\text{I}} = \psi_{\text{I}+} + \psi_{\text{I}-}

% 第三區波函數 (僅包含穿透波)
\psi_{\text{III}+} = Fe^{ik_1 x}

\psi_{\text{III}} = \psi_{\text{III}+} = Fe^{ik_1 x}

% 穿隧機率 (Transmission Probability) 定義
T = \frac{|\psi_{\text{III}+}|^2 v_{\text{III}+}}{|\psi_{\text{I}+}|^2 v_{\text{I}+}} = \frac{FF^* v_{\text{III}+}}{AA^* v_{\text{I}+}}




% 第二區薛丁格方程式
\frac{d^2 \psi_{\text{II}}}{dx^2} + \frac{2m}{\hbar^2}(E - U)\psi_{\text{II}} = \frac{d^2 \psi_{\text{II}}}{dx^2} - \frac{2m}{\hbar^2}(U - E)\psi_{\text{II}} = 0

% 第二區通解
\psi_{\text{II}} = Ce^{-k_2 x} + De^{k_2 x}

% 衰減係數定義
k_2 = \frac{\sqrt{2m(U - E)}}{\hbar}

% x = 0 處的邊界條件
\left. 
\begin{aligned}
\psi_{\text{I}} &= \psi_{\text{II}} \\
\frac{d\psi_{\text{I}}}{dx} &= \frac{d\psi_{\text{II}}}{dx}
\end{aligned} 
\right\} x = 0

% x = L 處的邊界條件
- x = L \text{ 處的邊界條件 } \quad
\left. 
\begin{aligned}
\psi_{\text{II}} &= \psi_{\text{III}} \\
\frac{d\psi_{\text{II}}}{dx} &= \frac{d\psi_{\text{III}}}{dx}
\end{aligned} 
\right\} x = L

% 聯立方程式組
A + B = C + D

ik_1 A - ik_1 B = -k_2 C + k_2 D

Ce^{-k_2 L} + De^{k_2 L} = Fe^{ik_1 L}

-k_2 Ce^{-k_2 L} + k_2 De^{k_2 L} = ik_1 Fe^{ik_1 L}






% 振幅比 A/F 的精確形式
\left( \frac{A}{F} \right) = \left[ \frac{1}{2} + \frac{i}{4} \left( \frac{k_2}{k_1} - \frac{k_1}{k_2} \right) \right] e^{(ik_1 + k_2)L} + \left[ \frac{1}{2} - \frac{i}{4} \left( \frac{k_2}{k_1} - \frac{k_1}{k_2} \right) \right] e^{(ik_1 - k_2)L}

% 近似條件
\frac{k_2}{k_1} - \frac{k_1}{k_2} \approx \frac{k_2}{k_1}

% 近似後的 A/F 及其共軛
\left( \frac{A}{F} \right) = \left( \frac{1}{2} + \frac{ik_2}{4k_1} \right) e^{(ik_1 + k_2)L}

\left( \frac{A}{F} \right)^* = \left( \frac{1}{2} - \frac{ik_2}{4k_1} \right) e^{(-ik_1 + k_2)L}

% 穿隧機率 T 的推導
- \text{ 穿隧機率 } T = \frac{FF^* v_{\text{III}+}}{AA^* v_{\text{I}+}} = \left( \frac{AA^*}{FF^*} \right)^{-1} = \left[ \frac{16}{4 + (k_2/k_1)^2} \right] e^{-2k_2 L}

% 波數比與能量的關係
\left( \frac{k_2}{k_1} \right)^2 = \frac{2m(U - E)/\hbar^2}{2mE/\hbar^2} = \frac{U}{E} - 1

% 最終近似公式
- \text{ 近似穿隧機率 } \quad T = e^{-2k_2 L}







### **例題 5.7：** 穿隧機率

具有能量 1.0 eV 和 2.0 eV 的電子射向 10.0 eV，寬 0.5nm 的障。(a) 求它們個別的穿隧機率。(b) 如果障的寬度加倍會有怎樣的影響？

(a) 對於 1.0 eV 的電子而言

$$
k_2 = \frac{\sqrt{2m(U - E)}}{\hbar} \\
= \frac{\sqrt{(2)(9.1 \times 10^{-31} \text{ kg})[(10.0 - 1.0) \text{ eV}](1.6 \times 10^{-19} \text{ J/eV})}}{1.054 \times 10^{-34} \text{ J} \cdot \text{s}} \\
= 1.6 \times 10^{10} \text{ m}^{-1}

L = 0.50 \text{ nm} = 5.0 \times 10^{-10} \text{ m}

2k_2L = (2)(1.6 \times 10^{10} \text{ m}^{-1})(5.0 \times 10^{-10} \text{ m}) = 16
$$

近似的穿隧機率為
$T_1 = e^{-2k_2 L} = e^{-16} = 1.1 \times 10^{-7}$


平均每八百九十萬個 1.0 eV 的電子會有一個可以穿隧過 10 eV 的能障。2.0 eV 電子的計算相似，可以得到$T_2 = 2.4 \times 10^{-7}$
這些電子有兩倍的機率可以穿隧障。

(b) 如果障的寬度加倍變成 1.0 nm，穿隧的機率變成

$T_1' = 1.3 \times 10^{-14}$

$T_2' = 5.1 \times 10^{-14}$

物理意義: 明顯地 T 對於障寬度的變化比對粒子能量的改變來得敏感

這個結果 \(1.1 \times 10^{-7}\) 告訴我們，在目前的設定下（勢壘高度 \(10.0 \text{ eV}\)、電子能量 \(1.0 \text{ eV}\)、寬度 \(0.5 \text{ nm}\)），電子穿透過去的可能性大約是千萬分之一。雖然這個數字看起來很小，但在量子尺度的實驗中，這是一個可以被觀測到的顯著效應。如果這是在古典物理世界，機率會是絕對的 0。

如果我們把牆壁的厚度 \(L\) 增加一倍（變成 \(1.0 \text{ nm}\)），指數項會從 \(16\) 變成 \(32\)，那麼穿隧機率會變成 \(e^{-32} \approx 1.2 \times 10^{-14}\)，機率直接縮小了一千萬倍！這展示了為什麼穿隧效應對「距離」極其敏感。

## 5.8 諧振子 (Harmonic Oscillator)

其能階為相等間隔
(什麼意思?)


分子的振動：

1. 運動方程：描述了一個物體受到的力與位移成正比，且方向相反（虎克定律）。
虎克定律
F = -kx

\frac{d^2x}{dt^2} + \frac{k}{m}x = 0


x = A \cos (2\pi \nu t + \phi)

2. 頻率 \(\nu \)：這是系統的自然頻率。在量子力學中，能量會與這個頻率掛鉤（\(E = (n + 1/2)h\nu\)）。

\nu = \frac{1}{2\pi} \sqrt{\frac{k}{m}}

3. 位能 \(U(x) = \frac{1}{2}kx^2\)：這是一個拋物線形狀的位能井。與之前的「方井」不同，這裡的牆壁不是垂直的，而是平滑彎曲的。
U(x) = -\int_{0}^{x} F(x) dx = k \int_{0}^{x} x dx = \frac{1}{2}kx^2



去因次化 (Non-dimensionalization)

$$

\frac{d^2\psi}{dx^2} + \frac{2m}{\hbar^2} \left( E - \frac{1}{2}kx^2 \right)\psi = 0

y = \left( \frac{1}{\hbar} \sqrt{km} \right)^{1/2} x = \sqrt{\frac{2\pi m \nu}{\hbar}} x

\alpha = \frac{2E}{\hbar} \sqrt{\frac{m}{k}} = \frac{2E}{h\nu}

\frac{d^2\psi}{dy^2} + (\alpha - y^2)\psi = 0

$$
最後一行是一個標準的微分方程 它的解會涉及到 Hermite 多項式 ，只有在 α = 2n + 1會有解



從這個簡化後的方程式可以推導出，為了讓波函數在無窮遠處趨近於零，\(\alpha \) 必須等於某些特定的數值（即 \(\alpha = 2n + 1\)）。這就是為什麼能量 \(E\) 必須是 \(h\nu\) 的半整數倍。


諧振子的能階

* E_n = (n + \frac{1}{2}) h \nu \quad n = 0, 1, 2, \dots 
* E_0 = \frac{1}{2} h \nu

* 零點能量 (Zero-Point Energy)：這是量子力學與古典力學最大的不同。在古典力學中，物體最低能量可以是 0（靜止在原點）。但在量子力學中，當 \(n=0\) 時，能量依然有 \(E_0 = \frac{1}{2}h\nu\)。這代表粒子永遠不會停止振動，即使在絕對零度下也是如此。這也符合不確定性原理：如果粒子靜止，位置和動量就同時確定了，這是被禁止的。


* 等間距能階：你會發現相鄰能階的差值 \(\Delta E\) 永遠等於 \(h\nu\)。這與之前的「無限深位能井」（能量與 \(n^{2}\) 成正比）不同。在諧振子模型中，能量是等距跳階的，這在分子振動光譜（紅外光譜）的研究中極其重要。




波函數 諧振子 

$$\psi_n = \left( \frac{2mv}{\hbar} \right)^{1/4} (2^n n!)^{-1/2} H_n(y) e^{-y^2/2}$$

1. 歸一化係數 \(\left(\frac{2mv}{\hbar }\right)^{1/4}(2^{n}n!)^{-1/2}\)
這串複雜的常數是為了確保粒子在空間中出現的總機率等於 \(1\)。其中包含階乘 \(n!\)，這與能階的高低有關。

2. Hermite polynomial
 \(H_n(y)\)：這是公式的「骨架」，決定了波函數在中間區域如何震盪。當 \(n=0\) 時，\(H_0(y) = 1\)。當 \(n=1\) 時，\(H_1(y) = 2y\)。隨階數 \(n\) 增加，波函數的節點（Node）也會隨之增加。

3. 高斯衰減項 \(e^{-y^{2}/2}\)：
這是公式的「包絡線」。它確保當粒子遠離平衡位置時（\(y\) 很大時），波函數會快速趨近於零，代表粒子被束縛在位能井內。

| n | $H_n(y)$ | $\alpha_n$ | $E_n$ |
|---|---|---|---|
| 0 | 1 | 1 | $\frac{1}{2}h\nu$ |
| 1 | $2y$ | 3 | $\frac{3}{2}h\nu$ |
| 2 | $4y^2 - 2$ | 5 | $\frac{5}{2}h\nu$ |
| 3 | $8y^3 - 12y$ | 7 | $\frac{7}{2}h\nu$ |
| 4 | $16y^4 - 48y^2 + 12$ | 9 | $\frac{9}{2}h\nu$ |
| 5 | $32y^5 - 160y^3 + 120y$ | 11 | $\frac{11}{2}h\nu$ |




### **例題 5.8：**


求出諧振子前兩個態的期望值〈x 〉，動量期望值 <P>、位置的平方期望值 \(\langle y^2 \rangle\)（這通常用來計算不確定性 \(\Delta x\)）

(a)位置

$$\langle x \rangle = \int_{-\infty}^{\infty} x |\psi|^2 dx$$

\psi_0 = \left( \frac{2m\nu}{\hbar} \right)^{1/4} e^{-y^2/2}
* 基態 \(\psi _{0}\) (Ground State)：
* 這是一個單純的高斯函數（鐘形曲線）。
* 特性：函數在全區域都沒有穿過零點（無節點），機率最高的地方就在中心點 \(y=0\)
\psi_1 = \left( \frac{2m\nu}{\hbar} \right)^{1/4} \left( \frac{1}{2} \right)^{1/2} (2y) e^{-y^2/2}
* 多了一個 \((2y)\) 項，這就是第一階厄米多項式 \(H_1(y)\)。
* 特性：這是一個奇函數。在 \(y=0\) 處，波函數剛好等於 \(0\)（稱為節點 Node）。
* 這意味著粒子在第一激發態時，絕對不會出現在中心平衡位置。

n = 0: \quad \int_{-\infty}^{\infty} y|\psi_0|^2 dy = \int_{-\infty}^{\infty} ye^{-y^2} dy = -\left[ \frac{1}{2}e^{-y^2} \right]_{-\infty}^{\infty} = 0

* 在 \(n=0\) 的情況中，積分項是 \(ye^{-y^{2}}\)。當 \(y \to \pm \infty\) 時，\(e^{-y^{2}}\) 衰減得極快，導致上限與下限代入後都是 \(0\)，相減結果自然是 \(0\)。

n = 1: \quad \int_{-\infty}^{\infty} y|\psi_1|^2 dy = \int_{-\infty}^{\infty} y^3e^{-y^2} dy = -\left[ \left( \frac{1}{4} + \frac{y^2}{2} \right) e^{-y^2} \right]_{-\infty}^{\infty} = 0
* 在 \(n=1\) 的情況中，雖然裡面變成了 \(y^{3}\)（因為 \(\psi _{1}\) 包含了一個 \(y\)，平方後變 \(y^{2}\)，再乘以期望值公式裡的 \(y\) 變成 \(y^{3}\)），但同樣因為指數衰減項 \(e^{-y^{2}}\) 的主導，最後結果依舊是 \(0\)。


(b) 動量

\hat{p} = \frac{\hbar}{i} \frac{d}{dx} \quad (\text{動量算符})

\langle p \rangle = \int_{-\infty}^{\infty} \psi_0^* \left( \frac{\hbar}{i} \frac{d}{dx} \right) \psi_0 dx

\text{由於 } \psi_0 \propto e^{-y^2/2} \text{，其微分為：}
\frac{d}{dy}(e^{-y^2/2}) = -y e^{-y^2/2}

\langle p \rangle \propto \frac{\hbar}{i} \int_{-\infty}^{\infty} (e^{-y^2/2}) (-y e^{-y^2/2}) dy = \frac{\hbar}{i} \int_{-\infty}^{\infty} -y e^{-y^2} dy

\text{因為 } -y e^{-y^2} \text{ 是奇函數，在對稱區間積分結果為：}
\langle p \rangle = 0

雖然期望值是 0，但這並不代表動量是 0。如果你去算動量的平方期望值 \(\langle p^2 \rangle\)，結果就會是一個大於 0 的數值，這代表粒子確實有動能，只是平均方向抵銷了。


(c)不確定性


\text{已知基態波函數為：}
\psi_0(y) = \left( \frac{1}{\pi} \right)^{1/4} e^{-y^2/2} \quad (\text{已歸一化形式})

\langle y^2 \rangle = \int_{-\infty}^{\infty} \psi_0^* y^2 \psi_0 dy

\langle y^2 \rangle = \int_{-\infty}^{\infty} \left( \frac{1}{\pi} \right)^{1/2} e^{-y^2/2} \cdot y^2 \cdot e^{-y^2/2} dy

\langle y^2 \rangle = \frac{1}{\sqrt{\pi}} \int_{-\infty}^{\infty} y^2 e^{-y^2} dy

\text{代入高斯積分公式 } \int_{-\infty}^{\infty} y^2 e^{-y^2} dy = \frac{\sqrt{\pi}}{2} \text{：}

\langle y^2 \rangle = \frac{1}{\sqrt{\pi}} \cdot \frac{\sqrt{\pi}}{2} = \frac{1}{2}


* 非零結果：與 \(\langle y \rangle = 0\) 不同，\(\langle y^2 \rangle\) 的結果是 \(1/2\)。這告訴我們，雖然平均位置在原點，但粒子並非靜止，而是「散開」在原點附近的
* 不確定性：由於 \(\langle y \rangle = 0\)，位置的不確定性（標準差）直接就是 \(\Delta y = \sqrt{\langle y^2 \rangle - \langle y \rangle^2} = \sqrt{1/2}\)。

要計算無因次單位下的 $\langle p^2 \rangle$ 並推導出不確定性原理的最小值，我們可以延續基態波函數 $\psi_0(y) = (\pi)^{-1/4} e^{-y^2/2}$ 的推導。
 1. 計算 $\langle p^2 \rangle$
在無因次單位系統中，動量算符簡化為 $\hat{p} = -i \frac{d}{dy}$（這裡我們設定 $\hbar=1$ 以方便觀察結構）。因此 $\hat{p}^2 = (-i \frac{d}{dy})^2 = -\frac{d^2}{dy^2}$。
 第一步：計算二階微分
對 $\psi_0 = C e^{-y^2/2}$ 進行兩次微分：

* 一階微分：$\frac{d}{dy} e^{-y^2/2} = -y e^{-y^2/2}$
* 二階微分：$\frac{d^2}{dy^2} e^{-y^2/2} = \frac{d}{dy}(-y e^{-y^2/2}) = (-1 + y^2) e^{-y^2/2}$

第二步：代入期望值積分
$$\langle p^2 \rangle = \int_{-\infty}^{\infty} \psi_0^* \left( -\frac{d^2}{dy^2} \right) \psi_0 dy$$ $$\langle p^2 \rangle = \int_{-\infty}^{\infty} \psi_0^* \left( 1 - y^2 \right) \psi_0 dy$$ $$\langle p^2 \rangle = \int \psi_0^* \psi_0 dy - \int \psi_0^* y^2 \psi_0 dy$$ 
第三步：利用已知結果

* 第一個積分是波函數的歸一化，等於 1。
* 第二個積分是我們剛才算過的 $\langle y^2 \rangle = \mathbf{1/2}$。
* 所以：$\langle p^2 \rangle = 1 - 1/2 = \mathbf{1/2}$。

 推導海森堡不確定性最小值
根據不確定性的定義，標準差（不確定度）為：

* $\Delta y = \sqrt{\langle y^2 \rangle - \langle y \rangle^2}$
* $\Delta p = \sqrt{\langle p^2 \rangle - \langle p \rangle^2}$

我們已知對於諧振子基態：$\langle y \rangle = 0$ 且 $\langle p \rangle = 0$。
因此：

* $\Delta y = \sqrt{1/2}$
* $\Delta p = \sqrt{1/2}$

乘積計算：
$$\Delta y \cdot \Delta p = \sqrt{1/2} \cdot \sqrt{1/2} = \mathbf{\frac{1}{2}}$$ 
若我們換回帶有物理單位（包含 $\hbar$）的真實單位，這個結果就是：
$$\Delta x \cdot \Delta p_x = \mathbf{\frac{\hbar}{2}}$$
