## 5.1 量子力學基礎概念 

### 量子力學與古典力學的差異

在探討量子力學之前，我們必須先釐清古典力學與近代物理在適用範圍與基本哲學上的根本差異。古典力學在描述巨觀、低速物體時非常精準，但當我們探測微觀世界（原子、電子尺度）或極高速運動時，古典理論便會失效，必須引入相對論與量子力學。

* **古典力學 (Classical Mechanics):** * 
  * 決定論（Determinism）：只要給定一個粒子的初始狀態（初始位置 $x_0$、初始動量 $p_0$）以及其受力隨時間的變化 $F(t)$，我們就能精確預測它在未來任何時刻的位置 $x(t)$ 與動量 $p(t)$
  * 適用於日常巨觀世界的物理現象
* **近代物理 (Modern Physics):**
  * **相對論 (Relativity):** 當物體速度 $v$ 接近光速 $c$ 時，古典力學的時空觀念失效，需引入狹義相對論。
  * **量子力學 (Quantum Mechanics):** 處理質量極小、尺度極小（如電子、質子）的系統。在量子力學中，「決定論」被「機率論」取代，我們無法同時確知粒子的精確位置與動量（海森堡測不準原理），所有物理量的測量結果皆以機率分佈呈現。
    * **對應原理 (Correspondence Principle):** 古典力學可視為量子力學在巨觀尺度（量子數極大）下的極限近似。

### 量子力學基本規則 (Fundamental Rules of Quantum Mechanics)

在量子力學中，粒子的狀態不再由確定的軌跡描述，而是由**波函數 (Wave function)** $\Psi$ 來完整描述。

1.  **波函數 $\Psi$ (Wave Function):**
    * 波函數 $\Psi(x,t)$ 是一個複數函數， **本身沒有直接的物理測量意義** ，但它包含了該量子系統的所有資訊。
    * $\Psi$ 及其複數共軛 $\Psi^*$ 皆屬於複數系 $\mathbb{C}$。例如若 $\Psi = A + Bi$，則 $\Psi^* = A - Bi$。
    * 同時滿足 **它是一個波動** 且 **自由粒子的機率分佈必須是均勻常數** 這兩個嚴苛的條件在數學上最簡單、最標準的複數平面波函數就是 $\Psi = Ae^{-i \omega(t- \frac{x}{v})}$ 
  
2.  **機率密度 (Probability Density):**
    
    $$\begin{gather*}
    P_{x_1 x_2} &=& \int_{x_1}^{x_2} \Psi \Psi^* dx \\
    &=& \int_{x_1}^{x_2} |\Psi|^2 dx \\
    \end{gather*}$$

    * 空間中某處發現粒子的「機率密度」為 $|\Psi|^2 = \Psi \Psi^*$
    * $P_{x_1 x_2}$ : 粒子在區間 $x_1$ 到 $x_2$ 內出現的機率 $[\text{dimensionless}]$
    * $x_1, x_2$: 空間座標的起點與終點 ，$[\text{m}]$
    * $\Psi$: 一維波函數， $[\text{m}^{-\frac{1}{2}}]$
    * $|\Psi|^2$: 一維機率密度 (Probability Density) ，$[\text{m}^{-1}]$
    * $dx$ : 位置微元 $[\text{m}]$

3.  **歸一化條件 (Normalization Condition):**
  * 既然粒子必然存在於全空間的某個角落，因此在全空間內找到該粒子的總機率必須為 $100\%$
  
    $$\int_{-\infty}^{\infty} |\Psi|^2 dV = 1$$
  
    * $\Psi$: 三維波函數 (Wave Function)，$[\text{m}^{-\frac{3}{2}}]$
    * $|\Psi|^2$: 三維機率密度 (Probability Density)，$[\text{m}^{-3}]$
    * $dV$ : 體積微元 (在三維空間中 $dV = dx \, dy \, dz$) $[\text{m}^3]$
    * 對於一維空間，歸一化條件簡化為 $\int_{-\infty}^{\infty} |\Psi|^2 dx = 1$。
  

4.  **波函數的連續性與單一值要求:**
  
  * 為了讓波函數符合物理現實（特別是要符合二階微分的波動方程式），$\Psi$ 本身及其對空間的一階微分 $\frac{\partial \Psi}{\partial x}$ 必須 **連續** 且為 **單一值**
  * **為什麼波函數本身必須連續？：** 若 $\Psi$ 在某點發生斷層，意味著粒子在相鄰兩點的機率有極大差異，這在數學上會導致機率流密度（Probability current）變成無窮大，暗示粒子能瞬間移動，違反因果律與物理現實。
  * **為什麼一階微分必須連續？：** 一階微分代表粒子的動量。若一階微分不連續代表動量在該點發生瞬間且無窮大的改變，這需要無窮大的力學位能才能達成。除非遇到理想化的「無限大位能牆」，否則在真實物理場中（位能皆為平滑連續），一階微分必須連續。（註：薛丁格方程式為二階微分方程，故只需一階可導，二階微分不必須連續，但必須可積分）。

**【進階探討：物理世界的時空真的連續嗎？】**
古典物理與廣義相對論皆假設時空是平滑連續的「流形 (Manifold)」，但在極端條件（如黑洞奇異點）下，此假設會崩潰。現代量子引力理論（如圈量子重力 LQG、弦論）提出了「時空量子化」的概念，認為在普朗克尺度（Planck scale）下，空間存在最小單位，不再連續。此外，全像原理（Holographic Principle）亦暗示宇宙資訊量是離散有限的。儘管目前尚無實驗能直接觀測普朗克尺度，但理論物理學界傾向認為「連續性」只是巨觀尺度下的一種近似錯覺。

## 5.2 薛丁格方程式：含時型式 (Schrödinger’s Equation: Time-Dependent Form)

薛丁格方程式不能被嚴格地從其他更基礎的古典力學原理中「數學推導」出來，所以我們只能 **構造（Heuristic construction）** 它，或者說它本身就是量子力學的一條**基本假設（Postulate）**，它之所以被確立為真理是因為它在描述氫原子光譜、穿隧效應等無數次實驗中，預測結果與觀測數據達到了驚人的吻合，其地位等同於古典力學中的牛頓第二定律（$F=ma$）


$$i\hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U(x,t)\Psi$$

* $E$: 總能量 (Total Energy)，$[\text{J}] , [\text{eV}]$
* $p$: 動量 (Momentum)，$[\text{kg} \cdot \text{m} \cdot \text{s}^{-1}]$
* $m$: 粒子質量 (Mass)，$[\text{kg}]$
* $U, U(x,t)$: 系統的位能函數 (Potential Energy)，$[\text{J}]$
* $i$: 虛數單位 (Imaginary Unit)，滿足 $i^2 = -1$
* $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 標準的複數平面波函數 ：**
  
  $$\Psi = Ae^{-i \omega(t- \frac{x}{v})}$$

  * $\Psi$: 一維波函數， $[\text{m}^{-\frac{1}{2}}]$
  * $A$: 振幅最大值 (Maximum Amplitude)，決定波的強度或歸一化常數， $[\text{m}^{-\frac{1}{2}}]$
  * $\omega$: 角頻率 (Angular Frequency)，$[\text{rad} \cdot \text{s}^{-1}]$
  * $v$: 波速 (Wave Velocity)，$[\text{m} \cdot \text{s}^{-1}]$
  * $x$: 位置 (Position)，$[\text{m}]$
  * $t$: 時間 (Time)，$[\text{s}]$

* **【定義 1】 [波的基本性質數學定義](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#describing-a-wave)：**
  
  $$\omega \overset{\text{def}}{=} 2\pi \nu , k \overset{\text{def}}{=} \frac{2\pi}{\lambda} = \frac{\omega}{v_p}$$

  * $\omega $: 角頻率 (Angular Frequency)， $[\text{rad} \cdot \text{s}^{-1}]$
  * $\nu $: 頻率 (Frequency)， $[\text{Hz}],[\text{s}^{-1}]$
  * $k$: 波數 (Wave Number)，代表單位長度內的相位變化， $[\text{rad} \cdot \text{m}^{-1}]$
  * $\lambda $: 波長 (Wavelength)，單位為 $[\text{m}]$
  * $v_{p}$: 相速度 (Phase Velocity) 波動相位在空間中傳播的速度， $[\text{m} \cdot \text{s}^{-1}]$，

* **【已知 2】 古典波動角頻率與波速關係 ：**
  
  $$\omega = 2\pi\nu \quad , \quad \quad v = \nu\lambda$$

  * $\omega$: 角頻率 (Angular Frequency)，$[\text{rad} \cdot \text{s}^{-1}]$
  * $\nu$: 頻率 (Frequency)，代表每秒振動次數， $[\text{s}^{-1}] , [\text{Hz}]$
  * $v$: 波速 (Wave Velocity)，$[\text{m} \cdot \text{s}^{-1}]$
  * $\lambda$ : 波長 (Wavelength)，$[\text{m}]$

* **【已知 3】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：**
  
  $$\begin{gather*}
  \lambda &\overset{\text{def}}{=}& \frac{h}{p} \\
  p &=& \frac{h}{\lambda}\\
  p &=& \frac{h}{\lambda}  \frac{2 \pi \nu }{2 \pi \nu  }  \\
  p &=& \frac{h}{2 \pi}  \frac{2 \pi \nu }{ \lambda \nu  }  \\
  \frac{p}{\hbar} &=&   \frac{2 \pi \nu }{ \lambda \nu  }  \\
  \frac{p}{\hbar} &\overset{\text{已知 2}}{=}&   \frac{\omega}{ v }  \\
  \end{gather*}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $\hbar$ : 約化普朗克常數 $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$

* **【已知 4】 [愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis) ：**
  
  $$\begin{gather*}
  E &=& h\nu \\
  E&\overset{\text{定義 1}}{=}& h \frac{\omega}{2\pi} \\
  E&=& \hbar \omega\\
  \frac{E}{\hbar} &=& \omega\\
  \end{gather*}$$

  * $E$ : 粒子總能量 $[\text{J}]$
  * $\hbar$ : 約化普朗克常數 $\hbar = \frac{h}{2\pi} \approx 1.054 \times 10^{-34} \text{ J}\cdot\text{s}$
  * $\nu$: 光子頻率 (Frequency)，$[\text{s}^{-1}] , [\text{Hz}]$
  * $\omega$: 光子角頻率 (Angular Frequency)，$[\text{rad} \cdot \text{s}^{-1}]$

* **【已知 5】 標準的複數平面波函數-能量動量 ：**
  
  $$\begin{gather*}
  \Psi(x,t) &\overset{\text{已知 1}}{=}& Ae^{-i\omega(t-\frac{x}{v})}\\
  &=& Ae^{-i(\omega t-\omega\frac{x}{v})}\\
  &\overset{\text{已知 3}}{=}& Ae^{-i(\omega t-\frac{p}{\hbar} x )}\\
  &\overset{\text{已知 4}}{=}& Ae^{-i(\frac{E}{\hbar} t-\frac{p}{\hbar} x )}\\
  &=& Ae^{-\frac{i}{\hbar}(Et - px)}
  \end{gather*}$$

  * $\Psi$: 一維波函數， $[\text{m}^{-\frac{1}{2}}]$
  * $A$: 振幅最大值 (Maximum Amplitude)，決定波的強度或歸一化常數， $[\text{m}^{-\frac{1}{2}}]$
  * $\omega$: 角頻率 (Angular Frequency)，$[\text{rad} \cdot \text{s}^{-1}]$
  * $v$: 波速 (Wave Velocity)，$[\text{m} \cdot \text{s}^{-1}]$
  * $E$ : 粒子總能量 $[\text{J}]$
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


* **【已知 6】 波函數對空間 $x$ 進行兩次微分 ：**
  
  $$\begin{gather*}
  \frac{\partial \Psi}{\partial x} &\overset{\text{已知 5}}{=}& \frac{\partial}{\partial x} \left[ Ae^{-\frac{i}{\hbar}(Et - px)} \right]\\
  \frac{\partial \Psi}{\partial x}&=& \left( \frac{i}{\hbar}p \right) Ae^{-\frac{i}{\hbar}(Et - px)} \\
  \frac{\partial \Psi}{\partial x}&\overset{\text{已知 5}}{=}& \left( \frac{i}{\hbar}p \right) \Psi \\
  \\ \\ \\ \\
  \end{gather*}  \quad , \quad
  \begin{gather*}
  \frac{\partial^2 \Psi}{\partial x^2} &=& \frac{\partial}{\partial x} \left[ \frac{\partial \Psi}{\partial x} \right] \\
  \frac{\partial^2 \Psi}{\partial x^2}&=& \frac{\partial}{\partial x} \left[ \frac{i}{\hbar}p \Psi \right] \\
  \frac{\partial^2 \Psi}{\partial x^2}&=& \left( \frac{i}{\hbar}p \right)^2 \Psi \\
  \frac{\partial^2 \Psi}{\partial x^2}&=& -\frac{p^2}{\hbar^2} \Psi \\
  p^2 \Psi &=& -\hbar^2 \frac{\partial^2 \Psi}{\partial x^2}
  \end{gather*}$$

  *  $p$ 打從一開始就不是位置 $x$ 的函數，而是一個被我們人為給定的 **常數參數**，故 $\frac{\partial}{\partial x} [p] = 0$

* **【已知 7】 波函數對時間 $t$ 進行一次微分 ：**
  
  $$\begin{gather*}
  \frac{\partial \Psi}{\partial t} &\overset{\text{已知 5}}{=}& \frac{\partial}{\partial t} \left[ Ae^{-\frac{i}{\hbar}(Et - px)} \right] \\
  \frac{\partial \Psi}{\partial t} &=& \left( -\frac{i}{\hbar}E \right) Ae^{-\frac{i}{\hbar}(Et - px)} \\
  \frac{\partial \Psi}{\partial t} &\overset{\text{已知 5}}{=}& \left( -\frac{i}{\hbar}E \right) \Psi \\
  E \Psi &=& -\frac{\hbar}{i} \frac{\partial \Psi}{\partial t} \\
  E \Psi &=& i\hbar \frac{\partial \Psi}{\partial t}
  \end{gather*}$$

  *  $E$ 打從一開始就不是時間 $t$ 的函數，而是一個被我們人為給定的 **常數參數**，故 $\frac{\partial}{\partial t} [E] = 0$

* **【已知 8】能量守恆 ：**
  
  $$E = KE + U(x, t)$$

  
* **【已知 9】動能的動量形式 ：**
  
  $$KE = \frac{p^2}{2m}$$

  

### 推導 


$$\begin{gather*}
E &\overset{\text{已知 8}}{=}& KE + U(x, t) \\
E &\overset{\text{已知 9}}{=}& \frac{p^2}{2m} + U(x, t) \\
E\Psi &=& \frac{p^2}{2m}\Psi + U(x, t)\Psi \\
i\hbar \frac{\partial \Psi}{\partial t} &\overset{\text{已知 7}}{=}& \frac{p^2}{2m}\Psi + U(x, t)\Psi \\
i\hbar \frac{\partial \Psi}{\partial t} &\overset{\text{已知 6}}{=}& -\frac{1}{2m}\hbar^2 \frac{\partial^2 \Psi}{\partial x^2} + U(x, t)\Psi \\
\end{gather*}$$


$$i\hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U(x,t)\Psi$$

* $\Psi$ : 系統的波函數
* $m$ : 粒子質量 $[\text{kg}]$
* $U(x,t)$ : 粒子所處的位能場 $[\text{J}]$

若推廣至三維空間，則空間微分項變為拉普拉斯算符（Laplacian）：

$$\begin{gather*}
i\hbar \frac{\partial \Psi}{\partial t} &=& -\frac{\hbar^2}{2m} \left( \frac{\partial^2 \Psi}{\partial x^2} + \frac{\partial^2 \Psi}{\partial y^2} + \frac{\partial^2 \Psi}{\partial z^2} \right) &+ U(x,y,z,t)\Psi \\
i\hbar \frac{\partial \Psi}{\partial t} &=& -\frac{\hbar^2}{2m} \nabla^2 \Psi &+ U(x,y,z,t)\Psi
\end{gather*}$$



### 線性與疊加原理 (Linearity and Superposition)

薛丁格方程式是一個**線性微分方程式**。在數學上，**齊次線性(Homogeneous Linear)** 意味著如果 $\Psi_1$ 和 $\Psi_2$ 皆為該方程式的合法解，那麼它們的線性組合 $\Psi = c_1\Psi_1 + c_2\Psi_2$（$c_1, c_2$ 為常數）必定也是一個合法解。

**舉例:**
* $t^x i x^{100} \frac{\partial \Psi}{\partial t} = t \ln x^t \frac{\partial \Psi}{\partial x} +U \Psi$ 是線性的
* $\frac{\partial \Psi}{\partial t} =  \frac{\partial^2 \Psi}{\partial x^2} + U\Psi +C$ ，不是線性的，因為C+C=2C會破壞疊加原理
* $\frac{\partial \Psi^2}{\partial t} =  \frac{\partial^2 \Psi}{\partial x^2} + U\Psi $ ，不是線性的，因為 $\Psi^2$

方程式不包含 $\Psi^2$ 或任何常數項（如 $+ C$），這確保了波函數疊加的特性。
* 物理意義：波函數的疊加是 $\Psi = \Psi_1 + \Psi_2$。
* 觀測機率則是波函數絕對值的平方：$|\Psi|^2 = |\Psi_1 + \Psi_2|^2 = |\Psi_1|^2 + |\Psi_2|^2 + (\Psi_1^*\Psi_2 + \Psi_1\Psi_2^*)$，後面的交叉項即為量子力學中產生「干涉現象（Interference）」的來源（例如雙狹縫實驗中的干涉條紋），這證明了**機率本身不能簡單相加**（$P \neq P_1 + P_2$），而是**波的振幅相加**。


### 期望值 (Expectation Value)

在量子力學中，因為粒子的位置是機率分佈的，我們無法預測單次測量的確切結果，但可以預測大量相同系統重複測量後的平均值，稱為**期望值 (Expectation Value)**。

對於單一粒子的位置期望值 $\langle x \rangle$：

$$\begin{gather*}
\langle x \rangle &\overset{\text{def}}{=}& \frac{\int_{-\infty}^{\infty} x |\Psi|^2 dx}{\int_{-\infty}^{\infty} |\Psi|^2 dx} \\
&\overset{\text{歸一化條件}}{=}& \int_{-\infty}^{\infty} x |\Psi|^2 dx
\end{gather*}$$

若要計算任何與位置 $x$ 相關的物理量 $G(x)$ 的期望值：

$$\begin{gather*}
\langle G(x) \rangle &\overset{\text{def}}{=}&  \int_{-\infty}^{\infty} G(x) |\Psi|^2 dx \\
&=& \int_{-\infty}^{\infty} \Psi^* G(x) \Psi dx
\end{gather*}$$

### **例題 5.2：機率與期望值計算**

**題目：** 一個限制於 $x$ 軸上的粒子，在 $x=0$ 與 $x=1$ 之間的波函數為 $\Psi = ax$ ；在其他位置時則為 $\Psi = 0$ 。(a) 求出粒子在 $x=0.45$ 與 $x=0.55$ 之間出現的機率。(b) 求出粒子位置的期望值。

**解答：**

(a) 量子力學基本規則-利用機率密度積分：

$$\begin{gather*}
P_{x_1 x_2} &=& \int_{x_1}^{x_2} \Psi \Psi^* dx \\
P_{0.45 \to 0.55} &=& \int_{0.45}^{0.55} |\Psi|^2 dx \\
&=& \int_{0.45}^{0.55} (ax)^2 dx \\
&=& a^2 \int_{0.45}^{0.55} x^2 dx \\
&=& a^2 \left[ \frac{x^3}{3} \right]_{0.45}^{0.55} \\
&=& \frac{a^2}{3} \left( 0.55^3 - 0.45^3 \right) \\
&=& \frac{a^2}{3} (0.166375 - 0.091125) \\
&=& 0.025083 a^2 \\
&\approx& 0.0251 a^2
\end{gather*}$$

(b) 單一粒子的位置期望值-利用期望值公式：

$$\begin{gather*}
\langle x \rangle &=& \int_{0}^{1} x |\Psi|^2 dx \\
&=& \int_{-\infty}^{0} x |\Psi|^2 dx + \int_{0}^{1} x |\Psi|^2 dx + \int_{1}^{\infty} x |\Psi|^2 dx \\
&=& \int_{-\infty}^{0} x |0|^2 dx + \int_{0}^{1} x (ax)^2 dx + \int_{1}^{\infty} x |0|^2 dx \\
&=& a^2 \int_{0}^{1} x^3 dx \\
&=& a^2 \left[ \frac{x^4}{4} \right]_{0}^{1} \\
&=& \frac{a^2}{4}
\end{gather*}$$

**【物理知識回顧】**
期望值 $\langle x \rangle$ 是「平均測量值」，而非「出現機率最高」的眾數位置。以此題為例，波函數在 $x=1$ 時 $|\Psi|^2$ 最大（粒子最常在此出現），但由於整個分佈涵蓋了 $0$ 到 $1$ 之間的區域，加權平均後，粒子的期望位置會落在 $x=1$ 的左側（即 $a^2/4$，若以歸一化條件求出 $a=\sqrt{3}$，則 $\langle x \rangle = 3/4$）。

## 5.3 運算子 (Operators)

在古典力學中，物理量（如動量、能量）都是確實的數值，但在先前的推導中我們發現，若要從波函數 $\Psi$ 中「萃取」出物理資訊，必須對其進行特定的微分運算：

$$\begin{gather*}
p \Psi &=& \frac{\hbar}{i} \frac{\partial \Psi}{\partial x} \\
E \Psi &=& i\hbar \frac{\partial \Psi}{\partial t}
\end{gather*}$$

在量子力學中，我們將這些執行微分動作的「數學指令」定義為**算符 (Operator)**，並在符號上方加上 $\hat{\cdot}$（Hat）來表示，當算符作用於波函數就能提取出對應的觀測值。

* **動量算符 (Momentum Operator):** 
  
  $$\begin{gather*}
  \hat{p} &\overset{\text{def}}{=}& \frac{\hbar}{i} \frac{\partial}{\partial x} \\
  &=& -i\hbar \frac{\partial}{\partial x}
  \end{gather*}$$

* **能量算符 (Energy Operator):** 
  
  $$\hat{E} \overset{\text{def}}{=} i\hbar \frac{\partial}{\partial t}$$

* **動能算符 (Kinetic Energy Operator):** 
  
  $$\begin{gather*}
  \hat{KE} &\overset{\text{def}}{=}& \frac{\hat{p}^2}{2m} \\
  &=& \frac{1}{2m} \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right)^2 \\
  &=& -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2}
  \end{gather*}$$

透過算符的定義，我們可以將薛丁格方程式以極度簡潔的方式重新表達，這充分展現了 **量子力學架構其實就是古典能量守恆定律的「算符化」**

$$\begin{gather*}
i\hbar \frac{\partial \Psi}{\partial t} &=& -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U\Psi \\
\hat{E}\Psi &=& \hat{KE}\Psi + \hat{U}\Psi
\end{gather*}$$



### 實際操作與期望值

當我們要求解任意物理量 $G$ 的期望值時，其嚴格的數學形式必須透過算符 $\hat{G}$ 來計算：

$$\langle G(x, t) \rangle = \int_{-\infty}^{\infty} \Psi^* \hat{G} \Psi dx$$

* **動量期望值:** $\langle p \rangle = \int_{-\infty}^{\infty} \Psi^* \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right) \Psi dx = \frac{\hbar}{i} \int_{-\infty}^{\infty} \Psi^* \frac{\partial \Psi}{\partial x} dx$
* **能量期望值:** $\langle E \rangle = \int_{-\infty}^{\infty} \Psi^* \left( i\hbar \frac{\partial}{\partial t} \right) \Psi dx = i\hbar \int_{-\infty}^{\infty} \Psi^* \frac{\partial \Psi}{\partial t} dx$

**為什麼順序必須是 $\int \Psi^* \hat{G} \Psi dx$？**
算符 $\hat{G}$ 是包含微分指令的數學物件，它只對「寫在它右邊」的函數作用。
1.  如果寫成 $\int \hat{G} (\Psi^* \Psi) dx$，算符會根據乘法連鎖律同時微分 $\Psi^*$ 與 $\Psi$，這會產生物理上毫無意義的額外項。
2.  $\Psi^*$ 在左側的作用，是為了透過複數共軛的內積性質，保證最終算出的期望值是一個「純實數」。因為在真實世界中，儀器量測到的物理數值（如長度、能量）絕不可能是帶有虛部的複數。

## 5.4 薛丁格方程式：穩態型式 (Schrödinger’s Equation: Steady-State Form)

$$\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi = 0$$

* $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
* $E$ : 系統的總能量常數 $[\text{J}]$
* $U(x)$ : 位能的空間分布 $[\text{J}]$
* $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
* 三維形式則為：$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial z^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 薛丁格方程式：含時型式 (Schrödinger’s Equation: Time-Dependent Form) ：**
  
  $$i\hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U(x,t)\Psi$$

  * $E$: 總能量 (Total Energy)，$[\text{J}] , [\text{eV}]$
  * $p$: 動量 (Momentum)，$[\text{kg} \cdot \text{m} \cdot \text{s}^{-1}]$
  * $m$: 粒子質量 (Mass)，$[\text{kg}]$
  * $U, U(x,t)$: 系統的位能函數 (Potential Energy)，$[\text{J}]$
  * $i$: 虛數單位 (Imaginary Unit)，滿足 $i^2 = -1$
  * $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx  1.054 \times 10^{-34} \text{J}\cdot \text{s} $

* **【假設 1】 系統處於穩態 (Steady-state)假設分離變數法 (Separation of Variables)拆解為空間部分與時間部分的乘積 ：**
  
  $$\Psi(x,t) = \psi(x) e^{-i\frac{E}{\hbar}t }$$

  * 系統處於**穩態 (Steady-state)**，即系統所在的位能場 $U$ 不隨時間改變（$U$ 僅為位置 $x$ 的函數，$U=U(x)$）



### 推導 



$$\begin{gather*}
i\hbar \frac{\partial \Psi}{\partial t} &\overset{\text{已知 1}}{=}& -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U(x,t)\Psi \\
i\hbar \frac{\partial}{\partial t} \left[ \psi e^{-i\frac{E}{\hbar}t } \right] &\overset{\text{假設 1}}{=}& -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} \left[ \psi e^{-i\frac{E}{\hbar}t } \right]+ U(x)\left( \psi e^{-iEt \hbar} \right) \\
i\hbar \left( -\frac{iE}{\hbar} \right) \psi e^{-i\frac{E}{\hbar}t } &=& -\frac{\hbar^2}{2m} \frac{d^2 \psi}{dx^2} e^{-i\frac{E}{\hbar}t } + U\psi e^{-i\frac{E}{\hbar}t }\\
E \psi  &=& -\frac{\hbar^2}{2m} \frac{d^2 \psi}{dx^2}  + U\psi \\
\frac{\hbar^2}{2m} \frac{d^2 \psi}{dx^2} + E \psi - U\psi &=& 0 \\
\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &=& 0
\end{gather*}$$


* $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
* $E$ : 系統的總能量常數 $[\text{J}]$
* $U(x)$ : 位能的空間分布 $[\text{J}]$
* $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
* 三維形式則為：$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial z^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$

### 本徵值與本徵函數 (Eigenvalues and Eigenfunctions)

在線性代數與量子力學中，若一個算符 $\hat{G}$ 作用在某特定函數 $\psi_n$ 上，其結果只是該函數乘上一個純量常數 $G_n$，則這被稱為一個**本徵值問題**：

$$\hat{G}\psi_n = G_n\psi_n$$

* $\psi_n$ 稱為算符 $\hat{G}$ 的**本徵函數 (Eigenfunction)**。
* $G_n$ 稱為對應的**本徵值 (Eigenvalue)**，即物理測量上的確定數值。這解釋了量子力學中「量子化 (一份一份)」的來源：只有特定的波函數形狀（$\psi_n$）能滿足這項等式，因此只會得到對應的特定觀測值（$G_n$）。

我們定義**漢彌爾頓算符 (Hamiltonian operator)** $\hat{H}$ 為系統總能（動能加位能）的算符，不含時薛丁格方程式即可極簡略為標準的本徵值方程形式$\hat{H}\psi_n = E_n\psi_n$

$$\begin{gather*}
&\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &=& 0 \\
&E\psi &=& -\frac{\hbar^2}{2m} \frac{d^2 \psi}{dx^2} + U(x)\psi \\
\hat{H} \overset{\text{def}}{=}& E\psi &=&\left( -\frac{\hbar^2}{2m} \frac{d^2}{dx^2} + U(x) \right) \psi \\
&\hat{H}\psi_n &=& E_n\psi_n
\end{gather*}$$


在量子力學系統（如位能井或原子）能滿足這個方程式的解通常是離散的，只有特定的波函數形狀才能成立並對應到特定的能量值，為了標示這些不同的 **量子態** 物理學家會加上標註下標 $n$（代表第 $n$ 個能階與對應的波函數），最終寫成我們熟知的標準形式 $\hat{H}\psi_n = E_n\psi_n$ 。透過這個過程，我們就把一個看起來複雜的二階微分方程式，完美包裝成了線性代數中最優雅的本徵值方程式：算符 $\hat{H}$ 作用在函數 $\psi_n$ 上，等於該函數自己乘上一個常數 $E_n$，這代表當我們將「總能量算符 $\hat{H}$」作用於系統特定的穩態波函數 $\psi_n$ 時，即可得到該系統特定且離散的「能階 $E_n$」。

#### **例題 5.4：本徵值計算**

**題目：** 運算子 $\hat{G} = \frac{d^2}{dx^2}$ 的本徵函數是 $\psi = e^{2x}$，求出對應的本徵值。


$$\begin{gather*}
\hat{G}\psi &=& \frac{d^2}{dx^2}\left[e^{2x} \right]  \\
\hat{G}\psi &=& \frac{d}{dx} \left[ \frac{d}{dx}(e^{2x}) \right] \\
\hat{G}\psi &=& \frac{d}{dx}\left[2e^{2x} \right]  \\
\hat{G}\psi &=& 4e^{2x} \\
\hat{G}\psi &=& 4\psi
\end{gather*}$$

因此，對應的本徵值 $G = 4$。

**【物理知識回顧】**
本徵方程式展示了量子測量的精髓：當系統處於某個算符的本徵態時，對該物理量的測量結果將100%是其本徵值，不會有任何測量的不確定性或機率分佈。

| 物理量 | 對應的運算子 (Operator) |
| :--- | :--- |
| 位置 $x$ | $x$ |
| 線性動量 $p$ | $-i\hbar \frac{\partial}{\partial x}$ |
| 位能 $U(x)$ | $U(x)$ |
| 動能 $K = \frac{p^2}{2m}$ | $-\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2}$ |
| 總能量 $E$ (含時) | $i\hbar \frac{\partial}{\partial t}$ |
| 總能量 $H$ (漢彌爾頓形式) | $-\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + U(x)$ |

## 5.5 一維無限深位能井：箱子中的粒子 (Particle in a Box)

這是一個最基礎但能完美展現「量子化」來源的系統模型，假設一個質量為 $m$ 的粒子被困在寬度為 $L$ 的一維堅硬箱子中。

**波函數**

$$
\psi_n(x) = 
\left\{\begin{matrix}
0 &\quad x \le 0  \\
\sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right) &\quad 0 \le x \le L & \quad n = 1, 2, 3, \dots \\
0 &\quad x \ge 0  \\
\end{matrix}\right.
$$



* $\psi_n(x)$ : 處於第 $n$ 能階的定態一維波函數 (Wave Function of the $n$-th state)， $[\text{m}^{-1/2}]$
* $x$ : 粒子在箱中的空間位置座標 (Position)， $[\text{m}]$
* $m$ : 粒子質量 (Mass)， $[\text{kg}]$
* $L$ : 一維箱子的寬度 (Width of the box)， $[\text{m}]$
* $n$ : 主量子數 (Principal Quantum Number)， $n = 1, 2, 3, \dots$ $[\text{dimensionless}]$


**量子化能階 (Energy Levels)：**
  

$$
E_n = \frac{n^2 \pi^2 \hbar^2}{2mL^2}  \quad n = 1, 2, 3, \dots 
$$

* $E_{n}$ : 箱中粒子處於第 $n$ 能階的定態總能量 (Energy of the $n$-th level)， $[\text{J}]$
* $L$ : 一維箱子的寬度 (Width of the box)， $[\text{m}]$
* $n$ : 主量子數 (Principal Quantum Number)， $[\text{dimensionless}]$
* $m$ : 粒子質量 (Mass)， $[\text{kg}]$
* $\hbar $ : 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s}$

### 假設與已知 (Assumptions & Preliminaries)

* **【假設 1】 無限高位能牆 ：**
  
  $$
  U(x) = 
  \left\{\begin{matrix}
  \infty &\quad x \le 0  \\
  0 &\quad 0 \le x \le L  \\
  \infty &\quad x \ge 0  \\
  \end{matrix}\right.
  $$


  * $U(x)$: 系統的位能隨x的函數 (Potential Energy)，$[\text{J}]$
  * 粒子絕對不可能出現在箱外，因此外部波函數為 $\Psi = 0$

* **【假設 2】 邊界條件 (Boundary Conditions) ：**
  
  $$\psi(0) = 0, \quad \psi(L) = 0$$

  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  * 由於波函數必須連續，因此在邊界處波函數必須為零

* **【已知 1】 不含時薛丁格方程式 (Schrödinger’s Equation: Steady-State Form) ：**

  $$\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi = 0$$

  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  * $E$ : 系統的總能量常數 $[\text{J}]$
  * $U(x)$ : 位能的空間分布 $[\text{J}]$
  * $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
  * 三維形式則為：$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial z^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$


* **【定義 1】 亞變數:二階常微分方程式(簡諧振動的形式)的波數 (Wave number) $k$ ：**

  $$k \overset{\text{def}}{=} \frac{\sqrt{2mE}}{\hbar}$$

  * $m$: 粒子質量 (Mass)，$[\text{kg}]$
  * $E$ : 系統的總能量常數 $[\text{J}]$
  * $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
  * $k$ : 波數 (Wave number)，$[\text{m}^{-1}]$

* **【已知 2】 二階常微分方程式(ODE)通解：**

  $$\begin{gather*}
  &\frac{d^2 \psi}{dx^2} + k^2 \psi &=& 0 \\
  \Rightarrow & \psi &=& A \sin(kx) + B \cos(kx) \\
  \end{gather*}$$


  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  * $k$ : 波數 (Wave number)，$[\text{m}^{-1}]$
  * $A,B$ : 待定常數（或振幅係數），$[\text{m}^{-1/2}]$


* **【已知 3】 依照邊界條件求出ODE待定常數：**

  * (a)
  
  $$\begin{gather*}
  \psi(x) &\overset{\text{已知 2}}{=}& A \sin(kx) + B \cos(kx) \\
  \psi(0) &=& A \sin(k \times 0) + B \cos(k \times 0) \\
  0 &\overset{\text{假設 2}}{=}& A \sin(0) + B \cos(0) \\
  0 &=& B \\
  \end{gather*}$$

  * (b)
  
  $$\begin{gather*}
  &\psi(x) &\overset{\text{已知 2}}{=}& A \sin(kx) + B \cos(kx) \\
  &\psi(x) &\overset{\text{已知 3(a)}}{=}& A \sin(kx) + 0 \cos(kx) \\
  &\psi(x) &=& A \sin(kx) \\
  &\psi(L) &=& A \sin(k L)\\
  &0 &\overset{\text{假設 2}}{=}& A \sin(kL)\\
  &0 &=& \sin(kL) \\
  \Rightarrow & kL &=& n \pi & \quad n = 1, 2, 3, \dots \\
  & k  &=& \frac{n \pi}{L}  & \quad n = 1, 2, 3, \dots \\
  \end{gather*}$$



  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  * $k$ : 波數 (Wave number)，$[\text{m}^{-1}]$
  * $A,B$ : 待定常數（或振幅係數），$[\text{m}^{-1/2}]$

* **【已知 4】 倍角公式（Double Angle Formula）：**
  
  $$\begin{gather*}
  \sin^2(x) &=& \frac{1 - \cos(2x)}{2} \\
  \end{gather*}$$

  * 再本次證明中 $\frac{n\pi x}{L}$ 為該公式中的 $x$



* **【已知 5】 波函數歸一化 (Normalization)：** 

  $$\begin{gather*}
  \int_{-\infty}^{\infty} |\psi_n|^2 dx &=& 1 \\
  \int_{-\infty}^{0} |\psi_n|^2 dx + \int_{0}^{L} |\psi_n|^2 dx +\int_{L}^{\infty} |\psi_n|^2 dx &=& 1 \\
  0 + \int_{0}^{L} |\psi_n|^2 dx +0 &\overset{\text{假設 1}}{=}& 1 \\
  \int_{0}^{L} A^2 \sin^2 \left( \frac{n\pi x}{L} \right) dx &=& 1 \\
  \frac{A^2}{2} \int_{0}^{L} \left[ 1 - \cos \left( \frac{2n\pi x}{L} \right) \right] dx  &\overset{\text{已知 4}}{=}& 1 \\
  \frac{A^2}{2} \left[ x - \frac{L}{2n\pi} \sin \left( \frac{2n\pi x}{L} \right) \right]_0^L &=& 1 \\
  \frac{A^2}{2} \left( L - 0 \right) &=& 1 \\
  A^2 \frac{L}{2} &=& 1 \quad \\
  A &=& \sqrt{\frac{2}{L}}
  \end{gather*}$$

  




### **波函數** 和 **能階** 推導

#### **波函數**

討論不是無限位能的波函數區域，並假設亞變數 $k$ 解ODE並帶入邊界條件，解出該區的波函數：

$$\begin{gather*}
&\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &\overset{\text{已知 1}}{=}& 0 \\
&\left. \frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2} E\psi \right|_{0\le x \le L}&\overset{\text{假設 1}}{=}& 0 \\
&\left. \frac{d^2 \psi}{dx^2} + k^2 \psi \right|_{0\le x \le L}&\overset{\text{定義 1}}{=}& 0 \\
\overset{\text{已知 2}}{\Rightarrow} & \left. \psi(x) \right|_{0\le x \le L} &=& A \sin(kx) + B \cos(kx) \\
& \left. \psi(x) \right|_{0\le x \le L} &\overset{\text{已知 3(a)}}{=}& A \sin\left(kx\right)  \\
& \left. \psi_n(x) \right|_{0\le x \le L} &\overset{\text{已知 3(b)}}{=}& A \sin\left(\frac{n\pi x}{L}\right)  & \quad n = 1, 2, 3, \dots\\
& \left. \psi_n(x) \right|_{0\le x \le L} &\overset{\text{已知 5}}{=}& \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)  & \quad n = 1, 2, 3, \dots\\
\end{gather*}$$

* $\psi(x)$ : 一維波函數的空間部分 (Spatial Wave Function)， $[\text{m}^{-1/2}]$
* $\psi_n(x)$ : 處於第 $n$ 能階的定態一維波函數 (Wave Function of the $n$-th state)， $[\text{m}^{-1/2}]$
* $x$ : 粒子在箱中的空間位置座標 (Position)， $[\text{m}]$
* $m$ : 粒子質量 (Mass)， $[\text{kg}]$
* $\hbar $ : 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s}$
* $E$ : 系統的總能量常數 (Total Energy)， $[\text{J}]$
* $U(x)$ : 位能的空間分布 (Potential Energy Profile)， $[\text{J}]$
* $k$ : 波數 (Wave number)， $[\text{m}^{-1}]$
* $A, B$ : 待定之微分方程式通解常數係數 (Constants of Integration)
* $\sqrt{\frac{2}{L}}$ : 經過波函數歸一化 (Normalization) 後求得的振幅係數， $[\text{m}^{-1/2}]$
* $L$ : 一維箱子的寬度 (Width of the box)， $[\text{m}]$
* $n$ : 主量子數 (Principal Quantum Number)， $n = 1, 2, 3, \dots$ $[\text{dimensionless}]$（無單位）



#### **量子化能階 (Energy Levels)：**
  
將 $k = \frac{n\pi}{L}$ 代回 $k$ 的定義式 $\frac{\sqrt{2mE}}{\hbar} = \frac{n\pi}{L}$ 並將兩邊平方整理：

$$\begin{gather*}
k &\overset{\text{定義 1}}{=}& \frac{\sqrt{2mE}}{\hbar} \\
\frac{n \pi}{L} &\overset{\text{已知 3(b)}}{=}& \frac{\sqrt{2mE_n}}{\hbar} & \quad n = 1, 2, 3, \dots \\
E_n &=& \frac{n^2 \pi^2 \hbar^2}{2mL^2} & \quad n = 1, 2, 3, \dots \\
\end{gather*}$$

* $E_{n}$ : 箱中粒子處於第 $n$ 能階的定態總能量 (Energy of the $n$-th level)， $[\text{J}]$
* $L$ : 一維箱子的寬度 (Width of the box)， $[\text{m}]$
* $n$ : 主量子數 (Principal Quantum Number)， $[\text{dimensionless}]$（無單位）
* $m$ : 粒子質量 (Mass)， $[\text{kg}]$
* $k$ : 波數 (Wave number)， $[\text{m}^{-1}]$
* $E$ : 系統的總能量常數 (Total Energy)， $[\text{J}]$
* $\hbar $ : 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s}$





### **例題 5.5.1：粒子在特定區域的出現機率**

**題目：** 求出陷在寬度為 $L$ 的一維位能井中的粒子，在基態 ($n=1$) 與第一受激態 ($n=2$) 時，出現在 $0.45L$ 和 $0.55L$ 之間的機率。

**解答：** 

* 第 $n$ 個態的粒子在 $x_1$ 到 $x_2$ 內出現的機率為：

$$\begin{gather*}
P_{x_1, x_2} &\overset{\text{ch 5.1}}{=}& \int_{x_1}^{x_2} |\psi_n|^2 dx \\
&\overset{\text{ch 5.5}}{=}& \frac{2}{L} \int_{x_1}^{x_2} \sin^2 \left( \frac{n\pi x}{L} \right) dx \\
&=& \frac{2}{L} \int_{x_1}^{x_2} \frac{1 - \cos(2\frac{n\pi x}{L})}{2} dx \\
&=& \frac{1}{L} \left[ x - \frac{L}{2n\pi} \sin \left( \frac{2n\pi x}{L} \right) \right]_{x_1}^{x_2} \\
&=& \left[ \frac{x}{L} - \frac{1}{2n\pi} \sin \left( \frac{2n\pi x}{L} \right) \right]_{0.45L}^{0.55L} \\
&=& \left( 0.55 - \frac{1}{2 n\pi } \sin(1.1n\pi ) \right) - \left( 0.45 - \frac{1}{2n\pi} \sin(0.9n\pi) \right) \\
&=& 0.10 - \frac{1}{2n\pi} ( \sin(1.1n\pi ) - \sin(0.9n\pi)) \\
\end{gather*}$$

* 對於基態 ($n=1$)：

$$\begin{gather*}
P_1 &=& \left. 0.10 - \frac{1}{2n\pi} ( \sin(1.1n\pi ) - \sin(0.9n\pi)) \right|_{n=1} \\
&\approx& 0.10 - \frac{1}{2\pi} (-0.3090 - 0.3090) \\
&=& 0.10 + \frac{0.618}{2\pi} \\
&\approx& 0.10 + 0.0984 \\
&=& 0.1984 \quad (19.8\%)
\end{gather*}$$

* 對於第一受激態 ($n=2$)：
  
$$\begin{gather*}
P_2 &=& \left. 0.10 - \frac{1}{2n\pi} ( \sin(1.1n\pi ) - \sin(0.9n\pi)) \right|_{n=2} \\
&\approx& 0.10 - \frac{1}{4\pi} (0.5878 - (-0.5878)) \\
&=& 0.10 - \frac{1.1756}{4\pi} \\
&\approx& 0.10 - 0.0935 \\
&=& 0.0065 \quad (0.65\%)
\end{gather*}$$


**【物理知識回顧】**

所計算的區間佔據箱子寬度的十分之一，且位居正中央。古典力學認為粒子在箱內做均勻反彈，所以在此區間出現的機率理應是 $10\%$。然而，量子力學的計算顯示基態粒子在中央的機率高達近 $20\%$，而第一受激態（因為 $L/2$ 處恰好是波節 Node）出現機率銳減至不到 $1\%$。這彰顯了微觀下機率與狀態量子數的強烈關聯性。

### **例題 5.5.2：粒子位置的期望值**

**題目：** 求出在寬度為 $L$ 的一維箱中，粒子位置的期望值 $\langle x \rangle$。

**解答：**

$$\begin{gather*}
\langle x \rangle &=& \int_{-\infty}^{\infty} x |\psi_n|^2 dx \\
&=& \frac{2}{L} \int_{0}^{L} x \sin^2 \left( \frac{n\pi x}{L} \right) dx \\
&=& \frac{2}{L} \left[ \frac{x^2}{4} - \frac{x \sin(2n\pi x/L)}{4n\pi/L} - \frac{\cos(2n\pi x/L)}{8(n\pi/L)^2} \right]_0^L \\
&=& \frac{2}{L} \left( \left(\frac{L^2}{4} - \frac{L \sin(2n\pi L/L)}{4n\pi/L} - \frac{\cos(2n\pi L/L)}{8(n\pi/L)^2}\right) -  \left( \frac{0^2}{4} - \frac{0 \sin(2n\pi 0/L)}{4n\pi/L} - \frac{\cos(2n\pi 0/L)}{8(n\pi/L)^2}\right) \right) \\
&=& \frac{2}{L} \left( \left( \frac{L^2}{4} - 0 - \frac{1}{8(n\pi/L)^2} \right) - \left( 0 - 0 - \frac{1}{8(n\pi/L)^2} \right) \right) \\
&=& \frac{2}{L} \left( \frac{L^2}{4} \right) \\
&=& \frac{L}{2}
\end{gather*}$$

**【物理知識回顧】**

不論粒子處於哪一個能階 $n$，其平均位置 $\langle x \rangle$ 永遠落在箱子的正中央 $L/2$。這非常直觀，因為 $\sin^2$ 代表的機率密度分佈以箱子中心為絕對對稱軸，長期觀測的幾何平均自然就落在中點。

**【動量期望值的特別說明】**

值得注意的是，上述的 $\psi_n$ **不是**動量算符的本徵態。依據尤拉公式 $\sin \theta = \frac{1}{2i} (e^{i\theta} - e^{-i\theta})$，該駐波實際上是兩道行進波（向右動量 $+p$ 與向左動量 $-p$）的疊加：

$$\psi_n = \frac{1}{2i} \sqrt{\frac{2}{L}} e^{in\pi x/L} - \frac{1}{2i} \sqrt{\frac{2}{L}} e^{-in\pi x/L}$$

這導致動量的期望值 $\langle p \rangle$ 在這個系統中恆為 $0$（向左與向右的機率相互抵消）。

## 5.6 一維有限深位能井 (Finite Potential Well)

若箱子的牆壁高度不是無限大，而是有限的位能 $U_0$。當粒子的總能量 $E < U_0$ 時，雖然古典力學依然認為粒子被困在井中，但量子力學容許波函數「滲透」定位能壁內。我們將空間分為三個區域來討論。

分區波函數通解

$$\psi(x) = 
\left\{\begin{matrix}
C e^{ax} &\quad x < 0 & \quad (\text{第一區 I}) \\
A \sin(kx) + B \cos(kx) &\quad 0 \le x \le L & \quad (\text{第二區 II}) \\
G e^{-ax} &\quad x > L & \quad (\text{第三區 III}) \\
\end{matrix}\right.$$

* $\psi(x)$ : 一維波函數的空間部分 (Spatial Wave Function)， $[\text{m}^{-1/2}]$
* $x$ : 粒子在箱中的空間位置座標 (Position)， $[\text{m}]$
* $k$ : 中央井內（區 II）的波數 (Wave number)， $[\text{m}^{-1}]$
* $a$ : 外部區域（區 I 與 III）的衰減常數 (Decay constant)， $[\text{m}^{-1}]$
* $A, B$ : 中央井內的待定振幅常數， $[\text{m}^{-1/2}]$
* $C, G$ : 外部區域的待定振幅常數， $[\text{m}^{-1/2}]$

量子化能階 (Energy Levels) 的定性結果：
由於波函數可以向外滲透，這意味著粒子活動的「有效箱寬度」變大了。根據德布羅意波長與能量的關係，這會導致有限位能井內的能階 $E_n$ 比無限深位能井來得低。精確的 $E_n$ 需透過解超越方程式（Transcendental equations）求得，無法寫成簡單的代數公式。

### 假設與已知 (Assumptions & Preliminaries)

* **【假設 1】 有限高位能牆 (Finite Potential Wall)**：

  $$U(x) = 
  \left\{\begin{matrix}
  U_0 &\quad x < 0  \\
  0 &\quad 0 \le x \le L  \\
  U_0 &\quad x > L  \\
  \end{matrix}\right.$$


  * $U(x)$: 系統的位能隨x的函數 (Potential Energy)，$[\text{J}]$
  * $U_0$: 位能井的深度（高度），$[\text{J}]$


* **【假設 2】 物理合理性篩選 (Finite Everywhere)：** 波函數在無窮遠處不能發散至無限大，否則機率為無限大，無法完成歸一化。

  $$\psi(x) \to 0 \quad \text{as} \quad x \to \pm \infty$$
  
  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  



* **【假設 3】 邊界連續性條件 (Continuity Conditions)：** 波函數 $\psi(x)$ 及其一階導數 $\frac{d\psi}{dx}$ 在任何邊界（$x=0$ 與 $x=L$）都必須是連續的。
  
  $$\lim_{\delta x  \to 0^+}   \psi(x+\delta x) = \psi(x)$$

  $$\lim_{\delta x  \to 0^+}   \frac{d\psi(x+\delta x)}{dx} = \frac{d\psi(x)}{dx}$$

  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 




* **【已知 1】 不含時薛丁格方程式 (Schrödinger’s Equation: Steady-State Form) ：**

  $$\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi = 0$$

  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  * $E$ : 系統的總能量常數 $[\text{J}]$
  * $U(x)$ : 位能的空間分布 $[\text{J}]$
  * $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
  * 三維形式則為：$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial z^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$



* **【定義 1】 中央井內（區 II）亞變數:二階常微分方程式(簡諧振動的形式)的波數 (Wave number) $k$ ：**

  $$k \overset{\text{def}}{=} \frac{\sqrt{2mE}}{\hbar}$$

  * $m$: 粒子質量 (Mass)，$[\text{kg}]$
  * $E$ : 系統的總能量常數 $[\text{J}]$
  * $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
  * $k$ : 波數 (Wave number)，$[\text{m}^{-1}]$

* **【定義 2】 外部區域（區 I 與 III）亞變數:衰減常數 $a$：**
  
  $$a \overset{\text{def}}{=} \frac{\sqrt{2m(U_0 - E)}}{\hbar}$$

  * $m$: 粒子質量 (Mass)，$[\text{kg}]$
  * $E$ : 系統的總能量常數 $[\text{J}]$
  * $U_0$ : 位能井的高度，$[\text{J}]$
  * $\hbar$: 約化普朗克常數 (Reduced Planck Constant)， $\hbar \approx 1.054 \times 10^{-34} \text{J}\cdot \text{s} $
  * $a$ : 衰減常數 (Decay Constant)，$[\text{m}^{-1}]$
  * 因為 $U_0 > E$，所以 $U_0 - E$ 為正值，保證 $a$ 是一個實數常數。


* **【已知 2】 二階常微分方程式(ODE)通解：**

  * (a)

  $$\begin{gather*}
  &\frac{d^2 \psi}{dx^2} + k^2 \psi &=& 0 \\
  \Rightarrow & \psi_{\text{II}} &=& A \sin(kx) + B \cos(kx) \\
  \end{gather*}$$

  * (b)
  
  $$\begin{gather*}
  &\frac{d^2 \psi}{dx^2} + k^2 \psi &=& 0 \\
  \Rightarrow & \psi_{\text{out}} &=& c_1 e^{ax} + c_2 e^{-ax} \\
  \end{gather*}$$

  * $\psi(x)$ : 一維波函數的空間部分 $[\text{m}^{-1/2}]$ 
  * $k$ : 波數 (Wave number)，$[\text{m}^{-1}]$
  * $A,B,c_1,c_2$ : 待定常數，$[\text{m}^{-1/2}]$


* **【已知 3】 帶入邊界條件並使常數有物理合理性：**
  
  * (a)對於在 $\text{I}$ 區的波函數 在負無窮遠處不能發散至無限大
  
  $$\begin{gather*}
  \psi_{\text{I}}(x) &=& C e^{ax} + D e^{-ax} \\
  \lim_{ x  \to -\infty}  \psi_{\text{I}}(x)  &=& \lim_{ x  \to -\infty} \left(C e^{ax} + D e^{-ax} \right) \\
  0 &\overset{\text{假設 2}}{=}& \lim_{ x  \to -\infty} \left(C e^{ax} + D e^{-ax} \right)\\
  0 &=& \lim_{ x  \to -\infty} C e^{ax} + \lim_{ x  \to -\infty} D e^{-ax} \\
  0 &=& 0 + \lim_{ x  \to -\infty} D e^{-ax} \\
  D &=& 0
  \end{gather*}$$

  * (b)
  
  $$\begin{gather*}
  \psi_{\text{III}}(x) &=& F e^{ax} + G e^{-ax} \\
  \lim_{ x  \to \infty}  \psi_{\text{III}}(x)  &=& \lim_{ x  \to \infty} \left(F e^{ax} + G e^{-ax} \right) \\
  0 &\overset{\text{假設 2}}{=}& \lim_{ x  \to \infty} \left(F e^{ax} + G e^{-ax} \right)\\
  0 &=& \lim_{ x  \to \infty} F e^{ax} + \lim_{ x  \to \infty} G e^{-ax} \\
  0 &=&  \lim_{ x  \to \infty} F e^{ax} + 0\\
  F &=& 0
  \end{gather*}$$

* **【已知 4】 在邊界 $x=0$ 處的連續性推導：**
  
  * **(a) 波函數值連續：**
  
  $$\begin{gather*}
  \psi_{\text{I}}(0) &\overset{\text{假設 3}}{=}& \psi_{\text{II}}(0) \\
  \left.C e^{ax} \right|_{x=0}&=& \left. (A \sin(k x) + B \cos(k x)) \right|_{x=0} \\
  C e^{a(0)} &\overset{x=0}{=}& A \sin(k \times 0) + B \cos(k \times 0) \\
  C \cdot 1 &=& A \cdot 0 + B \cdot 1 \\
  C &=& B
  \end{gather*}$$

  * **(b) 一階導數連續：**

  $$\begin{gather*}
  \frac{d\psi_{\text{I}}}{dx} &\overset{\text{假設 3}}{=}&  \frac{d\psi_{\text{II}}}{dx} \\
  \left.\frac{d}{dx}\left[C e^{ax} \right]\right|_{x=0} &=& \left.\frac{d}{dx}\left[A \sin(k x) + B \cos(k x) \right] \right|_{x=0}\\
  \left.aC e^{ax} \right|_{x=0}&=&\left. (kA \cos(kx) - kB \sin(kx) )\right|_{x=0}\\
  aC e^{a(0)} &\overset{x=0}{=}& kA \cos(0) - kB \sin(0) \\
  aC &=& kA \\
  aB &\overset{\text{已知 4(a)}}{=}& kA \\
  A &=& \frac{a}{k}B
  \end{gather*}$$

* **【已知 5】 在邊界 $x=L$ 處的連續性推導：**
  
  * **(a) 波函數值連續：**
  
  $$\begin{gather*}
  \psi_{\text{II}}(L) &\overset{\text{假設 3}}{=}& \psi_{\text{III}}(L) \\
  \left. A \sin(k x) + B \cos(k x) \right|_{x=L} &=& \left. G e^{-ax}\right|_{x=L} \\
  A \sin(k L) + B \cos(k L) &\overset{x=L}{=}& G e^{-aL} \\
  G &=& e^{aL} (A \sin(k L) + B \cos(k L))\\
  G &\overset{\text{已知 4(b)}}{=}& e^{aL} (\frac{a}{k}B \sin(k L) + B \cos(k L))\\ 
  G &=& e^{aL}B (\frac{a}{k} \sin(k L) +  \cos(k L))\\ 
  \end{gather*}$$

  * **(b) 一階導數連續：**

  $$\begin{gather*}
  \frac{d\psi_{\text{II}}}{dx} &\overset{\text{假設 3}}{=}&  \frac{d\psi_{\text{III}}}{dx} \\
  \left.\frac{d}{dx}\left[A \sin(k x) + B \cos(k x) \right]\right|_{x=L} &=& \left.\frac{d}{dx}\left[G e^{-aL} \right] \right|_{x=L}\\
  \left.(kA \cos(kx) - kB \sin(kx)) \right|_{x=L}&=&\left. -aG e^{-ax} \right|_{x=L}\\
  kA \cos(kL) - kB \sin(kL) &\overset{x=L}{=}& -aG e^{-aL}  \\
  G &=& e^{aL}\frac{k}{-a}( A \cos(kL) - B \sin(kL)) \\
  G &\overset{\text{已知 4(b)}}{=}& -e^{aL}\frac{k}{a}( \frac{a}{k} B\cos(kL) -  B\sin(kL)) \\
  G &=& -e^{aL}\frac{k}{a}B( \frac{a}{k} \cos(kL) -  \sin(kL)) \\
  \end{gather*}$$

  * **(c) 超越方程式 (Transcendental Equation) 能量量子化（Quantization）**
  
  $$\begin{gather*}
  &\frac{G}{G} &\overset{\text{已知 5(a)(b)}}{=}& -\frac{e^{aL} B(\frac{a}{k} \sin(k L) +  \cos(k L))}{e^{aL}\frac{k}{a}B( \frac{a}{k} \cos(kL) -  \sin(kL))} \\
  &-1 &=& \frac{\frac{a}{k}\sin(k L) + \cos(k L) }{\cos(k L)- \frac{k}{a} \sin(kL)} \\
  &-\cos(k L) + \frac{k}{a} \sin(kL) &=&  \frac{a}{k}\sin(k L) + \cos(k L) \\
  &(\frac{k}{a} - \frac{a}{k})\sin(kL) &=& 2 \cos(k L) \\
  &\frac{\sin(kL)}{\cos(k L)} &=& \frac{2}{\frac{k}{a} - \frac{a}{k}}  \\
  &\tan(kL) &=& \frac{2ak}{k^2 - a^2}  & \Rightarrow & \left\{\begin{matrix}\sin(kL)  &=& \frac{2ak}{k^2+a^2} \\ \cos(kL)  &=& \frac{k^2-a^2}{k^2+a^2} \\ \end{matrix}\right. \\
  \end{gather*}$$




* **【已知 6】 三角函數倍角公式 (Double Angle Formulas)：**

  * (a) 
  
  $$\sin^2(\theta) = \frac{1 - \cos(2\theta)}{2} $$

  * (b)
  
  $$\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2} $$

  * (c)
  
  $$2\sin(\theta)\cos(\theta) = \sin(2\theta)$$



* **【已知 7】歸一化推導:每個區域準備：** 
  
  
  1. 第一區積分 ($I_1$：左側牆壁內的滲透)

  $$\begin{gather*}
  I_1 &=& \int_{-\infty}^{0} (\psi_{\text{I}}(x))^2 dx \\
  &=& \int_{-\infty}^{0} (B e^{ax})^2 dx \\
  &=& B^2 \int_{-\infty}^{0} e^{2ax} dx \\
  &=& B^2 \left. \frac{1}{2a} e^{2ax} \right|^{-\infty}_{0} \\
  &=& B^2 \left( \frac{1}{2a} e^{0} - \frac{1}{2a} \lim_{x \to -\infty} e^{2ax} \right) \\
  &=& B^2 \left( \frac{1}{2a} \cdot 1 - 0 \right) \\
  &=& \frac{B^2}{2a}
  \end{gather*}$$

  2. 第三區積分 ($I_3$：右側牆壁內的滲透)

  $$\begin{gather*}
  I_3 &=& \int_{L}^{\infty} (\psi_{\text{III}}(x))^2 dx \\
  &=& \int_{L}^{\infty} (Be^{-a(x-L)})^2 dx \\
  &=& B^2   \int_{L}^{\infty} e^{-2a(x-L)} dx \\
  &\underset{du = dx}{\overset{u \overset{\text{let}}{=} x-L}{=}}& B^2  \int_{0}^{\infty} e^{-2au} du \\
  &=& B^2 \left[ \frac{-1}{2a} e^{-2au} \right]_{0}^{\infty} \\
  &=& B^2  \left( 0 - \left( \frac{-1}{2a} e^{0} \right) \right) \\
  &=& B^2  \left( \frac{1}{2a} \right) \\
  &=& \frac{B^2}{2a} \\
  \end{gather*}$$


  3. 第二區積分 ($I_2$：中央位能井內的震盪)

  $$\begin{gather*}
  I_2 &=& \int_{0}^{L} \left( \psi_{\text{II}}(x) \right)^2 dx \\
  &=& \int_{0}^{L} B^2 \left( \frac{a}{k} \sin(kx) + \cos(kx) \right)^2 dx \\
  &=& B^2 \int_{0}^{L} \left[ \frac{a^2}{k^2} \sin^2(kx) + \frac{2a}{k} \sin(kx)\cos(kx) + \cos^2(kx) \right] dx \\
  &\overset{\text{已知 6(a)(b)(c)}}{=}& B^2 \int_{0}^{L} \left[ \frac{a^2}{k^2} \left( \frac{1 - \cos(2kx)}{2} \right) + \frac{a}{k} \sin(2kx) + \left( \frac{1 + \cos(2kx)}{2} \right) \right] dx \\
  &=& B^2 \int_{0}^{L} \left[ \left( \frac{a^2}{2k^2} + \frac{1}{2} \right) + \left( \frac{1}{2} - \frac{a^2}{2k^2} \right) \cos(2kx) + \frac{a}{k} \sin(2kx) \right] dx \\
  &=& B^2 \int_{0}^{L} \left[ \frac{a^2+k^2}{2k^2} + \frac{k^2-a^2}{2k^2} \cos(2kx) + \frac{a}{k} \sin(2kx) \right] dx \\
  &=& B^2 \left[ \frac{a^2+k^2}{2k^2}x + \frac{k^2-a^2}{2k^2} \left( \frac{\sin(2kx)}{2k} \right) + \frac{a}{k} \left( \frac{-\cos(2kx)}{2k} \right) \right]_0^L \\
  &=& B^2 \left[ \frac{a^2+k^2}{2k^2}x + \frac{k^2-a^2}{4k^3} \sin(2kx) - \frac{a}{2k^2} \cos(2kx) \right]_0^L \\
  &=& B^2 \left[ \left[ \frac{a^2+k^2}{2k^2}L + \frac{k^2-a^2}{4k^3} \sin(2kL) - \frac{a}{2k^2} \cos(2kL) \right] - \left[ 0 + 0 - \frac{a}{2k^2} \cos(0) \right] \right) \\
  &=& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{k^2-a^2}{4k^3} \sin(2kL) + \frac{a}{2k^2} \left( 1 - \cos(2kL) \right) \right) \\
  &\overset{\text{已知 6(a)(c)}}{=}& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{k^2-a^2}{4k^3} 2\sin(kL)\cos(kL) + \frac{a}{2k^2} 2\sin^2(kL) \right) \\
  &\overset{\text{已知 5(c)}}{=}& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{k^2-a^2}{4k^3} 2\frac{2ak}{k^2+a^2}\frac{k^2-a^2}{k^2+a^2} + \frac{a}{2k^2} 2 \left(\frac{2ak}{k^2+a^2} \right)^2 \right) \\
  &=& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{k^{2}-a^{2}}{4k^{3}}\left(\frac{4ak(k^{2}-a^{2})}{(k^{2}+a^{2})^{2}}\right)+\frac{a}{2k^{2}}\left(\frac{8a^{2}k^{2}}{(k^{2}+a^{2})^{2}}\right) \right) \\
  &=& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{a(k^{2}-a^{2})^{2}}{k^{2}(k^{2}+a^{2})^{2}}+\frac{4a^{3}k^{2}}{k^{2}(k^{2}+a^{2})^{2}} \right) \\
  &=& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{a(k^{4}-2a^{2}k^{2}+a^{4}+4a^{2}k^{2})}{k^{2}(k^{2}+a^{2})^{2}} \right) \\
  &=& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{a(k^{2}+a^{2})^{2}}{k^{2}(k^{2}+a^{2})^{2}} \right) \\
  &=& B^2 \left( \frac{a^2+k^2}{2k^2}L + \frac{a}{k^{2}} \right) \\
  \end{gather*}$$





* **【已知 8】 波函數歸一化 (Normalization)：**

  $$\begin{gather*}
  \int_{-\infty}^{\infty} |\psi(x)|^2 dx &=& 1 \\
  \int_{-\infty}^{0} \psi_{\text{I}}^2 dx + \int_{0}^{L} \psi_{\text{II}}^2 dx + \int_{L}^{\infty} \psi_{\text{III}}^2 dx &=& 1 \\
  I_1 + I_2 + I_3 &=& 1 \\
  \frac{B^2}{2a} + \left(B^2 \left[ \frac{a^2+k^2}{2k^2}L + \frac{a}{k^{2}} \right]\right) + \frac{B^2}{2a}  &\overset{\text{已知 7}}{=}& 1 \\
  \frac{B^2}{2a} + \left(B^2 \left[ \frac{a^2+k^2}{2k^2}L + \frac{a}{k^{2}} \right]\right) + \frac{B^2}{2a}  &\overset{\text{已知 7}}{=}& 1 \\
  \frac{B^{2}}{a}+B^{2}\left[\frac{a^{2}+k^{2}}{2k^{2}}L+\frac{a}{k^{2}}\right]&=&1 \\
  B^{2}\left[\frac{1}{a}+\frac{a}{k^{2}}+\frac{a^{2}+k^{2}}{2k^{2}}L\right]&=&1 \\
  B^{2}\left[\frac{k^{2}+a^{2}}{ak^{2}}+\frac{k^{2}+a^{2}}{2k^{2}}L\right]&=&1 \\
  B^{2}\cdot \frac{k^{2}+a^{2}}{2k^{2}}\left[\frac{2}{a}+L\right]&=&1 \\
  B^{2}&=&\frac{2k^{2}}{(k^{2}+a^{2})\left(L+\frac{2}{a}\right)}  \\
  B&=&\frac{k}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}}  \\
  \end{gather*}$$



### 波函數證明


* **第一區 (區 I, $x < 0$)：**
  
$$\begin{gather*}
& \frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &\overset{\text{已知 1}}{=}& 0 \\
\overset{\text{區 I}}{\Rightarrow} & \frac{d^2 \psi_{\text{I}}}{dx^2} + \frac{2m}{\hbar^2} (E-U(x)) \psi_{\text{I}} &=& 0 \\
& \frac{d^2 \psi_{\text{I}}}{dx^2} + \frac{2m}{\hbar^2} (E-U_0) \psi_{\text{I}} &\overset{\text{假設 1}}{=}& 0 \\
& \frac{d^2 \psi_{\text{I}}}{dx^2} + a^2 \psi_{\text{I}} &\overset{\text{定義 2}}{=}& 0 \\
&\psi_{\text{I}}(x) &\overset{\text{已知 2(b)}}{=}& C e^{ax} + D e^{-ax} \\
&\psi_{\text{I}}(x) &\overset{\text{已知 3(a)}}{=}& C e^{ax} \\
&\psi_{\text{I}}(x) &\overset{\text{已知 4(a)}}{=}& B e^{ax} \\
&\psi_{\text{I}}(x) &\overset{\text{已知 8}}{=}& \frac{k}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}}  e^{ax} \\
\end{gather*}$$



* **第三區 (區 III, $x > L$)：**

$$\begin{gather*}
& \frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &\overset{\text{已知 1}}{=}& 0 \\
\overset{\text{區 III}}{\Rightarrow} & \frac{d^2 \psi_{\text{III}}}{dx^2} + \frac{2m}{\hbar^2} (E-U(x)) \psi_{\text{III}} &=& 0 \\
& \frac{d^2 \psi_{\text{III}}}{dx^2} + \frac{2m}{\hbar^2} (E-U_0) \psi_{\text{III}} &\overset{\text{假設 1}}{=}& 0 \\
& \frac{d^2 \psi_{\text{III}}}{dx^2} + a^2 \psi_{\text{III}} &\overset{\text{定義 2}}{=}& 0 \\
&\psi_{\text{III}}(x) &\overset{\text{已知 2(b)}}{=}& F e^{ax} + G e^{-ax} \\
&\psi_{\text{III}}(x) &\overset{\text{已知 3(b)}}{=}& G e^{-ax} \\
&\psi_{\text{III}}(x) &\overset{\text{已知 5(a)}}{=}&  e^{aL} B(\frac{a}{k} \sin(k L) +  \cos(k L)) e^{-ax} \\
&\psi_{\text{III}}(x) &=& B(\frac{a}{k} \sin(k L) +  \cos(k L)) e^{-a(x-L)} \\
&\psi_{\text{III}}(x) &\overset{\text{已知 5(c)}}{=}& B\left( \frac{a}{k} \frac{2ak}{k^2+a^2} + \frac{k^2-a^2}{k^2+a^2} \right) e^{-a(x-L)} \\
&\psi_{\text{III}}(x) &=& B\left(\frac{2a^2}{k^2+a^2} + \frac{k^2-a^2}{k^2+a^2} \right) e^{-a(x-L)} \\
&\psi_{\text{III}}(x) &=& B\left(\frac{2a^2+k^2-a^2}{k^2+a^2} \right) e^{-a(x-L)} \\
&\psi_{\text{III}}(x) &=& B\left(\frac{a^2+k^2}{k^2+a^2} \right) e^{-a(x-L)} \\
&\psi_{\text{III}}(x) &=& Be^{-a(x-L)} \\
&\psi_{\text{III}}(x) &\overset{\text{已知 8}}{=}& \frac{k}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}} e^{-a(x-L)} \\
\end{gather*}$$



* **第二區 (區 III, $0 \le x \le L$)：**

$$\begin{gather*}
&\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &\overset{\text{已知 1}}{=}& 0 \\
\overset{\text{區 II}}{\Rightarrow} & \frac{d^2 \psi_{\text{II}}}{dx^2} + \frac{2m}{\hbar^2} (E - U(x))\psi_{\text{II}} &=& 0 \\
& \frac{d^2 \psi_{\text{II}}}{dx^2} + \frac{2m}{\hbar^2} E\psi_{\text{II}} &\overset{\text{假設 1}}{=}& 0 \\
& \frac{d^2 \psi_{\text{II}}}{dx^2} + k^2 \psi_{\text{II}} &\overset{\text{定義 1}}{=}& 0 \\
&  \psi_{\text{II}}(x)  &\overset{\text{已知 2}}{=}& A \sin(kx) + B \cos(kx) \\
&  \psi_{\text{II}}(x)  &\overset{\text{已知 4(b)}}{=}& \frac{a}{k}B \sin(kx) + B \cos(kx) \\
&  \psi_{\text{II}}(x)  &=& \frac{a}{k}B \left(\sin(kx) +  \cos(kx)\right) \\
&  \psi_{\text{II}}(x)  &\overset{\text{已知 8}}{=}& \frac{a}{k}\frac{k}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}} \left(\sin(kx) +  \cos(kx)\right) \\
&  \psi_{\text{II}}(x)  &=& \frac{a}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}} \left(\sin(kx) +  \cos(kx)\right) \\
\end{gather*}$$




* 合併 **第一區、第二區、第三區**

$$\begin{gather*}
\psi(x) &=& \left\{\begin{matrix} \psi_{\text{I}} &\quad x < 0  \\ \psi_{\text{II}} &\quad 0 \le x \le L  \\ \psi_{\text{III}} &\quad x > L \end{matrix}\right.  \\
\psi(x) &=& \left\{\begin{matrix} 
\frac{k}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}}  e^{ax} &\quad x < 0  \\ 
\frac{a}{k}B \left(\sin(kx) +  \cos(kx)\right) &\quad 0 \le x \le L  \\ 
\frac{k}{\sqrt{k^{2}+a^{2}}}\sqrt{\frac{2}{L+\frac{2}{a}}} e^{-a(x-L)} &\quad x > L \end{matrix}\right.  \\
\end{gather*}$$

在物理的推導中，當我們遇到擁有多個未知常數的微分方程組時，目標就是盡可能「用最少的常數來表達整個系統」。我們將運用剛才得到的線索：$C = B$ 以及 $A = \frac{a}{k}B$。

#### 4. 聯立求解與超越方程式 (Transcendental Equation)

$$\begin{gather*}
2ak \cos(kL) &=& (k^2 - a^2) \sin(kL) \\
\frac{2ak}{k^2 - a^2} &=& \frac{\sin(kL)}{\cos(kL)} \\
\tan(kL) &=& \frac{2ak}{k^2 - a^2}
\end{gather*}$$

* **能階的物理意義**：這個方程式被稱為**超越方程式 (Transcendental Equation)**。因為 $k$ 和 $a$ 都隱含了能量 $E$ 變數（回顧 $k = \frac{\sqrt{2mE}}{\hbar}$ 和 $a = \frac{\sqrt{2m(U_0 - E)}}{\hbar}$），這條方程式無法像無限深位能井那樣直接移項寫成 $E_n = \dots$ 的完美代數解。在實際的研究與工程計算中，我們通常會利用數值分析方法（例如在 Python 中呼叫 SciPy 的 `fsolve` 模組）或圖解法來求出允許的 $E$ 值。

#### 5. 完整的波函數形式 (Complete Wave Function)

* $\psi(x)$ : 系統的一維定態波函數， $[\text{m}^{-1/2}]$
* $B$ : 最終的待定常數，可透過【已知 5】波函數歸一化 $\int_{-\infty}^{\infty} |\psi|^2 dx = 1$ 的積分求得， $[\text{m}^{-1/2}]$
* $k$ : 內部波數，與粒子動能有關， $[\text{m}^{-1}]$
* $a$ : 外部衰減常數，決定了波函數「滲透」牆壁的深度 (Penetration depth, 通常定義為 $\delta = \frac{1}{a}$)， $[\text{m}^{-1}]$

這樣我們就以嚴謹的邏輯，完美建構出有限位能井的數學模型了。這三個區域的波函數在邊界上平滑過渡，沒有斷層，也沒有不合理的折角，這正是量子力學中波函數所要求的美麗性質。


## 5.6 一維有限深位能井 (Finite Potential Well)

若箱子的牆壁高度不是無限大，而是有限的位能 $U_0$。當粒子的總能量 $E < U_0$ 時，雖然古典力學依然認為粒子被困在井中，但量子力學容許波函數「滲透」定位能壁內。



將空間分為三區：
* **第一區 ($x < 0$):** $U = U_0$
* **第二區 ($0 < x < L$):** $U = 0$
* **第三區 ($x > L$):** $U = U_0$

在外部區域（區 I 與 III），薛丁格方程為：

$$\begin{gather*}
\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2} (E - U_0)\psi &=& 0 \\
\frac{d^2 \psi}{dx^2} - a^2 \psi &=& 0 \quad \left( \text{其中 } a = \frac{\sqrt{2m(U_0 - E)}}{\hbar} \right)
\end{gather*}$$
因為 $U_0 > E$，$a$ 是一個實數常數。此微分方程的通解為指數函數：
$$\begin{gather*}
\psi_{\text{I}} &=& C e^{ax} + D e^{-ax} \\
\psi_{\text{III}} &=& F e^{ax} + G e^{-ax}
\end{gather*}$$

**物理合理性篩選 (Finite everywhere):**
波函數在無窮遠處不能發散至無限大（否則機率為無限大無法歸一化）。
* 在第一區 ($x \to -\infty$)，若保留 $D e^{-ax}$ 會發散，故 $D = 0$，留下 $\psi_{\text{I}} = C e^{ax}$。
* 在第三區 ($x \to +\infty$)，若保留 $F e^{ax}$ 會發散，故 $F = 0$，留下 $\psi_{\text{III}} = G e^{-ax}$。

在中央井內（區 II，與無限深位能井相同）：
$$\psi_{\text{II}} = A \sin \left( \frac{\sqrt{2mE}}{\hbar} x \right) + B \cos \left( \frac{\sqrt{2mE}}{\hbar} x \right)$$

最後，藉由在邊界 $x=0$ 與 $x=L$ 處，波函數 $\psi$ 及其一階導數 $d\psi/dx$ 必須連續的四大條件，便可聯立解出所有常數，並得出粒子的能階。因為波函數可以稍微滲透牆壁，實際上的有效箱寬度變大了，導致有限位能井內的能階 $E_n$ 會比無限深位能井來得低。

## 5.7 穿隧效應 (Tunnel Effect)

量子力學中最違反直覺的現象之一：若粒子能量 $E$ 小於位能障壁高度 $U$，在古典力學中粒子會被 $100\%$ 彈回；但在量子力學中，因為波函數在障壁內是以指數衰減而非瞬間歸零，只要牆壁夠薄，波函數就有機會在另一端「穿透」出來，使粒子具備非零的機率跨越障壁。

### 穿隧機率精確解推導

給定一個高度為 $U$，寬度為 $L$（從 $x=0$ 到 $x=L$）的位能障壁。

**1. 分區薛丁格方程式與解：**
* **第一區 (入射區 $x < 0$, $U=0$):**
    $$\frac{d^2 \psi_{\text{I}}}{dx^2} + \frac{2m}{\hbar^2} E \psi_{\text{I}} = 0$$
    定義入射波數 $k_1 = \frac{\sqrt{2mE}}{\hbar}$，通解包含入射波（$e^{ik_1 x}$）與反射波（$e^{-ik_1 x}$）：
    $$\psi_{\text{I}} = A e^{ik_1 x} + B e^{-ik_1 x}$$
* **第二區 (障礙區 $0 < x < L$, $U > E$):**
    $$\frac{d^2 \psi_{\text{II}}}{dx^2} + \frac{2m}{\hbar^2} (E - U) \psi_{\text{II}} = 0$$
    定義衰減係數 $k_2 = \frac{\sqrt{2m(U - E)}}{\hbar}$，通解為實數指數形式：
    $$\psi_{\text{II}} = C e^{-k_2 x} + D e^{k_2 x}$$
* **第三區 (穿透區 $x > L$, $U=0$):**
    $$\frac{d^2 \psi_{\text{III}}}{dx^2} + \frac{2m}{\hbar^2} E \psi_{\text{III}} = 0$$
    因為障礙物後方沒有其他東西反射它，我們只保留向右行進的穿透波：
    $$\psi_{\text{III}} = F e^{ik_1 x}$$

**2. 連續邊界條件 (Boundary Conditions):**
在 $x = 0$ 處波與斜率連續：
$$\begin{gather*}
\psi_{\text{I}}(0) &=& \psi_{\text{II}}(0) \implies A + B = C + D \\
\psi_{\text{I}}'(0) &=& \psi_{\text{II}}'(0) \implies ik_1 A - ik_1 B = -k_2 C + k_2 D
\end{gather*}$$

在 $x = L$ 處波與斜率連續：
$$\begin{gather*}
\psi_{\text{II}}(L) &=& \psi_{\text{III}}(L) \implies C e^{-k_2 L} + D e^{k_2 L} = F e^{ik_1 L} \\
\psi_{\text{II}}'(L) &=& \psi_{\text{III}}'(L) \implies -k_2 C e^{-k_2 L} + k_2 D e^{k_2 L} = ik_1 F e^{ik_1 L}
\end{gather*}$$

**3. 穿隧機率 (Transmission Probability) $T$ 求解：**
穿隧機率定義為穿透波強度（機率流）除以入射波強度。在速度相同的情況下，此為振幅平方比：
$$T = \frac{|\psi_{\text{III}+}|^2 v_{\text{III}+}}{|\psi_{\text{I}+}|^2 v_{\text{I}+}} = \frac{FF^*}{AA^*} = \left| \frac{F}{A} \right|^2$$

透過上述四大方程式進行繁瑣的代數消去 $B, C, D$ 之後，可得到精確的穿隧率公式：
$$T = \left[ 1 + \frac{1}{4} \left( \frac{k_1}{k_2} + \frac{k_2}{k_1} \right)^2 \sinh^2(k_2 L) \right]^{-1}$$

**4. 寬障壁的近似解：**
通常實驗中障壁不會極度薄（即 $k_2 L \gg 1$），此時 $\sinh(k_2 L) \approx \frac{1}{2} e^{k_2 L}$，公式大幅簡化：
$$\begin{gather*}
T &\approx& \left[ \frac{1}{4} \left( \frac{k_1^2 + k_2^2}{k_1 k_2} \right)^2 \left( \frac{1}{2} e^{k_2 L} \right)^2 \right]^{-1} \\
&=& \left[ \frac{16 k_1^2 k_2^2}{(k_1^2 + k_2^2)^2} \right] e^{-2k_2 L}
\end{gather*}$$
而最前端的中括號因子數量級通常落在1附近，真正決定穿隧機率數量級的「主宰項」是後方的指數衰減，故最終常用的**近似穿隧機率**為：

$$T \approx e^{-2k_2 L}$$
其中：
$$k_2 = \frac{\sqrt{2m(U - E)}}{\hbar}$$

* $T$ : 穿隧機率 $[\text{dimensionless}]$
* $U$ : 位能障壁高度 $[\text{J}]$
* $E$ : 粒子入射總能量 $[\text{J}]$
* $L$ : 位能障壁寬度 $[\text{m}]$

#### **例題 5.7：穿隧機率的極端敏感性**

**題目：** 具有能量 $1.0\text{ eV}$ 和 $2.0\text{ eV}$ 的電子射向高達 $10.0\text{ eV}$、寬 $0.5\text{ nm}$ 的位能障壁。(a) 求它們個別的穿隧機率。(b) 如果障壁的寬度加倍，會有怎樣的影響？電子質量 $m_e = 9.1 \times 10^{-31}\text{ kg}$，約化普朗克常數 $\hbar = 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$。

**解答：**

(a) 對於 $1.0\text{ eV}$ 的電子：
首先計算障壁內的衰減波數 $k_2$：
$$\begin{gather*}
k_2 &=& \frac{\sqrt{2m(U - E)}}{\hbar} \\
&=& \frac{\sqrt{2(9.1 \times 10^{-31} \text{ kg})[(10.0 - 1.0) \text{ eV}](1.6 \times 10^{-19} \text{ J/eV})}}{1.054 \times 10^{-34} \text{ J}\cdot\text{s}} \\
&\approx& 1.6 \times 10^{10} \text{ m}^{-1}
\end{gather*}$$

計算指數項：
$$\begin{gather*}
L &=& 0.50 \text{ nm} = 5.0 \times 10^{-10} \text{ m} \\
2k_2 L &=& 2(1.6 \times 10^{10} \text{ m}^{-1})(5.0 \times 10^{-10} \text{ m}) \\
&=& 16
\end{gather*}$$

近似穿隧機率為：
$$T_1 = e^{-2k_2 L} = e^{-16} \approx 1.1 \times 10^{-7}$$

（平均每八百九十萬個電子會有一個成功穿隧）

對 $2.0\text{ eV}$ 電子套用相同計算：
可以得到 $2k_2 L \approx 15.24$
$$T_2 = e^{-15.24} \approx 2.4 \times 10^{-7}$$
提升粒子的能量使得穿透機率提升了約兩倍多。

(b) 障壁寬度加倍（$L = 1.0\text{ nm}$）：
此時 $2k_2 L$ 直接翻倍：
$$\begin{gather*}
T_1' &=& e^{-32} \approx 1.3 \times 10^{-14} \\
T_2' &=& e^{-30.48} \approx 5.1 \times 10^{-14}
\end{gather*}$$

**【物理知識回顧】**
在 $1.0\text{ eV}$ 的條件下，原本百萬分之一機率可觀測到的穿隧現象，僅僅因為把障壁厚度加厚了 $0.5\text{ nm}$（約只有幾個原子的寬度），穿透機率就發生了雪崩式暴跌，縮小了一千萬倍！這展現了穿隧效應對於「距離 $L$」具有指數級的極端敏感性。這項特質被科學家完美利用發明了**掃描式穿隧電子顯微鏡 (STM)**，藉由偵測微小的穿隧電流變化，就能達到觀測單個原子表面起伏的超級解析度。
