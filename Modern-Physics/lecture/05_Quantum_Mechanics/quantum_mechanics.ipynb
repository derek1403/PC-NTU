### 量子力學基本規則

* 波函數 $\Psi$
* 在那個位子和時間發現物體的機率 $\Psi \Psi^*$
  * $\Psi , \Psi^* \in \mathbb{C}$ 是共軛負數
  * 例如 $\Psi= A+Bi$ 則 $\Psi^*=A-Bi$
* 歸一化條件（Normalization Condition） $\int_{-\infty}^{\infty} |\Psi|^2 dV = 1$
  * $|\Psi|^2$ 是 機率密度
  * $dV$ 體積微元
  * $= 1$ 代表全空間找到粒子的總機率為 100%。
* 波函數 $\Psi$ 本身和其對空間的一階微分要連續而且有單一值
  * $\Psi$ 本身要和 $\frac{d}{dt} [\Psi]$ $\frac{d}{dx} [\Psi]$ 要連續且唯一
    * 因為要符合波動方程式 $\frac{\partial^2 y}{\partial x^2} = \frac{1}{v^2} \frac{\partial^2 y}{\partial t^2}$
    * 只要一接微分連續就可，二街微分不需要連續(一般理想的位能井都不會連續)，但要可積分
    * 波函數不連續：意味著粒子在A點出現的機率與緊鄰的B點有巨大的斷層，這會導致機率流變成無窮大（粒子瞬間移動），這違反了因果律。
    * 一階微分不連續：這代表動量在瞬間發生無窮大的改變。在現實中，任何力場（位能）都有一個緩衝的空間，所以微分項通常是平滑的。
    * 物理世界真的連續嗎?
      * 廣義相對論與時空的「平滑性」:廣義相對論認為時空是連續且可微的「流形」（Manifold）。這意味著你可以無限地分割空間，且在任何一點都能定義切平面（微分）。
        * 危機：當我們把大量的質量壓縮到極小空間時（如黑洞中心或大霹靂初期），時空曲率會趨向無窮大（奇異點）。這顯示廣義相對論這套「連續」的語言在極限下失效了，我們需要一種更底層的物理架構。
      * 量子引力中的「時空量子化」: 許多量子引力理論認為，時空並非如絲綢般連續，而是像織布一樣，由細小的「纖維」編織而成：
        * 圈量子重力（Loop Quantum Gravity, LQG）：這是最直接挑戰連續性的理論。它預言空間本身是量子化的，存在最小的體積與面積單位。在普朗克尺度下，你無法再往下分割，這就像電腦螢幕的像素，遠看連續，近看則是一個個點。
        * 弦論（String Theory）：雖然弦論在數學上仍使用連續背景，但它認為基本的物理實體不是「點」，而是「弦」。因為弦有最小長度（普朗克長度），我們無法探測比弦更小的尺度，這在實際上為空間的「連續分割」設定了一個天花板。
      * 非交換幾何（Non-commutative Geometry）:有一些觀點認為，在極小尺度下，座標 x和y 可能不滿足乘法交換律（類似量子力學中的位置與動量）。如果 xy \ne yx ，那麼「點」的概念就消失了。在這種尺度下，空間不再是點的集合，而是一種模糊的幾何結構，連續性自然無從談起。
      * 物理直覺的挑戰：全像原理
        * 全像原理（Holographic Principle） 認為，一個區域內的所有資訊可以編碼在其邊界表面上。這暗示了資訊量是有限的（離散的），而非無限的（連續的）。如果資訊是離散的，那麼承載資訊的時空背景也很難是絕對連續的。
      * 總結觀點
        * 雖然我們目前還沒有直接的實驗證據（因為普朗克尺度太小，目前的對撞機根本測不到），但物理學界的共識趨向於：連續性只是巨觀下的錯覺。

#### **例題 5.2：**

**題目：** 一個限制於 x 軸上的粒子，在x=0 與x=1 之間的波函數 $\Phi=ax$ ；在其他位置時則為$\Phi=0$ 。(a) 求 出粒子在x=0.45 與x=0.55 之間出現的機率。(b) 求出粒子位置的期望值。

## 5.3 運算子 Operators

$\Psi = Ae^{-(i/\hbar)(Et - px)}$


$\frac{\partial \Psi}{\partial x} = \frac{i}{\hbar} p \Psi$

$\frac{\partial \Psi}{\partial t} = - \frac{i}{\hbar} E \Psi$


可以得到

$p \Psi = \frac{\hbar}{i} \frac{\partial}{\partial x} \Psi$

$E \Psi = i\hbar \frac{\partial}{\partial t} \Psi$


所以就定義 運算子。當運算子碰到 $\Phi$，就可以的到 運算子\hat下面的東西*$\Phi$


$\hat{p} \overset{\text{def}}{=} \frac{\hbar}{i} \frac{\partial}{\partial x}$

$\hat{E}\overset{\text{def}}{=}  i\hbar \frac{\partial}{\partial t}$

$\hat{E} \overset{\text{def}}{=} \hat{\text{KE}} + \hat{U}$

$\hat{\text{KE}} \overset{\text{def}}{=} \frac{\hat{p}^2}{2m} = \frac{1}{2m} \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right)^2 = -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2}$


組合後可以變成剛剛eqution的operator 

$$\begin{gather*}
i\hbar \frac{\partial}{\partial t} &=& -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + U\\
i\hbar \frac{\partial \Psi}{\partial t}&=&  -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U\Psi\\
\end{gather*}$$

* 總能
* 動能
* 位能

* 透過這樣的組合我們不用知道 G(x)這個物理量本身對x的關係(通常我們都不知道)，只要知道他的算子 $\hat{G}$就可以操作了


### 實際操作

$$\begin{gather*}
\langle p \rangle &=& \int_{-\infty}^{\infty} \Psi^* \hat{p} \Psi dx \\
&=&\int_{-\infty}^{\infty} \Psi^* \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right) \Psi dx \\
&=& \frac{\hbar}{i} \int_{-\infty}^{\infty} \Psi^* \frac{\partial \Psi}{\partial x} dx \\
\end{gather*}$$

$$\begin{gather*}
\langle E \rangle &=& \int_{-\infty}^{\infty} \Psi^* \hat{E} \Psi dx \\
&=& \int_{-\infty}^{\infty} \Psi^* \left( i\hbar \frac{\partial}{\partial t} \right) \Psi dx \\
&=& i\hbar \int_{-\infty}^{\infty} \Psi^* \frac{\partial \Psi}{\partial t} dx
\end{gather*}$$

$$\begin{gather*}
\langle G(x, t) \rangle = \int_{-\infty}^{\infty} \Psi^* \hat{G} \Psi dx
\end{gather*}$$




物理意義: 在相同的量子態 $\Psi $ 下，重複測量物理量 $G$ 無數次後得到的平均值。


. 為什麼一定要是 \(\int \Psi^* \hat{G} \Psi \, dx\)？在量子力學中，物理量（如位置、動量、能量）並不是簡單的數值，而是算符（通常包含微分）。\(\Psi \) 是「被操作者」：算符 \(\^{G}\) 必須作用在波函數 \(\Psi \) 上才有意義。就像 \(\frac{d}{dx}\) 必須寫在 \(x^{2}\) 前面（即 \(\frac{d}{dx}x^2\)），不能寫成 \(x^2 \frac{d}{dx}\)。\(\Psi ^{*}\) 的角色：為了讓計算結果是一個實數（物理測量值必須是實數），我們需要利用複數共軛 \(\Psi ^{*}\) 與原函數 \(\Psi \) 相乘（類似向量內積的概念）。

如果交換順序會怎麼樣？如果我們隨便調換位置，例如寫成 \(\int \hat{G} (\Psi^* \Psi) \, dx\) 或 \(\int \Psi \Psi^* \hat{G} \, dx\)：數學錯誤：如果 \(\^{G}\) 是動量算符 \(\hat{p} = -i\hbar \frac{\partial}{\partial x}\)，它會去微分它後面的所有東西。在 \(\int \Psi^* \hat{G} \Psi \, dx\) 中，它只微分 \(\Psi \)。如果你寫成 \(\int \hat{G} (\Psi^* \Psi) \, dx\)，它會同時微分 \(\Psi ^{*}\) 和 \(\Psi \)，根據微分的乘法法則（Product rule），這會多出多餘的項，算出來的結果會完全錯誤。複數問題：隨便交換順序通常會導致結果變成複數，而現實世界中你不可能量測到一個「長度是 \(3 + 2i\) 公尺」的距離。

## 5.4 薛丁格方程式：穩態型式 Schrödinger’s Equation: Steady-State Form


本徵值及本徵函數


$$\Psi = Ae^{-(i/\hbar)(Et-px)} = Ae^{-(iE/\hbar)t} e^{(ip/\hbar)x} = \psi e^{-(iE/\hbar)t}$$

$$i\hbar \frac{\partial \Psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi}{\partial x^2} + U\Psi$$

$$\frac{\partial^2 \psi}{\partial x^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$$

$$\frac{\partial^2 \psi}{\partial x^2} + \frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial z^2} + \frac{2m}{\hbar^2}(E - U)\psi = 0$$


本徵的意思是 hat 放的operator 會變成數字

$\hat{G}\psi_n = G_n\psi_n$

* 這個 $G_n$就是數字(純量)
* 這在數學上的意思是：當算符 \(\^{G}\) 作用在特定的函數 \(\psi _{n}\) 上時，結果居然只是把這個函數縮放了 \(G_{n}\) 倍，而函數本身的形狀完全沒變。這時候，我們稱 \(G_{n}\) 是這個系統可能的觀測值

漢彌爾頓運算子 定義

$\hat{H} \overset{\text{def}}{=} -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + U$
* 它代表的是系統的總能量算符（動能 + 位能）。你可以把它想像成一個「能量處理機」。



薛丁格方程式
$\hat{H}\psi_n = E_n\psi_n$

薛丁格方程式 (\(\hat{H}\psi_n = E_n\psi_n\))：這是最著名的應用。當你把「能量處理機」作用在波函數上，吐出來的縮放倍率 \(E_{n}\) 就是該量子態下的能量等級。這也說明了為什麼能量是「一份一份」的（量子化），因為只有特定的 \(\psi _{n}\) 才能滿足這個等號。

* 只有H作用在上面的形式才叫薛丁格方程式
* En能量量子化

### **例題 5.4：**

**題目：** 運算子 \( d^2/dx^2 \) 的本徵函數是 \( \psi = e^{2x} \)。求出對應的本徵值。

$\hat{G} = d^2/dx^2 $

$\hat{G}\psi = 4\psi$
$G = 4$

## 4.6 原子核運動 (Nuclear Motion)

前面的波耳模型都假設原子核是「無限重」而靜止不動的。然而實際上，質子的質量雖然是電子的 1836 倍，兩者實際上是共同繞著系統的**質心 (Center of mass)** 旋轉的。為了解決這個雙體問題，我們在物理上引入**減縮質量 (Reduced mass)** 的概念。我們可以將系統等效視為：一個質量無限大的原子核，被一個質量為 $m'$ 的等效電子繞行。


$$m' \overset{\text{def}}{=} \frac{mM}{m + M}$$

* $m'$ : 減縮質量 $[\text{kg}]$
* $m$ : 電子質量 $[\text{kg}]$
* $M$ : 原子核質量 $[\text{kg}]$

![](./pic/Nuclear_Motion_Reduce_Mass.png)

修正後的能階公式，只需將電子質量 $m$ 替換為減縮質量 $m'$：

$$\begin{gather*}
E'_n &=& -\frac{m'e^4}{8\epsilon_0^2 h^2} \left( \frac{1}{n^2} \right) &=& \left( \frac{m'}{m} \right) \left( \frac{E_1}{n^2} \right) &=& \frac{m'}{m} E_n \\
r'_n &=& \frac{n^2 h^2 \epsilon_0}{\pi m' e^2} &=& \left( \frac{m}{m'} \right) \left( n^2  r_1\right) &=&  \frac{m}{m'} r_n
\end{gather*}$$

這個微小的修正:
* $\frac{m'}{m} <1$
  * $E'_n < E_n$ 
  * $r'_n > r_n$ 
* 完美解釋了同位素之間光譜的微小差異（同位素位移效應）
* 當原子核質量 $M$ 越大，$m'$ 越接近 $m$，能階的偏移就越不明顯
* 當兩種粒子質量相近時，這項修正將帶來決定性的影響


### **例題 4.6.1：氫原子的能階修正**

**題目：** 針對氫原子，電子質量 $m \approx 9.109 \times 10^{-31}\text{ kg}$，質子質量 $M \approx 1.672 \times 10^{-27}\text{ kg}$，質量比 $M/m \approx 1836.15$。計算其減縮質量與原本電子的質量比值。


$$\begin{gather*}
m' &\overset{\text{def}}{=}& \frac{mM}{m + M} \\
\frac{m'}{m} &=& \frac{M}{m+M} \\
\frac{m'}{m}&=& \frac{M/m}{1 + M/m} \\
\frac{m'}{m}&=&\frac{1836.15}{1 + 1836.15} \\
\frac{m'}{m}&\approx& 0.99945
\end{gather*}$$

**【物理知識回顧】**
這個比例誤差極小（約 $0.05\%$），說明波耳「原子核不動」的假設是一個非常精準的近似。然而，對於含有中子的氘原子（重氫，$M \approx 3670m$），其 $m'/m \approx 0.99973$。光譜學家正是透過觀測到氫與氘這萬分之幾的光譜頻率差異，證實了重氫的存在。

### **例題 4.6.2：正電子偶 (Positronium)**

**題目：** 正電子偶（Positronium）為一個正電子 (positron，電子的反物質，質量與電子相同，帶正電) 和一個電子相互繞行的系統。試比較正電子偶的光譜線波長與一般氫原子光譜線波長的關係。


$$\begin{gather*}
E'_n &\overset{\text{ch4.6}}{=}& \left( \frac{m'}{m} \right) \frac{E_1}{n^2} \\
&\overset{\text{ch4.6}}{=}& \left( \frac{M}{m + M} \right) \frac{E_1}{n^2} \\
&=& \left( \frac{m}{2m} \right) \frac{E_1}{n^2} \\
&=& \frac{1}{2} \left( \frac{E_1}{n^2} \right)
\end{gather*}$$


* 正電子偶的雷德堡常數 $R'$ 為一般氫原子 $R$ 的一半
* 因為 $1/\lambda \propto R$，所以波長 $\lambda \propto 1/R$
* 正電子偶的所有光譜線波長，均為氫原子光譜線波長的**兩倍**。

### **例題 4.6.3：緲子原子 (Muonic Atom)**

**題目：** 緲子 (Muon, $\mu^-$) 是一種不穩定的基本粒子，質量約為電子質量 ($m_e$) 的 207 倍 ($m_{\mu^-} \approx 207 m_e$)，帶一單位的負電，一個負緲子可以被質子捕獲形成「緲子氫原子」，質子質量 $M \approx 1836m_e$ 。(a) 求出這個原子的第一軌道半徑（即修改後的波耳半徑）。(b) 求出這個原子的基態游離能。


(a) 先求減縮質量（）：

軌道半徑公式中，$r \propto 1/m$。因此新的基態半徑 $r'_1$ 為：

$$\begin{gather*}
r'_1 &\overset{\text{ch4.6}}{=}& \frac{m_{\mu^-}}{m'} \left(  r_{\mu^-1} \right)\\
&=& \frac{m_{\mu^-}}{m'} \left(  \frac{m_e}{m_{\mu^-}}  a_0 \right)\\
&=& \left( \frac{m_e}{m'} \right) a_0 \\
&=& \left( \frac{m_e}{\frac{mM}{m + M}} \right) a_0 \\
&=& \left( \frac{m_e}{\frac{(207m_e)(1836m_e)}{207m_e + 1836m_e}} \right) a_0 \\
&\overset{\text{ch4.4}}{=}& \left( \frac{1}{186} \right) (5.29 \times 10^{-11} \text{ m}) 
&\approx& 2.84 \times 10^{-13} \text{ m}
\end{gather*}$$

(b) 能階公式中，$E \propto m$。因此新的基態能量 $E'_1$ 為：

$$\begin{gather*}
E'_1  &\overset{\text{ch4.6}}{=}& \frac{m'}{m_{\mu^-}} \left(  \frac{m_{\mu^-}}{m_e}  E_1 \right)  \\
&=& \left( \frac{m'}{m_e} \right) E_1 \\
&\overset{\text{ch4.4}}{=}&  186 \times (-13.6 \text{ eV}) \\
&=& -2530 \text{ eV} &=& -2.53 \text{ keV}
\end{gather*}$$

* 游離能為其絕對值，即 $2.53\text{ keV}$。

**【物理知識回顧】**
因為緲子質量大，它繞行質子的軌道極端靠近原子核（半徑縮小近 200 倍），這使得緲子原子成為研究原子核內部電荷分佈的極佳探針。同時，其躍遷釋放的光子能量擴增了近 200 倍，發射的光譜不再是可見光，而是高能的 X 射線。

## 4.7 原子激發 (Atomic Excitation)

處於基態的原子是如何獲得能量躍遷到高能階的呢？主要有三種機制：
1. **熱碰撞 (Thermal Collisions)**：在高溫氣體中，原子間的劇烈動能轉換為內部激發能。
2. **電子撞擊 (Electron Impact)**：外來高能電子撞擊原子，將動能傳遞給原子的束縛電子。
3. **光子吸收 (Photon Absorption)**：原子吸收特定頻率的光子，光子能量 $h\nu$ 必須嚴格等於能階差。

原子受激發後並不穩定，通常會在極短時間內（約 $10^{-8}$ 秒）釋放光子並躍遷回低能階。

### [法蘭克-赫茲實驗 (Franck-Hertz Experiment)](https://zh.wikipedia.org/zh-tw/%E6%B3%95%E8%98%AD%E5%85%8B-%E8%B5%AB%E8%8C%B2%E5%AF%A6%E9%A9%97)

1914 年，法蘭克與赫茲透過電子撞擊水銀蒸氣的實驗，直接證實了原子能階的量子化。
* 他們發現，當加速電壓低於某個**臨界電位 (Critical Potential)**（如水銀的 4.9V）時，電子與原子發生彈性碰撞，不損失能量。
* 一旦達到臨界電位，電流突然驟降，代表電子將 $4.9\text{ eV}$ 的能量完全轉移給了水銀原子，使其激發，這剛好對應水銀光譜中 $254\text{ nm}$ 紫外光的躍遷能量！這證明了「吸收能量」與「放射能量」的量子化行為是一致的。

### 雷射 (LASER)

雷射全名為「輻射的受激發射光放大」(Light Amplification by Stimulated Emission of Radiation)，它具有以下獨特的光學特性：
* **單色性 (Monochromatic)**：幾乎只有單一波長。
* **同調性 (Coherence)**：所有光子的相位一致。
* **方向性高 (Low divergence)**：光束集中，不易色散擴散。
* **能量密度高 (High intensity)**：能在極小面積內聚集巨大能量。

#### 雷射的產生機制

要產生雷射，原子系統必須具備**亞穩態 (Metastable state)**，與一般受激態（壽命約 $10^{-8}\text{ s}$）不同，電子在亞穩態可以停留相對長的時間（例如 $10^{-3}\text{ s}$）。
這允許我們透過**光激昇 (Optical pumping)** 將大量電子「囤積」在這個高能階，形成**居量反轉 (Population inversion)**（高能階電子數多於低能階）。

當一顆能量為 $h\nu$ 的光子經過時，會誘發（受激發射）亞穩態的電子躍遷，並釋放出另一顆與入射光子相位、頻率、方向完全相同的新光子。一顆變兩顆、兩顆變四顆，形成雪崩式的連鎖反應，這就是雷射的放大原理。

常見的雷射系統有：
* **三能階雷射 (Three-Level Laser)**：如紅寶石雷射 (Ruby Laser)。
* **四能階雷射 (Four-Level Laser)**：如氦氖雷射 (He-Ne Laser)，四能階系統更容易達成居量反轉，因此能以連續波 (Continuous wave) 模式運作。
