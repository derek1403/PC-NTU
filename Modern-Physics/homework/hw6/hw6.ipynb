## Question 1 


Compare the angular momentum of a ground-state electron in the Bohr model of the hydrogen atom with its value in the quantum theory.

## Question 1 - Answer



### 假設與已知 (Assumptions & Preliminaries)


* **【已知 1】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：** 任何具有動量 $p$ 的粒子都伴隨一個波長為 $\lambda$ 的物質波。

  $$\lambda = \frac{h}{p}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


* **【已知 2】 量子化條件（駐波條件）(Standing-wave quantization condition)  ：** 電子的德布羅依波必須沿圓形軌道首尾相接形成駐波，軌道週長須為波長的整數倍。

  $$n\lambda = 2\pi r_{n}$$

  * $n$ : 主量子數 (Principal quantum number) $n = 1, 2, 3, \dots$ [無單位]
  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $r_n$ : 第 $n$ 能階的軌道半徑 (Orbital radius of the $n$-th level) $[\text{m}]$


* **【假設 1】 動量方向與半徑方向垂直 (Momentum perpendicular to radius)  ：** 電子沿圓形軌道運動，切線方向的動量與半徑向量恆正交，故角動量的外積可直接寫成純量乘積。

  $$\begin{gather*}
  \mathbf{L} &=& \mathbf{r} \times \mathbf{p} \\
  L &\overset{\text{垂直 } (\mathbf{r} \perp \mathbf{p})}{=}& rp \\
  \end{gather*}$$

  * $\mathbf{L}$ : 角動量向量 (Angular momentum vector) $[\text{kg}\cdot\text{m}^2\cdot\text{s}^{-1}]$ 或 $[\text{J}\cdot\text{s}]$
  * $\mathbf{r}$ : 位置向量 (Position vector) $[\text{m}]$
  * $\mathbf{p}$ : 動量向量 (Momentum vector) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $L, r, p$ : 對應上述向量的純量大小 (Magnitudes)


* **【已知 3】 量子力學的軌道角動量大小 (Magnitude of orbital angular momentum)  ：** 量子力學中軌道角動量的大小由角量子數 $l$ 決定。

  $$L = \sqrt{l(l+1)}\hbar$$

  * $L$ : 軌道角動量大小 (Magnitude of orbital angular momentum) $[\text{J}\cdot\text{s}]$
  * $l$ : 角量子數 (Azimuthal/Orbital quantum number) $l = 0, 1, 2, \dots, n-1$，基態的情況 $l=0$ [無單位]
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$


* **【推導 1】 波耳模型的角動量量子化 (Bohr angular momentum quantization)  ：** 把德布羅依波長代入駐波條件，先把波耳模型的角動量算成 $n\hbar$，讓後面的比較乾淨。

  $$\begin{gather*}
  n\lambda &\overset{\text{已知 2}}{=}& 2\pi r_{n} \\
  n\frac{h}{p_n} &\overset{\text{已知 1}}{=}& 2\pi r_{n} \\
  n \frac{h}{2\pi} &=&  r_{n} p_n \\
  n \hbar &\overset{\text{假設 1}}{=}& L_n \\
  \end{gather*}$$

  * $n$ : 主量子數 (Principal quantum number) $n = 1, 2, 3, \dots$ [無單位]
  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $r_n$ : 第 $n$ 能階的軌道半徑 (Orbital radius of the $n$-th level) $[\text{m}]$
  * $p_n$ : 第 $n$ 能階的電子動量 (Momentum at the $n$-th level) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $L_n$ : 第 $n$ 能階的角動量 (Angular momentum at the $n$-th level) $[\text{J}\cdot\text{s}]$




### compare 波耳模型與量子力學的氫原子基態電子角動量

1. 波耳模型

  $$\begin{gather*}
  L_n  &\overset{\text{推導 1}}{=}& n \hbar \\
  L_1 &\overset{\text{基態 } n=1}{=}& \hbar \\
  \end{gather*}$$

2. 量子力學

  $$\begin{gather*}
  L &\overset{\text{已知 3}}{=}& \sqrt{l(l+1)}\hbar \\
  \left. L \right|_{l=0}  &\overset{\text{基態 } l=0}{=}& 0 \\
  \end{gather*}$$


* **波耳模型的盲點（行星類比）**： 波耳假定電子必須要在一個二維平面的「圓周軌道」上跑動，如果電子沒有角動量（也就是不轉了）就會受到原子核的庫侖引力吸引，直接 **墜毀** 在原子核上，所以波耳模型中電子必須保持一個非零的最小角動量（$\hbar$）來維持軌道。
* **量子力學的真相（雲霧類比）**： 在量子力學中基態（$\text{1s}$ 軌域）的電子不再是小鋼珠，而是一團包圍著原子核的 **機率雲** ，這團 $\text{1s}$ 機率雲呈現完美對稱的球體、這團雲霧平均地分佈在原子核周圍，並沒有朝任何特定方向旋轉或流動的趨勢，既然沒有整體的旋轉趨勢，它的淨角動量自然就是 $0$ 了，這也證明了波耳模型將電子視為經典粒子的侷限性。

## Question 2 

* (a) What is Schrödinger’s equation for a particle of mass $m$ that is constrained to move in a circle of radius $R$, so that $\psi$ depends only on $\phi$? 
* (b) Solve this equation for $\psi$ and evaluate the normalization constant. (Hint: Review the solution of Schrödinger’s equation for the hydrogen atom.) 
* (c) Find the possible energies of the particle. 
* (d) Find the possible angular momenta of the particle.

## Question 2 - Answer


### 假設與已知 (Assumptions & Preliminaries)



* **【已知 1】[一維不含時薛丁格方程式 (1D Time-Independent Schrödinger Equation)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/05_Quantum_Mechanics/quantum_mechanics.html#schrodingers-equation-steady-state-form):**

  $$\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi = 0$$

  * $\psi(x)$ : 一維波函數的空間部分 (Spatial part of the wave function) $[\text{m}^{-1/2}]$
  * $m$ : 粒子的質量 (Mass of the particle) $[\text{kg}]$
  * $E$ : 系統的總能量常數 (Total energy) $[\text{J}]$
  * $U(x)$ : 位能的空間分布 (Potential energy) $[\text{J}]$
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34} \text{ J}\cdot \text{s} $


* **【已知 2】 座標轉換（從直線到圓周）(Arc-length coordinate transformation) :** 粒子被限制在半徑 $R$ 的圓環上，故一維直線座標可改以弧長 $s$ 描述，再換成角座標 $\theta$。

  $$\begin{gather*}
  s &=& R\theta \\
  ds &=& R \, d\theta \\
  \frac{d}{ds} &=& \frac{1}{R}\frac{d}{d\theta} \\
  \end{gather*}$$

  * $s$ : 圓弧長度 (Arc length) $[\text{m}]$
  * $R$ : 圓環半徑 (Radius of the circle) $[\text{m}]$
  * $\theta$ : 圓心角／角位置座標 (Angular coordinate) $[\text{rad}]$


* **【已知 3】 [歸一化條件 (Normalization)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/05_Quantum_Mechanics/quantum_mechanics.html#fundamental-rules-of-quantum-mechanics) :** 粒子必然存在於圓環上，機率密度繞一圈的定積分為 1。

  $$1 = \int_{0}^{2\pi} |\psi(\theta)|^2 \, d\theta$$

  * $1$ : 總機率 (Total probability)，代表粒子必然存在於系統中 [無單位]
  * $\int_{0}^{2\pi} \dots d\theta$ : 對整個圓環一圈進行定積分 (Definite integral over a full circle)
  * $|\psi(\theta)|^2$ : 機率密度 (Probability density) $[\text{rad}^{-1}]$
  * $\theta$ : 角位置座標 (Angular coordinate) $[\text{rad}]$


* **【假設 1】 自由粒子 (Free particle) :** 圓環上不受任何外力，位能 $U(\theta)$ 處處相等，取其零點為 $0$。

  $$U(\theta) \overset{\text{let}}{=} 0$$

  * $U$ : 位能 (Potential energy) $[\text{J}]$
  * $\theta$ : 角位置座標 (Angular coordinate) $[\text{rad}]$

* **【假設 2】圓環上週期性邊界條件 (Periodic boundary condition) :** 繞行一圈回到原處，波函數必須單值且連續。

  $$\psi(\theta) = \psi(\theta + 2\pi)$$

  * $\psi(\theta)$ : 在角度 $\theta$ 的波函數 (Wave function) $[\text{rad}^{-1/2}]$
  * $\theta$ : 角位置座標 (Angular coordinate) $[\text{rad}]$

* **【定義 1】分離常數 $k$ (Separation constant) :** 把方程中的常數係數整包命名，使方程化為簡諧振子形式。

  $$k^2 \overset{\text{def}}{=} \frac{2mR^2 E}{\hbar^2}$$

  * $k$ : 分離常數／無因次波數 (Separation constant / dimensionless wave number) [無單位]
  * $m$ : 粒子的質量 (Mass of the particle) $[\text{kg}]$
  * $E$ : 系統的總能量常數 (Total energy) $[\text{J}]$
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34} \text{ J}\cdot \text{s} $
  * $R$ : 圓環半徑 (Radius of the circle) $[\text{m}]$

* **【定義 2】角動量算符 $\hat{L}_z$ (Angular momentum operator) :** 繞著 $z$ 軸（垂直於圓環平面）旋轉的系統，其角動量算符定義為

  $$\hat{L}_z \overset{\text{def}}{=} -i\hbar \frac{d}{d\theta}$$

  * $\hat{L}_z$ : $z$ 方向的角動量算符 (z-component angular momentum operator) $[\text{J}\cdot\text{s}]$
  * $i$ : 虛數單位 (Imaginary unit) $i^2 = -1$ [無單位]
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $[\text{J}\cdot\text{s}]$
  * $\theta$ : 角位置座標 (Angular coordinate) $[\text{rad}]$

* **【推導 1】$k$ 的量子化 (Quantization of $k$) :** 對形如 $\psi(\theta) = C e^{ik\theta}$ 的解套用週期性邊界條件，$k$ 只能取整數。

  $$\begin{gather*}
  &\psi(\theta) &\overset{\text{假設 2}}{=}& \psi(\theta + 2\pi)  \\
  &C e^{ik\theta} &=& C e^{ik(\theta + 2\pi)}  \\
  &1 &=& e^{i 2\pi k} \\
  &1 &=& \cos(2\pi k) + i\sin(2\pi k) \\
  \Rightarrow &k &=& 0, \pm 1, \pm 2, \dots \\
  \end{gather*}$$

  * $\psi(\theta)$ : 在角度 $\theta$ 的波函數 (Wave function) $[\text{rad}^{-1/2}]$
  * $\theta$ : 角位置座標 (Angular coordinate) $[\text{rad}]$
  * $C$ : 未知的常數係數 (Constant coefficient) $[\text{rad}^{-1/2}]$
  * $k$ : 分離常數 (Separation constant)，$k \in \mathbb{Z}$ [無單位]


* **【推導 2】歸一化常數 $C$ (Normalization constant) :** 把【已知 3】的歸一化條件套到 $\psi(\theta) = C e^{ik\theta}$ 上，定出係數 $C$。

  $$\begin{gather*}
  1&\overset{\text{已知 3}}{=}&\int_{0}^{2\pi} |\psi(\theta)|^2 \, d\theta  \\
  1&=&\int_{0}^{2\pi} |C e^{ik\theta}|^2 \, d\theta  \\
  1&=&|C|^2 \int_{0}^{2\pi}  |e^{ik\theta}|^2\, d\theta \\
  1&=&|C|^2 \int_{0}^{2\pi}  \left(e^{ik\theta}  e^{-ik\theta} \right)\, d\theta \\
  1&=&|C|^2 \int_{0}^{2\pi}  1\, d\theta \\
  1&=& |C|^2 \times (2\pi) \\
  \frac{1}{\sqrt{2\pi}} &=& C \\
  \end{gather*}$$

  * $C$ : 歸一化常數 (Normalization constant)，確保總機率為 1 的縮放因子 $[\text{rad}^{-1/2}]$
  * $|\psi(\theta)|^2$ : 機率密度 (Probability density) $[\text{rad}^{-1}]$
  * $k$ : 分離常數 (Separation constant)，$k \in \mathbb{Z}$ [無單位]
  * $\theta$ : 角位置座標 (Angular coordinate) $[\text{rad}]$

### (a) proof 圓環上的薛丁格方程



$$\begin{gather*}
\frac{d^2 \psi}{dx^2} + \frac{2m}{\hbar^2}(E - U(x))\psi &\overset{\text{已知 1}}{=}& 0 \\
\frac{d^2 \psi}{ds^2} + \frac{2m}{\hbar^2}(E - U(s))\psi &\overset{\text{已知 2}}{=}& 0 \\
\frac{d^2 \psi}{ds^2} + \frac{2m}{\hbar^2} E \psi &\overset{\text{假設 1}}{=}& 0 \\
\frac{1}{R^2}\frac{d^2 \psi }{d\theta^2} + \frac{2m}{\hbar^2} E \psi &\overset{\text{已知 2}}{=}& 0 \\
\frac{d^2 \psi}{d\theta^2} + \frac{2mR^2}{\hbar^2}E\psi &=& 0
\end{gather*}$$

### (b) solve 波函數 $\psi$ 與歸一化常數

$$\begin{gather*}
&\frac{d^2 \psi}{d\theta^2} + \frac{2mR^2}{\hbar^2}E\psi &\overset{\text{Q2(a)}}{=}& 0 \\
&\frac{d^2 \psi}{d\theta^2} + k^2 \psi &\overset{\text{定義 1}}{=}& 0 \\
\overset{\text{解 ODE}}{\Rightarrow} &\psi(\theta) &=& C e^{ik\theta} \\
&\psi(\theta) &\overset{\text{推導 1,2}}{=}& \frac{1}{\sqrt{2\pi}} e^{ik\theta} &\quad (k = 0, \pm 1, \pm 2, \dots)\\
\end{gather*}$$

### (c) solve 可能的能量 $E$

$$\begin{gather*}
k^2 &\overset{\text{定義 1}}{=}&  \frac{2mR^2 E}{\hbar^2} \\
E &=& \frac{k^2 \hbar^2}{2mR^2} &\quad (k = 0, \pm 1, \pm 2, \dots)\\
\end{gather*}$$

* 因為 $k$ 是整數，所以能量 $E$ 不能是連續的任意值，只能是一階一階往上跳的 **階梯**

### (d) solve 可能的角動量 $L_z$

$$\begin{gather*}
&\hat{L}_z \psi &\overset{\text{定義 2}}{=}&  -i\hbar \frac{d}{d\theta}  \psi \\
&\hat{L}_z \psi &\overset{\text{Q2(b)}}{=}& -i\hbar \frac{d}{d\theta} \left[ \frac{1}{\sqrt{2\pi}} e^{ik\theta} \right]\\
&\hat{L}_z \psi &=&  -i\hbar (ik) \left( \frac{1}{\sqrt{2\pi}} e^{ik\theta} \right) \\
&\hat{L}_z \psi &=&  -i^2 k\hbar \psi \\
&\hat{L}_z \psi &=& k\hbar \psi \\
\Rightarrow&L_z  &=& k\hbar &\quad (k = 0, \pm 1, \pm 2, \dots)\\
\end{gather*}$$

* 因為 $k$ 是整數，所以 $z$ 方向的角動量 $L_z$ 不能是連續的任意值，只能是一階一階往上跳的 **階梯**

## Question 3

Find the percentage difference between $L$ and the maximum value of $L_z$ for an atomic electron in $p$ , $d$ , and $f$ states.

## Question 3 - Answer



### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 軌道角動量大小 (Magnitude of orbital angular momentum) ：**

  $$L = \sqrt{l(l+1)}\hbar$$

  * $L$ : 總軌道角動量大小 (Magnitude of total orbital angular momentum) $[\text{J}\cdot\text{s}]$
  * $l$ : 角量子數 (Azimuthal quantum number) $l = 0, 1, 2, \dots, n-1$ [無單位]
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $[\text{J}\cdot\text{s}]$


* **【已知 2】 角動量的 z 分量 (z-component of angular momentum) ：** 磁量子數 $m_l$ 的上限為 $l$，故 $L_z$ 的最大值為 $l\hbar$。

   $$\begin{gather*}
   L_z &=& m_l \hbar \\
   L_{z,\text{max}} &=& \text{max}(m_l) \, \hbar \\
   L_{z,\text{max}} &\overset{\text{取 } m_l = l}{=}& l\hbar \\
   \end{gather*}$$

   *  $L_z$ : 角動量在 z 軸的分量 (z-component of angular momentum) $[\text{J}\cdot\text{s}]$
   *  $L_{z,\text{max}}$ : $L_z$ 的最大值 (Maximum of $L_z$) $[\text{J}\cdot\text{s}]$
   *  $m_l$ : 磁量子數 (Magnetic quantum number) $m_l \in \{-l, -l+1, \dots, l-1, l \}$ [無單位]
   *  $l$ : 角量子數 (Azimuthal quantum number) [無單位]
   *  $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $[\text{J}\cdot\text{s}]$


* **【定義 1】百分比差異 (Percentage difference) ：**

  $$\text{Percentage Difference} \overset{\text{def}}{=} \frac{L - L_{z,\text{max}}}{L} \times 100\%$$

  * $\text{Percentage Difference}$ : $L$ 與 $L_{z,\text{max}}$ 的百分比差異 (Percentage difference) [無單位]
  * $L$ : 總軌道角動量大小 (Magnitude of total orbital angular momentum) $[\text{J}\cdot\text{s}]$
  * $L_{z,\text{max}}$ : $L_z$ 的最大值 (Maximum of $L_z$) $[\text{J}\cdot\text{s}]$


* **【推導 1】百分比差異的通式 (General form of the percentage difference) ：** 把【已知 1】【已知 2】代入【定義 1】，先把冗長的代數化簡算完，只留下角量子數 $l$ 的函數。

  $$\begin{gather*}
  \text{Percentage Difference} &\overset{\text{定義 1}}{=}& \frac{L - L_{z,\text{max}}}{L} \times 100\% \\
  &=& \left( 1 - \frac{L_{z,\text{max}}}{L} \right) \times 100\% \\
  &\overset{\text{已知 1,2}}{=}& \left( 1 - \frac{l\hbar}{\sqrt{l(l+1)}\hbar} \right) \times 100\% \\
  &=& \left( 1 - \frac{l}{\sqrt{l(l+1)}} \right) \times 100\% \\
  &=& \left( 1 - \sqrt{\frac{l^2}{l(l+1)}} \right) \times 100\% \\
  &=& \left( 1 - \sqrt{\frac{l}{l+1}} \right) \times 100\% \\
  \end{gather*}$$

  * $l$ : 角量子數 (Azimuthal quantum number) [無單位]


### compute

1. $\text{p}$ states ， $l = 1$

$$\begin{gather*}
\text{Percentage Difference}_{\text{p}} &\overset{\text{推導 1}}{=}& \left( 1 - \sqrt{\frac{l}{l+1}} \right) \times 100\%  \\
&\overset{\text{Q3.1}}{=}& \left( 1 - \sqrt{\frac{1}{1+1}} \right) \times 100\% \\
&\approx& (1 - 0.7071) \times 100\% \\
&\approx& 29.3\%
\end{gather*}$$


2. $\text{d}$ states ， $l = 2$

$$\begin{gather*}
\text{Percentage Difference}_{\text{d}} &\overset{\text{推導 1}}{=}& \left( 1 - \sqrt{\frac{l}{l+1}} \right) \times 100\%  \\
&\overset{\text{Q3.2}}{=}& \left( 1 - \sqrt{\frac{2}{2+1}} \right) \times 100\% \\
&\approx& (1 - 0.8165) \times 100\% \\
&\approx& 18.4\%
\end{gather*}$$


3. $\text{f}$ states ， $l = 3$

$$\begin{gather*}
\text{Percentage Difference}_{\text{f}} &\overset{\text{推導 1}}{=}& \left( 1 - \sqrt{\frac{l}{l+1}} \right) \times 100\%  \\
&\overset{\text{Q3.3}}{=}& \left( 1 - \sqrt{\frac{3}{3+1}} \right) \times 100\% \\
&\approx& (1 - 0.8660) \times 100\% \\
&\approx& 13.4\%
\end{gather*}$$


#### 物理觀察

**當角量子數 $l$ 越來越大時，百分比差異越來越小。**

物理學中 **對應原理 (Correspondence Principle)** 說 **當量子數極大時，量子系統的行為會越來越趨近於古典物理系統** ，當 $l$ 趨近於無限大時，角動量向量就會越來越貼近 z 軸，量子力學造成的「進動錐角」效應就越來越微乎其微了！

## Question 4


A hydrogen atom is in the $\text{4p}$ state ($n_i=4 , l_i=1$). To what state or states can it go by radiating a photon in an allowed transition?

## Question 4 - Answer



### 假設與已知 (Assumptions & Preliminaries)



* **【已知 1】 選擇定則 (Selection Rules)  ：** 電偶極輻射躍遷時，角量子數只能改變 $\pm 1$。

  $$\Delta l = \pm 1$$

  * $\Delta l$ : 角量子數的變化量 (Change in azimuthal quantum number) $\Delta l = l_f - l_i$ [無單位]


* **【已知 2】 主量子數與角量子數的約束 (Constraint between $n$ and $l$)  ：** 角量子數必須嚴格小於主量子數。

  $$l \le n - 1$$

  * $n$ : 主量子數 (Principal quantum number) $n = 1, 2, 3, \dots$ [無單位]
  * $l$ : 角量子數 (Azimuthal quantum number) [無單位]


* **【已知 3】 氫原子能階的簡併 (Degeneracy of hydrogen energy levels)  ：** 氫原子的能量只與主量子數 $n$ 有關而與 $l$ 無關，相同 $n$ 的軌域沒有能量差、無法輻射光子躍遷，故終態必須落在更低的 $n$。

  $$E_n = -\frac{13.6 \text{ eV}}{n^2}$$

  * $E_n$ : 第 $n$ 能階的能量 (Energy of the $n$-th level) $[\text{eV}]$
  * $n$ : 主量子數 (Principal quantum number) [無單位]




### solve 允許的終態 (allowed final states)

先由【已知 1】定出終態的角量子數 $l_f$，再用【已知 2】【已知 3】夾出可能的主量子數 $n_f$。

$$\begin{gather*}
l_f &\overset{\text{已知 1}}{=}& l_i \pm 1 \\
l_f &\overset{\text{題目 } l_i = 1}{=}& 0 \ \text{ 或 } \ 2 \\
n_f &\overset{\text{已知 3}}{\le}& n_i - 1 = 3 \\
\end{gather*}$$

* (a) $l_f = 0$（$\text{s}$ 軌域）

  $$\begin{gather*}
  n_f &\overset{\text{已知 2}}{\ge}& l_f + 1 = 1 \\
  n_f &\overset{\text{已知 3}}{\le}& 3 \\
  n_f &=& 1, 2, 3 \\
  \end{gather*}$$

  對應 $\text{1s}, \text{2s}, \text{3s}$

* (b) $l_f = 2$（$\text{d}$ 軌域）

  $$\begin{gather*}
  n_f &\overset{\text{已知 2}}{\ge}& l_f + 1 = 3 \\
  n_f &\overset{\text{已知 3}}{\le}& 3 \\
  n_f &=& 3 \\
  \end{gather*}$$

  對應 $\text{3d}$


所以可能的軌域是  $\text{1s}, \text{2s}, \text{3s}$ , $\text{3d}$

## Question 5


In the Bohr model of the hydrogen atom ($z_{\text{p}}$ = 1), what is the magnitude of the orbital magnetic moment of an electron in the nth energy level?

## Question 5 - Answer


### 假設與已知 (Assumptions & Preliminaries)


* **【定義 1】 古典電磁學的磁矩 (Classical magnetic dipole moment) :**

  $$\mu \overset{\text{def}}{=} I A$$

  * $\mu$ : 磁矩 (Magnetic dipole moment) $[\text{A}\cdot\text{m}^2] , [\text{J}\cdot\text{T}^{-1}]$
  * $I$ : 電流 (Electric current) $[\text{A}]$
  * $A$ : 線圈所圍繞的面積 (Area of the current loop) $[\text{m}^2]$


* **【假設 1】 等速圓周軌道 (Uniform circular orbit) :** 波耳模型中電子以固定速率 $v$ 在半徑 $r$ 的圓形軌道上繞行，繞行一圈的週期為

  $$T = \frac{2\pi r}{v}$$

  * $T$ : 電子繞行一圈的時間／週期 (Period) $[\text{s}]$
  * $r$ : 軌道半徑 (Orbital radius) $[\text{m}]$
  * $v$ : 電子的切線速率 (Tangential speed of electron) $[\text{m}\cdot\text{s}^{-1}]$


* **【已知 1】 電流就是單位時間內通過的電荷量 (Electric current) :**

  $$\begin{gather*}
  I &=& \frac{e}{T} \\
  &\overset{\text{假設 1}}{=}& \frac{ev}{2\pi r}
  \end{gather*}$$

  * $I$ : 電流 (Electric current) $[\text{A}]$
  * $e$ : 基本電荷 (Elementary charge) $e \approx 1.602 \times 10^{-19} \text{ C}$
  * $T$ : 電子繞行一圈的時間／週期 (Period) $[\text{s}]$
  * $v$ : 電子的切線速率 (Tangential speed of electron) $[\text{m}\cdot\text{s}^{-1}]$
  * $r$ : 軌道半徑 (Orbital radius) $[\text{m}]$

* **【已知 2】 圓形軌道面積 (Area of the circular orbit) :**

  $$A \overset{\text{假設 1}}{=} \pi r^2$$

  * $A$ : 圓形軌道的面積 (Area of the circular orbit) $[\text{m}^2]$
  * $r$ : 軌道半徑 (Orbital radius) $[\text{m}]$


* **【已知 3】角動量 (Angular momentum) :**

  * (a) 古典角動量（切線動量 $\times$ 半徑）

  $$L \overset{\text{假設 1}}{=} m_e v r$$

  * (b) 波耳模型的角動量量子化

  $$L \overset{\text{Q1 推導 1}}{=} n\hbar$$

  * $L$ : 角動量 (Angular momentum) $[\text{kg}\cdot\text{m}^2\cdot\text{s}^{-1}]$ 或 $[\text{J}\cdot\text{s}]$
  * $m_e$ : 電子的靜止質量 (Electron mass) $m_e \approx 9.109 \times 10^{-31} \text{ kg}$
  * $v$ : 電子的切線速率 (Tangential speed) $[\text{m}\cdot\text{s}^{-1}]$
  * $r$ : 軌道半徑 (Orbital radius) $[\text{m}]$
  * $n$ : 主量子數 (Principal quantum number) $n = 1, 2, 3, \dots$ [無單位]
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $[\text{J}\cdot\text{s}]$


* **【定義 2】波耳磁子 (Bohr magneton) $\mu_B$ :**

  $$\mu_B \overset{\text{def}}{=} \frac{e\hbar}{2m_e}$$

  * $\mu_B$ : 波耳磁子 (Bohr magneton) $\mu_B \approx 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1}$
  * $e$ : 基本電荷 (Elementary charge) $[\text{C}]$
  * $\hbar$ : 約化普朗克常數 (Reduced Planck constant) $[\text{J}\cdot\text{s}]$
  * $m_e$ : 電子質量 (Electron mass) $[\text{kg}]$


### proof 

$$\begin{gather*}
\mu &\overset{\text{定義 1}}{=}& IA \\
&\overset{\text{已知 1,2}}{=}& \left(\frac{ev}{2\pi r}\right) (\pi r^2) \\
&=& \frac{evr}{2} \\
&\overset{\text{已知 3(a),(b)}}{=}& \frac{e}{2} \left(\frac{n\hbar}{m_e}\right) \\
&=& n \left(\frac{e\hbar}{2m_e}\right) \\
&\overset{\text{定義 2}}{=}& n \mu_B
\end{gather*}$$
