## Question 1 : Density of standing waves in cavity

Consider the radiation inside a cavity of absolute temperature $T$ whose walls are perfect reflectors to be a series of standing em waves(電磁波). This is a three-dimensional generalization of standing waves in a stretched string. The condition for standing waves in such a cavity is that the path length from wall to wall, whatever the direction, must be a whole number of half-wavelengths, so that a node occurs at each reflecting surface. Proof that the number of independent standing waves $G(\nu)d\nu$ in the frequency interval between $\nu$ and $\nu+d\nu$ per unit volume in the cavity is
   

$$G(\nu)d\nu = \frac{8\pi\nu^2}{c^3} d\nu $$

## Question 1 - Answer

邏輯主軸:
邊界條件限制了波長 $\to$ 波長離散化形成網格點 $\to$ 計算球體體積求總數 $\to$ 轉成頻率求微分 $\to$ 乘以2（偏振）得到密度。


$$G(\nu)d\nu = \frac{8\pi\nu^2}{c^3} d\nu$$

* $\nu$ : 頻率 (frequency) $[\text{s}^{-1}]=[\text{Hz}]$
* $d\nu$ : 頻率區間 (frequency interval)，從 $\nu$ 到 $\nu+d\nu$ 的極小範圍 $[\text{s}^{-1}]$
* $c$ : 光速 (speed of light)，$c \approx 2.99 \times 10^8 \text{ m}\cdot\text{s}^{-1}$ $[\text{m}\cdot\text{s}^{-1}]$
* $G(\nu)$ : 駐波密度／狀態密度 (density of states) $[\text{m}^{-3}\cdot\text{s}]$
* $G(\nu)d\nu$ : 總駐波數密度 (number of standing waves per unit volume)，頻率介於 $\nu$ 與 $\nu + d\nu$ 之間單位體積內的駐波模式數 $[\text{m}^{-3}]$

---

### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 三維立方體空腔與駐波條件 (3D cubic cavity & standing-wave condition)：** 邊界節點條件使各軸波數離散化。

  $$k_x = n_x \frac{\pi}{L}, \quad k_y = n_y \frac{\pi}{L}, \quad k_z = n_z \frac{\pi}{L}$$

  * $L$ : 空腔邊長 (side length of cavity) $[\text{m}]$
  * $V$ : 空腔體積 (volume)，$V = L^3$ $[\text{m}^3]$
  * $k_x, k_y, k_z$ : 波向量在各軸的分量 (wave-vector components) $[\text{m}^{-1}]$
  * $n_x, n_y, n_z$ : 駐波模式量子數，正整數 $n_x, n_y, n_z \in \mathbb{N}$ (mode quantum numbers) $[\text{無單位}]$

* **【已知 2】 波向量長度平方 (squared magnitude of wave vector)：**

  $$\begin{gather*} 
  k^2 &=& k_x^2 + k_y^2 + k_z^2 \\
  &\overset{\text{已知 1}}{=}& \left(\frac{\pi}{L}\right)^2 (n_x^2 + n_y^2 + n_z^2)
  \end{gather*}$$

  * $k$ : 波向量大小 (magnitude of wave vector) $[\text{m}^{-1}]$

* **【定義 1】 模式空間半徑 (radius in mode space)：** 在 $(n_x, n_y, n_z)$ 三維網格中，每個座標點代表一個合法駐波模式。

  $$R^2 \overset{\text{def}}{=} n_x^2 + n_y^2 + n_z^2$$

  * $R$ : 模式空間中的半徑 (radius in $n$-space) $[\text{無單位}]$

* **【已知 3】 波數與頻率關係 (wavenumber–frequency relation)：**

  $$k = \frac{\omega}{c} = \frac{2\pi\nu}{c}$$

  * $\omega$ : 角頻率 (angular frequency)，$\omega = ck = 2\pi\nu$ $[\text{s}^{-1}]$

* **【已知 4】 半徑 $R$ 球體內的整數格點數 (lattice-point count ≈ sphere volume)：** 大半徑下，球內格點數可用球體體積近似。

  $$\text{半徑 } R \text{ 內的整數格點數} \approx \frac{4}{3}\pi R^3$$

  * 估計：$R = \frac{2L\nu}{c} \sim 3.33 \times 10^5$（取 $L = 0.1 \text{ m}$、$\nu = 5 \times 10^{14} \text{ Hz}$、$c = 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$）
  * 表面誤差占比 $\propto \dfrac{R^2}{R^3} = \dfrac{1}{R} \approx \dfrac{1}{333{,}333} \approx 0.0003\%$，故球體體積近似成立

* **【定義 2】 半徑 $R$ 內的合法模式總數 $N(R)$ (total number of allowed modes)：**

  $$\begin{gather*}
  N(R) &\overset{\text{def}}{=}& \text{半徑 } R \text{ 內的合法整數格點數} \\
  &=& \left(\text{合法格點比例}\right) \times \left(\text{半徑 } R \text{ 內的整數格點數}\right) \\
  &\overset{\text{已知 4}}{\approx}& \left(\text{合法格點比例}\right) \times \frac{4}{3}\pi R^3 \\
  &\overset{\text{已知 1}}{=}& \frac{1}{8} \times \frac{4}{3}\pi R^3 \\
  &=& \frac{\pi}{6} R^3
  \end{gather*}$$

  * $N(R)$ : 半徑 $R$ 以內的合法駐波模式總數 (number of modes within radius $R$) $[\text{無單位}]$
  * 合法格點比例 $= \frac{1}{8}$：因 $n_x, n_y, n_z > 0$（已知 1），只取第一卦限。

* **【推導 1】 模式空間半徑對頻率的轉換 (mode-space radius as a function of frequency)：** 結合定義與已知，把 $R$ 寫成 $\nu$ 的函數。

  $$\begin{gather*}
  R &\overset{\text{定義 1}}{=}& \sqrt{n_x^2 + n_y^2 + n_z^2} \\
  &\overset{\text{已知 2}}{=}& \sqrt{\left(\frac{L}{\pi}\right)^2 k^2} \\
  &=& \frac{L}{\pi} k \\
  &\overset{\text{已知 3}}{=}& \frac{L}{\pi} \left( \frac{2\pi\nu}{c} \right) \\
  &=& \frac{2L\nu}{c}
  \end{gather*}$$

* **【已知 5】 電磁波的極化 (polarization of EM waves)：** 給定一個波向量 $\mathbf{k}$，橫波有兩種互相垂直的獨立偏振，故模式數須乘以 2。

  $$dN_{\text{total}} = 2\, dN$$

  * $dN$ : 單一偏振下、頻率區間 $d\nu$ 內的模式數 (mode count for one polarization) $[\text{無單位}]$
  * $dN_{\text{total}}$ : 計入兩種獨立偏振後的總模式數 (total mode count) $[\text{無單位}]$

* **【定義 3】 單位體積內的駐波密度 $G(\nu)$ (density of standing waves per unit volume)：**

  $$G(\nu)\,d\nu \overset{\text{def}}{=} \frac{dN_{\text{total}}}{V}$$

  * $G(\nu)$ : 單位體積、單位頻率的駐波模式數 (density of states) $[\text{m}^{-3}\cdot\text{s}]$



### proof

$$\begin{gather*}
N &\overset{\text{定義 2}}{=}& \frac{\pi}{6} R^3 \\
N &\overset{\text{推導 1}}{=}& \frac{\pi}{6} \left( \frac{2L\nu}{c} \right)^3 \\
N &=& \frac{\pi}{6} \frac{8 L^3 \nu^3}{c^3} \\
N &=& \frac{4\pi L^3}{3 c^3} \nu^3 \\
N &\overset{\text{已知 1}}{=}& \frac{4\pi V}{3 c^3} \nu^3 \\
\frac{dN}{d\nu} &=& \frac{4\pi V}{c^3} \nu^2 \\
dN &=& \frac{4\pi V}{c^3} \nu^2\, d\nu \\
dN_{\text{total}} &\overset{\text{已知 5}}{=}& 2 \cdot \frac{4\pi V}{c^3} \nu^2\, d\nu \\
\frac{dN_{\text{total}}}{V} &=& \frac{8\pi}{c^3} \nu^2\, d\nu \\
G(\nu)\, d\nu &\overset{\text{定義 3}}{=}& \frac{8\pi}{c^3} \nu^2\, d\nu
\end{gather*}$$

## Question 2 Average energy per standing wave

Suppose the oscillators in the cavity walls do not have a continuous distribution of possible energies but have only the specific energies

$$\varepsilon_{n}=nh\nu$$

Proof that the average energy per oscillator in the cavity walls—and so per standing wave—is

$$\bar{\varepsilon} = \frac{h\nu}{e^{h\nu/kT}-1}$$

(Hint: Maxwell-Boltzmann distribution function $f_{MB} (\varepsilon)=Ae^{-\varepsilon/kT} $ )

## Question 2 - Answer


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 無窮等比級數求和 (sum of infinite geometric series)：**

   $$\begin{gather*}
   \sum_{n=0}^{\infty} x^n &=& 1 + x + x^2 + \dots \\
   &=& \frac{1}{1-x}
   \end{gather*}   $$

   * $x$ : 公比／啞變數 (common ratio / dummy variable)，需 $|x| < 1$；本題 $x = e^{-h\nu/kT}$ 符合條件 $[\text{無單位}]$
   * $n$ : 求和指標 (summation index) $[\text{無單位}]$

* **【已知 2】 加權幾何級數 (weighted geometric series)：** 對等比級數兩邊作用 $x\frac{d}{dx}$。

   $$\begin{gather*}
   \sum_{n=0}^{\infty} x^n &\overset{\text{已知 1}}{=}& \frac{1}{1-x} \\
   x \frac{d}{dx} \left[ \sum_{n=0}^{\infty} x^n \right] &=& x \frac{d}{dx} \left[ \frac{1}{1-x} \right] \\
   \sum_{n=0}^{\infty} n x^n &=& \frac{x}{(1-x)^2}
   \end{gather*}$$

* **【已知 3】 能量量子化 (energy quantization)：**

   $$\varepsilon_{n}=nh\nu$$

   * $\varepsilon_n$ : 第 $n$ 個能階的能量 (energy of $n$-th level) $[\text{J}]$
   * $h$ : 普朗克常數 (Planck constant)，$h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s}$ $[\text{J}\cdot\text{s}]$
   * $\nu$ : 振子頻率 (oscillator frequency) $[\text{s}^{-1}]$
   * $n$ : 量子數，非負整數 (quantum number) $[\text{無單位}]$

* **【已知 4】 馬克士威–波茲曼分佈 (Maxwell–Boltzmann distribution)：**

   $$f_{\text{MB}}(\varepsilon)=Ae^{-\varepsilon/kT}$$

   * $f_{\text{MB}}$ : 能量為 $\varepsilon$ 時的相對機率因子 (Boltzmann factor) $[\text{無單位}]$
   * $A$ : 歸一化常數 (normalization constant) $[\text{無單位}]$
   * $k$ : 波茲曼常數 (Boltzmann constant)，$k \approx 1.38 \times 10^{-23} \text{ J}\cdot\text{K}^{-1}$ $[\text{J}\cdot\text{K}^{-1}]$
   * $T$ : 絕對溫度 (absolute temperature) $[\text{K}]$

* **【定義 1】 啞變數 (dummy variable)：** 為化簡級數而引入。

   $$x \overset{\text{def}}{=} e^{-h\nu/kT}$$

   * $x$ : 啞變數 (dummy variable) $[\text{無單位}]$


### proof

$$\begin{gather*}
\bar{\varepsilon} &=& \frac{\sum_{n=0}^{\infty} \varepsilon_n f_{\text{MB}}(\varepsilon_n)}{\sum_{n=0}^{\infty} f_{\text{MB}}(\varepsilon_n)}\\
&\overset{\text{已知 3,4}}{=}&\frac{\sum_{n=0}^{\infty} (nh\nu) A e^{-nh\nu/kT}}{\sum_{n=0}^{\infty} A e^{-nh\nu/kT}}\\
&\overset{\text{定義 1}}{=}& \frac{h\nu \sum_{n=0}^{\infty} n x^n}{\sum_{n=0}^{\infty} x^n}\\
&\overset{\text{已知 2}}{=}& \frac{h\nu \frac{x}{(1-x)^2}}{\frac{1}{1-x}}\\
&=& h\nu \frac{x}{1-x}\\
&\overset{\text{定義 1}}{=}&h\nu \frac{e^{-h\nu/kT}}{1 - e^{-h\nu/kT}}\\
&=& \frac{h\nu \cdot e^{-h\nu/kT} \cdot e^{h\nu/kT}}{(1 - e^{-h\nu/kT}) \cdot e^{h\nu/kT}}\\
&=& \frac{h\nu}{e^{h\nu/kT} - 1}
\end{gather*}$$

## Question 3

Express the Planck radiation formula in terms of wavelength.

## Question 3 - Answer


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 頻率與波長關係 (frequency–wavelength relation)：** 微分後能量恆正故取絕對值，$d\nu = -\frac{c}{\lambda^2} d\lambda$。

   $$\nu = \frac{c}{\lambda}, \qquad |d\nu| = \frac{c}{\lambda^2} |d\lambda|$$

   * $\lambda$ : 波長 (wavelength) $[\text{m}]$
   * $\nu$ : 頻率 (frequency) $[\text{s}^{-1}]$
   * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$

* **【已知 2】 能量密度的變數變換 (Jacobian relation for spectral density)：** 同一段頻譜的能量不因變數選擇而改變。

   $$|u(\nu)\,d\nu| = |u(\lambda)\,d\lambda|$$

   * $u(\nu)$ : 以頻率表示的能量密度 (spectral energy density per frequency) $[\text{J}\cdot\text{m}^{-3}\cdot\text{s}]$
   * $u(\lambda)$ : 以波長表示的能量密度 (spectral energy density per wavelength) $[\text{J}\cdot\text{m}^{-4}]$

* **【已知 3】 普朗克輻射能量密度 (Planck spectral energy density)：** 能量密度 = 單位體積駐波數密度 $\times$ 每模式平均能量。

   $$u(\nu)\,d\nu = G(\nu)\,\bar{\varepsilon}\,d\nu \overset{\text{Q1, Q2}}{=} \left( \frac{8\pi \nu^2}{c^3} \right) \left( \frac{h\nu}{e^{h\nu/kT} - 1} \right) d\nu$$

   * $G(\nu)$ : 駐波密度，來自 Q1 (density of states) $[\text{m}^{-3}\cdot\text{s}]$
   * $\bar{\varepsilon}$ : 每駐波平均能量，來自 Q2 (mean energy per mode) $[\text{J}]$


### proof

$$\begin{gather*}
u(\nu)\,d\nu &\overset{\text{已知 3}}{=}& \frac{8\pi h \nu^3}{c^3 (e^{h\nu/kT} - 1)} d\nu \\
u(\lambda)\,d\lambda &\overset{\text{已知 1,2}}{=}& \frac{8\pi h \left( \frac{c}{\lambda} \right)^3}{c^3 \left( e^{\frac{h(c/\lambda)}{kT}} - 1 \right)} \left( \frac{c}{\lambda^2} d\lambda \right)\\
&=& \frac{8\pi hc}{\lambda^5} \frac{1}{e^{hc/\lambda kT} - 1} d\lambda
\end{gather*}$$

## Question 4

A photon of frequency 𝜈 is scattered by an electron initially at rest. Verify that the maximum kinetic energy of the recoil electron is

$$KE_{max}=\frac{2h^{2}\nu^{2}/mc^{2}}{1+2h\nu/mc^{2}}$$

## Question 4 - Answer


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 能量守恆 (Conservation of Energy)：** 光子–電子散射前後總能量守恆。

   $$E_{\text{光子,初}} + E_{\text{電子,初}} = E_{\text{光子,末}} + E_{\text{電子,末}}$$

   * $E_{\text{光子,初}}, E_{\text{光子,末}}$ : 散射前／後光子能量 (photon energy before/after) $[\text{J}]$
   * $E_{\text{電子,初}}, E_{\text{電子,末}}$ : 散射前／後電子能量 (electron energy before/after) $[\text{J}]$

* **【已知 2】 [康普頓波長位移公式 (Compton shift formula)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#proof)：** 由能量守恆與動量守恆導出。

   $$\Delta \lambda = \lambda' - \lambda = \frac{h}{mc}(1 - \cos\theta)$$

   * $\lambda, \lambda'$ : 散射前／後光子波長 (photon wavelength before/after) $[\text{m}]$
   * $h$ : 普朗克常數 (Planck constant) $[\text{J}\cdot\text{s}]$
   * $m$ : 電子靜止質量 (electron rest mass)，$m \approx 9.11 \times 10^{-31} \text{ kg}$ $[\text{kg}]$
   * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$
   * $\theta$ : 散射角 (scattering angle) $[\text{rad}]$

* **【已知 3】 頻率與波長關係 (frequency–wavelength relation)：**

   $$\lambda = \frac{c}{\nu}$$

   * $\lambda$ : 光子波長 (photon wavelength) $[\text{m}]$
   * $\nu$ : 光子頻率 (photon frequency) $[\text{s}^{-1}]$
   * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$

* **【推導 1】 散射後頻率 $\nu'$ (scattered frequency at $\theta=180^\circ$)：** 電子獲最大動能對應 $\theta = 180^\circ$，代入康普頓公式解出 $\nu'$。

    $$\begin{gather*}
    \lambda' - \lambda &\overset{\text{已知 2}}{=}& \frac{h}{mc}(1 - \cos 180^\circ) \\
    \lambda' - \lambda &=& \frac{2h}{mc}\\
    \frac{c}{\nu'} - \frac{c}{\nu} &\overset{\text{已知 3}}{=}& \frac{2h}{mc}\\
    \frac{1}{\nu'} - \frac{1}{\nu} &=& \frac{2h}{mc^2}\\
    \frac{1}{\nu'} &=& \frac{1}{\nu} + \frac{2h}{mc^2}  \\
    \frac{1}{\nu'} &=& \frac{mc^2 + 2h\nu}{\nu mc^2}\\
    \nu' &=& \frac{\nu mc^2}{mc^2 + 2h\nu}
    \end{gather*}$$

   * $\nu, \nu'$ : 散射前／後光子頻率 (photon frequency before/after) $[\text{s}^{-1}]$


### proof

$$\begin{gather*}
h\nu + mc^2 &\overset{\text{已知 1}}{=}& h\nu' + (KE_{\text{max}} + mc^2)\\
KE_{\text{max}} &=& h\nu - h\nu' \\
&\overset{\text{推導 1}}{=}& h\nu - h\left( \frac{\nu mc^2}{mc^2 + 2h\nu} \right) \\
&=& h\nu \left( 1 - \frac{mc^2}{mc^2 + 2h\nu} \right)\\
&=& h\nu \left( \frac{mc^2 + 2h\nu - mc^2}{mc^2 + 2h\nu} \right) \\
&=& h\nu \left( \frac{2h\nu}{mc^2 + 2h\nu} \right) \\
&=& \frac{2h^2\nu^2}{mc^2 + 2h\nu}\\
&=& \frac{2h^2\nu^2 / mc^2}{1 + 2h\nu / mc^2}
\end{gather*}$$

## Question 5

The gravitational potential energy $U$ relative to infinity of a body of mass m at a distance $R$ from the center of a body of mass $M$ is $U = -GmM/R$. 

(a) If $R$ is the radius of the body of mass $M$ , find the escape speed $v_e$ of the body, which is the minimum speed needed to leave it permanently. 

(b) Obtain a formula for the Schwarzschild radius of the body by setting $v_e=c$, the speed of light, and solving for $R$. (Of course, a relativistic calculation is correct here, but it is interesting to see what a classical calculation produces.)

## Question 5 - Answer


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 力學能守恆定律 (Conservation of Mechanical Energy)：** 逃逸過程僅重力作功，力學能守恆。

   $$E_{\text{total}} = K + U = \text{常數}$$

   * $E_{\text{total}}$ : 總力學能 (total mechanical energy) $[\text{J}]$
   * $K$ : 動能 (kinetic energy) $[\text{J}]$
   * $U$ : 重力位能 (gravitational potential energy)，$U = -\frac{GmM}{R}$ $[\text{J}]$
   * $G$ : 重力常數 (gravitational constant)，$G \approx 6.67 \times 10^{-11} \text{ m}^3\cdot\text{kg}^{-1}\cdot\text{s}^{-2}$ $[\text{m}^3\cdot\text{kg}^{-1}\cdot\text{s}^{-2}]$
   * $m$ : 逃逸物體質量 (mass of escaping body) $[\text{kg}]$
   * $M$ : 中心天體質量 (mass of central body) $[\text{kg}]$
   * $R$ : 中心天體半徑／初始距離 (radius of central body) $[\text{m}]$
   * $v_e$ : 逃逸速度 (escape speed) $[\text{m}\cdot\text{s}^{-1}]$
   * 初態：$E_{\text{i}} = \frac{1}{2}mv_e^2 - \frac{GmM}{R}$（$K_{\text{i}} = \frac{1}{2}mv_e^2$、$U_{\text{i}} = -\frac{GmM}{R}$）
   * 末態（恰抵無窮遠且停下）：$E_{\text{f}} = 0$（$K_{\text{f}} = 0$、$U_{\text{f}} = -\frac{GmM}{\infty} = 0$）


### proof (a) 逃逸速度 $v_e$ (escape speed)

$$\begin{gather*}
\frac{1}{2}mv_e^2 - \frac{GmM}{R} &\overset{\text{已知 1}}{=}& 0 \\
\frac{1}{2}mv_e^2 &=& \frac{GmM}{R}\\
v_e^2 &=& \frac{2GM}{R}\\
v_e &=& \sqrt{\frac{2GM}{R}}
\end{gather*}$$


### proof (b) 史瓦西半徑 $R_S$ (Schwarzschild radius)

$$\begin{gather*}
v_e &\overset{\text{Q5(a)}}{=}& \sqrt{\frac{2GM}{R}} \\
c &\overset{\text{Q5}}{=}& \sqrt{\frac{2GM}{R_S}}\\
R_S &=& \frac{2GM}{c^2}
\end{gather*}$$
