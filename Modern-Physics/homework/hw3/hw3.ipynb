## Question 1 : Density of standing waves in cavity

Show that the de Broglie wavelength of a particle of mass $m$ and kinetic energy $KE$ is given by
   
$$\lambda = \frac{hc}{\sqrt{KE(KE + 2mc^2)}}$$




## Question 1 - Answer


* **гҖҗе·ІзҹҘ 2гҖ‘ [иҖғж…®зӣёе°Қи«–ж•ҲжҮүзҡ„еӢ•йҮҸ](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)пјҡ** 

  $$p = \gamma m_0 v$$

  * $p$ : зү©й«”еӢ•йҮҸ пјҢ$[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : еӢһдҫ–иҢІеӣ еӯҗ (Lorentz factor)
  * $m_0$ : зү©й«”йқңжӯўиіӘйҮҸ (Rest mass)пјҢ$[\text{kg}]$
  * $v$ : зү©й«”йҖҹеәҰ (Velocity)пјҢ$[\text{m}\cdot\text{s}^{-1}]$


### еҒҮиЁӯиҲҮе·ІзҹҘ (Assumptions & Preliminaries)

* **гҖҗе·ІзҹҘ 1гҖ‘ [зү©й«”зҡ„иғҪйҮҸиҲҮеӢ•йҮҸй—ңдҝӮејҸ](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#energy-and-momentum)пјҡ** 

  $$E^2 = (m_0 c^2)^2 + (pc)^2 $$

  * $E$ : зёҪиғҪйҮҸ (Total Energy) $[\text{J}]$
  * $m_0$ : зү©й«”йқңжӯўиіӘйҮҸ (Rest mass)пјҢ$[\text{kg}]$
  * $c$ : зңҹз©әдёӯе…үйҖҹ (Speed of light)пјҢ$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$
  * $p$ : зү©й«”еӢ•йҮҸ пјҢ$[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$



* **гҖҗе®ҡзҫ© 1гҖ‘ еҫ·еёғзҫ…дҫқжіў (De Broglie Wave)  пјҡ**
  
  $$\lambda \overset{\text{def}}{=} \frac{h}{p}$$

  * $\lambda$ : еҫ·еёғзҫ…дҫқжіўй•· (de Broglie wavelength) $[\text{m}]$
  * $h$ : жҷ®жң—е…Ӣеёёж•ё (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : зү©й«”еӢ•йҮҸ (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


$$\begin{gather*}
E^2 &\overset{\text{е·ІзҹҘ 1}}{=}& (pc)^2 + (m_0 c^2)^2 \\
E^2 - (m_0 c^2)^2 &=& (pc)^2 \\
\frac{1}{c}\sqrt{E^2 - (m_0 c^2)^2} &=& p \\
\frac{1}{c}\sqrt{(KE+E_0)^2 - (m_0 c^2)^2} &=& p \\
\frac{1}{c}\sqrt{{KE}^2+2KE \cdot E_0} &=& p \\
\frac{1}{c}\sqrt{KE(KE + 2m_0c^2)} &=& p \\
\frac{1}{c}\sqrt{KE(KE + 2m_0c^2)} &\overset{\text{е®ҡзҫ© 1}}{=}& \frac{h}{\lambda} \\
\lambda &=&\frac{hc}{\sqrt{KE(KE + 2m_0c^2)}}  \\
\end{gather*}$$

  $$m_0 = 0$$

   E^2 &\overset{\text{е·ІзҹҘ 1}}{=}& (pc)^2 + (m_0 c^2)^2 \\

   \end{gather*}$$






### proof



$$\begin{gather*}
\lambda_{\text{particle}} &\overset{\text{Q1}}{=}&\frac{hc}{\sqrt{KE(KE + 2mc^2)}} \\
\lambda_{\text{particle}} &=&\frac{hc}{\sqrt{KE(KE + 2E_0)}} \\

\lambda_{\text{particle}} &=&\frac{hc}{KE} \\
\lambda_{\text{particle}} &\overset{\text{еҒҮиЁӯ 1}}{=}&\frac{hc}{E} &\overset{\text{е·ІзҹҘ 3}}{=}& \lambda_{\text{photon}}\\
\end{gather*}$$

$E \gg mc^2$ зҡ„жғ…жіҒдёӢпјҢ$\lambda_{\text{particle}} \approx  \lambda_{\text{photon}}$

### е·ІзҹҘиҲҮеҒҮиЁӯ

* **гҖҗеҒҮиЁӯ 1гҖ‘еӢ•иғҪзӣёеҗҢ:**
  
   $$KE_{\text{electron}} = KE_{\text{proton}}$$

* **гҖҗе·ІзҹҘ 1гҖ‘еҸӨе…ёеӢ•иғҪ:**
  
   $$KE = \frac{1}{2}mv^2$$

   * еңЁ **дёҚеӨұдёҖиҲ¬жҖ§** зҡ„жғ…жіҒдёӢпјҢдҫқйЎҢзӣ®жҲ‘еҖ‘еҸҜд»ҘзӣҙжҺҘиҖғж…® **еҸӨе…ёеӢ•иғҪ**

  $$\lambda \overset{\text{def}}{=} \frac{h}{p}$$

  * $\lambda$ : еҫ·еёғзҫ…дҫқжіўй•· (de Broglie wavelength) $[\text{m}]$
  * $h$ : жҷ®жң—е…Ӣеёёж•ё (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : зү©й«”еӢ•йҮҸ (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$

* **гҖҗе·ІзҹҘ 2гҖ‘иіӘеӯҗиіӘйҮҸйҒ еӨ§ж–јйӣ»еӯҗиіӘйҮҸ:**
   
   $$m_e \ll m_p$$

   * йӣ»еӯҗиіӘйҮҸ : $m_e \sim 9.11 \times 10^{-31} \text{kg}$
   * иіӘеӯҗиіӘйҮҸ : $m_p \sim 1.67 \times 10^{-27} \text{kg}$

* **гҖҗе·ІзҹҘ 3гҖ‘зҫӨйҖҹеәҰжҳҜзү©й«”йҖҹеәҰ:**
   
   $$v_{\text{g}}=v$$

* **гҖҗе·ІзҹҘ 4гҖ‘зҫӨйҖҹеәҰд№ҳзӣёйҖҹеәҰжҳҜе…үйҖҹе№іж–№:**
   
   $$v_{\text{g}} v_{\text{p}}=c^2$$

   * $v_{\text{g}}$ жҳҜзҫӨйҖҹеәҰ
   * $v_{\text{p}}$ жҳҜзӣёйҖҹеәҰ

   
### 1. еҫ·еёғзҫ…дҫқжіўй•· ($\lambda$)

$$\begin{gather*}
KE_{\text{electron}} &\overset{\text{еҒҮиЁӯ 1}}{=}& KE_{\text{proton}} \\
m_{\text{electron}} KE_{\text{electron}} &\overset{\text{е·ІзҹҘ 2}}{\ll}& m_{\text{proton}}  KE_{\text{proton}} \\
\sqrt{2m_{\text{electron}} KE_{\text{electron}}} &\ll& \sqrt{2m_{\text{proton}}  KE_{\text{proton}}} \\
p_{\text{electron}}  &\ll& p_{\text{proton}} \\
\frac{1}{p_{\text{electron}}} &\gg& \frac{1}{p_{\text{proton}}} \\
\frac{h}{p_{\text{electron}}} &\gg& \frac{h}{p_{\text{proton}}} \\
\lambda_{\text{electron}} &\overset{\text{е®ҡзҫ© 1}}{\gg}& \lambda_{\text{proton}} \\
\end{gather*}$$


### 2. зҫӨйҖҹеәҰ ($v_{\text{g}}$)

$$\begin{gather*}
KE_{\text{electron}} &\overset{\text{еҒҮиЁӯ 1}}{=}& KE_{\text{proton}} \\
\frac{1}{2}m_{\text{electron}}v_{\text{electron}}^2 &=& \frac{1}{2}m_{\text{proton}}v_{\text{proton}}^2 \\
\frac{m_{\text{electron}}}{m_{\text{proton}}}v_{\text{electron}}^2 &=& v_{\text{proton}}^2 \\
v_{\text{electron}} &\overset{\text{е·ІзҹҘ 2}}{\gg}& v_{\text{proton}}\\
v_{\text{g,electron}} &\overset{\text{е·ІзҹҘ 3}}{\gg}& v_{\text{g,proton}}\\
\end{gather*}$$

### 3. зӣёйҖҹеәҰ ($v_{\text{p}}$)


$$\begin{gather*}
v_{\text{g,electron}} v_{\text{p,electron}} &\overset{\text{е·ІзҹҘ 4}}{=}& v_{\text{g,proton}} v_{\text{p,proton}}\\

## Question 4 - Answer



### е·ІзҹҘиҲҮеҒҮиЁӯ

* гҖҗе·ІзҹҘ 1гҖ‘[еә·жҷ®й “жіўй•·дҪҚз§»е…¬ејҸпјҲCompton shift formulaпјү](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#proof):
  
   $$\Delta \lambda = \lambda' - \lambda = \frac{h}{mc}(1 - \cos\theta)$$

   * иғҪйҮҸе®ҲжҒҶгҖҒеӢ•йҮҸе®ҲжҒҶ
   * и®“йӣ»еӯҗзҚІеҫ—жңҖеӨ§еӢ•иғҪ $\theta = 180^\circ$ 

* гҖҗе·ІзҹҘ 2гҖ‘ж•Је°„еҫҢй »зҺҮ $\nu'$ :
    
    $$\begin{gather*}
    \lambda' - \lambda &=& \frac{h}{mc}(1 - (-1)) \\
    \lambda' - \lambda &=& \frac{2h}{mc}\\
    \frac{c}{\nu'} - \frac{c}{\nu} &=& \frac{2h}{mc}\\
    \frac{1}{\nu'} - \frac{1}{\nu} &=& \frac{2h}{mc^2}\\
    \frac{1}{\nu'} &=& \frac{1}{\nu} + \frac{2h}{mc^2}  \\
    \frac{1}{\nu'} &=& \frac{mc^2 + 2h\nu}{\nu mc^2}\\
    \nu' &=& \frac{\nu mc^2}{mc^2 + 2h\nu}\\
    \end{gather*}$$



### proof



$$\begin{gather*}
E_{\text{е…үеӯҗ,еҲқ}} + E_{\text{йӣ»еӯҗ,еҲқ}} &\overset{\text{иғҪйҮҸе®ҲжҒҶ}}{=}& E_{\text{е…үеӯҗ,жң«}} + E_{\text{йӣ»еӯҗ,жң«}}\\
h\nu + mc^2 &=& h\nu' + (KE_{max} + mc^2)\\
KE_{max} &=& h\nu - h\nu' \\
KE_{max}&\overset{\text{е·ІзҹҘ 2}}{=}& h\nu - h\left( \frac{\nu mc^2}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& h\nu \left( 1 - \frac{mc^2}{mc^2 + 2h\nu} \right)\\
KE_{max}&=& h\nu \left( \frac{mc^2 + 2h\nu - mc^2}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& h\nu \left( \frac{2h\nu}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& \frac{2h^2\nu^2}{mc^2 + 2h\nu}\\
KE_{max} &=& \frac{2h^2\nu^2 / mc^2}{1 + 2h\nu / mc^2}
\end{gather*}$$


## Question 5

The gravitational potential energy $U$ relative to infinity of a body of mass m at a distance $R$ from the center of a body of mass $M$ is $U = -GmM/R$. 

(a) If $R$ is the radius of the body of mass $M$ , find the escape speed $v_e$ of the body, which is the minimum speed needed to leave it permanently. 

(b) Obtain a formula for the Schwarzschild radius of the body by setting $v_e=c$, the speed of light, and solving for $R$. (Of course, a relativistic calculation is correct here, but it is interesting to see what a classical calculation produces.)


## Question 5 - Answer


### е·ІзҹҘиҲҮеҒҮиЁӯ

* гҖҗе·ІзҹҘ 1гҖ‘еҠӣеӯёиғҪе®ҲжҒҶе®ҡеҫӢпјҲConservation of Mechanical Energyпјү:
  
   $$E_{total} = K + U = \text{еёёж•ё}$$

   
   * еҲқе§ӢзёҪиғҪйҮҸпјҡ$E_i = \frac{1}{2}mv_e^2 - \frac{GmM}{R}$
     * еҲқе§ӢеӢ•иғҪпјҡ$K_i = \frac{1}{2}mv_e^2$
     * еҲқе§ӢдҪҚиғҪпјҡ$U_i = -\frac{GmM}{R}$
   * жңҖзөӮзёҪиғҪйҮҸпјҡ$E_f = 0$
     * жңҖзөӮеӢ•иғҪпјҡ$K_f = 0$пјҲеӣ зӮәеҒңдёӢдҫҶдәҶпјү
     * жңҖзөӮдҪҚиғҪпјҡ$U_f = -\frac{GmM}{\infty} = 0$





### proof (a) е°ӢжүҫйҖғйҖёйҖҹеәҰ $v_e$



$$\begin{gather*}
\frac{1}{2}mv_e^2 - \frac{GmM}{R} &\overset{\text{е·ІзҹҘ 1}}{=}& 0 \\
\frac{1}{2}mv_e^2 &=& \frac{GmM}{R}\\
v_e^2 &=& \frac{2GM}{R}\\
v_e &=& \sqrt{\frac{2GM}{R}}
\end{gather*}$$



### proof (b)жҺЁе°ҺеҸІз“ҰиҘҝеҚҠеҫ‘ $R_S$



$$\begin{gather*}
v_e &\overset{\text{Q5.(a)}}{=}& \sqrt{\frac{2GM}{R}} \\
c &=& \sqrt{\frac{2GM}{R_s}}\\
R_s &=& \frac{2GM}{c^2}
\end{gather*}$$
