## Question 1 : Density of standing waves in cavity

Show that the de Broglie wavelength of a particle of mass $m$ and kinetic energy $KE$ is given by
   
$$\lambda = \frac{hc}{\sqrt{KE(KE + 2mc^2)}}$$




## Question 1 - Answer


* **【已知 2】 [考慮相對論效應的動量](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)：** 

  $$p = \gamma m_0 v$$

  * $p$ : 物體動量 ，$[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : 勞侖茲因子 (Lorentz factor)
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$


### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 [物體的能量與動量關係式](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#energy-and-momentum)：** 

  $$E^2 = (m_0 c^2)^2 + (pc)^2 $$

  * $E$ : 總能量 (Total Energy) $[\text{J}]$
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$
  * $p$ : 物體動量 ，$[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$



* **【定義 1】 德布羅依波 (De Broglie Wave)  ：**
  
  $$\lambda \overset{\text{def}}{=} \frac{h}{p}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


$$\begin{gather*}
E^2 &\overset{\text{已知 1}}{=}& (pc)^2 + (m_0 c^2)^2 \\
E^2 - (m_0 c^2)^2 &=& (pc)^2 \\
\frac{1}{c}\sqrt{E^2 - (m_0 c^2)^2} &=& p \\
\frac{1}{c}\sqrt{(KE+E_0)^2 - (m_0 c^2)^2} &=& p \\
\frac{1}{c}\sqrt{{KE}^2+2KE \cdot E_0} &=& p \\
\frac{1}{c}\sqrt{KE(KE + 2mc^2)} &=& p \\
\frac{1}{c}\sqrt{KE(KE + 2mc^2)} &\overset{\text{定義 1}}{=}& \frac{h}{\lambda} \\
\lambda &=&\frac{hc}{\sqrt{KE(KE + 2mc^2)}}  \\
\end{gather*}$$

## Question 2 

Show that if the total energy $E$ of a moving particle greatly exceeds its rest energy $E_0$ ($E \gg E_0$), its de Broglie wavelength $\lambda$ is nearly the same as the wavelength of a photon with the same total energy.
Proof:

   1. The relativistic relationship between total energy $E$, momentum $p$, and rest mass $m$ is:
   $$E^2 = (pc)^2 + (mc^2)^2$$ where $mc^2$ is the rest energy $E_0$.
   2. According to the given condition, $E \gg mc^2$. Therefore, the term $(mc^2)^2$ becomes negligible compared to $E^2$:
   $$E^2 \approx (pc)^2 \implies E \approx pc$$ 
   3. Solving for momentum $p$:
   $$p \approx \frac{E}{c}$$ 
   4. The de Broglie wavelength $\lambda$ is defined as:
   $$\lambda = \frac{h}{p} \approx \frac{h}{E/c} = \frac{hc}{E}$$ 
   5. For a photon with the same total energy $E$, its wavelength $\lambda_{photon}$ is given by:
   $$E = h\nu = \frac{hc}{\lambda_{photon}} \implies \lambda_{photon} = \frac{hc}{E}$$ 

Conclusion:
Comparing the two results, we see that $\lambda \approx \lambda_{photon}$ when $E \gg mc^2$. Q.E.D.

## Question 3

Express the Planck radiation formula in terms of wavelength.


## Question 3 - Answer



### 已知與假設

* 【已知 1】頻率 $\nu$ 與波長 $\lambda$ 的關係式:
  
   $$\nu = \frac{c}{\lambda}$$

   $$d\nu =-\frac{c}{\lambda^2} d\lambda$$

   * 能量是正的 $|d\nu| = \frac{c}{\lambda^2} |d\lambda|$

* 【已知 2】Jacobiany 在 $u(\nu)$ 與 $u(\lambda)$ 的關係式:
   
   $$|u(\nu)d\nu| = |u(\lambda)d\lambda|$$

   
### proof


$$\begin{gather*}
u(\nu)d\nu &=& G(\nu) \bar{\epsilon} d\nu \\
u(\nu)d\nu&\overset{\text{Q 1,2}}{=}& \left( \frac{8\pi \nu^2}{c^3} \right) \left( \frac{h\nu}{e^{h\nu/kT} - 1} \right) d\nu \\
u(\nu)d\nu&=& \frac{8\pi h \nu^3}{c^3 (e^{h\nu/kT} - 1)} d\nu  \\
u(\lambda)d\lambda &\overset{\text{已知 1,2}}{=}& \frac{8\pi h \left( \frac{c}{\lambda} \right)^3}{c^3 \left( e^{\frac{h(c/\lambda)}{kT}} - 1 \right)} \left( \frac{c}{\lambda^2} d\lambda \right)\\
u(\lambda)d\lambda &=& \frac{8\pi hc}{\lambda^5} \frac{1}{e^{hc/\lambda kT} - 1} d\lambda
\end{gather*}$$


## Question 4

A photon of frequency 𝜈 is scattered by an electron initially at rest. Verify that the maximum kinetic energy of the recoil electron is

$$KE_{max}=\frac{2h^{2}\nu^{2}/mc^{2}}{1+2h\nu/mc^{2}}$$



## Question 4 - Answer



### 已知與假設

* 【已知 1】[康普頓波長位移公式（Compton shift formula）](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#proof):
  
   $$\Delta \lambda = \lambda' - \lambda = \frac{h}{mc}(1 - \cos\theta)$$

   * 能量守恆、動量守恆
   * 讓電子獲得最大動能 $\theta = 180^\circ$ 

* 【已知 2】散射後頻率 $\nu'$ :
    
    $$\begin{gather*}
    \lambda' - \lambda &=& \frac{h}{mc}(1 - (-1)) \\
    \lambda' - \lambda &=& \frac{2h}{mc}\\
    \frac{c}{\nu'} - \frac{c}{\nu} &=& \frac{2h}{mc}\\
    \frac{1}{\nu'} - \frac{1}{\nu} &=& \frac{2h}{mc^2}\\
    \frac{1}{\nu'} &=& \frac{1}{\nu} + \frac{2h}{mc^2}  \\
    \frac{1}{\nu'} &=& \frac{mc^2 + 2h\nu}{\nu mc^2}\\
    \nu' &=& \frac{\nu mc^2}{mc^2 + 2h\nu}\\
    \end{gather*}$$



### proof



$$\begin{gather*}
E_{\text{光子,初}} + E_{\text{電子,初}} &\overset{\text{能量守恆}}{=}& E_{\text{光子,末}} + E_{\text{電子,末}}\\
h\nu + mc^2 &=& h\nu' + (KE_{max} + mc^2)\\
KE_{max} &=& h\nu - h\nu' \\
KE_{max}&\overset{\text{已知 2}}{=}& h\nu - h\left( \frac{\nu mc^2}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& h\nu \left( 1 - \frac{mc^2}{mc^2 + 2h\nu} \right)\\
KE_{max}&=& h\nu \left( \frac{mc^2 + 2h\nu - mc^2}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& h\nu \left( \frac{2h\nu}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& \frac{2h^2\nu^2}{mc^2 + 2h\nu}\\
KE_{max} &=& \frac{2h^2\nu^2 / mc^2}{1 + 2h\nu / mc^2}
\end{gather*}$$


## Question 5

The gravitational potential energy $U$ relative to infinity of a body of mass m at a distance $R$ from the center of a body of mass $M$ is $U = -GmM/R$. 

(a) If $R$ is the radius of the body of mass $M$ , find the escape speed $v_e$ of the body, which is the minimum speed needed to leave it permanently. 

(b) Obtain a formula for the Schwarzschild radius of the body by setting $v_e=c$, the speed of light, and solving for $R$. (Of course, a relativistic calculation is correct here, but it is interesting to see what a classical calculation produces.)


## Question 5 - Answer


### 已知與假設

* 【已知 1】力學能守恆定律（Conservation of Mechanical Energy）:
  
   $$E_{total} = K + U = \text{常數}$$

   
   * 初始總能量：$E_i = \frac{1}{2}mv_e^2 - \frac{GmM}{R}$
     * 初始動能：$K_i = \frac{1}{2}mv_e^2$
     * 初始位能：$U_i = -\frac{GmM}{R}$
   * 最終總能量：$E_f = 0$
     * 最終動能：$K_f = 0$（因為停下來了）
     * 最終位能：$U_f = -\frac{GmM}{\infty} = 0$





### proof (a) 尋找逃逸速度 $v_e$



$$\begin{gather*}
\frac{1}{2}mv_e^2 - \frac{GmM}{R} &\overset{\text{已知 1}}{=}& 0 \\
\frac{1}{2}mv_e^2 &=& \frac{GmM}{R}\\
v_e^2 &=& \frac{2GM}{R}\\
v_e &=& \sqrt{\frac{2GM}{R}}
\end{gather*}$$



### proof (b)推導史瓦西半徑 $R_S$



$$\begin{gather*}
v_e &\overset{\text{Q5.(a)}}{=}& \sqrt{\frac{2GM}{R}} \\
c &=& \sqrt{\frac{2GM}{R_s}}\\
R_s &=& \frac{2GM}{c^2}
\end{gather*}$$
