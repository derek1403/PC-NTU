## Question 1 


The frequency of vibration of the $\text{H}_2$ molecule is $1.32 \times 10^{14}\text{ Hz}$. 

* (a) Find the relative populations of the $v = 0, 1, 2, 3,$ and $4$ vibrational states at $5000\text{ K}$. 

* (b) Can the populations of the $v = 2$ and $v = 3$ states ever be equal? If so, at what temperature does this occur?




## Question 1 - Answer


### 假設與已知 (Assumptions & Preliminaries)




* **【已知 1】  波茲曼分佈（Boltzmann distribution）  ：** 在熱平衡時，處於能階 $E_v$ 的粒子數 $N_v$ 會與 $\exp(-E_v/kT)$ 成正比
   
  $$\frac{N_v}{N_0} = \frac{e^{-\frac{E_v}{kT}}}{e^{-\frac{E_0}{kT}}}$$

  * $N_{v}$ : 處於振動量子數為 $v$ 的激發態粒子數 (Number of particles in the excited state $v$) [無單位]
  * $N_{0}$ : 處於基態（$v=0$）的粒子數 (Number of particles in the ground state) [無單位]
  * $E_{v}$ : 振動量子數為 $v$ 的能階能量 (Energy of the state $v$) $[\text{J}]$
  * $E_{0}$ : 基態（零點振動）的能階能量 (Ground state energy) $[\text{J}]$
  * $k$ : 波茲曼常數 (Boltzmann constant) $k \approx 1.38 \times 10^{-23} \text{ J}\cdot\text{K}^{-1}$
  * $T$ : 系統的絕對溫度 (Absolute temperature) $[\text{K}]$

* **【已知 2】 量子諧振子的子能階與能量差（Vibrational energy levels and difference）  ：**
  
  $$\begin{gather*}
  E_v &=& (v + \frac{1}{2})h\nu \\
  E_v - E_0 &=&  (v + \frac{1}{2})h\nu  - (0 + \frac{1}{2})h\nu  \\
  E_v - E_0 &=&  vh\nu \\
  \end{gather*}$$

  * $v$ : 振動量子數 (Vibrational quantum number) $v = 0, 1, 2, \dots$ [無單位]
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s}$
  * $\nu $ : 分子的特徵振動頻率 (Characteristic vibrational frequency) $[\text{Hz}],[\text{s}^{-1}]$

* **【推導 1】 特定溫度與頻率下的激發態粒子比例（Derivation of excited state population ratio）  ：** 代入已知條件與特定物理量（例如 $\nu = 1.32 \times 10^{14} \text{ Hz}$, $T = 5000 \text{ K}$），計算出相對於基態的粒子分佈比例。
  
  $$\begin{gather*}
  \frac{N_v}{N_0} &\overset{\text{已知 1}}{=}& \frac{e^{-\frac{E_v}{kT}}}{e^{-\frac{E_0}{kT}}} \\
  &=& e^{-\frac{E_v - E_0}{kT}} \\
  &\overset{\text{已知 2}}{=}& e^{-\frac{vh\nu}{kT}} \\
  &\overset{\text{Q1,已知1,已知2}}{\approx} &  e^{-\frac{v \times (6.626 \times 10^{-34} \text{ J s})(1.32 \times 10^{14} \text{ s}) }{(1.38 \times 10^{-23} \text{ J/K})(5000 \text{ K})}} \\
  &\approx&  e^{-v \cdot 1.2675} \\
  \end{gather*}$$

  * $N_{v}$ : 處於振動量子數為 $v$ 的激發態粒子數 (Number of particles in the excited state $v$) [無單位]
  * $N_{0}$ : 處於基態（$v=0$）的粒子數 (Number of particles in the ground state) [無單位]
  * $E_{v}$ : 振動量子數為 $v$ 的能階能量 (Energy of the state $v$) $[\text{J}]$
  * $E_{0}$ : 基態（零點振動）的能階能量 (Ground state energy) $[\text{J}]$
  * $k$ : 波茲曼常數 (Boltzmann constant) $k \approx 1.38 \times 10^{-23} \text{ J}\cdot\text{K}^{-1}$
  * $T$ : 系統的絕對溫度 (Absolute temperature) $[\text{K}]$
  * $v$ : 振動量子數 (Vibrational quantum number) $v = 0, 1, 2, \dots$ [無單位]
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s}$
  * $\nu $ : 分子的特徵振動頻率 (Characteristic vibrational frequency) $[\text{Hz}],[\text{s}^{-1}]$
  

### solve 分子佔據基態 ($v=0$) 到第四激發態 ($v=4$) 的相對比例是多少

1. $v = 0$

$$\begin{gather*}
\frac{N_v}{N_0} &\overset{\text{推導 1}}{=}& e^{-v \cdot 1.2675} \\
&\overset{\text{Q1.1.1}}{=}& e^{-0 \times 1.2675} \\
&=& e^{0} \\
&=& 1 \\
\end{gather*}$$


2. $v = 1$

$$\begin{gather*}
\frac{N_v}{N_0} &\overset{\text{推導 1}}{=}& e^{-v \cdot 1.2675} \\
&\overset{\text{Q1.1.2}}{=}& e^{-1 \times 1.2675} \\
&\approx& e^{-1.2675} \\
&\approx& 0.2815 \\
\end{gather*}$$

3. $v = 2$

$$\begin{gather*}
\frac{N_v}{N_0} &\overset{\text{推導 1}}{=}& e^{-v \cdot 1.2675} \\
&\overset{\text{Q1.1.3}}{=}& e^{-2 \times 1.2675} \\
&\approx& e^{-2.5350} \\
&\approx& 0.0793 \\
\end{gather*}$$

4. $v = 3$

$$\begin{gather*}
\frac{N_v}{N_0} &\overset{\text{推導 1}}{=}& e^{-v \cdot 1.2675} \\
&\overset{\text{Q1.1.4}}{=}& e^{-3 \times 1.2675} \\
&\approx& e^{-3.8025} \\
&\approx& 0.0223 \\
\end{gather*}$$

5. $v = 4$

$$\begin{gather*}
\frac{N_v}{N_0} &\overset{\text{推導 1}}{=}& e^{-v \cdot 1.2675} \\
&\overset{\text{Q1.1.5}}{=}& e^{-4 \times 1.2675} \\
&\approx& e^{-5.0700} \\
&\approx& 0.0063 \\
\end{gather*}$$

### $v=2$ 與 $v=3$ 粒子數相等的情況

$$\begin{gather*}
N_2 &=& N_3 \\
\frac{N_2}{N_0} &=& \frac{N_3}{N_0} \\
e^{-\frac{2h\nu}{kT}} &\overset{\text{推導 1}}{=}& e^{-\frac{3h\nu}{kT}} \\
-\frac{2h\nu}{kT} &=& -\frac{3h\nu}{kT} \\
\frac{h\nu}{kT} &=& 0 \\
\end{gather*}$$

由於 $h$ 與 $\nu$ 皆為不等於零的正實數常數，若要使上式成立，唯有當分母的溫度 $T \to \infty$ 時才有可能。

因此結論是：在任何實際的有限溫度下，這兩個態的粒子數是不可能相等的。

## Question 2 


* (a) Verify that the average value of $1/v$ for an ideal-gas molecule is $\sqrt{\frac{2m}{\pi k T}}$.
* (b) Verify that the most probable speed of an ideal-gas molecule is $\sqrt{\frac{2kT}{m}}$.



## Question 2 - Answer


### 已知與假設

* **【已知 1】馬克士威速率分佈（Maxwell-Boltzmann speed distribution） :** 
  
  $$f(v) = 4\pi \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} v^2 e^{-\frac{mv^2}{2kT}}$$

  * $f(v)$ : 速率分佈函數 (Speed distribution function) $[\text{s}\cdot\text{m}^{-1}]$
  * $m$ : 氣體分子質量 (Mass of a single gas molecule) $[\text{kg}]$
  * $k$ : 波茲曼常數 (Boltzmann constant) $k \approx 1.38 \times 10^{-23} \text{ J}\cdot\text{K}^{-1}$
  * $T$ : 絕對溫度 (Absolute temperature) $[\text{K}]$
  * $v$ : 分子的熱運動速率 (Molecular speed) $[\text{m}\cdot\text{s}^{-1}]$
  
* **【已知 2】速率函數的統計平均值定義（Statistical expectation value definition） :** 
  
  $$\langle g(v) \rangle = \int_0^\infty g(v) f(v) dv$$

  * $\langle g(v) \rangle$ : 函數 $g(v)$ 的統計期望值 (Expectation value / Average value)
  * $g(v)$ : 關於速率 $v$ 的任意連續物理量函數 (Any continuous function of speed v)
  * $f(v)$ : 速率分佈函數 (Speed distribution function) $[\text{s}\cdot\text{m}^{-1}]$

* **【推導 1】積分變數變換 :** 

  * (a) 令亞變數 $x$ 與常數 $a$
  
  $$ \begin{gather*}
  x &\overset{\text{let}}{=}& av^2 \\
  dx &=& 2av \ dv \\
  \frac{1}{2a} \ dx &=& v \ dv \\
  \end{gather*}$$

  * (b) 利用變數變換求解特定積分項 

  $$\begin{gather*}
  \int_0^\infty v e^{-av^2} dv  &\overset{\text{推導 1(a)}}{=}& \frac{1}{2a} \int_0^\infty e^{-x} dx \\
  &=&  \frac{1}{2a} \int_0^\infty e^{-x} dx \\
  &=&  \frac{1}{2a} \left[ -e^{-x} \right]_0^\infty \\
  &=&  \frac{1}{2a}  \\
  \end{gather*}$$

  * $a$ : 簡化運算引入的常數係數 (Substitution constant) $a = \frac{m}{2kT}$ $[\text{s}^2\cdot\text{m}^{-2}]$
  * $x$ : 無因次變數變換代號 (Dimensionless dummy variable) [無單位]

### (a) proof $\frac{1}{v}$ 的平均值為 $\sqrt{\frac{2m}{\pi kT}}$  

$$\begin{gather*}
\langle \frac{1}{v} \rangle &\overset{\text{已知 2}}{=}& \int_0^\infty \frac{1}{v} f(v) dv \\
&\overset{\text{已知 1}}{=}& \int_0^\infty \frac{1}{v} 4\pi \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} v^2 e^{-\frac{mv^2}{2kT}} dv \\
&=& 4\pi \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} \int_0^\infty v e^{-\frac{mv^2}{2kT}} dv \\
&\overset{\text{推導 1}}{=}& 4\pi \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} \frac{2kT}{2m} \\
&=& 4\pi \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} \frac{kT}{m} \\
&=& \sqrt{\frac{2}{\pi}} \sqrt{\frac{m}{kT}} \\
&=&  \sqrt{\frac{2m}{\pi kT}} \\
\end{gather*}$$


### (b) proof 最有可能的速度 $\sqrt{\frac{2kT}{m}}$ 

$$\begin{gather*}
0 &=& \frac{d}{dv} \left[ f(v) \right]  \\
0 &\overset{\text{已知 1}}{=}& \frac{d}{dv} \left[ 4\pi \left(\frac{m}{2\pi kT}\right)^{\frac{3}{2}} v^2 e^{-\frac{mv^2}{2kT}} \right]  \\
0 &=& \frac{d}{dv} \left[ v^2 e^{-\frac{mv^2}{2kT}} \right]  \\
0 &=& 2ve^{-\frac{mv^2}{2kT}} + v^2  \left( e^{-\frac{mv^2}{2kT}} \left( - \frac{m2v}{2kT} \right) \right)\\
0 &=&  v e^{-\frac{mv^2}{2kT}} \left( 2 - \frac{mv^2}{kT} \right)  \\
0 &=&  2 - \frac{mv^2}{kT}   \\
\frac{mv^2}{kT}&=&  2     \\
v &=&  \sqrt{\frac{2kT}{m}}    \\
\end{gather*}$$

## Question 3

Show that the median energy $\epsilon_{\text{med}}$ in a free-electron gas at $T=0\text{ K}$ equal to $\frac{\epsilon_F}{2^{\frac{2}{3}}} = 0.630\epsilon_F$.


## Question 3 - Answer



### 已知與假設

* **【已知 1】 **

  $$N =\int_0^{\infty} g(\varepsilon) f_{\text{Fb}}(\varepsilon)  d\varepsilon$$


* **【已知 2】 三維自由電子氣體的狀態密度函數**

  $$g(\epsilon) = C \varepsilon^{1/2}$$

* **【已知 3】 FB分布 **

  $$f_{\text{Fb}}(\varepsilon) &=& \frac{1}{e^{\frac{\varepsilon - \mu}{kT}} + 1}$$

* **【推導 1】 :**
  
  $$\begin{gather*}
  f_{\text{Fb}}(\varepsilon) &\overset{\text{已知 3}}{=}& \frac{1}{e^{\frac{\varepsilon - \mu}{kT}} + 1} \\
  &\overset{\text{假設 1}}{=}& \lim\limits_{T \to 0} \frac{1}{e^{\frac{\varepsilon - \mu}{kT}} + 1} \\
  &=&  \frac{1}{0 + 1} \\
  &=&  1 \\
  \end{gather*}$$

* **【假設 1】 $T=0\text{ K}$**

  * (a) 費米面（Fermi surface）變成 step-function
    
  $$\begin{gather*}
  \int_0^{\infty} g(\varepsilon) f_{\text{Fb}}(\varepsilon)  d\varepsilon  &=& \int_0^{\varepsilon_F} g(\varepsilon) f_{\text{Fb}}(\varepsilon)  d\varepsilon\\
  \end{gather*}$$


### proof 

$$\begin{gather*}
N_{\text{med}} &=& \frac{1}{2} N\\
&\overset{\text{已知 1}}{=}& \frac{1}{2} \int_0^{\infty} g(\varepsilon) f_{\text{Fb}}(\varepsilon)  d\varepsilon \\
&\overset{\text{假設 1(a)}}{=}& \frac{1}{2} \int_0^{\\varepsilon_F} g(\varepsilon) f_{\text{Fb}}(\varepsilon)  d\varepsilon \\
&\overset{\text{已知 3}}{=}& \frac{1}{2} \int_0^{\varepsilon_F} g(\varepsilon) d\varepsilon \\
&\overset{\text{已知 2}}{=}& \frac{1}{2} \int_0^{\varepsilon_F} C \varepsilon^{\frac{1}{2}} d\varepsilon \\
&=&\frac{1}{2} \int_0^{\varepsilon_F} C \varepsilon^{1/2} d\varepsilon \\

\end{gather*}$$


## Question 4

Assume that the $\text{H}_2$ molecule behaves exactly like a harmonic oscillator with a force constant of $573\text{ N/m}$. 

* (a) Find the energy (in $\text{eV}$) of its ground and first excited vibrational states. 
* (b) Find the vibrational quantum number that approximately corresponds to its $4.5\text{ eV}$ dissociation energy.




## Question 4 - Answer



### 假設與已知 (Assumptions & Preliminaries)


* **【假設 1】 建立一維座標系 One-dimensional coordinate system for diatomic vibration） ：**

  1. $m_1$ 的位置在 $x_1$
  2. $m_2$ 的位置在 $x_2$（假設 $x_2 > x_1$）
  3. 彈簧的自然長度（沒有被拉伸或壓縮時的長度）為 $L_0$
  4. 兩個原子之間的實際距離為 $x = x_2 - x_1$
  5. 彈簧被拉伸或壓縮的形變量可以表示為 $\Delta x = x - L_0$
 

  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $x_1, x_2$ : 兩原子在一維軸上的絕對位置 (Positions of the two atoms) $[\text{m}]$
  * $L_{0}$ : 鍵的平衡長度／自然長度 (Equilibrium bond length) $[\text{m}]$
  * $x$ : 雙原子間的瞬時實際鍵長 (Instantaneous bond length) $[\text{m}]$
  * $\Delta x$ : 遠離平衡位置的位移／形變量 (Displacement from equilibrium) $[\text{m}]$



* **【已知 1】 個別原子的牛頓第二運動定律與受力（Newton's second law for individual atoms）  ：**

  * 對於 $m_1$（受到向右的拉力）：
  
  $$\begin{gather*}
  \vec{F} &=& \vec{F}_{\text{spring}} \\
  m_1 \vec{a_1} &=& -k \Delta \vec{x} \\
  m_1 a_1 &=& k\Delta x  \\
  m_1 \frac{d^2 x_1}{dt^2} &=& k\Delta x  \\
  \frac{d^2 x_1}{dt^2} &=& \frac{k}{m_1}\Delta x \\
  \end{gather*}$$

  * 對於 $m_2$（受到向左的拉力）：
  
  $$\begin{gather*}
  \vec{F} &=& \vec{F}_{\text{spring}} \\
  m_2 \vec{a_2} &=& -k \Delta \vec{x} \\
  m_2 a_2 &=& -k\Delta x  \\
  m_2 \frac{d^2 x_2}{dt^2} &=& -k\Delta x  \\
  \frac{d^2 x_2}{dt^2} &=& -\frac{k}{m_2}\Delta x \\
  \end{gather*}$$

  * $k$ : 鍵的力常數／彈簧力常數 (Force constant of the chemical bond) $[\text{N}\cdot\text{m}^{-1}],[\text{kg}\cdot\text{s}^{-2}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $x$ : 雙原子間的瞬時實際鍵長 (Instantaneous bond length) $[\text{m}]$
  * $\Delta x$ : 遠離平衡位置的位移／形變量 (Displacement from equilibrium) $[\text{m}]$
  * $a_1, a_2$ : 兩原子各自的加速度大小 (Accelerations of individual atoms) $[\text{m}\cdot\text{s}^{-2}]$

* **【定義 1】縮減質量 (Reduce mass , $m'$) :**

  $$\begin{gather*}
  m' &\overset{\text{def}}{=}& \frac{m_1 m_2}{m_1 + m_2}  \\
  &\overset{\text{Q4}}{=}& \frac{m_p m_p}{m_p + m_p} \\
  &=& \frac{1}{2} m_p\\
  \end{gather*}$$

  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

* **【定義 2】系統的特徵角頻率 (Characteristic angular frequency of the system, $\omega$) :**

  $$\begin{gather*}
  \omega^2 &\overset{\text{def}}{=}& \frac{k}{m'} \\
  &\overset{\text{定義 1}}{=}& \frac{k}{\frac{1}{2} m_p}\\
  &=& 2\frac{k}{m_p}\\
  \end{gather*}$$

  * $\omega $ : 簡諧振動的特徵角頻率 (Characteristic angular frequency) $[\text{rad}\cdot\text{s}^{-1}]$
  * $k$ : 鍵的力常數／彈簧力常數 (Force constant of the chemical bond) $[\text{N}\cdot\text{m}^{-1}],[\text{kg}\cdot\text{s}^{-2}]$
  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

* **【已知 2】 相對運動的簡諧運動微分方程式（Differential equation of relative motion）  ：**

  $$\begin{gather*}
  \frac{d^2 x}{dt^2} &=& \frac{d^2 (x_2 - x_1)}{dt^2} \\
  \frac{d^2 x}{dt^2} &=& \frac{d^2 x_2}{dt^2} - \frac{d^2 x_1}{dt^2}  \\
  \frac{d^2 x}{dt^2} &=& -\frac{k}{m_2}\Delta x - \frac{k}{m_1}\Delta x  \\
  \frac{d^2 x}{dt^2} &=& -k \left( \frac{1}{m_2}+\frac{1}{m_1}  \right) \Delta x  \\
  \frac{d^2 x}{dt^2} &=& -k \left( \frac{m_1+ m_2}{m_2 m_1}  \right) \Delta x  \\
  \frac{d^2 x}{dt^2} &\overset{\text{定義 1}}{=}&  \frac{-k}{m'} \Delta x  \\
  \frac{d^2 }{dt^2} \left[ \Delta x + L_0 \right]&\overset{\text{假設 1}}{=}&  \frac{-k}{m'} \Delta x  \\
  \frac{d^2 \Delta x}{dt^2} &=&  \frac{-k}{m'} \Delta x  \\
  \frac{d^2 \Delta x}{dt^2} +  \frac{k}{m'} \Delta x  &=&0\\
  \frac{d^2 \Delta x}{dt^2} +  \omega^2 \Delta x  &\overset{\text{定義 2}}{=}&0\\
  \end{gather*}$$



* **【已知 3】 量子化能階  ：**
  
  $$\begin{gather*}
  E_v = \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
  \end{gather*}$$

  * $E_{v}$ : 量子化後的振動能階 (Vibrational energy levels) $[\text{J}]$
  * $v$ : 振動量子數 (Vibrational quantum number) $v = 0, 1, 2, \dots$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $\frac{1}{2}\hbar\omega$ : 零點能量 (Zero-point energy)，即基態（$v=0$）時的系統能量
  

### (a) vibrational states 能階對應的能量

1. 基態能量 ($v=0$)：

$$\begin{gather*}
E_v &\overset{\text{定義 1}}{=}& \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
E_0 &\overset{v=0}{=}& \left( 0 + \frac{1}{2} \right) \hbar \omega \\
E_0 &\overset{\text{定義 2}}{=}& \frac{1}{2} \hbar \sqrt{2\frac{k}{m_p}} \\
E_0 &=& \frac{1}{\sqrt{2}} \hbar \sqrt{\frac{k}{m_p}} \\
E_0 &\overset{\text{Q4}}{=}& \frac{1}{\sqrt{2}} 1.054 \times 10^{-34}\text{ J}\cdot\text{s}  \sqrt{\frac{573\text{ N/m}}{1.6726 \times 10^{-27}\text{ kg}}} \\
E_0 &=&  4.36 \times 10^{-20} \text{ J} \\
E_0 &\approx& 0.272 \text{ eV} \\
\end{gather*}$$

2. 第一激發態能量 ($v=1$)：

$$\begin{gather*}
E_v &\overset{\text{定義 1}}{=}& \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
E_1 &\overset{v=1}{=}& \left( 1 + \frac{1}{2} \right) \hbar \omega \\
E_1 &\overset{\text{定義 2}}{=}& \frac{3}{2} \hbar \sqrt{2\frac{k}{m_p}} \\
E_1 &=& \frac{3}{\sqrt{2}} \hbar \sqrt{\frac{k}{m_p}} \\
E_1 &\overset{\text{Q4}}{=}& \frac{3}{\sqrt{2}} 1.054 \times 10^{-34}\text{ J}\cdot\text{s}  \sqrt{\frac{573\text{ N/m}}{1.6726 \times 10^{-27}\text{ kg}}} \\
E_1 &=&  1.31 \times 10^{-19} \text{ J} \\
E_1 &\approx& 0.817 \text{ eV} \\
\end{gather*}$$

### (b) 解離能對應到的 vibrational quantum number




$$\begin{gather*}
E_v &\overset{\text{定義 1}}{=}& \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
4.5 \text{ eV} &\overset{\text{Q4}}{=}& \left( v + \frac{1}{2} \right) \hbar \omega & v \in \{0, 1, 2, \dots\} \\
v &=& \frac{4.5 \text{ eV}}{\hbar \omega}-\frac{1}{2} & v \in \{0, 1, 2, \dots\} \\
v  &\overset{\text{定義 2}}{=}& \frac{4.5 \text{ eV}}{\hbar \sqrt{2\frac{k}{m_p}}}-\frac{1}{2} & v \in \{0, 1, 2, \dots\} \\
v  &\overset{\text{Q4}}{=}& \frac{4.5 \text{ eV}}{1.054 \times 10^{-34}\text{ J}\cdot\text{s} \sqrt{2\frac{573\text{ N/m}}{1.6726 \times 10^{-27}\text{ kg}}}}-\frac{1}{2} & v \in \{0, 1, 2, \dots\} \\
v  &\approx& 7.763 & v \in \{0, 1, 2, \dots\} \\
v  &=& 8  \\
\end{gather*}$$

## Question 5

The lowest vibrational states of the $^{23}\text{Na}^{35}\text{Cl}$ molecule are $0.063\text{ eV}$ apart. Find the approximate force constant of this molecule.


## Question 5 - Answer


### 已知與假設




* **【已知 1】 量子化能階  ：**
  
  $$\begin{gather*}
  E_v = \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
  \end{gather*}$$

  * $E_{v}$ : 量子化後的振動能階 (Vibrational energy levels) $[\text{J}]$
  * $v$ : 振動量子數 (Vibrational quantum number) $v = 0, 1, 2, \dots$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $\frac{1}{2}\hbar\omega$ : 零點能量 (Zero-point energy)，即基態（$v=0$）時的系統能量
  

* **【定義 1】縮減質量 (Reduce mass , $m'$) :**

  $$\begin{gather*}
  m' &\overset{\text{def}}{=}& \frac{m_1 m_2}{m_1 + m_2}  \\
  &\overset{\text{Q5}}{=}& \frac{23 \text{u} \cdot 35\text{u}}{23\text{u} + 35\text{u}} \\
  &=& \frac{805}{58} \text{u}\\
  &\approx& \frac{805}{58} m_p\\
  \end{gather*}$$

  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $\text{u}$ : 原子質量單位 (Atomic Mass Unit) $1\text{u} \approx 1 m_p $ ，$[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

* **【定義 2】系統的特徵角頻率 (Characteristic angular frequency of the system, $\omega$) :**

  $$\begin{gather*}
  \omega^2 &\overset{\text{def}}{=}& \frac{k}{m'} \\
  &\overset{\text{定義 1}}{=}& \frac{k}{\frac{805}{58} m_p}\\
  &=& \frac{58}{805}\frac{k}{m_p}\\
  \end{gather*}$$

  * $\omega $ : 簡諧振動的特徵角頻率 (Characteristic angular frequency) $[\text{rad}\cdot\text{s}^{-1}]$
  * $k$ : 鍵的力常數／彈簧力常數 (Force constant of the chemical bond) $[\text{N}\cdot\text{m}^{-1}],[\text{kg}\cdot\text{s}^{-2}]$
  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

###  solve force constant

$$\begin{gather*}
\Delta E &=& E_{v+1} - E_v \\
\Delta E &\overset{\text{已知 1}}{=}& \left( v + 1 + \frac{1}{2} \right) \hbar \omega - \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
\Delta E &=& \hbar \omega  \\
\Delta E &\overset{\text{定義 2}}{=}& \hbar \sqrt{\frac{58}{805}\frac{k}{m_p}}  \\
k &=& \frac{805}{58} m_p \left( \frac{\Delta E }{\hbar} \right)^2 \\
k &=& \frac{805}{58} \cdot 1.6726 \times 10^{-27}\text{ kg} \left( \frac{0.063\text{ eV}}{1.054 \times 10^{-34}\text{ J}\cdot\text{s}} \right)^2   \\
k &=& 212.85 \text{ N}\cdot\text{m}^{-1} \\
\end{gather*}$$
