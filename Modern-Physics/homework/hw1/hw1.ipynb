## Question 1

Prove time dilation
   
$$t = \frac{2L_0/c}{\sqrt{1-v^2/c^2}} = \frac{t_0}{\sqrt{1-v^2/c^2}}$$

## Question 1 - Answer

雖然題目是寫 $t$，但寫成 $\Delta t$（時間間隔）更精準，以下推導採 $\Delta t$。

![](./pic/Light_Clock_Derivation.png)

### 假設與已知 (Assumptions & Preliminaries)

* **【假設 1】光鐘模型 (Light-clock model)：** 飛船（$S'$ 系）上有一光鐘，光子在相距 $L_0$ 的兩面平行鏡之間沿**垂直於運動方向**往返反射。

  $$\text{在 } S' \text{ 系中，光子單程路徑長} = L_0 \perp v$$

  * $L_0$ : 鏡距 (mirror separation) $[\text{m}]$
  * $v$ : 飛船相對地球的速度 (relative speed of the spaceship) $[\text{m}\cdot\text{s}^{-1}]$
  * $S'$ : 光鐘靜止座標系 (rest frame of the clock) $[\text{無單位}]$
  * $S$ : 地球座標系 (earth frame) $[\text{無單位}]$

* **【已知 1】光速不變原理 (Constancy of the speed of light)：** 光速 $c$ 在所有慣性座標系中皆相同。

  $$c = \text{constant}$$

  * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$，$c \approx 3.0 \times 10^{8} \text{ m}\cdot\text{s}^{-1}$

* **【已知 2】畢氏定理 (Pythagorean theorem)：** 直角三角形斜邊平方等於兩股平方和。

  $$h^2 = a^2 + b^2$$

  * $h$ : 斜邊 (hypotenuse) $[\text{m}]$
  * $a,\, b$ : 兩直角股 (two legs) $[\text{m}]$

* **【定義 1】原時 (Proper time)：** 在光鐘靜止系 $S'$ 中量得光子往返一次的時間。

  $$\Delta t_0 \overset{\text{def}}{=} (S' \text{ 系中光子往返一週期所需的時間})$$

  * $\Delta t_0$ : 原時 (proper time) $[\text{s}]$

* **【定義 2】膨脹時間 (Dilated time)：** 在地球系 $S$ 中量得光子往返一次的時間。

  $$\Delta t \overset{\text{def}}{=} (S \text{ 系中光子往返一週期所需的時間})$$

  * $\Delta t$ : 地球系量得的時間 (dilated time) $[\text{s}]$

* **【推導 1】原時與鏡距關係：** 由定義 1 原時 $\Delta t_0$（$S'$ 系往返一週期），其半程為光子以光速 $c$ 沿單程 $L_0$ 直線通過。

  $$\begin{gather*}
  \frac{\Delta t_0}{2} &\overset{\text{圖(a)}}{=}& \frac{L_0}{c} \\
  L_0 &=& c\,\frac{\Delta t_0}{2} \\
  \Delta t_0 &=& \frac{2L_0}{c}
  \end{gather*}$$

* **【推導 2】地球系斜線路徑：** 由定義 2 膨脹時間 $\Delta t$（$S$ 系往返一週期），光子走斜線：鉛直分量為 $L_0$、水平分量為 $v\frac{\Delta t}{2}$、斜邊為 $c\frac{\Delta t}{2}$，代入畢氏定理。

  $$\left( c \frac{\Delta t}{2} \right)^2 \overset{\text{圖(c)}}{=} L_0^2 + \left( v \frac{\Delta t}{2} \right)^2$$

### proof

由地球系光子路徑出發，代入原時與鏡距關係即得時間膨脹。

$$\begin{gather*}
\left( c \frac{\Delta t}{2} \right)^2
&\overset{\text{推導 2}}{=}& L_0^2 + \left( v \frac{\Delta t}{2} \right)^2 \\
\left( c \frac{\Delta t}{2} \right)^2 &\overset{\text{推導 1}}{=}& \left( c \frac{\Delta t_0}{2} \right)^2 + \left( v \frac{\Delta t}{2} \right)^2 \\
c^2 (\Delta t)^2 &=& c^2 (\Delta t_0)^2 + v^2 (\Delta t)^2 \\
(\Delta t)^2 (c^2 - v^2) &=& c^2 (\Delta t_0)^2 \\
(\Delta t)^2 &=& \frac{1}{1 - v^2/c^2} (\Delta t_0)^2 \\
\Delta t &=& \frac{1}{\sqrt{1 - v^2/c^2}}\, \Delta t_0 \\
\Delta t &\overset{\text{推導 1}}{=}& \frac{1}{\sqrt{1 - v^2/c^2}} \cdot \frac{2L_0}{c}
\end{gather*}$$

故靜止觀察者測得運動時鐘變慢，得證。

## Question 2

Prove that when $v$ is much smaller than $c$ , the kinetic energy reduces to the classical definition

$$
KE = \gamma m_0c^2-m_0c^2 = (\gamma-1) m_0c^2
$$

## Question 2 - Answer

### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】古典動能公式 (Classical kinetic energy)：**

  $$KE_{\text{classical}} = \frac{1}{2} mv^2$$

  * $KE_{\text{classical}}$ : 古典動能 (classical kinetic energy) $[\text{J}]$
  * $m$ : 質量 (mass) $[\text{kg}]$
  * $v$ : 速率 (speed) $[\text{m}\cdot\text{s}^{-1}]$

* **【假設 1】低速極限 (Low-speed / non-relativistic limit)：** 物體速率遠小於光速。

  $$\begin{gather*}
  v &\ll& c \\
  \frac{v^2}{c^2} &\ll& 1
  \end{gather*}$$

  * $v$ : 速率 (speed) $[\text{m}\cdot\text{s}^{-1}]$
  * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$

* **【已知 2】[勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)：** 含定義 (a) 與低速質量接合 (b)。

  * (a) 定義 (definition)：

  $$\gamma = \frac{1}{\sqrt{1-v^2/c^2}}$$

  * (b) 低速質量接合 (mass bridge)：由相對論質量 $m=\gamma m_0$，在 $v\ll c$ 下取**零階** $\gamma\approx 1$，故古典質量等於靜止質量。

  $$\begin{gather*}
  m &=& \gamma m_0 \\
  &\overset{\text{假設 1}}{=}& m_0\\
  \end{gather*}$$

  * $\gamma$ : 勞侖茲因子 (Lorentz factor) $[\text{無單位}]$
  * $m$ : 古典力學中的質量 (mass) $[\text{kg}]$
  * $m_0$ : 靜止質量 (rest mass) $[\text{kg}]$
  * $v$ : 速率 (speed) $[\text{m}\cdot\text{s}^{-1}]$
  * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$

  > ※ 註：(b) 取零階 $\gamma\approx 1$ **僅用於接合質量符號** $m\!\leftrightarrow\! m_0$；主推導中的 $(\gamma-1)$ 仍透過已知 3 保留**一階** $\tfrac{1}{2}v^2/c^2$，兩處對 $\gamma$ 取的精度不同。

* **【已知 3】泰勒級數一階展開 (First-order Taylor expansion)：**

  $$(1+\varepsilon)^n \approx 1+n \varepsilon \quad (|\varepsilon| \ll 1)$$

  * $\varepsilon$ : 微小量 (small quantity) $[\text{無單位}]$
  * $n$ : 冪次 (exponent) $[\text{無單位}]$

* **【已知 4】分母微擾近似 (Denominator perturbation approximation)：**

  $$\frac{a}{b+\varepsilon} \approx \frac{a}{b} \quad (|\varepsilon| \ll |b|)$$

  * $a$ : 分子 (numerator) $[\text{無單位}]$
  * $b$ : 分母主項 (leading denominator term) $[\text{無單位}]$
  * $\varepsilon$ : 微小量 (small quantity) $[\text{無單位}]$

### proof

由相對論動能（以靜止質量 $m_0$ 表示）出發，在 $v \ll c$ 下逐步近似至古典極限。

$$\begin{gather*}
KE &=& (\gamma-1)m_0 c^2 \\
&\overset{\text{已知 2(a)}}{=}& \left(\frac{1}{\sqrt{1-v^2/c^2}}-1\right)m_0 c^2 \\
&=& \frac{ 1-\sqrt{1-v^2/c^2} }{\sqrt{1-v^2/c^2}}\,m_0 c^2 \\
&\overset{\text{已知 3}}{=}& \frac{ 1-(1-\frac{1}{2}v^2/c^2) }{1-\frac{1}{2}v^2/c^2}\,m_0 c^2 \\
&=& \frac{ \frac{1}{2}v^2/c^2 }{1-\frac{1}{2}v^2/c^2}\,m_0 c^2 \\
&\overset{\text{已知 4}}{=}& \frac{1}{2}\frac{v^2}{c^2}\,m_0 c^2 \\
&=& \frac{1}{2} m_0 v^2 \\
&\overset{\text{已知 2(b)}}{=}& \frac{1}{2} m v^2 \\
&\overset{\text{已知 1}}{=}& KE_{\text{classical}}
\end{gather*}$$

## Question 3

Prove equation:

$$E^2 = (m_0c^2)^2+p^2c^2$$

## Question 3 - Answer

我們希望找出不包含速度變數 $v$ 的能量與動量關係。

### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】[物體總能量 (Total energy)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#mass-and-energy)：**

  $$E = \gamma m_0 c^2$$

  * $E$ : 總能量 (total energy) $[\text{J}]$
  * $\gamma$ : 勞侖茲因子 (Lorentz factor) $[\text{無單位}]$
  * $m_0$ : 靜止質量 (rest mass) $[\text{kg}]$
  * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$

* **【已知 2】[相對論動量 (Relativistic momentum)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)：**

  $$p = \gamma m_0 v$$

  * $p$ : 動量 (momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : 勞侖茲因子 (Lorentz factor) $[\text{無單位}]$
  * $m_0$ : 靜止質量 (rest mass) $[\text{kg}]$
  * $v$ : 速率 (speed) $[\text{m}\cdot\text{s}^{-1}]$

* **【已知 3】[勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)：** 為便於後續代換，同時整理出 $\gamma^{-2}$ 形式。

  $$\begin{gather*}
  \gamma &=& \frac{1}{\sqrt{1-v^2/c^2}} \\
  \gamma^{-2} &=& 1-\frac{v^2}{c^2}
  \end{gather*}$$

  * $\gamma$ : 勞侖茲因子 (Lorentz factor) $[\text{無單位}]$
  * $v$ : 速率 (speed) $[\text{m}\cdot\text{s}^{-1}]$
  * $c$ : 光速 (speed of light) $[\text{m}\cdot\text{s}^{-1}]$

### proof

將總能量與動量平方後相減，消去 $\gamma$ 與 $v$。

$$\begin{gather*}
E^2 - (pc)^2 &\overset{\text{已知 1,2}}{=}& (\gamma m_0 c^2)^2 - (\gamma m_0 v c)^2 \\
E^2 - (pc)^2 &=& \gamma^2 m_0^2 c^4 - \gamma^2 m_0^2 v^2 c^2 \\
E^2 - (pc)^2 &=& \gamma^2 m_0^2 c^4 \left( 1 - \frac{v^2}{c^2} \right) \\
E^2 - (pc)^2 &\overset{\text{已知 3}}{=}& \gamma^2 m_0^2 c^4 \cdot \gamma^{-2} \\
E^2 - (pc)^2 &=& m_0^2 c^4 \\
E^2 &=& (m_0 c^2)^2 + (pc)^2
\end{gather*}$$
