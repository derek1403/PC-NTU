## 3.1 德布羅依波 (De Broglie Wave)

德布羅意波(de Broglie wave)又稱物質波所有物質（電子、原子等）皆具有波動性，是波粒二象性的延伸。波長 $\lambda$ 與動量 $P$ 有關，公式為 $\lambda = \frac{h}{ p}$，其中 $h$ 為普朗克常數。

### 光子

一個光子如果頻率是ν的話，光動量為$p = h\nu/c = h/λ$，故其對應 $\lambda =  \frac{h}{ p}$

$$(h \nu)^2 = E^2 = (pc)^2 + (m_0 c^2)^2$$

### 粒子

粒子動量 $p = \gamma mv$

粒子德布羅依波長 $\lambda = \frac{h}{ p}= \frac{h}{ \gamma m_0v}$

* $\lambda$ :德布羅依波長 ， $[\text{m}]$
* $h$ :普朗克常數 $h \sim 6.63*10^-34 \text{m}$
* $p$ :物體動量 $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
* $\gamma$ : 
* $v$ :物體速度 $[\text{m}\cdot\text{s}^{-1}]$
* $m_0$ :物體靜止質量 $[\text{kg}]$

## 3.5 箱子中的粒子 Particle in a Box

為何一個被捕捉的粒子之能量會被量子化

* 當運動粒子被限制於某些區域而非自由移動時，其波性質會導致相當重要的結果。
* 從波的觀點來看，一個陷於箱中的粒子正如同在箱壁間拉伸繩索中的駐波一樣，在這兩個狀況中（繩索的橫向位移及運動物體的波函數 ），波函數在壁上必須為0，因為波會在壁上停止。
> 意思是牆壁上出現的機率為0?
* 被補陷粒子的德布羅依波長 
* 每個允許的能量被稱為能階(energy level)，而決定能階En的整數 n 被稱為量子數(quantum number)。


1. 被捕捉粒子的德布羅依波長
$$\lambda = \frac{2L}{n} \quad n = 1, 2, 3, \dots $$

2. 箱中的粒子（能量量子化）
$$E_n = \frac{n^2 h^2}{8mL^2} \quad n = 1, 2, 3, \dots $$
參數說明：

* $L$：位能井（箱子）的寬度。
* $n$：量子數（代表不同的能階）。
* $h$：普朗克常數。
* $m$：粒子的質量。

這組公式描述了當粒子被限制在一個空間內時，其能量不再是連續的，而是量子化的。


**例題**
這個箱子是原子，粒子是電子
一個在寬為0.10 nm箱中的電子，這個箱子大小和原子尺寸相當，求出允許的能量。


電子能量量子化計算
在此 $m = 9.1 \times 10^{-31} \text{ kg}$ 且 $L = 0.1 \text{ nm} = 1.0 \times 10^{-10} \text{ m}$，故允許電子能量為：
$$E_n = \frac{(n^2)(6.63 \times 10^{-34} \text{ J} \cdot \text{s})^2}{(8)(9.1 \times 10^{-31} \text{ kg})(1.0 \times 10^{-10} \text{ m})^2} = 6.0 \times 10^{-18} n^2 \text{ J}$$ 
$$E_n = 38n^2 \text{ eV}$$ 

電子的最小能量為 38 eV，對應的 $n$ 為 1。能階序列為：

* $E_2 = 152 \text{ eV}$
* $E_3 = 342 \text{ eV}$
* $E_4 = 608 \text{ eV}$

以此類推（圖 3.11）。如果這樣的箱子存在，捕陷電子能量的量子化將是系統的特徵之一（能量量子化也的確是原子電子中最顯著的特色之一）。



這裡展示了當空間限制在 0.1 奈米（大約是一個原子的尺度）時，電子的能量會出現非常明顯的跳躍。
如果把箱子寬度 $L$ 增加 10 倍，能階的能量會發生什麼變化嗎？


如果我們將箱子寬度 $L$ 增加 10 倍（從 $0.1\text{ nm}$ 變為 $1.0\text{ nm}$），電子的能量會急劇下降。
根據公式：
$$E_n = \frac{n^2 h^2}{8mL^2}$$
你可以發現能量 $E_n$ 與寬度的平方 ($L^2$) 成反比。

計算結果
當 $L$ 變為原來的 10 倍，能量就會縮減為原來的 1/100 ($10^2 = 100$)：

* 原本的最小能量 ($n=1$): $38\text{ eV}$
* 增加 10 倍後的最小能量 ($n=1$): $38 \times \frac{1}{100} = \mathbf{0.38\text{ eV}}$

物理意義：為什麼會這樣？

   1. 空間越寬敞，波動越平緩：
   當空間變大，電子的波長 $\lambda$ 可以變得更長，根據德布羅依公式，波長越長，動量和能量就越低。
   2. 量子化效應變模糊：
   隨著 $L$ 越來越大（例如變成公分或公尺等級），能階之間的差距（如 $0.38\text{ eV}$ 與 $1.52\text{ eV}$ 的差距）會小到我們幾乎感覺不到，這就是為什麼在日常生活的宏觀世界中，我們感覺不到能量是「一階一階」的原因。

簡單來說：箱子越大，電子就越「自由」，能量也就越低且越接近連續。
你想了解為什麼這個公式能解釋原子中電子的穩定性，還是想看看如果換成質量更重的質子，能量會如何變化？


電子在原子裡必須「剛好繞一圈（或半圈）能首尾相接」，否則波會產生干涉而互相抵銷。這個「剛好能住下來」的大小，剛好就是我們看到的原子尺度。


**例題**
一個10 g的彈珠位於10 cm寬的箱中，求出其允許能量。


宏觀物體（彈珠）的能量量子化計算
當 $m = 10 \text{ g} = 1.0 \times 10^{-2} \text{ kg}$ 且 $L = 10 \text{ cm} = 1.0 \times 10^{-1} \text{ m}$ 時：
$$E_n = \frac{(n^2)(6.63 \times 10^{-34} \text{ J} \cdot \text{s})^2}{(8)(1.0 \times 10^{-2} \text{ kg})(1.0 \times 10^{-1} \text{ m})^2} = 5.5 \times 10^{-64} n^2 \text{ J}$$


彈珠的極小能量為 $5.5 \times 10^{-64} \text{ J}$，對應的 $n$ 為 $1$。具有此動能的彈珠速度為 $3.3 \times 10^{-31} \text{ m/s}$，因此並無法由實驗分辨它和靜止彈珠的差異。
彈珠合理的速度可能為 $\frac{1}{3} \text{ m/s}$ ——對應能階的量子數為 $n = 10^{30}$！允許的能階非常接近，故沒有方法能決定彈珠是否位於式 (3.18) 所預測的能量或是別的能量。
因此在我們日常生活經驗中，量子效應是無法被察覺到的，這也解釋了牛頓力學 (Newtonian mechanics) 在日常生活中之所以成功的原因。

重點解析：
這段話非常經典地解釋了「對應原理」：

   1. 量子數極大：當物體很大（如 10 克的彈珠）時，即便速度很慢，量子數 $n$ 也高達 $10^{30}$。
   2. 能階連續化：相鄰能階（$n$ 與 $n+1$）之間的能量差極小，小到儀器完全無法分辨，看起來就像是「連續」的。
   3. 結論：在巨觀尺度下，量子力學的結果會趨近於牛頓力學。

## 3.6 測不準原理 (Uncertainty Principle)
測不準原理：

* 不可能同時知道一個物體的正確位置和正確動量。 
* 這個原理於1927年被在美國的海森堡(Heisenberg)和在德國的薛丁格發現，並且為物理定律的重要發現之一。 
* 在某一時間 t 時，波群函數 (x)可以傅利葉積分(Fourier integral)來表示

$$\Psi(x) = \int_{0}^{\infty} g(k) \cos kx \, dk$$
物理意義簡述：
這個公式描述了如何將許多不同波數（$k$）的單頻波（餘弦波）疊加起來，形成一個在空間中具有特定形狀的波包（Wave Packet）：

* $\Psi(x)$：代表在位置 $x$ 的波函數（位移）。
* $g(k)$：代表不同波數 $k$ 的振幅分布函數（或稱**權重**），它決定了每一種頻率的波對總和的貢獻程度。
* $\cos kx$：基本的簡諧波成分。

這實際上是傅立葉變換（Fourier Transform）的一種形式，用於說明量子力學中，一個具有空間定域性的粒子（如電子）是如何由一組連續分布的德布羅依波疊加而成的。

這個公式在量子力學中非常重要，因為它通常用來描述一個波包的最理想狀態：

 1. 德布羅依關係式 (De Broglie Relations)

* 德布羅依波長
$$\lambda = h/p$$
* 波數與動量的關係
$$k = 2\pi / \lambda = 2\pi p/h$$
$$p = hk/2\pi$$
* 動量不確定度與波數不確定度的關係
$$\Delta p = h \Delta k / 2\pi$$

 1. 測不準原理 (Uncertainty Principle)

* 空間與波數的不確定度關係
$$\Delta x \Delta k \ge \frac{1}{2} $$
* 測不準原理（位置與動量）
$$\Delta x \Delta p \ge \frac{h}{4\pi} $$

關鍵筆記：
公式 $\Delta x \Delta p \ge \frac{h}{4\pi} $ 是將 $\Delta p = \frac{h \Delta k}{2\pi}$ 代入 $\Delta x \Delta k \ge \frac{1}{2}$ 推導而來的。這代表你對一個粒子的位置 ($\Delta x$) 知道得越精確，你對它的動量 ($\Delta p$) 了解就越模糊。
如果你想更精簡地寫，物理學家通常定義約化普朗克常數 $\hbar = \frac{h}{2\pi}$，這樣測不準原理就可以寫成更常見的：
$$\Delta x \Delta p \ge \frac{\hbar}{2}$$

**例題**某次對質子的位置測量準確度為 ± 1.00 × 10-11 m。求出一秒後質子位置的測不準量。假設 v << c。
 質子位置與速度的不確定度推導
設時間 $t=0$ 時，質子位置的測不準量為 $\Delta x_0$。此時它的動量測不準量可從求出，
$$\Delta p \ge \frac{\hbar}{2\Delta x_0}$$
因為 $v \ll c$（非相對論情況），動量的測不準量為 $\Delta p = \Delta(mv) = m \Delta v$，所以質子速度的測不準量為：
$$\Delta v = \frac{\Delta p}{m} \ge \frac{\hbar}{2m\Delta x_0}$$
在時間 $t$ 內質子出現範圍的準確度不可能超過：
$$\Delta x = t \Delta v \ge \frac{\hbar t}{2m\Delta x_0}$$

因此 $\Delta x$ 會反比於 $\Delta x_0$：我們對於 $t = 0$ 的質子位置知道的越清楚，我們就會對於 $t > 0$ 的位置知道的越少。在 $t = 1.00 \text{ s}$ 時 $\Delta x$ 的值為：
$$\Delta x \ge \frac{(1.054 \times 10^{-34} \text{ J} \cdot \text{s})(1.00 \text{ s})}{(2)(1.672 \times 10^{-27} \text{ kg})(1.00 \times 10^{-11} \text{ m})}$$ 
$$\ge 3.15 \times 10^3 \text{ m}$$ 

答案是 $3.15 \text{ km}$ ——接近 2 哩！原本的波群竟然散布到這麼寬的範圍（圖 3.16）。發生的原因在於組合波的相速度隨波數而改變，而原本狹窄的波群需要很大範圍的波數才得以表達出來。

物理觀點：
這個範例非常震撼地展示了微觀世界的特性：如果你在初始時刻把質子關在一個極小的空間內（$10^{-11} \text{ m}$，約為原子大小的十分之一），僅僅過了 1 秒鐘，它在空間中可能出現的範圍就擴散到了 3 公里 這麼大！
這就是為什麼在量子力學中，我們無法像經典力學那樣預測一個粒子在長時間後的確切位置。



核心概念：波包的擴散 (Wave Packet Spreading)
這段公式推導出一個非常有趣的物理現象：隨時間增加，粒子的位置會變得越來越不確定。
即便你在初始時刻（$t=0$）非常精確地鎖定了質子的位置（$\Delta x_0$ 很小），這也會導致速度的不確定度 $\Delta v$ 變得很大。隨著時間流逝，這個速度的不確定性會累積，讓粒子在空間中「散開」得越來越嚴重。
這說明在量子力學中，微觀粒子並不像經典的小球那樣有固定的軌跡，而是像一團會隨時間膨脹的雲霧。
對應到一個移動包的波包由許多個別的波組成，如圖3.13所示。個別波的相速度隨波長而改變。結果是，當粒子移動時，波包在空間中散開。原始波包越窄——也就是我們對當時它的位置知道的越精準——它就會分散地越開，因為它是由大量不同相速度的波所組成

## 觀念補充：
這組公式展示了海森堡測不準原理的一個直觀物理模型：

   1. 觀測限制：如果你想用光子（波長為 $\lambda$）來測量粒子的位置，你所能達到的解析度極限大約就是波長本身（$\Delta x \ge \lambda$）。
   2. 動量干擾：光子與粒子碰撞後，會將其動量（$h/\lambda$）傳遞給粒子，造成粒子動量的不確定度（$\Delta p \approx h/\lambda$）。
   3. 結論：將兩者相乘，波長 $\lambda$ 會被約掉，剩下 $\Delta x \Delta p \ge h$。雖然這與正式的嚴格公式（$\hbar/2$）在常數項上有所不同，但它完美解釋了為什麼位置和動量無法同時被精確測量。

## 3.7 測不準原理應用 (Applying the Uncertainty Principle)

測不準原理一個有用的工具，而不只是一個負面的陳述。


能量與時間

* 能量與時間的測不準
$$\Delta \nu \sim 1/\Delta t$$
* 能量不確定度與頻率的關係
$$\Delta E \sim h \Delta \nu$$
* 推導結果
$$\Delta E \sim h/\Delta t \quad \text{or} \quad \Delta E \Delta t \sim h$$
* 比較嚴謹的推導會得到
$$\Delta E \Delta t \ge \frac{\hbar}{2} $$

物理意義簡述：
這個公式稱為「能量－時間測不準原理」。它告訴我們：

   1. 生存時間越短，能量越不確定：如果一個量子態（例如原子激發態）存在的時間非常短（$\Delta t$ 很小），那麼它的能量就會變得很模糊（$\Delta E$ 很大）。
   2. 能量守恆的短暫違背：在極短的時間內，能量可以發生劇烈的波動，這甚至允許「虛粒子」在真空中短暫出現後又消失。


**例題** 原子核中是否含有電子的推論
一個典型原子核的半徑約為 $5.0 \times 10^{-15} \text{ m}$。利用測不準原理求出電子為原子核的一部分時所需要的能量下限。
1. 動量不確定度計算
令 $\Delta x = 5.0 \times 10^{-15} \text{ m}$，則得到：
$$\Delta p \ge \frac{\hbar}{2\Delta x} \ge \frac{1.054 \times 10^{-34} \text{ J} \cdot \text{s}}{(2)(5.0 \times 10^{-15} \text{ m})} \ge 1.1 \times 10^{-20} \text{ kg} \cdot \text{m/s}$$
如果這是核電子動量的測不準量，則動量 $p$ 本身必須具有與其相當的強度。一個有這樣動量的電子動能 $KE$ 比它的靜止能量 $mc^2$ 大了好幾倍。從式 (1.24) 中我們看到可以假設 $KE = pc$，準確度不會差太多。
1. 動能下限計算
因此：
$$KE = pc \ge (1.1 \times 10^{-20} \text{ kg} \cdot \text{m/s})(3.0 \times 10^8 \text{ m/s}) \ge 3.3 \times 10^{-12} \text{ J}$$
因為 $1 \text{ eV} = 1.6 \times 10^{-19} \text{ J}$，在核中的電子動能必須超過 $20 \text{ MeV}$。
結論
實驗證明由某些不穩定核中射出的電子能量遠小於此值，從這裡我們可以得到結論就是原子核不可能含有電子。不穩定核所發射的電子，是因為那時原子核衰變所造成（見 11.3 和 12.5 節）。

這段推論非常重要，它解釋了為什麼 $\beta$ 衰變產生的電子並不是「原本就住在原子核裡」，而是衰變瞬間才產生的。
*  $\beta$ 衰變 具體是如何在沒有電子的原子核中「變」出電子來的?



**例題**  氫原子電子的極小能量估算
一個氫原子的半徑為 $5.3 \times 10^{-11} \text{ m}$。運用測不準原理估算在此原子中電子所需具備的極小能量。
解答過程
這裡我們發現 $\Delta x = 5.3 \times 10^{-11} \text{ m}$。

* 動量不確定度：
$$\Delta p \ge \frac{\hbar}{2\Delta x} \ge 9.9 \times 10^{-25} \text{ kg} \cdot \text{m/s}$$

具有這種大小等級動量的電子表現和古典的粒子相似，它的動能為：

* 動能計算：
$$KE = \frac{p^2}{2m} \ge \frac{(9.9 \times 10^{-25} \text{ kg} \cdot \text{m/s})^2}{(2)(9.1 \times 10^{-31} \text{ kg})} \ge 5.4 \times 10^{-19} \text{ J}$$
結論
計算結果約為 $3.4 \text{ eV}$。而實際上，在氫原子最低能階（基態）的電子動能為 $13.6 \text{ eV}$。

注意： 雖然測不準原理算出的只是「數量級」的估計，但它成功解釋了為什麼電子必須具備一定的動能，而不會靜止在原子核上。

**例題**  原子輻射頻率的測不準度
一個「激發的」原子將多餘的能量用於發射出某個特徵頻率的光子，如第 4 章所述。平均一個原子從激發至輻射所花的時間約為 $1.0 \times 10^{-8} \text{ s}$。求光子頻率中的天生的測不準。
解答過程

* 光子能量的測不準量為：
$$\Delta E \ge \frac{\hbar}{2\Delta t} \ge \frac{1.054 \times 10^{-34} \text{ J} \cdot \text{s}}{2(1.0 \times 10^{-8} \text{ s})} \ge 5.3 \times 10^{-27} \text{ J}$$
* 對應光頻率的測不準量為：
$$\Delta \nu \ge \frac{\Delta E}{h} \ge 8 \times 10^6 \text{ Hz}$$

物理意義
這就是用來決定原子所發射之輻射頻率的準確度一定不可減少的限制。如結果所示，從一群激發的原子不會以準確的頻率 $\nu$ 出現。假設一個光子的頻率為 $5.0 \times 10^{14} \text{ Hz}$，則相對不確定度為：
$$\Delta \nu / \nu = 1.6 \times 10^{-8}$$ 
實際上，其他現象如都卜勒效應對於譜線變寬（Spectral line broadening）的貢獻遠大於前者。
