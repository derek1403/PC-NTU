## 3.1 德布羅依波 (De Broglie Wave)

1924年，德布羅依 (Louis de Broglie) 提出了一個大膽的假說：既然光（電磁波）具有粒子性（光子），那麼所有具備質量的物質（如電子、原子等）也應該具有波動性。這稱為**物質波 (Matter wave)**，是波粒二象性的完美對稱。

物體的波長 $\lambda$ 與其動量 $p$ 之間的關係由以下公式給出：

$$\lambda \overset{\text{def}}{=} \frac{h}{p}$$


* $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
* $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
* $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$

### 光子與粒子的動量

**1. 光子 (Photon)**

* [物體的能量與動量關係式](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#energy-and-momentum) $E^2 = (m_0 c^2)^2 + (pc)^2 \quad (1.8)$
* 光子沒有靜止質量 ($m_0 = 0$)
* 若光子的頻率為 $\nu$，根據[愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis)，其能量為 $E = h\nu$
* $c=\nu \lambda$


$$\begin{gather*}
E^2 &\overset{\text{(1.8)}}{=}& (pc)^2 + (m_0 c^2)^2 \\
E^2&=& (pc)^2 \\
(h\nu)^2 &=& (pc)^2 \\
h\nu &=& pc\\
h\frac{c}{\lambda} &=& pc\\
\end{gather*}$$
 
$$\Rightarrow p = \frac{E}{c} = \frac{h\nu}{c} = \frac{h}{\lambda} \tag{3.1}$$

這與德布羅依公式 $\lambda = \frac{h}{p}$ 完全吻合。

**2. 粒子 (Particle)**

* [物體的能量與動量關係式](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#energy-and-momentum) $E^2 = (m_0 c^2)^2 + (pc)^2 \quad (1.8)$
* 有靜止質量 $m_0$ 的粒子
* 運動速度為 $v$，若 $v$ 接近光速 $c$ 則需考慮[相對論效應](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum) $p = \gamma m_0 v$
* [物體總能量 (Total Energy)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#mass-and-energy)：$E = \gamma m_0 c^2 $
* $1\text{eV} = 1.6\times 10^{-19} \text{J}$


$$\begin{gather*}
E^2 &\overset{\text{(1.8)}}{=}& (pc)^2 + (m_0 c^2)^2 \\
E^2 - (m_0 c^2)^2 &=& (pc)^2 \\
\frac{1}{c}\sqrt{E^2 - (m_0 c^2)^2} &=& p \\
\frac{1}{c}\sqrt{(KE+E_0)^2 - (m_0 c^2)^2} &=& p \\
\frac{1}{c}\sqrt{{KE}^2+2KE \cdot E_0} &=& p \\
\end{gather*}$$

$$\Rightarrow p =\sqrt{\gamma^2 -1} m_0 c  =  \frac{\sqrt{E^2 - (m_0 c^2)^2}}{c} = \frac{\sqrt{{KE}^2+2KE \cdot E_0}}{c} = \gamma m_0 v = \frac{h}{\lambda} \tag{3.2}$$


* $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)
* $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$
* $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
* $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$
* $E_0$ : 靜止能量 (Rest Energy) $E_0 = m_0 c^2$  $[\text{J}]$
* $KE$ : 動能 (Kinetic Energy)， $[\text{J}]$

### 例題：巨觀與微觀的物質波

#### **例題 3.1(a) 巨觀物體：高爾夫球**

**題目：** 求出質量 $46 \text{ g}$ 且速度為 $30 \text{ m/s}$ 的高爾夫球的德布羅依波長

* 高爾夫球速度遠小於光速 $\gamma \approx 1$
   
$$\begin{gather*}
\lambda &\overset{\text{(3.2)}}{=}&  \frac{h}{\gamma m_0 v} \\
&=& \frac{6.626 \times 10^{-34} \text{ J}\cdot\text{s}}{1\times 0.046 \text{ kg} \times 30 \text{ m/s}} \\
&=& 4.8 \times 10^{-34} \text{ m}
\end{gather*}$$


**物理解釋：** $10^{-34} \text{ m}$ 是一個小到無法想像的尺度（質子直徑約為 $10^{-15} \text{ m}$）。因為波長遠小於任何可測量或發生繞射的狹縫尺寸，故我們無法觀察到高爾夫球的波動行為。日常生活中感受不到物質波

> **思想實驗探討：** 若把高爾夫球切成一半 ($23 \text{ g}$)，以同速度丟出，波長會變為 $9.6 \times 10^{-34} \text{ m}$。若切的足夠小是不是就可以觀察到物質波了?


#### **例題 3.1(b) 微觀粒子：電子**

**題目：** 計算以速度 $10^7 \text{ m/s}$ 運動之電子的德布羅依波長，電子靜止質量 $m_e = 9.11 \times 10^{-31} \text{ kg}$

* $v = 10^7 \text{ m/s}\approx 0.033c$ 所以 $\gamma \approx 1$
  
$$\begin{gather*}
\lambda &\overset{\text{(3.2)}}{=}& \frac{h}{\gamma m_0 v} \\
&=& \frac{6.626 \times 10^{-34} \text{ J}\cdot\text{s}}{1\times 9.11 \times 10^{-31} \text{ kg} \times 10^7 \text{ m/s}} \\
&=& 7.3 \times 10^{-11} \text{ m}
\end{gather*}$$

**物理解釋：** $7.3 \times 10^{-11} \text{ m}$ 大約是 $0.073 \text{ nm}$，相當於氫原子的波耳半徑約為 $0.053 \text{ nm}$。當電子的波長與原子尺寸相當甚至更大時，電子的波動特性就會與原子結構產生強烈的交互作用（例如干涉、繞射），這就是為什麼古典物理無法解釋原子內部運作，必須引入量子力學。


#### **例題 3.1(c) 質子的相對論性能量與動量**

**題目：** 德布羅依波長為 $1 \text{ fm}$ ($1 \times 10^{-15} \text{ m}$) 的質子，此波長大約為質子的直徑，請計算其動能，質子靜止質量 $E_0 = m_p c^2 \approx 938 \text{ MeV}$

$$\begin{gather*}
\frac{\sqrt{{KE}^2+2KE \cdot E_0}}{c}  &\overset{\text{(3.2)}}{=}& \frac{h}{\lambda} \\
{KE}^2+2KE \cdot E_0 &=& \left(\frac{hc}{\lambda}\right)^2 \\
{KE}^2+2KE \cdot E_0 - \left(\frac{hc}{\lambda}\right)^2 &=& 0 \\
KE  &=& \frac{-2E_0 \pm \sqrt{(2 E_0)^2+4\left(\frac{hc}{\lambda}\right)^2}}{2}\\
KE  &=& -E_0 \pm \sqrt{(E_0)^2+\left(\frac{hc}{\lambda}\right)^2}\\
KE  &=& -938 \text{ MeV} \pm \sqrt{(938 \text{ MeV})^2+\left(\frac{ 6.6 \times 10^{-34} \text{ J}\cdot\text{s} \cdot 3 \times 10^8 \text{ m}\cdot\text{ s}^{-1}}{10^{-15} \text{ m}}\right)^2}\\
KE  &=& -938 \text{ MeV} \pm \sqrt{(938 \text{ MeV})^2+\left( 19.8 \times 10^{-11} \text{ J} \right)^2}\\
KE  &=& -938 \text{ MeV} \pm \sqrt{(938 \text{ MeV})^2+\left( \frac{19.8 \times 10^{-11} \text{ J}}{1.6\times 10^{-19} \text{ J} \cdot \text{ eV}^{-1}} \right)^2}\\
KE  &=& -938 \text{ MeV} \pm \sqrt{(938 \text{ MeV})^2+\left( 1240 \text{ MeV} \right)^2}\\
KE  &=& -938 \text{ MeV} \pm \sqrt{ 2417444 }\text{ MeV}\\
KE  &=& -938 \text{ MeV} \pm 1555 \text{ MeV} \\
KE  &\overset{KE>0}{=}& -938 \text{ MeV} + 1555 \text{ MeV} \\
KE  &=& 617 \text{ MeV} \\
\end{gather*}$$



**物理解釋：** 當我們想要探測原子核內部的結構（尺度約 $1 \text{ fm}$）時，探測粒子（如質子或電子）的波長必須小於或等於這個尺度才能發生有效的交互作用 (interaction)。由計算可知，這需要高達數百 $\text{MeV}$ 的動能，這也是為什麼研究核物理需要「高能粒子加速器」。

### 什麼波？機率波 (Probability Wave)

在古典物理中，波都有具體的物理量在震盪（水波是高度，聲波是壓力，光波是電磁場）。那物質波是什麼在震盪？
物理學家 Max Born 提出了**機率幅 (Probability Amplitude)** 的解釋：

* **波函數 (Wave function) $\Psi(x,y,z,t)$**：代表物質波的振幅（本身為複數，無直接物理意義）。
* **機率密度 (Probability density) $|\Psi(x,y,z,t)|^2$**：代表在時間 $t$、空間座標 $(x,y,z)$ 處找到該粒子的**機率**。

> "Life can be understood backwards, but must be lived forwards." —— 呼應了量子力學中，對於尚未測量的未來，粒子存在於各種狀態的機率疊加中；一旦測量（時間過去），波函數塌縮，結果才成為確定的事實。

## 3.2 描述波 (Describing a Wave)

### 數學定義

* 角頻率 $\omega \overset{\text{def}}{=} 2\pi \nu$
* 波數 $k \overset{\text{def}}{=} \frac{2\pi}{\lambda} = \frac{\omega}{v_p}$
* 一個向 $+x$ 方向傳播的簡諧波一般表達式為：
  
    $$\begin{gather*}
    y(x,t) &=& A\cos\left(2\pi \nu \left(t - \frac{x}{v_p}\right)\right)\\
    y(x,t) &=& A\cos(\omega t - kx)
    \end{gather*}$$





### 德布羅依相速度 (Phase Velocity)

* **【定義 1】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：**
  
  $$\lambda \overset{\text{def}}{=} \frac{h}{p}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  
* **【定義 2】 [愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis)  ：**
  
  $$E \overset{\text{def}}{=} h\nu$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $\nu$ : 光子的頻率  $[\text{s}^{-1}]$

* **【已知 1】 [總能量 (Total Energy)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#mass-and-energy)  ：**
  
  $$E = \gamma m_0 c^2$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$

* **【已知 2】 [相對論動量 (Relativistic Momentum)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)  ：**
  
  $$p = \gamma m_0 v $$

  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$

#### 證明

單一頻率波浪的「波峰」前進的速度稱為相速度 $v_p$

$$\begin{gather*}
v_p &=& \nu \lambda \\
&\overset{\text{定義 1,2}}{=}& \left(\frac{E}{h}\right) \left(\frac{h}{p}\right) \\
&=& \frac{E}{p} \\
&\overset{\text{已知 1,2}}{=}& \frac{\gamma m_0 c^2}{\gamma m_0 v} \\
&=& \frac{c^2}{v}
\end{gather*}$$

* **相速度超越光速 ($v_p > c$) 違反相對論嗎？**
  * 相對論說「攜帶能量或資訊的物體速度不能超過 $c$」。
  * **相速度 $v_p$ 只是數學上「純粹正弦波波峰」移動的速率。** 一個從 $-\infty$ 到 $+\infty$ 完美重複的正弦波，是處處存在的，它沒有開頭也沒有結尾，因此**無法用來傳遞任何資訊**

## 3.3 相速度與群速度 (Phase and Group Velocities)

真正的粒子（如電子）存在於局部的空間中，而不是像純粹正弦波那樣延伸到無限遠。因此，粒子必須被描述為一個**波包 (Wave packet) 或波群 (Wave group)**，這是由許多不同頻率、不同波數的波疊加而成的。群速度的本質，就是當兩個（或多個）頻率相近的波疊加時，它們產生的「干涉包絡線」的移動速度

### 節拍 (Beat) 與波群速度推導

* **【假設 1】 假設有兩個振幅相同、但頻率和波數略有差異的波 ：**
  
  $$\begin{gather*}
  y_1 &=& A\cos(\omega t - kx)\\
  y_2 &=& A\cos[(\omega + \Delta \omega)t - (k + \Delta k)x]
  \end{gather*}$$

* **【已知 1】 和差化積 ：**
  
  $$\cos \alpha + \cos \beta = 2 \cos \left(\frac{\alpha + \beta}{2}\right) \cos \left(\frac{\alpha - \beta}{2}\right)$$

  * $y_1$ 的括號內視為 $\beta$
  * $y_2$ 的括號內視為 $\alpha$
  * $\alpha + \beta = (2\omega + \Delta \omega)t - (2k + \Delta k)x$
  * $\alpha - \beta = \Delta \omega t - \Delta k x$

* **【已知 2】 近似 ：**
  
  $$2\omega + \Delta \omega \approx 2\omega$$

  * $\Delta \omega \ll \omega$
  * $\Delta k \ll k$



**兩個波疊加的表現：**

$$\begin{gather*}
y &=& y_1 + y_2 \\
&\overset{\text{已知 1}}{=}& 2A \cos \left( \frac{(2\omega + \Delta \omega)t - (2k + \Delta k)x}{2} \right) \cos \left( \frac{\Delta \omega t - \Delta k x}{2} \right)\\
&\overset{\text{已知 2}}{\approx}& \underbrace{2A \cos \left( \frac{\Delta \omega}{2} t - \frac{\Delta k}{2} x \right)}_{\text{振幅調變（包絡線）}} \cdot \underbrace{\cos(\omega t - kx)}_{\text{載波（高頻震盪）}}
\end{gather*}$$


**物理意義：**
1.  後面的 $\cos(\omega t - kx)$ 是原本的高頻波，它的移動速度是**相速度** $v_p = \frac{\omega}{k}$。
2.  前面的 $2A \cos \left( \frac{\Delta \omega}{2} t - \frac{\Delta k}{2} x \right)$ 形成了一個巨大的「封包」（包絡線），這個封包就是粒子的實體，這個封包移動的速度就是 **群速度** ，從括號內的相位可知，群速度為：

$$\begin{gather*}
v_g &=& \frac{\Delta \omega / 2}{\Delta k / 2} &=& \frac{\Delta \omega}{\Delta k}\\
v_g &\overset{\text{def}}{=}& \frac{d\omega}{dk}
\end{gather*}$$


### 證明：物質波的群速度等於粒子真實速度 ($v_g = v$)

* **【定義 1】 數學上波的關鍵變數 ：**
  
  * 角頻率 $\omega \overset{\text{def}}{=} 2\pi \nu$
  * 波數 $k \overset{\text{def}}{=} \frac{2\pi}{\lambda} = \frac{\omega}{v_p}$

* **【定義 2】 群波 ：**
  
  $$v_g \overset{\text{def}}{=} \frac{d\omega}{dk}$$


* **【已知 1】 [愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis)  ：**
  
  $$\begin{gather*}
  E &\overset{\text{def}}{=}& h\nu \\
  E &\overset{\text{定義 1}}{=}& h\frac{\omega}{2\pi} \\
  \frac{ 2\pi E}{h} &=& \omega \\
  \end{gather*}$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $\nu$ : 光子的頻率  $[\text{s}^{-1}]$

* **【已知 2】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：**
  
  $$\begin{gather*}
  \lambda &\overset{\text{def}}{=}& \frac{h}{p}\\
  frac{2\pi}{k} &\overset{\text{定義 1}}{=}& \frac{h}{p}\\
  k &=& \frac{2\pi p}{h}\\
  \end{gather*}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  


* **【已知 3】 [總能量 (Total Energy)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#mass-and-energy)  ：**
  
  $$E = \gamma m_0 c^2$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$

* **【已知 4】 [相對論動量 (Relativistic Momentum)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)  ：**
  
  $$p = \gamma m_0 v $$

  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$




**proof**

$$\begin{gather*}
v_g &\overset{\text{定義 2}}{=}& \frac{d\omega}{dk}\\
&=& \frac{\quad \frac{d\omega}{dv}\quad}{\frac{dk}{dv} }\\
&\overset{\text{已知 1,2}}{=}&\frac{\quad\frac{d}{dv}\left[\frac{E}{\hbar} \right]\quad}{\frac{d}{dv}\left[\frac{p}{\hbar} \right] }\\
&=&\frac{\quad\frac{d}{dv}\left[E \right]\quad}{\frac{d}{dv}\left[p \right] }\\
&\overset{\text{已知 3,4}}{=}&\frac{\quad\frac{d}{dv}\left[\gamma m_0 c^2 \right]\quad}{\frac{d}{dv}\left[\gamma m_0 v \right] }\\
&=&\frac{\quad\frac{d}{dv}\left[\frac{ c^2}{\sqrt{1 - \frac{v^2}{c^2}}} \right]\quad}{\frac{d}{dv}\left[\frac{ v}{\sqrt{1 - \frac{v^2}{c^2}}} \right] }\\
&=&\frac{\quad  c^2 \left(-\frac{1}{2}\right)\left(1-\frac{v^2}{c^2}\right)^{-3/2} \left(-\frac{2v}{c^2}\right) \quad}{\quad \left[ \left(1-\frac{v^2}{c^2}\right)^{-1/2} + v\left(-\frac{1}{2}\right)\left(1-\frac{v^2}{c^2}\right)^{-3/2}\left(-\frac{2v}{c^2}\right) \right] \quad}\\
&=&\frac{\quad \frac{ v}{(1-v^2/c^2)^{3/2}} \quad}{\frac{1}{(1-v^2/c^2)^{3/2}}}\\
&=&v\\
\end{gather*}$$

**結論：**
* 波包整體的移動速度（群速度）完美對應了粒子在巨觀世界中被測量到的運動速度 $v$
* 且由先前可知 $v_p = c^2/v$，故有優美的關係式：$v_p \cdot v_g = c^2$



---

**例題 3.3 電子的相速度與群速度**
**題目：** 一個電子的德布羅依波長為 $2 \text{ pm} = 2.00 \times 10^{-12} \text{ m}$，求出其動能和德布羅依波的相速度與群速度。

> **老師的解說：解答你對 $(pc)^2$ 的疑惑**
> 你問：一直把 $(pc)^2$ 和 $KE$ 搞混。
> 請記住核心公式：**$E^2 = E_0^2 + (pc)^2$**。這是一個直角三角形的畢氏定理關係！
> 斜邊是總能量 $E$，一股是靜止能量 $E_0 = m_0 c^2$，另一股是「動量對應的能量」$pc$。
> 而動能 $KE$ 的定義是「總能扣掉靜止能」，也就是 $KE = E - E_0$（也就是斜邊減掉其中一股）。
> $(pc)^2$ 不是動能！在非相對論極限下，$KE \approx p^2 / 2m$，但在高能下這不成立。正確的流程永遠是：算出 $pc$ $\rightarrow$ 算出總能 $E$ $\rightarrow$ 算出動能 $KE = E - E_0$。

**計算步驟：**
**(a) 第一個步驟為計算電子的 $pc$**
$$pc = \frac{hc}{\lambda} = \frac{1240 \text{ eV}\cdot\text{nm}}{0.002 \text{ nm}} = 6.20 \times 10^5 \text{ eV} = 620 \text{ keV}$$
電子的靜止能量為 $E_0 = m_e c^2 \approx 511 \text{ keV}$。
利用能量三角形求總能量 $E$：
$$E = \sqrt{E_0^2 + (pc)^2} = \sqrt{(511 \text{ keV})^2 + (620 \text{ keV})^2} = 803 \text{ keV}$$
動能 $KE$ 為總能量減去靜止能量：
$$KE = E - E_0 = 803 \text{ keV} - 511 \text{ keV} = 292 \text{ keV}$$

**(b) 計算速度**
由 $E = \gamma E_0 = \frac{E_0}{\sqrt{1 - v^2/c^2}}$ 移項可得：
$$v = c\sqrt{1 - \left(\frac{E_0}{E}\right)^2} = c\sqrt{1 - \left(\frac{511}{803}\right)^2} \approx 0.771c$$
* **群速度 (粒子實際速度)：** $v_g = v = 0.771c$
* **相速度 (波峰虛擬速度)：** $v_p = \frac{c^2}{v} = \frac{c^2}{0.771c} \approx 1.30c$

## 3.5 箱子中的粒子 (Particle in a Box)

當粒子被限制在特定空間中，其波動性會導致「能量量子化」(Energy Quantization)。

**邊界條件的物理意義：**
如果一個粒子被困在一個寬度為 $L$ 的堅硬箱子裡，由於粒子無法穿透箱壁，因此在牆壁上以及箱子外找到粒子的機率必須為零。這意味著波函數在牆壁處必須為零（即 $\Psi(0) = \Psi(L) = 0$）。
這就像是在兩端固定的吉他弦上產生**駐波 (Standing wave)**，能穩定存在的波，其半波長的整數倍必須剛好等於箱子寬度：
$$L = n \frac{\lambda}{2} \implies \lambda_n = \frac{2L}{n} \quad (n = 1, 2, 3, \dots)$$
將德布羅依關係式 $\lambda = h/p$ 轉換為動量 $p = h/\lambda = \frac{nh}{2L}$。
因為箱內沒有位能，粒子的總能即為動能（非相對論情況 $E = p^2/2m$）：
$$E_n = \frac{p^2}{2m} = \frac{(nh/2L)^2}{2m} = \frac{n^2 h^2}{8mL^2} \quad (n = 1, 2, 3, \dots)$$

**公式變數解釋：**
* $E_n$：第 $n$ 個能階的能量 $[\text{J}]$ 或 $[\text{eV}]$
* $n$：主量子數 $[\text{dimensionless}]$，必須為正整數。
* $m$：粒子質量 $[\text{kg}]$
* $L$：箱子寬度（一維位能井長度） $[\text{m}]$

**例題 3.5 宏觀與微觀的量子效應差異**
*(結合你筆記中的兩個例子)*
* **微觀：$0.1 \text{ nm}$ 箱中的電子** (原子尺度)
    代入質量 $m_e = 9.1 \times 10^{-31} \text{ kg}$，寬度 $L = 1.0 \times 10^{-10} \text{ m}$，計算得基態 ($n=1$) 能量約為 $38 \text{ eV}$。能階間距很大 ($152 \text{ eV}, 342 \text{ eV}\dots$)，能量呈明顯不連續階梯狀。
* **巨觀：$10 \text{ cm}$ 箱中的 $10 \text{ g}$ 彈珠** (日常尺度)
    代入質量 $10^{-2} \text{ kg}$，寬度 $0.1 \text{ m}$，計算得基態能量約為 $5.5 \times 10^{-64} \text{ J}$。如果彈珠以極慢的 $1/3 \text{ m/s}$ 滾動，對應的量子數 $n \sim 10^{30}$！相鄰能階的能量差小到任何儀器都測不出來，看起來就像是連續的。
**物理知識點回顧：** 這完美展現了玻耳的**對應原理 (Correspondence Principle)**：在巨觀極限下（量子數 $n$ 極大，或尺寸 $L$ 極大，或質量 $m$ 極大），量子力學的預測會平滑地過渡到古典牛頓力學的結果。

## 3.6 測不準原理 (Uncertainty Principle)

1927年，海森堡提出測不準原理：**不可能同時絕對精準地知道一個粒子的位置與動量。** 這不是測量儀器不夠好的問題，而是物質波本質上的數學限制。

### 數學推導：傅立葉轉換與高斯波包

一個局部的波包 $\Psi(x)$ 可以看作是無限多個不同波數 $k$ 的完美正弦波的疊加，這正是傅立葉轉換：
$$\Psi(x) = \int_{-\infty}^{\infty} g(k) e^{ikx} \, dk$$
* $g(k)$ 代表在波數 $k$ 上的權重分佈（頻譜）。

> **老師的解說：證明高斯函數的傅立葉轉換仍是高斯函數**
> 你在筆記中問到「須證明」。這是一個非常優美的數學性質。
> 假設空間中的波包是一個標準差為 $\sigma_x$ 的高斯函數：
> $$\Psi(x) = N \exp\left(-\frac{x^2}{2\sigma_x^2}\right)$$
> 其傅立葉轉換 $g(k)$ 為：
> $$g(k) \propto \int_{-\infty}^{\infty} \exp\left(-\frac{x^2}{2\sigma_x^2}\right) e^{-ikx} \, dx = \int_{-\infty}^{\infty} \exp\left(-\frac{x^2}{2\sigma_x^2} - ikx\right) \, dx$$
> 我們在指數部位「配方 (Completing the square)」：
> $$-\frac{1}{2\sigma_x^2} \left( x^2 + 2i\sigma_x^2 kx \right) = -\frac{1}{2\sigma_x^2} \left[ (x + i\sigma_x^2 k)^2 - (i\sigma_x^2 k)^2 \right] = -\frac{(x + i\sigma_x^2 k)^2}{2\sigma_x^2} - \frac{\sigma_x^2 k^2}{2}$$
> 積分中只與 $x$ 有關的部分會積出一個常數，剩下與 $k$ 有關的部分會被提出來：
> $$g(k) \propto \exp\left(-\frac{\sigma_x^2 k^2}{2}\right)$$
> 這依然是一個高斯函數！若我們將其標準差定義為 $\sigma_k$，則 $\sigma_k^2 = \frac{1}{\sigma_x^2}$，也就是 $\sigma_x \sigma_k = 1$。
> 結合動量 $p = \hbar k$，所以 $\Delta p = \hbar \Delta k$。因此：
> $$\Delta x \Delta p = \hbar (\sigma_x \sigma_k) = \hbar$$
> （註：嚴格定義下方差為 $\Delta x = \sigma_x / \sqrt{2}$ 時，極限值即為 $\hbar / 2$）。

**測不準關係式：**
$$\Delta x \Delta p \ge \frac{\hbar}{2}$$
其中 $\hbar = \frac{h}{2\pi} \approx 1.054 \times 10^{-34} \text{ J}\cdot\text{s}$，稱為約化普朗克常數。

### 波包擴散 (Wave Packet Spreading) 與色散關係

**例題 3.6 質子位置的擴散**
**題目：** 若時間 $t=0$ 時對質子位置測量準確度為 $\Delta x_0 = 1.00 \times 10^{-11} \text{ m}$，求一秒後位置的不確定度。（假設非相對論 $v \ll c$）。
**推導與計算：**
初始動量不確定度：
$$\Delta p \ge \frac{\hbar}{2\Delta x_0}$$
因為 $\Delta p = m \Delta v$，速度不確定度為：
$$\Delta v = \frac{\Delta p}{m} \ge \frac{\hbar}{2m\Delta x_0}$$
經過時間 $t$ 後，由於粒子實際速度不明確，位置的散佈範圍會增加：
$$\Delta x = t \Delta v \ge \frac{\hbar t}{2m\Delta x_0}$$
代入質子質量 $m \approx 1.67 \times 10^{-27} \text{ kg}$， $t=1 \text{ s}$：
$$\Delta x \ge \frac{(1.054 \times 10^{-34})(1)}{(2)(1.67 \times 10^{-27})(10^{-11})} \approx 3.15 \times 10^3 \text{ m} = 3.15 \text{ km}$$

> **老師的解說：連結色散關係 (Dispersion Relation)**
> 你問到了為什麼會這樣。因為物質波的角頻率 $\omega = \frac{\hbar k^2}{2m}$（由 $E = p^2/2m$ 推得）。
> 這代表相速度 $v_p = \omega / k = \frac{\hbar k}{2m}$ 是**波數 $k$ 的函數**。
> 也就是說，組成波包的各種不同波長的分波，它們「跑的速度不一樣」！高頻波跑得快，低頻波跑得慢。隨著時間過去，原本疊加在一起的波包就會像一群速度不一的賽跑選手一樣，越拉越開。這就是所謂的「色散 (Dispersion)」，導致了巨觀時間後位置的極度不確定。

## 3.7 測不準原理應用 (Applying the Uncertainty Principle)

測不準原理不僅適用於位置與動量，也適用於能量與時間：
$$\Delta E \Delta t \ge \frac{\hbar}{2}$$

**例題 3.7(a) 原子核內沒有電子**
**題目：** 典型原子核半徑 $R \approx 5.0 \times 10^{-15} \text{ m}$。若電子被束縛在核內，求其所需動能下限。
**計算：**
將不確定度視為直徑 $\Delta x = 2R = 10 \times 10^{-15} \text{ m}$。（或直接取數量級 $5 \times 10^{-15}$）
$$\Delta p \ge \frac{\hbar}{2\Delta x} \approx 1.1 \times 10^{-20} \text{ kg}\cdot\text{m/s}$$
因為這動量極大，必定為極端相對論情況，動能 $K \approx E \approx pc$：
$$K \approx pc = (1.1 \times 10^{-20} \text{ kg}\cdot\text{m/s})(3 \times 10^8 \text{ m/s}) = 3.3 \times 10^{-12} \text{ J} \approx 20 \text{ MeV}$$
**物理知識點回顧與解答：** 計算結果顯示，如果電子要穩定待在原子核內，它必須具備至少 $20 \text{ MeV}$ 的動能。但實驗觀測到 $\beta$ 衰變射出的電子能量只有幾 $\text{MeV}$，這反證了「電子平時並不存在於原子核內」。
* **你問：$\beta$ 衰變具體是如何在沒有電子的原子核中「變」出電子來的？**
    這涉及**弱交互作用 (Weak Interaction)**。在中子衰變成質子的瞬間，能量和電荷發生轉換：$n \rightarrow p^+ + e^- + \bar{\nu}_e$（反電子微中子）。
    **物理類比：** 就像你說話時發出的聲音。聲音並不是原本就「儲存」在你的肺部或喉嚨裡，而是氣流經過聲帶振動時，能量在瞬間「轉換」成了聲波。電子也是在核衰變的瞬間，由能量轉換而生成的。

**例題 3.7(b) 原子輻射頻率的測不準度（譜線變寬）**
**題目：** 原子從激發態躍遷發光，平均壽命 $\Delta t \approx 1.0 \times 10^{-8} \text{ s}$。求光子頻率的測不準量。
**計算：**
$$\Delta E \ge \frac{\hbar}{2\Delta t} = \frac{1.054 \times 10^{-34}}{2 \times 10^{-8}} \approx 5.3 \times 10^{-27} \text{ J}$$
由 $E = h\nu$，得 $\Delta E = h \Delta \nu$：
$$\Delta \nu \ge \frac{\Delta E}{h} \approx 8 \times 10^6 \text{ Hz}$$
**物理知識點回顧：** 即使是同種原子的同一個躍遷，發出的光也不會是絕對單一的頻率（完美的雷射是不存在的）。這被稱為**自然譜線變寬 (Natural line broadening)**，這是大自然給定的物理極限。
