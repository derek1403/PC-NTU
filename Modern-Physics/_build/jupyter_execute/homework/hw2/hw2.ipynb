## Question 1 : Density of standing waves in cavity

Consider the radiation inside a cavity of absolute temperature $T$ whose walls are perfect reflectors to be a series of standing em waves(電磁波). This is a three-dimensional generalization of standing waves in a stretched string. The condition for standing waves in such a cavity is that the path length from wall to wall, whatever the direction, must be a whole number of half-wavelengths, so that a node occurs at each reflecting surface. Proof that the number of independent standing waves $G(\nu)d\nu$ in the frequency interval between $\nu$ and $\nu+d\nu$ per unit volume in the cavity is
   

$$G(\nu)d\nu = \frac{8\pi\nu^2}{c^3} d\nu $$



## Question 1 - Answer

邏輯主軸:
邊界條件限制了波長 $\to$ 波長離散化形成網格點 $\to$ 計算球體體積求總數 $\to$ 轉成頻率求微分 $\to$ 乘以2（偏振）得到密度。


$$G(\nu)d\nu = \frac{8\pi\nu^2}{c^3} d\nu$$

* $\nu$ (頻率) $[\text{Hz}] = [\text{s}^{-1}]$
* $d\nu$ (頻率區間)：一個極小的頻率範圍（從 $\nu$ 到 $\nu + d\nu$）$[\text{Hz}] = [\text{s}^{-1}]$
* $c$ (光速)：真空中光速，$2.99 \times 10^8  [\text{m} \cdot\text{s}^{-1}]$
* $G(\nu)$ (駐波密度 / 狀態密度，Density of States)：在單位體積內、每單位頻率區間中，可以容納多少個駐波模式
* $G(\nu)d\nu$ (總駐波數密度)：代表在頻率介於 $\nu$ 與 $\nu + d\nu$ 之間，單位體積內共有多少個駐波模式 $[\text{m}^{-3}]$

---



### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 考慮3D正方體空腔情況：** 

  * 三維正方體:一個邊長為 $L$、體積為 $V = L^3$
  * 模式方向有3個 $x$、$y$、$z$
  * $k_x = n_x \frac{\pi}{L}, \quad k_y = n_y \frac{\pi}{L}, \quad k_z = n_z \frac{\pi}{L}$
  * $n_x, n_y, n_z$ 都是正整數 ($1, 2, 3, \dots$) ($n_x, n_y, n_z \in \mathbb{N}$)

* **【已知 2】 波向量的總長度平方 $k^2$ 為：** 

  $$k^2 = k_x^2 + k_y^2 + k_z^2 = \left(\frac{\pi}{L}\right)^2 (n_x^2 + n_y^2 + n_z^2)$$

* **【定義 1】 模式空間中的半徑 $R$ ：**
  
  $$R^2 \overset{\text{let}}{=} n_x^2 + n_y^2 + n_z^2$$

  * 「$n$ 空間」的三維網格系統裡。每一個座標點 $(n_x, n_y, n_z)$ 都代表一個合法的「駐波模式」， **如何計算特定半徑範圍內，總共有多少個點？**

* **【已知 3】 波向量 $k$ 轉換透過真空中電磁波的角頻率 $\omega$ ：**
  
  $$k = \frac{2\pi\nu}{c}$$

  * $\omega = c k$
  * $\omega = 2\pi\nu$


* **【已知 4】 大模式空間的半徑$R$內的整數點近似體積  ：**
  
  $$\text{半徑R內的整數點} = \frac{4}{3}\pi R^3$$
  
  * $R = \frac{2L\nu}{c} \sim 3.33 \cdot 10^5$
    * $L = 0.1 \text{ m}$
    * $\nu = 5 \times 10^{14} \text{ Hz}$
    * $c = 3 \times 10^8 \text{ m}\cdot \text{ s}^{-1} $
  * $\frac{\text{表面誤差的方塊數}}{\text{總體積的方塊數}} \propto \frac{R^2}{R^3} = \frac{1}{R} \approx \frac{1}{333,333} \approx 0.0003\%$

* **【定義 2】 模式空間的半徑在 $R$ 以內的總模式數 $N(R)$：**

  $$\begin{gather*}
  N(R) &=& \text{半徑R內的合法整數點} \\
  &=& \left(\text{合法整數點比例}\right) \times \left(\text{半徑R內的整數點}\right) \\
  &\overset{\text{已知 4}}{\sim}&\left(\text{合法整數點比例}\right) \times \frac{4}{3}\pi R^3 \\
  &\overset{\text{已知 1}}{=}& \frac{1}{8} \times \frac{4}{3}\pi R^3 \\
  &=& \frac{\pi}{6} R^3
  \end{gather*}$$

* **【已知 5】 模式空間的半徑 $R$ 對頻率轉換  ：**
  
  $$\begin{gather*}
  R &=& \sqrt{\left(\frac{L}{\pi}\right)^2 k^2}\\
  &=&\frac{L}{\pi}  k \\
  &=& \frac{L}{\pi} \left( \frac{2\pi\nu}{c} \right) \\
  &=& \frac{2L\nu}{c}\\
  \end{gather*}$$
  
  * 【定義 1】$R^2 \overset{\text{let}}{=} n_x^2 + n_y^2 + n_z^2$
  * 【已知 2】$k^2 = \left(\frac{\pi}{L}\right)^2 (n_x^2 + n_y^2 + n_z^2)$
  * 【已知 3】$k = \frac{2\pi\nu}{c}$

*  **【已知 6】 電磁波的極化 ：**
  
    $$dN_{\text{total}} = 2 dN$$

    * 於電磁波而言，給定一個行進方向（一個波向量 $\mathbf{k}$），它可以有**兩種互相垂直的獨立偏振方向**（例如 $x$ 偏振與 $y$ 偏振的橫波）。因此，實際的駐波數量必須乘以 2

*  **【定義 3】 單位體積內的駐波密度 $G(\nu)$ ：**
  
  $$G(\nu)d\nu = \frac{dN_{\text{total}}}{V}$$



### 證明

$$\begin{gather*}
N(\nu) &\overset{\text{定義 2}}{=}& \frac{\pi}{6} R^3 \\
N(\nu)&\overset{\text{已知 5}}{=}& \frac{\pi}{6} \left( \frac{2L\nu}{c} \right)^3 \\
N(\nu)&=& \frac{\pi}{6}  \frac{8 L^3 \nu^3}{c^3} \\
N(\nu)&=& \frac{4\pi L^3}{3 c^3} \nu^3 \\
N(\nu)&\overset{\text{已知 1}}{=}& \frac{4\pi V}{3 c^3} \nu^3 \\
\frac{dN}{d\nu} d\nu&=& \frac{d}{d\nu} \left[ \frac{4\pi V}{3 c^3} \nu^3 \right] d\nu \\
dN&=& \frac{4\pi V}{c^3} \nu^2 d\nu\\
dN_{\text{total}} &\overset{\text{已知 6}}{=}&2\frac{4\pi V}{c^3} \nu^2 d\nu \\
\frac{dN_{\text{total}}}{V} &=&\frac{8\pi }{c^3} \nu^2 d\nu \\
G(\nu) d\nu &\overset{\text{定義 3}}{=}&\frac{8\pi }{c^3} \nu^2 d\nu \\
\end{gather*}$$


## Question 2 Average energy per standing wave

Suppose the oscillators in the cavity walls do not have a continuous distribution of possible energies but have only the specific energies

$$\epsilon_{n}=nh\nu$$

Proof that the average energy per oscillator in the cavity walls—and so per standing wave—is

$$\bar{\epsilon} = \frac{h\nu}{e^{h\nu/kT}-1}$$

(Hint: Maxwell-Boltzmann distribution function $f_{MB} (\epsilon)=Ae^{-\epsilon/kT} $ )



## Question 2 - Answer


### 已知與假設



* 【已知 1】無窮等比級數求和:

   $$\sum_{n=0}^{\infty} x^n = 1 + x + x^2 + \dots = \frac{1}{1-x}$$

   * $|x| < 1$ ，本題中 $x = e^{-h\nu/kT}$ ，符合條件



* 【已知 2】等式:

   

   $$\begin{gather*}
   \sum_{n=0}^{\infty} x^n  &\overset{\text{已知 1}}{=}&  \frac{1}{1-x}  \\
   x \frac{d}{dx} \left[ \sum_{n=0}^{\infty} x^n \right] &=& x \frac{d}{dx} \left[ \frac{1}{1-x} \right] \\
   \sum_{n=0}^{\infty} n x^n &=& \frac{x}{(1-x)^2}
   \end{gather*}$$



* 【已知 3】能量量子化:

   $$\epsilon_{n}=nh\nu$$



* 【已知 4】Maxwell-Boltzmann distribution function:
 
   $$f_{MB}(\epsilon)=Ae^{-\epsilon/kT}$$



* 【定義 1】亞變數:
   
   $$x = e^{-h\nu/kT}$$





### proof



$$\begin{gather*}
\bar{\epsilon} &=& \frac{\sum_{n=0}^{\infty} \epsilon_n f_{MB}(\epsilon_n)}{\sum_{n=0}^{\infty} f_{MB}(\epsilon_n)}\\
&\overset{\text{已知 3,4}}{=}&\frac{\sum_{n=0}^{\infty} (nh\nu) A e^{-nh\nu/kT}}{\sum_{n=0}^{\infty} A e^{-nh\nu/kT}}\\
&\overset{\text{定義 1}}{=}& \frac{h\nu \sum_{n=0}^{\infty} n x^n}{\sum_{n=0}^{\infty} x^n}\\
&\overset{\text{已知 2}}{=}& \frac{h\nu \frac{x}{(1-x)^2}}{\frac{1}{1-x}}\\
&=& h\nu \frac{x}{1-x}\\
&\overset{\text{定義 1}}{=}&h\nu \frac{e^{-h\nu/kT}}{1 - e^{-h\nu/kT}}\\
&=& \frac{h\nu \cdot e^{-h\nu/kT} \cdot e^{h\nu/kT}}{(1 - e^{-h\nu/kT}) \cdot e^{h\nu/kT}}\\
&=& \frac{h\nu}{e^{h\nu/kT} - 1}
\end{gather*}$$

## Question 3

Express the Planck radiation formula in terms of wavelength.


## Question 3 - Answer



### 已知與假設

* 【已知 1】頻率 $\nu$ 與波長 $\lambda$ 的關係式:
  
   $$\nu = \frac{c}{\lambda}$$

   $$d\nu =-\frac{c}{\lambda^2} d\lambda$$

   * 能量是正的 $|d\nu| = \frac{c}{\lambda^2} |d\lambda|$

* 【已知 2】Jacobiany 在 $u(\nu)$ 與 $u(\lambda)$ 的關係式:
   
   $$|u(\nu)d\nu| = |u(\lambda)d\lambda|$$

   
### proof


$$\begin{gather*}
u(\nu)d\nu &=& G(\nu) \bar{\epsilon} d\nu \\
u(\nu)d\nu&\overset{\text{Q 1,2}}{=}& \left( \frac{8\pi \nu^2}{c^3} \right) \left( \frac{h\nu}{e^{h\nu/kT} - 1} \right) d\nu \\
u(\nu)d\nu&=& \frac{8\pi h \nu^3}{c^3 (e^{h\nu/kT} - 1)} d\nu  \\
u(\lambda)d\lambda &\overset{\text{已知 1,2}}{=}& \frac{8\pi h \left( \frac{c}{\lambda} \right)^3}{c^3 \left( e^{\frac{h(c/\lambda)}{kT}} - 1 \right)} \left( \frac{c}{\lambda^2} d\lambda \right)\\
u(\lambda)d\lambda &=& \frac{8\pi hc}{\lambda^5} \frac{1}{e^{hc/\lambda kT} - 1} d\lambda
\end{gather*}$$


## Question 4

A photon of frequency 𝜈 is scattered by an electron initially at rest. Verify that the maximum kinetic energy of the recoil electron is

$$KE_{max}=\frac{2h^{2}\nu^{2}/mc^{2}}{1+2h\nu/mc^{2}}$$



## Question 4 - Answer



### 已知與假設

* 【已知 1】[康普頓波長位移公式（Compton shift formula）](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#proof):
  
   $$\Delta \lambda = \lambda' - \lambda = \frac{h}{mc}(1 - \cos\theta)$$

   * 能量守恆、動量守恆
   * 讓電子獲得最大動能 $\theta = 180^\circ$ 

* 【已知 2】散射後頻率 $\nu'$ :
    
    $$\begin{gather*}
    \lambda' - \lambda &=& \frac{h}{mc}(1 - (-1)) \\
    \lambda' - \lambda &=& \frac{2h}{mc}\\
    \frac{c}{\nu'} - \frac{c}{\nu} &=& \frac{2h}{mc}\\
    \frac{1}{\nu'} - \frac{1}{\nu} &=& \frac{2h}{mc^2}\\
    \frac{1}{\nu'} &=& \frac{1}{\nu} + \frac{2h}{mc^2}  \\
    \frac{1}{\nu'} &=& \frac{mc^2 + 2h\nu}{\nu mc^2}\\
    \nu' &=& \frac{\nu mc^2}{mc^2 + 2h\nu}\\
    \end{gather*}$$



### proof



$$\begin{gather*}
E_{\text{光子,初}} + E_{\text{電子,初}} &\overset{\text{能量守恆}}{=}& E_{\text{光子,末}} + E_{\text{電子,末}}\\
h\nu + mc^2 &=& h\nu' + (KE_{max} + mc^2)\\
KE_{max} &=& h\nu - h\nu' \\
KE_{max}&\overset{\text{已知 2}}{=}& h\nu - h\left( \frac{\nu mc^2}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& h\nu \left( 1 - \frac{mc^2}{mc^2 + 2h\nu} \right)\\
KE_{max}&=& h\nu \left( \frac{mc^2 + 2h\nu - mc^2}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& h\nu \left( \frac{2h\nu}{mc^2 + 2h\nu} \right) \\
KE_{max}&=& \frac{2h^2\nu^2}{mc^2 + 2h\nu}\\
KE_{max} &=& \frac{2h^2\nu^2 / mc^2}{1 + 2h\nu / mc^2}
\end{gather*}$$


## Question 5

The gravitational potential energy $U$ relative to infinity of a body of mass m at a distance $R$ from the center of a body of mass $M$ is $U = -GmM/R$. 

(a) If $R$ is the radius of the body of mass $M$ , find the escape speed $v_e$ of the body, which is the minimum speed needed to leave it permanently. 

(b) Obtain a formula for the Schwarzschild radius of the body by setting $v_e=c$, the speed of light, and solving for $R$. (Of course, a relativistic calculation is correct here, but it is interesting to see what a classical calculation produces.)


## Question 5 - Answer


### 已知與假設

* 【已知 1】力學能守恆定律（Conservation of Mechanical Energy）:
  
   $$E_{total} = K + U = \text{常數}$$

   
   * 初始總能量：$E_i = \frac{1}{2}mv_e^2 - \frac{GmM}{R}$
     * 初始動能：$K_i = \frac{1}{2}mv_e^2$
     * 初始位能：$U_i = -\frac{GmM}{R}$
   * 最終總能量：$E_f = 0$
     * 最終動能：$K_f = 0$（因為停下來了）
     * 最終位能：$U_f = -\frac{GmM}{\infty} = 0$





### proof (a) 尋找逃逸速度 $v_e$



$$\begin{gather*}
\frac{1}{2}mv_e^2 - \frac{GmM}{R} &\overset{\text{已知 1}}{=}& 0 \\
\frac{1}{2}mv_e^2 &=& \frac{GmM}{R}\\
v_e^2 &=& \frac{2GM}{R}\\
v_e &=& \sqrt{\frac{2GM}{R}}
\end{gather*}$$



### proof (b)推導史瓦西半徑 $R_S$



$$\begin{gather*}
v_e &\overset{\text{Q5.(a)}}{=}& \sqrt{\frac{2GM}{R}} \\
c &=& \sqrt{\frac{2GM}{R_s}}\\
R_s &=& \frac{2GM}{c^2}
\end{gather*}$$
