## Question 1 


A beam of electrons enters a uniform $1.20 \text{ T}$ magnetic field. 
 
* (a) Find the energy difference between electrons whose spins are parallel and antiparallel to the field. 

* (b) Find the wavelength of the radiation that can cause the electrons whose spins are parallel to the field to flip so that their spins are antiparallel.




## Question 1 - Answer


### 假設與已知 (Assumptions & Preliminaries)


* **【定義 1】 波耳磁子（Bohr magneton, $\mu_B$）  ：**
    
  $$\begin{gather*}
  \mu_B &\overset{\text{def}}{=} & \frac{e\hbar}{2m_e} \\
  &=& \frac{1.602 \times 10^{-19} \text{ C} \times 1.054 \times 10^{-34}\text{ J}\cdot\text{s}}{2 \times 9.109 \times 10^{-31} \text{ kg}}\\
  &\approx& 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1}
  \end{gather*}$$


  * $\mu _{B}$ : 波耳磁子 (Bohr magneton) $\mu_B \approx 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1} ,  [\text{A} \cdot \text{m}^2]$
  * $e$ : 基本電荷 (Elementary charge) $e \approx 1.602 \times 10^{-19} \text{ C}$
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $m_{e}$ : 電子靜止質量 (Electron rest mass) $m_e \approx 9.109 \times 10^{-31} \text{ kg}$

* **【已知 1】 自旋角動量（Spin angular momentum, $S_z$）  ：**
  
  $$S_z = m_s\hbar$$

  * $S_{z}$ : 自旋角動量在 $z$ 軸上的分量 (z-component of spin angular momentum) $[\text{J}\cdot\text{s}]$
  * $m_{s}$ : 自旋磁量子數 (Spin magnetic quantum number) $m_s = \pm\frac{1}{2}$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$



* **【已知 2】 自旋磁矩（Spin magnetic moment, $\mu_s$）  ：**
  
  $$\begin{gather*}
  \vec{\mu}_s &=& -g_s\frac{e}{2m_e}\vec{S} \\
  \mu_{sz} &=& -2 \cdot \frac{e}{2m_e}S_z \\
  \mu_{sz} &=& -2 \frac{e}{2m_e}(m_s\hbar) \\
  \mu_{sz} &=& -2m_s\left(\frac{e\hbar}{2m_e}\right) \\
  \mu_{sz} &\overset{\text{定義 1}}{=} &  -2m_s \mu_B \\
  \end{gather*}$$


  * $\vec{\mu}_{s}$ : 自旋磁矩向量 (Spin magnetic moment vector) $[\text{J}\cdot\text{T}^{-1}]$
  * $\mu _{sz}$ : 自旋磁矩在 $z$ 軸上的分量 (z-component of spin magnetic moment) $[\text{J}\cdot\text{T}^{-1}]$
  * $g_{s}$ : 電子自旋 $g$ 因子 (Electron spin g-factor) $g_s \approx 2$ [無單位]
  * $\vec{S}$ : 自旋角動量向量 (Spin angular momentum vector)  ，$[\text{J}\cdot\text{s}]$
  * $S_{z}$ : 自旋角動量在 $z$ 軸上的分量 (z-component of spin angular momentum) $[\text{J}\cdot\text{s}]$
  * $m_{s}$ : 自旋磁量子數 (Spin magnetic quantum number) $m_s = \pm\frac{1}{2}$ [無單位]
  * 負號 ($-$) : 代表電子的自旋磁矩方向與其自旋角動量方向相反（因電子帶負電）

* **【已知 3】 磁偶極矩的位能 ：**
  
  $$\begin{gather*}
  U &=& -\vec{\mu } \cdot \vec{B} \\
  &=& -\mu_{sz}B \\
  &\overset{\text{已知 2}}{=} &  - \left(-2m_s \mu_B \right)B \\
  &=& 2m_s \mu_B B \\
  \end{gather*}$$

  * $\vec{\mu }$ : 磁偶極矩向量 (Magnetic dipole moment vector) $[\text{A}\cdot\text{m}^2]$ 或 $[\text{J}\cdot\text{T}^{-1}]$
  * $\vec{B}$ : 外加磁場向量 (External magnetic field vector) $[\text{T}]$
  * $U$ : 磁位能 (Magnetic potential energy) $[\text{J}]$
  * $m_{s}$ : 自旋磁量子數 (Spin magnetic quantum number) $m_s = \pm\frac{1}{2}$ [無單位]

* **【已知 4】 [愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis)  ：**
  
  $$\begin{gather*}
  E &\overset{\text{def}}{=}& h\nu \\
  E &=& h \frac{c}{\lambda} \\
  \lambda &=& hc \frac{1}{E}
  \end{gather*}$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$
  * $\nu$ : 光子的頻率  $\nu =\frac{c}{\lambda}  $，$[\text{s}^{-1}]$ 
  * $\lambda$ : 光子的波長  $[\text{m}]$

 

### (a) 在 $1.20\text{ T}$ 磁場下，電子自旋在塞曼效應（Zeeman Effect）中的能量分裂差（$\Delta E = g \mu_B B$）


$$\begin{gather*}
\Delta E &=& U_{\text{up}} - U_{\text{down}} \\
&\overset{\text{已知 3}}{=}& (2m_s \mu_B B)_{\text{up}} - (2m_s \mu_B B)_{\text{down}} \\
&\overset{\text{已知 3}}{=}& (2\frac{1}{2}\mu_B B) - (2\frac{-1}{2}\mu_B B) \\
&=& (+\mu_B B) - (-\mu_B B) \\
&=& 2\mu_B B \\
&\overset{\text{定義 1,Q1}}{=}& 2\times 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1} \times  1.20 \text{ T} \\
&\approx& 2.23 \times 10^{-23} \text{ J}
\end{gather*}$$


### (b) 計算輻射波長

$$\begin{gather*}
\lambda &\overset{\text{已知 4}}{=}& hc \frac{1}{E} \\
&\overset{\text{已知 4,Q1(a)}}{=}& 6.626 \times 10^{-34} \text{ J}\cdot\text{s} \times 3 \times 10^8 \text{ m}\cdot\text{s}^{-1} \frac{1}{2.22576 \times 10^{-23} \text{ J}} \\
&\approx& 8.93 \times 10^{-3} \text{ m}
\end{gather*}$$

大概屬於微波波段，或許可以應用在電子順磁共振（EPR）光譜儀，專門用來偵測與分析 **自由基** 和 **過渡金屬**

## Question 3

* (a) What are the possible values of $L$ (Total Orbital Angular Momentum) for a system of two electrons whose orbital quantum numbers are $l_1 = 1$ and $l_2 = 3$ ? 
* (b) What are the possible values of $S$ (Total Spin Angular Momentum)? 
* (c) What are the possible values of $J$ (Total Angular Momentum)?



## Question 3 - Answer

$l \in \{0, 1, \dots, n-1\}$ 是針對「單一電子」的軌域角動量，但當我們面對多電子原子時，這些電子之間的角動量會互相影響、互相疊加稱作 **角動量耦合（Angular Momentum Coupling）** 。

想像兩位花式溜冰選手（代表兩個電子）正在同一個溜冰場上表演。

1. 每位選手都有自己繞著場地中心滑行的 **公轉** 軌跡，這叫做個別的軌域角動量（$l_1$ 與 $l_2$）。
2. 每位選手同時也在原地 **自轉** ，這叫做個別的自旋角動量（$s_1$ 與 $s_2$）。

這四個動量在空間中會互相疊加形成整體的動量，對於較輕的原子，它們的疊加規律通常遵循 LS 耦合（LS Coupling），也就是：
* $L$ (Total Orbital Angular Momentum)： 兩位選手的 **公轉** 先加總起來
* $S$ (Total Spin Angular Momentum)： 兩位選手的 **自轉** 先加總起來
* $J$ (Total Angular Momentum)： 最後，整體的公轉 $L$ 與整體的自轉 $S$ 再加總起來，形成整個系統的「總角動量 $J$」


### 已知與假設

* **【已知 1】 角動量是量子化：**
  
  * 只能以整數為間距跳躍在可能的整數上形成的集合






### (a) possible values of $L$ (Total Orbital Angular Momentum)

1. 最小值（兩向量反向）：$|1 - 3| = 2$
2. 最大值（兩向量同向）：$1 + 3 = 4$
3. 在最小值與最大值之間以整數遞增 $\overset{\text{已知 1}}{:} L \in \{ 2, 3, 4 \}$

### (b) possible values of $S$ (Total Spin Angular Momentum)

1. 最小值（兩向量反向）：$|\frac{1}{2} - \frac{1}{2}| = 0$
2. 最大值（兩向量同向）：$\frac{1}{2} + \frac{1}{2} = 1$
3. 在最小值與最大值之間以整數遞增 $\overset{\text{已知 1}}{:} S \in\{ 0, 1 \}$


### (c) possible values of $J$ (Total Angular Momentum)

1. 最小值（兩向量反向）：$\text{min} \{|2 - 1| , |0-4|\} = 1$
2. 最大值（兩向量同向）：$4+1 = 5$
3. 在最小值與最大值之間以整數遞增 $\overset{\text{已知 1}}{:}J \in \{ 1 ,2 ,3 ,4 ,5 \}$
  
* $S = 0$ 
  * $L \in \{2, 3, 4\} \implies J \in \{2, 3, 4\}$
* $S = 1$
  * $L=2 \implies J \in \{1, 2, 3\}$
  * $L=3 \implies J \in \{2, 3, 4\}$
  * $L=4 \implies J \in \{3, 4, 5\}$
* 取所有可能值的聯集，得到 $J \in \{1, 2, 3, 4, 5\}$

## Question 5


The magnetic moment $\mu_J$ of an atom in which LS coupling holds has the magnitude

$$\mu_J = g_J \mu_B \sqrt{J(J + 1)}$$

where $\mu_B = e\hbar/2m$ is the Bohr magneton and

$$g_J = 1 + \frac{J(J + 1) - L(L + 1) + S(S + 1)}{2J(J + 1)}$$

is the Landé g factor.

* (a) Derive this result with the help of the law of cosines starting from the fact that averaged over time, only the components of $\mu_L$ and $\mu_S$ parallel to $\mathbf{J}$ contribute to $\mu_J$. 
* (b) Consider an atom that obeys LS coupling that is in a weak magnetic field $\mathbf{B}$ in which the coupling is preserved. How many substates are there for a given value of $\mathbf{J}$? What is the energy difference between different substates?




## Question 5 - Answer


### 已知與假設



* **【定義 1】 波耳磁子（Bohr magneton, $\mu_B$）  ：**
    
  $$\begin{gather*}
  \mu_B &\overset{\text{def}}{=} & \frac{e\hbar}{2m_e} \\
  &=& \frac{1.602 \times 10^{-19} \text{ C} \times 1.054 \times 10^{-34}\text{ J}\cdot\text{s}}{2 \times 9.109 \times 10^{-31} \text{ kg}}\\
  &\approx& 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1}
  \end{gather*}$$


  * $\mu _{B}$ : 波耳磁子 (Bohr magneton) $\mu_B \approx 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1} ,  [\text{A} \cdot \text{m}^2]$
  * $e$ : 基本電荷 (Elementary charge) $e \approx 1.602 \times 10^{-19} \text{ C}$
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $m_{e}$ : 電子靜止質量 (Electron rest mass) $m_e \approx 9.109 \times 10^{-31} \text{ kg}$



* **【已知 1】 軌域磁矩 :**
  
  $$\begin{gather*}
  \vec{\mu}_L &=& -g_l\frac{e}{2m}\vec{L} \\
  &\overset{g_l = 1}{=}& -\frac{e}{2m}\vec{L} \\
  &\overset{\text{定義 1}}{=}& -\frac{\mu_B}{\hbar}\vec{L}
  \end{gather*}$$

  * $\vec{\mu }_{L}$ : 軌域磁矩向量 (Orbital magnetic moment vector) $[\text{J}\cdot\text{T}^{-1}]$
  * $g_{l}$ : 軌域 $g$ 因子 (Orbital g-factor) $g_l = 1$ [無單位]
  * $e$ : 基本電荷 (Elementary charge) $e \approx 1.602 \times 10^{-19} \text{ C}$
  * $m_{e}$ : 電子靜止質量 (Electron rest mass) $m_e \approx 9.109 \times 10^{-31} \text{ kg}$
  * $\vec{L}$ : 軌域角動量向量 (Orbital angular momentum vector) $[\text{J}\cdot\text{s}]$
  * $\mu _{B}$ : 波耳磁子 (Bohr magneton) $\mu_B \approx 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1}$
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$


* **【已知 2】 自旋磁矩 :**
  
  $$\begin{gather*}
  \vec{\mu}_S &=& -g_s\frac{e}{2m}\vec{S} \\
  &\overset{g_s \approx 2}{=}& -2\frac{e}{2m}\vec{S} \\
  &\overset{\text{定義 1}}{=}& -2\frac{\mu_B}{\hbar}\vec{S}
  \end{gather*}$$

  * $\vec{\mu }_{S}$ : 自旋磁矩向量 (Spin magnetic moment vector) $[\text{J}\cdot\text{T}^{-1}]$
  * $g_{s}$ : 電子自旋 $g$ 因子 (Electron spin g-factor) $g_s \approx 2.0023 \approx 2$ [無單位]
  * $\vec{S}$ : 自旋角動量向量 (Spin angular momentum vector) $[\text{J}\cdot\text{s}]$

* **【定義 2】 總磁矩 :**

  $$\begin{gather*}
  \vec{\mu}_{total} &\overset{\text{def}}{=}& \vec{\mu}_J \\
  &=& -g_J \frac{\mu_B}{\hbar} \vec{J} \\ 
  \end{gather*}$$

  * $\vec{\mu}_{total} , \vec{\mu}_J$ : 總磁矩向量 (Total magnetic moment vector) $[\text{J}\cdot\text{T}^{-1}]$
  * $g_{J}$ : 朗德 $g$ 因子 (Lande g-factor) [無單位]
  * $\vec{J}$ : 總角動量向量 (Total angular momentum vector) $\vec{J} = \vec{L} + \vec{S}$ $[\text{J}\cdot\text{s}]$

* **【已知 3】 向量守恆關係 :**

  $$\vec{J} = \vec{L} + \vec{S}$$

  * $\vec{J}$ : 總角動量向量 (Total angular momentum vector) $[\text{J}\cdot\text{s}]$
  * $\vec{L}$ : 軌域角動量向量 (Orbital angular momentum vector) $[\text{J}\cdot\text{s}]$
  * $\vec{S}$ : 自旋角動量向量 (Spin angular momentum vector) $[\text{J}\cdot\text{s}]$

* **【已知 4】 量子力學的角動量平方期望值 :**

  * (a)

  $$\vec{J}^2 = J(J+1)\hbar^2$$


  * (b)
  
  $$\vec{L}^2 = L(L+1)\hbar^2$$


  * (c)
  
  $$\vec{S}^2 = S(S+1)\hbar^2$$

  * $\vec{J}^2, \vec{L}^2, \vec{S}^2$ : 角動量平方算符 (Squared angular momentum operators) $[\text{J}^2\cdot\text{s}^2]$
  * $J$ : 總角動量量子數 (Total angular momentum quantum number) [無單位]
  * $L$ : 軌域角動量量子數 (Orbital angular momentum quantum number) [無單位]
  * $S$ : 自旋角動量量子數 (Spin angular momentum quantum number) [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$


* **【推導 1】 利用餘弦定理求出內積項 :** 
  
  * (a) 對於 $\vec{J}\cdot\vec{L}$

  $$\begin{gather*}
  \vec{J} &\overset{\text{已知 3}}{=}& \vec{L} + \vec{S} \\
  \vec{J} - \vec{L} &=& \vec{S} \\
  (\vec{J} - \vec{L})\cdot (\vec{J} - \vec{L}) &=& \vec{S} \cdot \vec{S} \\
  \vec{J}^2 + \vec{L}^2 -2\vec{J} \cdot \vec{L} &=& \vec{S}^2 \\
  \vec{J}^2 + \vec{L}^2 - \vec{S}^2  &=& 2\vec{J}\cdot\vec{L} \\
  \end{gather*}$$

  * (b) 對於 $\vec{J}\cdot\vec{S}$
  
  $$\begin{gather*}
  \vec{J} &\overset{\text{已知 3}}{=}& \vec{L} + \vec{S} \\
  \vec{J} - \vec{S} &=& \vec{L} \\
  (\vec{J} - \vec{S})\cdot (\vec{J} - \vec{S}) &=& \vec{L} \cdot \vec{L} \\
  \vec{J}^2 + \vec{S}^2 -2\vec{J} \cdot \vec{S} &=& \vec{L}^2 \\
  \vec{J}^2 + \vec{S}^2 - \vec{L}^2  &=& 2\vec{J}\cdot\vec{S} \\
  \end{gather*}$$


  * $\vec{J}\cdot\vec{L} , \vec{J}\cdot\vec{S}$ : 算符的內積對應項 (Operator dot products) $[\text{J}^2\cdot\text{s}^2]$
  

* **【已知 5】 z 軸方向總磁偶極矩的位能 ：**
  
  $$\begin{gather*}
  U &=& -\vec{\mu } \cdot \vec{B} \\
  &\overset{\text{總角動量}}{=}& -\vec{\mu}_J \cdot \vec{B}\\
  &\overset{\vec{\mu}_J\parallel \vec{B}}{=}& -\mu_J B\\
  &\overset{\text{定義 2}}{=}& -(-g_J \frac{\mu_B}{\hbar} J_z) B\\
  &=& g_J \frac{\mu_B}{\hbar} (m_J \hbar)B \\
  &=& g_J \mu_B m_J B \\
  \end{gather*}$$

  * $U$ : 塞曼分裂位能 (Zeeman potential energy) $[\text{J}]$
  * $\vec{B}$ : 外加磁場向量，假設沿 $z$ 軸方向 $\vec{B} = (0, 0, B)$ $[\text{T}]$
  * $\mu _{Jz}$ : 總磁矩在 $z$ 軸上的分量 (z-component of total magnetic moment) $[\text{J}\cdot\text{T}^{-1}]$
  * $J_{z}$ : 總角動量在 $z$ 軸上的分量 (z-component of total angular momentum) $[\text{J}\cdot\text{s}]$
  * $m_{j}$ : 總磁量子數 (Total magnetic quantum number) $m_j = -j, -j+1, \dots, j$ [無單位]
  * $\mu _{B}$ : 波耳磁子 (Bohr magneton) $\mu_B \approx 9.274 \times 10^{-24} \text{ J}\cdot\text{T}^{-1}$
  * $g_{J}$ : 朗德 $g$ 因子 (Lande g-factor) [無單位]



### (a) proof 

$$\begin{gather*}
\vec{\mu}_{total} &=& \vec{\mu}_L + \vec{\mu}_S \\
\vec{\mu}_{total} &\overset{\text{已知 1,已知 2}}{=}& \left( -\frac{\mu_B}{\hbar}\vec{L} \right)+ \left( -2\frac{\mu_B}{\hbar}\vec{S}\right) \\
\vec{\mu}_{total}  &=& -\frac{\mu_B}{\hbar}(\vec{L} + 2\vec{S})\\
\vec{\mu}_{total} \cdot \vec{J} &=&-\frac{\mu_B}{\hbar}(\vec{L} + 2\vec{S})\cdot \vec{J}\\
\vec{\mu}_{total} \cdot \vec{J} &=&-\frac{\mu_B}{\hbar}(\vec{L}\cdot \vec{J} + 2\vec{S}\cdot \vec{J})\\
\vec{\mu}_{total} \cdot \vec{J} &\overset{\text{推導 1(a)(b)}}{=}&-\frac{\mu_B}{\hbar}(\frac{1}{2} \left( \vec{J}^2 + \vec{L}^2 - \vec{S}^2 \right) + \left( \vec{J}^2 + \vec{S}^2 - \vec{L}^2  \right) )\\
\vec{\mu}_{total} \cdot \vec{J} &=&-\frac{\mu_B}{\hbar}(\frac{3}{2}  \vec{J}^2 +\frac{1}{2} \vec{S}^2  - \frac{1}{2}\vec{L}^2   )\\
\vec{\mu}_{total} \cdot \vec{J} &=&-\frac{\mu_B}{2 \hbar}(3  \vec{J}^2 + \vec{S}^2  - \vec{L}^2   )\\
\vec{\mu}_J \cdot \vec{J} &\overset{\text{定義 2}}{=}& -\frac{\mu_B}{2 \hbar}(3  \vec{J}^2 + \vec{S}^2  - \vec{L}^2   )\\
-g_J \frac{\mu_B}{\hbar} \vec{J} \cdot \vec{J} &\overset{\text{定義 2}}{=}& -\frac{\mu_B}{2 \hbar}(3  \vec{J}^2 + \vec{S}^2  - \vec{L}^2   )\\
-g_J \frac{\mu_B}{\hbar} \vec{J}^2 &=& -\frac{\mu_B}{2 \hbar}(3  \vec{J}^2 + \vec{S}^2  - \vec{L}^2   )\\
g_J &=& \frac{3  \vec{J}^2 + \vec{S}^2  - \vec{L}^2 }{2\vec{J}^2} \\ 
g_J &=& 1 + \frac{1  \vec{J}^2 + \vec{S}^2  - \vec{L}^2 }{2\vec{J}^2} \\ 
g_J &\overset{\text{已知 4(a)(b)(c)}}{=}& 1 + \frac{J(J+1) - L(L+1) + S(S+1) }{2J(J+1)} \\ 
\end{gather*}$$
