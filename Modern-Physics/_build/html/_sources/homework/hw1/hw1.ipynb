## Question 1

Prove time dilation
   
$$t = \frac{2L_0/c}{\sqrt{1-v^2/c^2}} = \frac{t_0}{\sqrt{1-v^2/c^2}}$$



## Question 1 - Answer

雖然題目是寫 $t$ 但是我覺得寫 $\Delta t$ 更加精準


![](./pic/Light_Clock_Derivation.png)


假設飛船上有一個光鐘，光子在相距 $L_0$ 的兩面鏡子間上下反射。

* **在飛船上 ($S'$ 座標系)：** 觀察者看到光子直上直下，來回一次所需的時間為原時 $\Delta t_0$。

$$\frac{\Delta t_0}{2} = \frac{L_0}{c}$$


* **在地球上 ($S$ 座標系)：** 觀察者看到飛船以速度 $v$ 前進，因此光子走的是斜線路徑、設地球觀察者測量到的時間為 $\Delta t$，根據畢氏定理：

$$\left( c \frac{\Delta t}{2} \right)^2 = L_0^2 + \left( v \frac{\Delta t}{2} \right)^2$$



將 $L_0 = c \frac{\Delta t_0}{2}$ 代入：

$$\begin{aligned}
c^2 \left( \frac{\Delta t}{2} \right)^2 &= \left( c \frac{\Delta t_0}{2} \right)^2 + v^2 \left( \frac{\Delta t}{2} \right)^2 \\
c^2 (\Delta t)^2 &= c^2 (\Delta t_0)^2 + v^2 (\Delta t)^2 \\
(\Delta t)^2 (c^2 - v^2) &= c^2 (\Delta t_0)^2 \\
(\Delta t)^2 &= \frac{c^2}{c^2 - v^2} (\Delta t_0)^2 \\
(\Delta t)^2&= \frac{1}{1 - v^2/c^2} (\Delta t_0)^2 \\
\Delta t&= \frac{1}{\sqrt{1 - v^2/c^2}} \Delta t_0 \\
\Delta t&= \frac{1}{\sqrt{1 - v^2/c^2}} \frac{2L_0}{c}
\end{aligned}$$


核心概念： 靜止的觀察者會測量到，相對於他運動的時鐘，其滴答速度比他身旁靜止的時鐘還要慢。


## Question 2

Prove that when $v$ is much smaller than $c$ , the kinetic energy reduces to the classical definition

$$
KE = \gamma mc^2-mc^2 = (\gamma-1) mc^2
$$

## Question 2 - Answer

### 已知與假設

* 【已知 1】古典KE公式:
  
   $$KE_{\text{classical}} = \frac{1}{2} mv^2$$

* 【已知 2】[Lorentz factor $\gamma$](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation):
   
   $$ \gamma = \frac{1}{\sqrt{1-v^2/c^2}}$$

* 【已知 3】數學近似-泰勒級數1階展開:
   
   $$ (1+\varepsilon)^n \approx 1+n \varepsilon $$

* 【已知 4】數學近似:
   
   $$ \frac{a}{b+\varepsilon} \approx \frac{a}{b} $$

### proof 

$$\begin{gather*}
KE &=& (\gamma-1)mc^2 \\
&\overset{\text{已知 2}}{=}& (\frac{1}{\sqrt{1-v^2/c^2}}-1)mc^2 \\
&=& \frac{ 1-\sqrt{1-v^2/c^2} }{\sqrt{1-v^2/c^2}}mc^2 \\
&\overset{\text{已知 3}}{=}&\frac{ 1-(1-\frac{1}{2}v^2/c^2) }{(1-\frac{1}{2}v^2/c^2)}mc^2 \\
&=&\frac{ \frac{1}{2}v^2/c^2 }{(1-\frac{1}{2}v^2/c^2)}mc^2 \\
&\overset{\text{已知 4}}{=}& \frac{1}{2}v^2/c^2 mc^2 \\
&=& \frac{1}{2}v^2 m &\overset{\text{已知 1}}{=}& KE_{\text{classical}}
\end{gather*}$$


## Question 3

Prove equation:

$$E^2 = (mc^2)^2+p^2c^2$$


## Question 3 - Answer

我們希望找出不包含速度變數 $v$ 的能量與動量關係。

### 已知與假設

* 【已知 1】[物體總能量](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#mass-and-energy):
  
   $$E = \gamma m_0 c^2$$

* 【已知 2】[物體動量](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum):
   
   $$ p = \gamma m_0 v $$

* 【已知 3】[Lorentz factor $\gamma$](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation):
   
   $$ \gamma = \frac{1}{\sqrt{1-v^2/c^2}}$$


### proof

將總能量和動量平方後相減：

$$\begin{gather*}
E^2 - (pc)^2 &=& (\gamma m_0 c^2)^2 - (\gamma m_0 v c)^2 \\             
&=& \gamma^2 m_0^2 c^4 - \gamma^2 m_0^2 v^2 c^2 \\             
&=& \gamma^2 m_0^2 c^4 \left( 1 - \frac{v^2}{c^2} \right) \\            
&=& \gamma^2 m_0^2 c^4  \gamma^{-2}  \\
&=& m_0^2 c^4 \\
E^2 &=& (m_0 c^2)^2 + (pc)^2 
\end{gather*}$$
