## Question 1 


The protons in the $\text{H}_2^+$ molecular ion are $0.106\text{ nm}$ apart, and the binding energy of $\text{H}_2^+$ is $2.65\text{ eV}$.

What negative charge must be placed halfway between two protons this distance apart to give the same binding energy?





## Question 1 - Answer


### 假設與已知 (Assumptions & Preliminaries)


* **【已知 1】 建立系統的總位能模型  ：**
  
  * (a) $\text{p}_1$ 和 $\text{p}_2$ 的位能(排斥):

  $$\begin{gather*}
  U &\overset{\text{def}}{=} & \frac{kQq}{r} \\
  U_{\text{p}_1,\text{p}_2} &\overset{\text{(a)}}{=} & \frac{k\cdot e\cdot e}{r} \\
  &= & \frac{ke^2}{r} \\
  \end{gather*}$$


  * (b) $\text{p}_1$ 和 q  的位能(吸引):
  
  $$\begin{gather*}
  U &\overset{\text{def}}{=} & \frac{kQq}{r} \\
  U_{\text{p}_1,q} &\overset{\text{(b)}}{=} & \frac{k\cdot e\cdot q}{\frac{r}{2}} \\
  &= & \frac{2keq}{r} \\
  \end{gather*}$$


  * (c) $\text{p}_2$ 和 q  的位能(吸引):
  
  $$\begin{gather*}
  U &\overset{\text{def}}{=} & \frac{kQq}{r} \\
  U_{\text{p}_2,q} &\overset{\text{(c)}}{=} & \frac{k\cdot e\cdot q}{\frac{r}{2}} \\
  &= & \frac{2keq}{r} \\
  \end{gather*}$$

  * $U_{\text{p}_{1},\text{p}_{2}}$ : 兩質子間的靜電位能 (Electrostatic potential energy between protons) $[\text{J}]$
  * $U_{\text{p}_1,q} , U_{\text{p}_2,q}$ : 質子與電荷 $q$ 間的靜電位能 (Potential energy between proton and charge $q$) $[\text{J}]$
  * $k$ : 庫侖常數 (Coulomb's constant) $k \approx 8.988 \times 10^9 \text{ N}\cdot\text{m}^2\cdot\text{C}^{-2}$
  * $e$ : 質子電量／基本電荷 (Elementary charge) $e \approx 1.602 \times 10^{-19} \text{ C}$
  * $q$ : 中間電荷的電量 (Charge of the central particle) $[\text{C}]$
  * $r$ : 兩質子 $\text{p}_{1}$ 與 $\text{p}_{2}$ 之間的距離 (Distance between protons) $[\text{m}]$

### solve 鍵結能實驗觀測對應的電荷量

$$\begin{gather*}
E_b &=& -U_{total} \\
E_b &=& -\left( U_{\text{p}_1,\text{p}_2} + U_{\text{p}_1,q} +  U_{\text{p}_2,q} \right) \\
E_b &\overset{\text{已知 1(a)(b)(c)}}{=}& -\left( \frac{ke^2}{r} + \frac{2keq}{r} + \frac{2keq}{r} \right) \\
E_b &=& -\frac{ke}{r}\left( e + 4q \right) \\
q &=& -\frac{1}{4}\left( e + \frac{E_b r }{ke} \right) \\
q &\overset{\text{Q1,已知 1}}{\approx}&  -\frac{1}{4}\left( 1.602 \times 10^{-19} \text{ C} + \frac{2.65\text{ eV}\times 0.106\text{ nm} }{8.988 \times 10^9 \text{ N}\cdot\text{m}^2\cdot\text{C}^{-2} \times 1.602 \times 10^{-19} \text{ C}} \right) \\
q &\approx&  -4.786 \times 10^{-20} \text{ C} \\
q &\approx&  -0.299 e \\
\end{gather*}$$

* 古典視角：
  
  古典模型將電荷視為一個無限小的 **點**，精準地擺在距離兩側質子 $\frac{r}{2}$ 的最佳位置，發揮了 $100 \%$ 的最大吸引效率

* 量子現實：
  
  雖然它出現在兩個質子正中間的機率確實較高，但依然有很大一部分的機率分佈在離質子較遠的地方。這種分散的空間分佈會讓電子吸引質子的「整體平均效率」打折扣，因此需要更大的總電荷量 ($-1e$) 來彌補波函數分散所造成的吸引力衰減。

在真實的 $\text{H}_{2}^{+}$ 離子中電子並非侷限在正中央，而是以 **機率波（電子雲）** 的形式分佈在兩核之間，若要用古典模型模擬，只需要約 $0.3$ 個電子的電量集中在中心，就能達到相同的束縛效果。


## Question 2 


Calculate the energies of the four lowest non-zero rotational energy states of the $\text{H}_2$ and $\text{D}_2$ molecules, where $\text{D}$ represents the deuterium atom $_1^2\text{H}$.



## Question 2 - Answer


### 已知與假設

* **【假設 1】剛性轉子 (Rigid Rotor) 的模型 :** 假設兩個原子是點質量 $m_1$ 與 $m_2$，彼此之間的鍵長 $r$ 保持固定不變
  
  * (a) 幾何關係：
  
  $$r_1 + r_2 = r$$
  
  * (b)質心定義：
  
  $$m_1 r_1 = m_2 r_2$$

  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $r$ : 雙原子的固定鍵長 (Fixed bond length) $[\text{m}]$
  * $r_1, r_2$ : 兩原子各自到分子質心的距離 (Distances from each atom to the center of mass) $[\text{m}]$

* **【推導 1】原子到質心的距離 :** 解聯立方程式
  
  * (a) $r_1$：
  
  $$\begin{gather*}
  r_1 + r_2 &\overset{\text{假設 1(a)}}{=}& r \\
  r_1 + \frac{m_1}{m_2}r_1 &\overset{\text{假設 1(b)}}{=}& r \\
  r_1 \left( 1+ \frac{m_1}{m_2} \right) &=& r \\
  r_1 \left( \frac{m_2+m_1}{m_2} \right) &=& r \\
  r_1 &=& r \frac{m_2}{m_2+m_1}\\
  \end{gather*}$$
  
  * (b) $r_2$：
  
  $$\begin{gather*}
  r_1 + r_2 &\overset{\text{假設 1(a)}}{=}& r \\
  \frac{m_2}{m_1}r_2 + r_2  &\overset{\text{假設 1(b)}}{=}& r \\
  \left( \frac{m_2}{m_1} +1 \right)r_2  &=& r \\
  \left( \frac{m_2+m_1}{m_1} \right)r_2 &=& r \\
  r_2 &=& r \frac{m_1}{m_2+m_1}\\
  \end{gather*}$$

  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $r$ : 雙原子的固定鍵長 (Fixed bond length) $[\text{m}]$
  * $r_1, r_2$ : 兩原子各自到分子質心的距離 (Distances from each atom to the center of mass) $[\text{m}]$

* **【定義 1】縮減質量 (Reduce mass , $m'$) :**

  $$  m' \overset{\text{def}}{=} \frac{m_1 m_2}{m_1 + m_2}  $$

  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$

* **【已知 2】垂直於鍵軸的轉動 ($I_{\perp}$) :**

  $$\begin{gather*}
  I_{\perp} &=& m_1 r_1^2 + m_2 r_2^2 \\
  &\overset{\text{推導 1(a)(b)}}{=}& m_1 \left( \frac{m_2}{m_1 + m_2} r \right)^2 + m_2 \left( \frac{m_1}{m_1 + m_2} r \right)^2 \\
  &=& \frac{m_1 m_2^2 + m_2 m_1^2}{(m_1 + m_2)^2} r^2 \\
  &=& \frac{m_1 m_2 (m_2 + m_1)}{(m_1 + m_2)^2} r^2\\
  &=& \left( \frac{m_1 m_2}{m_1 + m_2} \right) r^2\\
  &\overset{\text{定義 1}}{=}& m' r^2\\
  \end{gather*}$$

  * $I_{\perp }$ : 垂直於鍵軸的轉動慣量 (Moment of inertia perpendicular to the bond axis) $[\text{kg}\cdot\text{m}^2]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $r$ : 雙原子的固定鍵長 (Fixed bond length) $[\text{m}]$
  * $r_1, r_2$ : 兩原子各自到分子質心的距離 (Distances from each atom to the center of mass) $[\text{m}]$

* **【已知 3】垂直於鍵軸的轉動動能 ($E_J$) :**
  
  $$\begin{gather*}
  KE &=& \frac{1}{2}I_{\perp}\omega^2 \\
  KE &=& \frac{(I\omega)^2}{2I_{\perp}} \\
  KE &=& \frac{L^2}{2I_{\perp}} \\
  E_J &\overset{\text{量子化}}{=}& \frac{L^2}{2I_{\perp}} \\
  E_J &=& \frac{J(J+1)\hbar^2}{2I_{\perp}} \\
  E_J &\overset{\text{已知 2}}{=}& \frac{J(J+1)\hbar^2}{2 m' r^2} \\
  \end{gather*}$$

  * $KE$ : 經典力學的轉動動能 (Classical rotational kinetic energy) $[\text{J}]$
  * $\omega $ : 轉動角速度 (Angular velocity) $[\text{rad}\cdot\text{s}^{-1}]$
  * $L$ : 角動量大小 (Magnitude of angular momentum) $[\text{J}\cdot\text{s}]$
  * $E_{J}$ : 量子化後的轉動能階 (Rotational energy levels) $[\text{J}]$
  * $J$ : 轉動量子數 (Rotational quantum number) $J = 0, 1, 2, \dots$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$

* **【已知 4】基本資訊 :**
  
  * (a) 氫氣分子 $\text{H}_{2}$ 的縮減質量：
  
  $$\begin{gather*}
  m'_{\text{H}_{2}}&\overset{\text{定義 1}}{=}&\frac{m_{p}\cdot m_{p}}{m_{p}+m_{p}} \\
  &=&\frac{1}{2}m_{p} \\
  &=& 0.5\times 1.6726\times 10^{-27}\text{ kg} \\
  &=& 8.363\times 10^{-28}\text{ kg} \\
  \end{gather*}$$

  * (b) 氘氣分子 $\text{D}_{2}$ 的縮減質量：
  
  $$\begin{gather*}
  m'_{\text{D}_{2}}&\overset{\text{定義 1}}{=}&\frac{2m_{p}\cdot 2m_{p}}{2m_{p}+2m_{p}} \\
  &=& m_{p} \\
  &=& 1.6726 \times 10^{-27}\text{ kg} \\
  \end{gather*}$$

  * (c) $r$ 觀測典型鍵長度
  
  $$r \overset{\text{obs}}{=} 7.4 \times 10^{-11} \text{ m}$$

  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27} \text{ kg}$
  * $r$ : 雙原子的觀測典型鍵長度 (Observation bond length) $[\text{m}]$
  * $\text{H}_{2}$ : 輕氫分子 (Hydrogen molecule)
  * $\text{D}_{2}$ : 重氫/氘分子 (Deuterium molecule)


### (a) estimate lowest non-zero rotational energy states of the $\text{H}_2$  

$$\begin{gather*}
&E_J &\overset{\text{已知 3}}{=}& \frac{J(J+1)\hbar^2}{2 m' r^2} \\
&&\overset{\text{已知 4(a)(c)}}{=}& \frac{J(J+1)(1.054 \times 10^{-34}\text{ J}\cdot\text{s})^2}{2 \times  8.363\times 10^{-28}\text{ kg} (7.4 \times 10^{-11} \text{ m})^2} \\
&&\approx & J(J+1) \cdot 1.213\times 10^{-21}\text{ J} \\
\Rightarrow&\begin{bmatrix}  E_1 \\  E_2 \\ E_3  \\ E_4  \end{bmatrix} &=& \begin{bmatrix}  1\cdot(1+1) \cdot 1.213\times 10^{-21}\text{ J} \\  2\cdot(2+1) \cdot 1.213\times 10^{-21}\text{ J} \\ 3\cdot(3+1) \cdot 1.213\times 10^{-21}\text{ J}  \\ 4\cdot(4+1) \cdot 1.213\times 10^{-21}\text{ J}  \end{bmatrix}  \\
&\begin{bmatrix}  E_1 \\  E_2 \\ E_3  \\ E_4  \end{bmatrix} &=& \begin{bmatrix}  2.43\times 10^{-21}\text{ J} \\ 7.28\times 10^{-21}\text{ J} \\ 1.46\times 10^{-20}\text{ J}  \\ 2.43\times 10^{-20}\text{ J}  \end{bmatrix}  \\
&\begin{bmatrix}  E_1 \\  E_2 \\ E_3  \\ E_4  \end{bmatrix} &=& \begin{bmatrix} 1.51 \times 10^{-2} \text{ eV}\\ 4.54 \times 10^{-2} \text{ eV}\\ 9.09 \times 10^{-2} \text{ eV}\\ 1.51 \times 10^{-1} \text{ eV}\end{bmatrix}  \\
\end{gather*}$$


### (b) estimate lowest non-zero rotational energy states of the $\text{D}_2$

$$\begin{gather*}
&E_J &\overset{\text{已知 3}}{=}& \frac{J(J+1)\hbar^2}{2 m' r^2} \\
&&\overset{\text{已知 4(b)(c)}}{=}& \frac{J(J+1)(1.054 \times 10^{-34}\text{ J}\cdot\text{s})^2}{2 \times  1.6726 \times 10^{-27}\text{ kg} (7.4 \times 10^{-11} \text{ m})^2} \\
&&\approx & J(J+1) \cdot 6.064\times 10^{-22}\text{ J} \\
\Rightarrow&\begin{bmatrix}  E_1 \\  E_2 \\ E_3  \\ E_4  \end{bmatrix} &=& \begin{bmatrix}  1\cdot(1+1) \cdot 6.064\times 10^{-22}\text{ J} \\  2\cdot(2+1) \cdot 6.064\times 10^{-22}\text{ J} \\ 3\cdot(3+1) \cdot 6.064\times 10^{-22}\text{ J}  \\ 4\cdot(4+1) \cdot 6.064\times 10^{-22}\text{ J}  \end{bmatrix}  \\
&\begin{bmatrix}  E_1 \\  E_2 \\ E_3  \\ E_4  \end{bmatrix} &=& \begin{bmatrix}  1.2128\times 10^{-21}\text{ J} \\ 3.6384\times 10^{-21}\text{ J} \\ 7.2768\times 10^{-21}\text{ J}  \\ 1.2128\times 10^{-20}\text{ J}  \end{bmatrix}  \\
&\begin{bmatrix}  E_1 \\  E_2 \\ E_3  \\ E_4  \end{bmatrix} &=& \begin{bmatrix} 7.57 \times 10^{-3} \text{ eV}\\ 2.27 \times 10^{-2} \text{ eV}\\ 4.54 \times 10^{-2} \text{ eV}\\ 7.57 \times 10^{-2} \text{ eV}\end{bmatrix}  \\
\end{gather*}$$

## Question 3

In Sec. 4.6 it was shown that, for large quantum numbers, the frequency of the radiation from a hydrogen atom that drops from an initial state of quantum number $n$ to a final state of quantum number $n - 1$ is equal to the classical frequency of revolution of an electron in the $n$th Bohr orbit. This is an example of Bohr's correspondence principle. 

Show that a similar correspondence holds for a diatomic molecule rotating about its center of mass.


## Question 3 - Answer



### 已知與假設




* **【假設 1】剛性轉子 (Rigid Rotor) 的模型 :** 假設兩個原子是點質量 $m_1$ 與 $m_2$，彼此之間的鍵長 $r$ 保持固定不變
  
  * (a) 幾何關係：
  
  $$r_1 + r_2 = r$$
  
  * (b)質心定義：
  
  $$m_1 r_1 = m_2 r_2$$

  * 可以固定 $I_{\perp}$ 為常數
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $r$ : 雙原子的固定鍵長 (Fixed bond length) $[\text{m}]$
  * $r_1, r_2$ : 兩原子各自到分子質心的距離 (Distances from each atom to the center of mass) $[\text{m}]$
  * $I_{\perp }$ : 垂直於鍵軸的轉動慣量 (Moment of inertia perpendicular to the bond axis) $[\text{kg}\cdot\text{m}^2]$


* **【已知 1】垂直於鍵軸的轉動動能 ($E_J$) :**
  
  $$\begin{gather*}
  KE &=& \frac{1}{2}I_{\perp}\omega^2 \\
  KE &=& \frac{(I_{\perp}\omega)^2}{2I_{\perp}} \\
  KE &=& \frac{L^2}{2I_{\perp}} \\
  E_J &\overset{\text{量子化}}{=}& \frac{L^2}{2I_{\perp}} \\
  E_J &=& \frac{J(J+1)\hbar^2}{2I_{\perp}} \\
  \end{gather*}$$

  * $KE$ : 經典力學的轉動動能 (Classical rotational kinetic energy) $[\text{J}]$
  * $\omega $ : 轉動角速度 (Angular velocity) $[\text{rad}\cdot\text{s}^{-1}]$
  * $L$ : 角動量大小 (Magnitude of angular momentum) $[\text{J}\cdot\text{s}]$
  * $E_{J}$ : 量子化後的轉動能階 (Rotational energy levels) $[\text{J}]$
  * $J$ : 轉動量子數 (Rotational quantum number) $J = 0, 1, 2, \dots$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$

* **【假設 2】 系統處於極高的激發態，旋轉量子數 $J$ 極大 ：**
  
  $$\begin{gather*}
  J &\gg& 1 \\
  J+1 &\approx& J \\
  \end{gather*}$$

  * $J$ : 轉動量子數 (Rotational quantum number) $J = 0, 1, 2, \dots$ [無單位]

* **【已知 2】 [愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis)  ：**
  
  $$\begin{gather*}
  E &\overset{\text{def}}{=}& h\nu \\
  \frac{E}{ h} &=& \nu \\
  \end{gather*}$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$
  * $\nu$ : 光子的頻率  $\nu =\frac{c}{\lambda}  $，$[\text{s}^{-1}]$ 



* **【已知 3】量子角動量 ($L$) :**

  $$\begin{gather*}
  L &=& \sqrt{J(J+1)}\hbar \\
  &\overset{\text{假設 2}}{\approx}& \sqrt{J (J)}\hbar \\
  &=& J \hbar
  \end{gather*}$$

  * $L$ : 角動量大小 (Magnitude of angular momentum) $[\text{J}\cdot\text{s}]$
  * $J$ : 轉動量子數 (Rotational quantum number) [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$

* **【已知 4】古典頻率 ($\nu_{\text{classical}}$) :**

  $$\begin{gather*}
  L &=& I \omega  \\
  L &=& 2\pi I \nu_{\text{classical}} \\
  \frac{L}{2\pi I} &=& \nu_{\text{classical}} \\
  \end{gather*}$$

  * $L$ : 角動量大小 (Magnitude of angular momentum) $[\text{J}\cdot\text{s}]$
  * $I_{\perp }$ : 垂直於鍵軸的轉動慣量 (Moment of inertia perpendicular to the bond axis) $[\text{kg}\cdot\text{m}^2]$
  * $\omega $ : 轉動角速度 (Angular velocity) $[\text{rad}\cdot\text{s}^{-1}]$
  * $\nu _{\text{classical}}$ : 古典轉動頻率 (Classical rotational frequency) $[\text{Hz}],[\text{s}^{-1}]$


### proof 

$$\begin{gather*}
\Delta E &=& E_J - E_{J-1} \\
\Delta E &=&\frac{J(J+1)\hbar^2}{2I_{\perp}} - \frac{(J-1)J\hbar^2}{2I_{\perp}} \\
\Delta E &=&\frac{J\hbar^2}{2I_{\perp}} \left( (J+1) - (J-1) \right) \\
\Delta E &=&\frac{J\hbar^2}{2I_{\perp}} ( 2 ) \\
\Delta E &=&\frac{J\hbar^2}{I_{\perp}} \\
\frac{\Delta E}{h} &=&\frac{1}{h} \frac{J\hbar^2}{I_{\perp}}  \\
\nu_{\text{quantum}} &\overset{\text{已知 2}}{=}& \frac{1}{h} \frac{J h\hbar}{2\pi I_{\perp}} \\
\nu_{\text{quantum}} &=&  \frac{J \hbar}{2\pi I_{\perp}} \\
\nu_{\text{quantum}} &\overset{\text{已知 3}}{=}&  \frac{L}{2\pi I_{\perp}} \\
\nu_{\text{quantum}} &\overset{\text{已知 4}}{=}&  \nu_{\text{classical}} \\
\end{gather*}$$


## Question 4

Assume that the $\text{H}_2$ molecule behaves exactly like a harmonic oscillator with a force constant of $573\text{ N/m}$. 

* (a) Find the energy (in $\text{eV}$) of its ground and first excited vibrational states. 
* (b) Find the vibrational quantum number that approximately corresponds to its $4.5\text{ eV}$ dissociation energy.




## Question 4 - Answer



### 假設與已知 (Assumptions & Preliminaries)


* **【假設 1】 建立一維座標系 One-dimensional coordinate system for diatomic vibration） ：**

  1. $m_1$ 的位置在 $x_1$
  2. $m_2$ 的位置在 $x_2$（假設 $x_2 > x_1$）
  3. 彈簧的自然長度（沒有被拉伸或壓縮時的長度）為 $L_0$
  4. 兩個原子之間的實際距離為 $x = x_2 - x_1$
  5. 彈簧被拉伸或壓縮的形變量可以表示為 $\Delta x = x - L_0$
 

  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $x_1, x_2$ : 兩原子在一維軸上的絕對位置 (Positions of the two atoms) $[\text{m}]$
  * $L_{0}$ : 鍵的平衡長度／自然長度 (Equilibrium bond length) $[\text{m}]$
  * $x$ : 雙原子間的瞬時實際鍵長 (Instantaneous bond length) $[\text{m}]$
  * $\Delta x$ : 遠離平衡位置的位移／形變量 (Displacement from equilibrium) $[\text{m}]$



* **【已知 1】 個別原子的牛頓第二運動定律與受力（Newton's second law for individual atoms）  ：**

  * 對於 $m_1$（受到向右的拉力）：
  
  $$\begin{gather*}
  \vec{F} &=& \vec{F}_{\text{spring}} \\
  m_1 \vec{a_1} &=& -k \Delta \vec{x} \\
  m_1 a_1 &=& k\Delta x  \\
  m_1 \frac{d^2 x_1}{dt^2} &=& k\Delta x  \\
  \frac{d^2 x_1}{dt^2} &=& \frac{k}{m_1}\Delta x \\
  \end{gather*}$$

  * 對於 $m_2$（受到向左的拉力）：
  
  $$\begin{gather*}
  \vec{F} &=& \vec{F}_{\text{spring}} \\
  m_2 \vec{a_2} &=& -k \Delta \vec{x} \\
  m_2 a_2 &=& -k\Delta x  \\
  m_2 \frac{d^2 x_2}{dt^2} &=& -k\Delta x  \\
  \frac{d^2 x_2}{dt^2} &=& -\frac{k}{m_2}\Delta x \\
  \end{gather*}$$

  * $k$ : 鍵的力常數／彈簧力常數 (Force constant of the chemical bond) $[\text{N}\cdot\text{m}^{-1}],[\text{kg}\cdot\text{s}^{-2}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $x$ : 雙原子間的瞬時實際鍵長 (Instantaneous bond length) $[\text{m}]$
  * $\Delta x$ : 遠離平衡位置的位移／形變量 (Displacement from equilibrium) $[\text{m}]$
  * $a_1, a_2$ : 兩原子各自的加速度大小 (Accelerations of individual atoms) $[\text{m}\cdot\text{s}^{-2}]$

* **【定義 1】縮減質量 (Reduce mass , $m'$) :**

  $$\begin{gather*}
  m' &\overset{\text{def}}{=}& \frac{m_1 m_2}{m_1 + m_2}  \\
  &\overset{\text{Q4}}{=}& \frac{m_p m_p}{m_p + m_p} \\
  &=& \frac{1}{2} m_p\\
  \end{gather*}$$

  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

* **【定義 2】系統的特徵角頻率 (Characteristic angular frequency of the system, $\omega$) :**

  $$\begin{gather*}
  \omega^2 &\overset{\text{def}}{=}& \frac{k}{m'} \\
  &\overset{\text{定義 1}}{=}& \frac{k}{\frac{1}{2} m_p}\\
  &=& 2\frac{k}{m_p}\\
  \end{gather*}$$

  * $\omega $ : 簡諧振動的特徵角頻率 (Characteristic angular frequency) $[\text{rad}\cdot\text{s}^{-1}]$
  * $k$ : 鍵的力常數／彈簧力常數 (Force constant of the chemical bond) $[\text{N}\cdot\text{m}^{-1}],[\text{kg}\cdot\text{s}^{-2}]$
  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

* **【已知 2】 相對運動的簡諧運動微分方程式（Differential equation of relative motion）  ：**

  $$\begin{gather*}
  \frac{d^2 x}{dt^2} &=& \frac{d^2 (x_2 - x_1)}{dt^2} \\
  \frac{d^2 x}{dt^2} &=& \frac{d^2 x_2}{dt^2} - \frac{d^2 x_1}{dt^2}  \\
  \frac{d^2 x}{dt^2} &=& -\frac{k}{m_2}\Delta x - \frac{k}{m_1}\Delta x  \\
  \frac{d^2 x}{dt^2} &=& -k \left( \frac{1}{m_2}+\frac{1}{m_1}  \right) \Delta x  \\
  \frac{d^2 x}{dt^2} &=& -k \left( \frac{m_1+ m_2}{m_2 m_1}  \right) \Delta x  \\
  \frac{d^2 x}{dt^2} &\overset{\text{定義 1}}{=}&  \frac{-k}{m'} \Delta x  \\
  \frac{d^2 }{dt^2} \left[ \Delta x + L_0 \right]&\overset{\text{假設 1}}{=}&  \frac{-k}{m'} \Delta x  \\
  \frac{d^2 \Delta x}{dt^2} &=&  \frac{-k}{m'} \Delta x  \\
  \frac{d^2 \Delta x}{dt^2} +  \frac{k}{m'} \Delta x  &=&0\\
  \frac{d^2 \Delta x}{dt^2} +  \omega^2 \Delta x  &\overset{\text{定義 2}}{=}&0\\
  \end{gather*}$$



* **【已知 3】 量子化能階  ：**
  
  $$\begin{gather*}
  E_v = \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
  \end{gather*}$$

  * $E_{v}$ : 量子化後的振動能階 (Vibrational energy levels) $[\text{J}]$
  * $v$ : 振動量子數 (Vibrational quantum number) $v = 0, 1, 2, \dots$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $\frac{1}{2}\hbar\omega$ : 零點能量 (Zero-point energy)，即基態（$v=0$）時的系統能量
  

### (a) vibrational states 能階對應的能量

1. 基態能量 ($v=0$)：

$$\begin{gather*}
E_v &\overset{\text{定義 1}}{=}& \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
E_0 &\overset{v=0}{=}& \left( 0 + \frac{1}{2} \right) \hbar \omega \\
E_0 &\overset{\text{定義 2}}{=}& \frac{1}{2} \hbar \sqrt{2\frac{k}{m_p}} \\
E_0 &=& \frac{1}{\sqrt{2}} \hbar \sqrt{\frac{k}{m_p}} \\
E_0 &\overset{\text{Q4}}{=}& \frac{1}{\sqrt{2}} 1.054 \times 10^{-34}\text{ J}\cdot\text{s}  \sqrt{\frac{573\text{ N/m}}{1.6726 \times 10^{-27}\text{ kg}}} \\
E_0 &=&  4.36 \times 10^{-20} \text{ J} \\
E_0 &\approx& 0.272 \text{ eV} \\
\end{gather*}$$

2. 第一激發態能量 ($v=1$)：

$$\begin{gather*}
E_v &\overset{\text{定義 1}}{=}& \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
E_1 &\overset{v=1}{=}& \left( 1 + \frac{1}{2} \right) \hbar \omega \\
E_1 &\overset{\text{定義 2}}{=}& \frac{3}{2} \hbar \sqrt{2\frac{k}{m_p}} \\
E_1 &=& \frac{3}{\sqrt{2}} \hbar \sqrt{\frac{k}{m_p}} \\
E_1 &\overset{\text{Q4}}{=}& \frac{3}{\sqrt{2}} 1.054 \times 10^{-34}\text{ J}\cdot\text{s}  \sqrt{\frac{573\text{ N/m}}{1.6726 \times 10^{-27}\text{ kg}}} \\
E_1 &=&  1.31 \times 10^{-19} \text{ J} \\
E_1 &\approx& 0.817 \text{ eV} \\
\end{gather*}$$

### (b) 解離能對應到的 vibrational quantum number




$$\begin{gather*}

## Question 5

The lowest vibrational states of the $^{23}\text{Na}^{35}\text{Cl}$ molecule are $0.063\text{ eV}$ apart. Find the approximate force constant of this molecule.


## Question 5 - Answer


### 已知與假設




* **【已知 1】 量子化能階  ：**
  
  $$\begin{gather*}
  E_v = \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
  \end{gather*}$$

  * $E_{v}$ : 量子化後的振動能階 (Vibrational energy levels) $[\text{J}]$
  * $v$ : 振動量子數 (Vibrational quantum number) $v = 0, 1, 2, \dots$ [無單位]
  * $\hbar $ : 約化普朗克常數 (Reduced Planck constant) $\hbar \approx 1.054 \times 10^{-34}\text{ J}\cdot\text{s}$
  * $\frac{1}{2}\hbar\omega$ : 零點能量 (Zero-point energy)，即基態（$v=0$）時的系統能量
  

* **【定義 1】縮減質量 (Reduce mass , $m'$) :**

  $$\begin{gather*}
  m' &\overset{\text{def}}{=}& \frac{m_1 m_2}{m_1 + m_2}  \\
  &\overset{\text{Q5}}{=}& \frac{23 \text{u} \cdot 35\text{u}}{23\text{u} + 35\text{u}} \\
  &=& \frac{805}{58} \text{u}\\
  &\approx& \frac{805}{58} m_p\\
  \end{gather*}$$

  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_1, m_2$ : 兩原子的質量 (Masses of the two atoms) $[\text{kg}]$
  * $\text{u}$ : 原子質量單位 (Atomic Mass Unit) $1\text{u} \approx 1 m_p $ ，$[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

* **【定義 2】系統的特徵角頻率 (Characteristic angular frequency of the system, $\omega$) :**

  $$\begin{gather*}
  \omega^2 &\overset{\text{def}}{=}& \frac{k}{m'} \\
  &\overset{\text{定義 1}}{=}& \frac{k}{\frac{805}{58} m_p}\\
  &=& \frac{58}{805}\frac{k}{m_p}\\
  \end{gather*}$$

  * $\omega $ : 簡諧振動的特徵角頻率 (Characteristic angular frequency) $[\text{rad}\cdot\text{s}^{-1}]$
  * $k$ : 鍵的力常數／彈簧力常數 (Force constant of the chemical bond) $[\text{N}\cdot\text{m}^{-1}],[\text{kg}\cdot\text{s}^{-2}]$
  * $m^{\prime }$ : 系統的縮減質量 (Reduced mass of the system) $[\text{kg}]$
  * $m_{p}$ : 質子質量 (Proton mass) $m_p \approx 1.6726 \times 10^{-27}\text{ kg}$

###  solve force constant

$$\begin{gather*}
\Delta E &=& E_{v+1} - E_v \\
\Delta E &\overset{\text{已知 1}}{=}& \left( v + 1 + \frac{1}{2} \right) \hbar \omega - \left( v + \frac{1}{2} \right) \hbar \omega & v = 0, 1, 2, \dots \\
\Delta E &=& \hbar \omega  \\
\Delta E &\overset{\text{定義 2}}{=}& \hbar \sqrt{\frac{58}{805}\frac{k}{m_p}}  \\
k &=& \frac{805}{58} m_p \left( \frac{\Delta E }{\hbar} \right)^2 \\
k &=& \frac{805}{58} \cdot 1.6726 \times 10^{-27}\text{ kg} \left( \frac{0.063\text{ eV}}{1.054 \times 10^{-34}\text{ J}\cdot\text{s}} \right)^2   \\
k &=& 212.85 \text{ N}\cdot\text{m}^{-1} \\
\end{gather*}$$
