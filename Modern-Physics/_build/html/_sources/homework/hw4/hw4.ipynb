## Question 1 


The **fine structure constant(精細結構常數)** is defined as $\alpha = \frac{e^2}{ 2\epsilon_0 hc}$. This quantity got its name because it first appeared in a theory by the German physicist Arnold Sommerfeld that tried to explain the fine structure in spectral lines (multiple lines close together instead of single lines) by assuming that elliptical as well as circular orbits are possible in the Bohr model. Sommerfeld’s approach was on the wrong track, but $\alpha$ has nevertheless turned out to be a useful quantity in atomic physics.


* (a) Show that $\alpha = \frac{v_1}{c}$, where $v_1$ is the velocity of the electron in the ground state of the Bohr atom.
* (b) Show that the value of $\alpha$ is very close to $\frac{1}{ 137}$ and is a pure number with no dimensions. Because the magnetic behavior of a moving charge depends on its velocity, the small value of $\alpha$ is representative of the relative magnitudes of the magnetic and electric aspects of electron behavior in an atom.
* (c) Show that $\alpha a_0 = \frac{\lambda_c}{ 2\pi}$, where $a_0$ is the radius of the ground-state Bohr orbit and $\lambda_c$ is the Compton wavelength of the electron.







## Question 1 - Answer



### 假設與已知 (Assumptions & Preliminaries)

* **【定義 1】 精細結構常數：** 

  $$\alpha = \frac{e^2}{ 2\epsilon_0 hc}$$

  * $e$ : 基本電荷 (Elementary charge)，$e \approx 1.602 \times 10^{-19} text{ C}$ 
  * $\epsilon_0$ : 真空電容率 (Vacuum permittivity)，$\epsilon_0 \approx 8.854 \times 10^{-12} \text{ F}\cdot\text{m}^{-1} : [\text{ C}^{2}\cdot\text{Nt}^{-1}\cdot\text{m}^{-2}]$
  * $c$ : 真空中光速 (Speed of light)，$c \approx 3 \times 10^8 \text{ m}\cdot\text{s}^{-1}$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} 


* **【定義 2】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：**
  
  $$\lambda \overset{\text{def}}{=} \frac{h}{p}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\lambda_{c}$  :  一個光子的波長，這個光子擁有的能量剛好等同於一個靜止質量 $\lambda_c = \frac{h}{m_0c}$ $[\text{m}]$
  

* **【已知 1】 力學平衡（靜電力 = 向心力）  ：**
  
  $$\begin{gather*}
  \mathbf{F}_{\text{Electrostatic}} &\overset{\text{balance}}{=}& \mathbf{F}_{\text{Centripetal}} \\
  -\frac{kQq}{r^{2}} \hat{r} &=& \frac{d\mathbf{p}}{dt} \\
  -\frac{kQq}{r^{2}} \hat{r} &=& \frac{d}{dt}[\gamma m_0 \mathbf{v}] \\ 
  -\frac{kQq}{r^{2}} \hat{r}  &=& m_0 \left[ \frac{d\gamma}{dt}\mathbf{v} + \gamma\frac{d\mathbf{v}}{dt} \right] \\
  -\frac{kQq}{r^{2}} \hat{r} &=& m_0 \left[ 0\mathbf{v} + \gamma\frac{d\mathbf{v}}{dt} \right] \\
  -\frac{kQq}{r^{2}} \hat{r} &=&  \gamma m_0 \mathbf{a} \\
  -\frac{kQq}{r^{2}} \hat{r} &=& \gamma m_0 \frac{v^2}{r} (-\hat{r}) \\
  \frac{kQq}{r^{2}} &=& \frac{\gamma m_0v^{2}}{r} \\ 
  \frac{1}{4\pi\epsilon_{0}} \frac{e^{2}}{r^{2}}   &=&\frac{\gamma m_0 v^{2}}{r} \\
  \frac{1}{4\pi\epsilon_{0}} e^{2}   &=&\gamma m_0 v^{2} r \\
  \end{gather*}$$

  * $e$ : 基本電荷 (Elementary charge)，$e \approx 1.602 \times 10^{-19} \text{ C}$ 
  * $\epsilon_0$ : 真空電容率 (Vacuum permittivity)，$\epsilon_0 \approx 8.854 \times 10^{-12} \text{ F}\cdot\text{m}^{-1} : [\text{ C}^{2}\cdot\text{Nt}^{-1}\cdot\text{m}^{-2}]$
  * $r$ : 軌道半徑 $[\text{m}]$
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


* **【已知 2】 [相對論動量 (Relativistic Momentum)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)  ：**
  
  $$p = \gamma m_0 v $$

  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation)
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$

* **【已知 3】 量子化條件（駐波條件）  ：**
  
  $$\begin{gather*}
  n\lambda &=& 2\pi r_{n} \\ 
  n\frac{h}{p_n} &\overset{\text{定義 2}}{=}& 2\pi r_{n} \\
  n\frac{h}{\gamma m_0 v_n} &\overset{\text{已知 2}}{=}& 2\pi r_{n} \\
  \gamma m_0v_nr_n &=& n \frac{h}{2\pi}\\
  \gamma m_0v_1r_1 &\overset{\text{基態 n=1}}{=}&  \frac{h}{2\pi}\\
  \end{gather*}$$

  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$
  * $r$ : 軌道半徑 $[\text{m}]$ 
  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * 波耳定義波耳半徑$a_0 \overset{\text{def}}{=} r_1 =  \frac{h}{2\pi m_0 v_{1}}$ ， $\gamma \sim 1$






### proof(a): 

$$\left\{\begin{matrix}
\gamma m_0 v_{1}^{2} r &\overset{\text{已知 1}}{=}& \frac{1}{4\pi\epsilon_{0}} e^{2} \\\gamma m_0v_1 r_1 &\overset{\text{已知 3}}{=}&  \frac{h}{2\pi}
\end{matrix}\right.$$

$$\begin{gather*}
\Rightarrow & \frac{mv_{1}^{2}r_{1}}{mv_{1}r_{1}} &=& \frac{\quad \frac{e^{2}}{4\pi\epsilon_{0}} \quad}{ \frac{h}{2\pi} } \\
&v_{1} &=& \frac{e^{2}}{\quad4\pi\epsilon_{0}\quad} \cdot \frac{2\pi}{h} \\
&v_{1}&=& \frac{e^{2}}{2\epsilon_{0}h} \\
&\frac{v_{1}}{c}&=& \frac{e^{2}}{2\epsilon_{0}h} \\
&\frac{v_{1}}{c}&=& \frac{e^{2}}{2\epsilon_{0}hc} &\overset{\text{定義 1}}{=}\alpha\\
\end{gather*}$$

### proof(b): 



$$\begin{gather*}
\alpha &\overset{\text{定義 1}}{=}& \frac{e^{2}}{2\epsilon_{0}hc}  \\
&\overset{\text{定義 1}}{\approx}&\frac{(1.602 \times 10^{-19} \text{ C})^{2}}{2 \times  8.854 \times 10^{-12} \text{ F/m} \times  6.626 \times 10^{-34} \text{ J}\cdot\text{s} \times  2.998 \times 10^{8} \text{ m/s} } \\
&\approx& 0.007294 \\
&\approx& \frac{1}{137}
\end{gather*}$$


### proof(c): 


$$\begin{gather*}
\alpha a_{0} &\overset{\text{Q1(a),已知 3}}{=}& \frac{v_{1}}{c} \frac{h}{2\pi m_0 v_{1}}  \\
&=& \frac{h}{2\pi m_0 c} \\
&\overset{\text{定義 2}}{=}&  \frac{\lambda_{c}}{2\pi}  \\
\end{gather*}$$

## Question 2 

An excited hydrogen atom emits a photon of wavelength $\lambda$ in returning to the ground state.

* (a) Derive a formula that gives the quantum number of the initial excited state in terms of $\lambda$ and Rydberg constant(芮得柏常數) $R$ .
* (b) Use this formula to find $n_i$ for a $102.55\text{-nm}$ photon.



## Question 2 - Answer


### 已知與假設



* **【已知 1】芮得柏公式 (Rydberg formula):**

  $$\frac{1}{\lambda} = R \left( \frac{1}{n_{f}^2} - \frac{1}{n_{i}^2} \right)$$

  * $\lambda$ 是發射光子的波長 $[\text{ m}]$
  * $R$ 是芮得柏常數 (Rydberg constant) $R \approx 1.097 \times 10^7\text{ m}^{-1}$
  * $n_f$ 是最終狀態的主量子數 (final state)，回到基態 $n_f = 1$
  * $n_i$ 是初始狀態的主量子數 (initial state)，且 $n_i > n_f$






### proof(a)



$$\begin{gather*}
\frac{1}{\lambda} &\overset{\text{已知 1}}{=}& R \left( \frac{1}{n_{f}^2} - \frac{1}{n_{i}^2} \right) \\
\frac{1}{\lambda} &\overset{\text{已知 1,Q2}}{=}& R \left( \frac{1}{1^2} - \frac{1}{n_{i}^2} \right) \\
\frac{1}{\lambda R} &=& 1 - \frac{1}{n_{i}^2}\\
\frac{1}{n_{i}^2} &=& 1 - \frac{1}{\lambda R}\\
\frac{1}{n_{i}^2} &=& \frac{\lambda R - 1}{\lambda R}\\
n_{i} &=& \sqrt{\frac{\lambda R}{\lambda R - 1}}\\
\end{gather*}$$

### compute(b)

$$\begin{gather*}
n_{i} &\overset{\text{Q2(a)}}{=}& \sqrt{\frac{\lambda R}{\lambda R - 1}}\\
n_{i} &\overset{\text{已知 1,Q2}}{=}& \sqrt{\frac{102.55 \times 10^{-9}\text{ m} \times 1.097 \times 10^7\text{ m}^{-1}}{102.55 \times 10^{-9}\text{ m} \times 1.097 \times 10^7\text{ m}^{-1} - 1}}\\
n_{i} &\approx& \sqrt{\frac{1.125}{0.125}}\\
&=&3
\end{gather*}$$

* 這顆光子是電子從 $n=3$ 的能階掉到 $n=1$ 的基態所發射出來的（來曼系 Lyman series 的第二條譜線）


## Question 3

An excited atom of mass $m$ and initial speed $v$ emits a photon in its direction of motion. If $v \ll c$, use the requirement that linear momentum and energy must both be conserved to show that the frequency of the photon is higher by $\frac{\Delta\nu}{\nu} \approx \frac{v}{c}$ than it would have been if the atom had been at rest. (See also Exercise 16 of Chap. 1.)





## Question 3 - Answer



### 已知與假設

* **【定義 1】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：**
  
  $$\lambda \overset{\text{def}}{=} \frac{h}{p}$$

  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


* **【已知 1】動量守恆:**
  
   $$\begin{gather*}
   mv &=& mv' + p_{\text{photon}} \\
   mv &\overset{\text{定義 1}}{=}& mv' + \frac{h\nu}{c} \\
   v' &=& v - \frac{h\nu}{mc} 
   \end{gather*}$$

* **【定義 2】 [愛因斯坦的光量子假說](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/02_Particle_Properties_of_Waves/particle_properties_of_waves.html#einsteins-photon-hypothesis)  ：**
  
  $$E \overset{\text{def}}{=} h\nu$$

  * $E$ : 總能量 (Total Energy)  $[\text{J}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $\nu$ : 光子的頻率  $[\text{s}^{-1}]$


* **【已知 2】能量守恆:**
  
   $$\begin{gather*}
   \frac{1}{2}mv^2 + E_i &=& \frac{1}{2}m(v')^2 + E_f + E_{\text{photon}}\\
   \frac{1}{2}mv^2 + E_i &\overset{\text{定義 2}}{=}&  \frac{1}{2}m(v')^2 + E_f + h\nu\\
   \frac{1}{2}mv^2 +  h\nu_0 &=&  \frac{1}{2}m(v')^2  + h\nu\\
   h\nu_0 &=& \frac{1}{2}m(v')^2 - \frac{1}{2}mv^2 + h\nu 
   \end{gather*}$$


* **【已知 3】尺度分析:**

   $$\begin{gather*}
   \text{ Ratio}  &=& \frac{ \frac{h^2\nu^2}{2mc^2} }{ \frac{v}{c}h\nu }\\
   &=& \frac{h\nu}{2mc^2} \times \frac{c}{v} \\
   &=& \frac{h\nu / c}{2mv}\\
   \end{gather*}$$

   * 題目說 $v \ll c$ 
   * 將真實物理數量級代入測試 $\frac{h\nu}{2mc^2} \times \frac{c}{v} \approx \frac{10\text{ eV}}{2 \times 10^9\text{ eV}} \frac{3 \times 10^8}{10^3} \approx 1.5 \times 10^{-3} $
   * 比例約為 $10^{-3}$ 倍所以可以忽略 $\frac{h^2\nu^2}{2mc^2}$
  


### proof


$$\begin{gather*}
h\nu_0 &\overset{\text{已知 2}}{=}& \frac{1}{2}m(v')^2 - \frac{1}{2}mv^2 + h\nu \\
h\nu_0 &\overset{\text{已知 1}}{=}& \frac{1}{2}m \left( v - \frac{h\nu}{mc} \right)^2 - \frac{1}{2}mv^2 + h\nu\\
h\nu_0 &=& \frac{1}{2}m \left( v^2 - 2v\frac{h\nu}{mc} + \left(\frac{h\nu}{mc}\right)^2 \right) - \frac{1}{2}mv^2 + h\nu\\
h\nu_0 &=& \frac{1}{2}mv^2 - \frac{mvh\nu}{mc} + \frac{m}{2}\left(\frac{h\nu}{mc}\right)^2 - \frac{1}{2}mv^2 + h\nu\\
h\nu_0 &=& - \frac{v}{c}h\nu + \frac{h^2\nu^2}{2mc^2} + h\nu\\
h\nu_0 &\overset{\text{已知 3}}{\approx}& - \frac{v}{c}h\nu + h\nu\\
\nu_0 &=& \nu - \nu\frac{v}{c}\\
\Delta\nu &=& \nu\frac{v}{c}\\
\frac{\Delta\nu}{\nu} &=& \frac{v}{c}\\
\end{gather*}$$


## Question 4

* **【定義 1】 [德布羅依波 (De Broglie Wave)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/03_Wave_Properties_of_Particles/wave_properties_of_particles.html#de-broglie-wave)  ：**

  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $

  -\frac{kQq}{r^{2}} \hat{r} &=& \gamma m_0 \frac{v^2}{r} (-\hat{r}) \\

  * $\epsilon_0$ : 真空電容率 (Vacuum permittivity)，$\epsilon_0 \approx 8.854 \times 10^{-12} \text{ F}\cdot\text{m}^{-1} : [\text{ C}^{2}\cdot\text{Nt}^{-1}\cdot\text{m}^{-2}]$

  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


* **【已知 2】 [相對論動量 (Relativistic Momentum)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#relativistic-momentum)  ：**
  
  $$p = \gamma m_0 v  = m_0 v$$

  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$
  * $\gamma$ : [勞侖茲因子 (Lorentz factor)](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/01_Relativity/relativity.html#lorentz-transformation) $\gamma \sim 1$
  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$

* **【已知 3】 量子化條件（駐波條件）  ：**
  
  $$\begin{gather*}
  n\lambda &=& 2\pi r_{n} \\ 
  n\frac{h}{p_n} &\overset{\text{定義 1}}{=}& 2\pi r_{n} \\
  n\frac{h}{ m_0 v_n} &\overset{\text{已知 2}}{=}& 2\pi r_{n} \\
  m_0v_nr_n &=& n \frac{h}{2\pi}\\
  m_0v_1r_1 &\overset{\text{基態 n=1}}{=}&  \frac{h}{2\pi}\\
  \end{gather*}$$

  * $m_0$ : 物體靜止質量 (Rest mass)，$[\text{kg}]$
  * $v$ : 物體速度 (Velocity)，$[\text{m}\cdot\text{s}^{-1}]$
  * $r$ : 軌道半徑 $[\text{m}]$ 
  * $\lambda$ : 德布羅依波長 (de Broglie wavelength) $[\text{m}]$
  * $h$ : 普朗克常數 (Planck's constant) $h \approx 6.626 \times 10^{-34} \text{ J}\cdot\text{s} $
  * $p$ : 物體動量 (Momentum) $[\text{kg}\cdot\text{m}\cdot\text{s}^{-1}]$


* **【已知 4】速度 $v_n$  ：**
  
  $$\begin{gather*}
  \frac{m v_n^2 r_n}{m v_n r_n} &\overset{\text{已知 1,3}}{=}& \frac{ \frac{e^2}{4\pi\epsilon_0} }{ n \frac{h}{2\pi} }\\
  v_n &=& \frac{e^2}{2\epsilon_0 n h}\\
  v_n &\propto& \frac{1}{n}
  \end{gather*}$$
  

* **【已知 5】軌道半徑 $r_n$  ：**
  
  $$\begin{gather*}
  m v_n r_n  &\overset{\text{已知 3}}{=}& n \frac{h}{2\pi}\\
  \frac{1}{n} \cdot r_n &\overset{\text{已知 4}}{\propto}& n \\
  r_n &\propto& n^2
  \end{gather*}$$




### relationship





1. 旋轉頻率 (Frequency of revolution, $f$)
   * 頻率是「速度除以圓周長」， $f = \frac{v_n}{2\pi r_n} \overset{\text{已知 4,5}}{\propto} \frac{1}{n^3}$
   * **物理意義：** $n$ 變大時，電子不但跑得變「慢」($v$ 變小)、要跑的「一圈距離」還變得超級長 ($r$ 變大)，繞一圈所花的時間變很久，頻率自然就大幅下降了
   * 當 $n$ 增加時，旋轉頻率 減少 (Decrease)
2. 電子速度 (Electron speed, $v$)
   * $v_n \overset{\text{已知 4}}{\propto} \frac{1}{n}$
   * **物理意義：** 電子離原子核越遠（樓層越高），原子核對它的吸引力越弱，所以它不需要跑那麼快來維持圓周運動
   * 當 $n$ 增加時，電子速度 減少 (Decrease)
3. 電子波長 (Electron wavelength, $\lambda$)
   * $\lambda = \frac{h}{mv_n}\overset{\text{已知 4}}{\propto} \frac{1}{1/n} = n$
   * **物理意義：** 速度越慢，物質波的波長就越長
   * 當 $n$ 增加時，電子波長 增加 (Increase)
4. 角動量 (Angular momentum, $L$)
   * $L = n\frac{h}{2\pi}$
   * **物理意義：** 軌道本身就是用角動量的整數倍來定義的，$n$ 越大 角動量當然直接跟著等比例變大
   * 當 $n$ 增加時，角動量 增加 (Increase)
5. 動能 (Kinetic energy, $KE$)
   * $KE = \frac{1}{2}mv_n^2 \overset{\text{已知 4}}{\propto} \left(\frac{1}{n}\right)^2 = \frac{1}{n^2}$
   * **物理意義：** 既然電子跑得變慢了，動能當然也就跟著變小
   * 當 $n$ 增加時，動能 減少 (Decrease)
6. 位能 (Potential energy, $U$)
   * $U = -\frac{1}{4\pi\epsilon_0} \frac{e^2}{r_n} \overset{\text{已知 5}}{\propto} -\frac{1}{n^2} $
   * **物理意義：** 當 $n$ 增加時，軌道變大、分母變大，所以數值的「絕對值」變小（例如從 $-10$ 變成 $-2$），在數學和物理上，從 $-10$ 爬升到 $-2$，代表位能是「上升」的！電子要往高樓層爬，必須克服原子核的吸引力作功，所以位能增加。
   * 當 $n$ 增加時，位能 增加 (Increase)
7. 總能量 (Total energy, $E$)
   * $E = KE + U = KE + (-2KE) \overset{\text{已知 4}}{\propto} -KE \propto -\frac{1}{n^2} $
   * **物理意義：** 和位能的邏輯一樣，隨著 $n$ 變大，總能量的數值從很負（例如基態 $-13.6 \text{ eV}$）變得比較不負（例如 $n=2$ 是 $-3.4 \text{ eV}$），這代表原子的總能量是爬升的。
   * 當 $n$ 增加時，總能量 增加 (Increase)
   




## Question 5


Show that the probability for a $2.0\text{-MeV}$ proton to be scattered by more than a given angle $\theta$ when it passes through a thin foil is the same as that for a $4.0\text{-MeV}$ alpha particle.




## Question 5 - Answer


### 已知與假設


* **【已知 1】 拉塞福散射 (Rutherford scattering)**

  $$b = \frac{z Z e^2}{4\pi\epsilon_0 KE} \cot\left(\frac{\theta}{2}\right)$$

  * $b$ : 碰撞參數 (Impact Parameter) 
  * $e$ : 基本電荷 (Elementary charge)，$e \approx 1.602 \times 10^{-19} text{ C}$ 
  * $\epsilon_0$ : 真空電容率 (Vacuum permittivity)，$\epsilon_0 \approx 8.854 \times 10^{-12} \text{ F}\cdot\text{m}^{-1} : [\text{ C}^{2}\cdot\text{Nt}^{-1}\cdot\text{m}^{-2}]$
  * $z$：入射粒子的原子序（也就是它帶了幾個基本電荷 $e$）
  * $Z$：金屬靶核的原子序
  * $KE$：入射粒子的動能 ($KE = \frac{1}{2}mv^2$)
  * $\theta$：散射角度。

* **【已知 2】 穿過薄箔發生這種大角度散射的機率 $P(\theta)$** 
  
  $$\begin{gather*}
  P(\theta) &\propto& \text{截面積} \\
  &\propto&\pi b^2 \\
  &\propto& \left( \frac{z}{KE} \right)^2
  \end{gather*}$$
  
  * 對於同一個金屬箔和同一個給定角度 $\theta$ 、金屬靶核的原子序 $Z$、$e$、$\epsilon_0$ 和 $\cot(\theta/2)$ 均為常數


* **【已知 3】 質子 (Proton) 基本資訊**
  
  $$\frac{z_{\text{p}}}{KE_{\text{p}}} = \frac{1}{2.0 \text{ MeV} } = 0.5\text{ MeV}^{-1}$$

  * 質子是氫原子核，帶 1 個正電，所以電荷數 $z_{\text{p}} = 1$
  * 題目給定質子的動能 $KE_{\text{p}} = 2.0\text{ MeV}$
  

* **【已知 4】$\alpha$ 粒子 (Alpha particle) 基本資訊**

  $$\frac{z_\alpha}{KE_\alpha} = \frac{2}{4.0 \text{ MeV} } = 0.5\text{ MeV}^{-1}$$

  * $\alpha$ 粒子是氦原子核（包含 2 個質子），帶 2 個正電，所以電荷數 $z_\alpha = 2$
  * 題目給定 $\alpha$ 粒子的動能 $KE_\alpha = 4.0\text{ MeV}$ 
  


### proof 

$$\begin{gather*}
\frac{P_{\text{p}}(\theta)}{ P_{\alpha}(\theta)} &\overset{\text{已知 2}}{=} & \frac{\quad\left( \frac{z_{\text{p}}}{K_{\text{p}}} \right)^2\quad}{\left( \frac{z_{\alpha}}{K_{\alpha}} \right)^2} \\
&\overset{\text{已知 3,4}}{=}&\frac{\quad \left( \frac{1}{2.0 \text{ MeV} } \right)^2 \quad}{\left(\frac{2}{4.0 \text{ MeV} }\right)^2}\\
&=& 1
\end{gather*}$$
