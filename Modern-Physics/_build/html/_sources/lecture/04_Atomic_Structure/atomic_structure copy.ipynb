## 4.6 原子核運動 Nuclear Motion

* 原子核質量影響光譜線的波長，原子核越重這個效應越不明顯
* 兩種粒子質量越接近放出的光能量越小 


減縮質量
$m' = \frac{mM}{m + M}$


考慮原子核運動後修正的能階
$E'_n = -\frac{m'e^4}{8\epsilon_0^2 h^2} \left( \frac{1}{n^2} \right) = \left( \frac{m'}{m} \right) \left( \frac{E_1}{n^2} \right) = \frac{m'}{m} E_n$

這導致修正後的能階 $E_n'$ 比原來的 $E_n$ 稍微淺（絕對值變小）一點。
這個修正成功解釋了同位素位移（例如氫與氘的光譜差異）

**例題4.6.1** 氫原子
在氫原子中，電子質量為 $m$，原子核（質子）質量為 $M$。根據實驗測量：

* $m \approx 9.109 \times 10^{-31} \text{ kg}$
* $M \approx 1.672 \times 10^{-27} \text{ kg}$
* 質子與電子的質量比 $M/m \approx 1836.15$

計算過程
減縮質量 $m'$ 與電子質量 $m$ 的比例公式為：
$$\frac{m'}{m} = \frac{\frac{mM}{m+M}}{m} = \frac{M}{m+M}$$ 
將分子分母同除以 $m$：
$$\frac{m'}{m} = \frac{M/m}{1 + M/m}$$ 
代入數值：
$$\frac{m'}{m} = \frac{1836.15}{1 + 1836.15} = \frac{1836.15}{1837.15} \approx \mathbf{0.99945}$$ 

物理意義

* 誤差極小：這個比例非常接近 $1$（誤差僅約 $0.05\%$），這說明波耳模型假設原子核靜止不動是一個非常好的近似。
* 能階修正：修正後的能階 $E'_n$ 會是原本的 $0.99945$ 倍。雖然差異很小，但在高解析度的光譜實驗中是可以觀察到的。
* 同位素對比：如果是氘原子（原子核含一個質子一個中子，$M \approx 3670m$），其 $m'/m$ 約為 $0.99973$。氫與氘之間這微小的比例差異，就是發現氘（重氫）的關鍵證據。

**例題4.6.2**  
**正電子「原子」** 為一個正電子(positron)和一個電子相互繞行的系統。試比較正電子原子的光譜線波長與一般氫原子光譜線的波長。

m' = \frac{mM}{m + M} = \frac{m^2}{2m} = \frac{m}{2}

E'_n = \left( \frac{m'}{m} \right) \frac{E_1}{n^2} = \frac{E_1}{2n^2}



這表示雷德堡常數在正電子原子中為一般氫原子的一半。因此正電子原子光譜線所有的波長均為氫原子光譜線波長的兩倍

**例題4.6.3**  

渺是一種不穩定的基本粒子，質量為電子質量(me)的207倍，帶一單位的正電或負電，一個負渺(mu )可以被一個原子核攫取而形成一個渺原子。(a) 一個質子攫取一個負渺。求出這個原子的第一波耳軌道半徑。(b) 求出這個原子的游離能

E'_1 = \left( \frac{m'}{m} \right) E_1 = 186E_1 = -2.53 \times 10^3 \text{ eV} = -2.53 \text{ keV}
