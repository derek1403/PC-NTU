## 1. Classical Physics (古典物理)

### 1.1 Mechanics (力學與粒子性)

古典力學主要探討具備質量的粒子行為，核心由重力（Gravitational force）與牛頓運動定律（Newton's laws of motion）建構。

* **第一運動定律 (慣性定律):** 靜者恆靜、動者恆動。
* **物理視角：** 為什麼要有第一定律？明明第一定律看起來只是第二定律在 $\mathbf{F}=0$ 時的特例（沒有力就沒有加速度）。事實上，第一定律的核心價值在於 **定義「慣性坐標系」**（Inertial frame of reference）。它宣告了宇宙中存在一種觀察視角，在這種視角下，不受力的物體會保持等速直線運動。後續所有的牛頓力學甚至愛因斯坦的狹義相對論，都是建立在慣性坐標系的假設之上。


* **第二運動定律:** $\mathbf{F} = m\mathbf{a}$
* **概念類比：** 這條定律其實是在定義 **質量（Mass）** 。就像電路學中的歐姆定律 $V = IR$ 定義了「電阻」是電壓與電流的比例常數；牛頓第二定律定義了「質量」是物體在受力（$\mathbf{F}$）與產生加速度（$\mathbf{a}$）之間的抵抗係數（慣性大小）。


* **第三運動定律 (作用力與反作用力):** 對於每一個作用力 $\mathbf{F}_{ij}$，必然存在一個大小相等、方向相反的反作用力 $\mathbf{F}_{ji} = -\mathbf{F}_{ij}$。

### 1.2 Electromagnetism (電磁學與波動性)

古典電磁學主要探討電磁力（Electromagnetic force）與波動行為，其核心為馬克士威方程組（Maxwell's Equations）。

1. 高斯定律 (Gauss's Law)
2. 高斯磁定律 (Gauss's Law for Magnetism)
3. 法拉第電磁感應定律 (Faraday's Law of Induction)
4. 安培-馬克士威定律 (Ampère-Maxwell Law)

$$\begin{cases} 
\nabla \cdot \mathbf{E} &=& \frac{\rho}{\varepsilon_0} \\ 
\nabla \cdot \mathbf{B} &=& 0 \\
\nabla \times \mathbf{E} &=& -\frac{\partial \mathbf{B}}{\partial t} \\
\nabla \times \mathbf{B} &=& \mu_0 \left( \mathbf{J} + \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t} \right)
\end{cases}$$

若考慮介質中的磁場與電場，可引入輔助場 $\mathbf{H}$ 與極化向量 $\mathbf{P}$（此處 $\mathbf{D} = \varepsilon_0 \mathbf{E} + \mathbf{P}$，在特定單位制下可寫作你筆記中的形式，但國際標準制 SI 如下）：

$$\nabla \times \mathbf{H} = \mathbf{J}_f + \frac{\partial \mathbf{D}}{\partial t}$$

**【符號精確定義】**

* $\mathbf{E}$: 電場 (Electric field) $[\text{V} \cdot \text{m}^{-1}]$ 
* $\mathbf{B}$: 磁場 (Magnetic field)，$[\text{T}] = [\text{Tesla}] $ 
* $\mathbf{P}$: 電極化強度 (Electric Polarization) $[\text{C} \cdot \text{m}^{-2}] $
* $\mathbf{D}$: 電位移場 (Electric Displacement Field) $[\text{C} \cdot \text{m}^{-2}] $
* $\mathbf{H}$: 磁場強度 (Magnetic Field Intensity) $[\text{A} \cdot \text{m}^{-1}] $
* $c$: 真空中的光速，精確值為 $299,792,458$ $[\text{m} \cdot \text{s}^{-1}] $ ，近似計算常取 $3 \times 10^8$ $[\text{m} \cdot \text{s}^{-1}] $
* $\rho$: 體電荷密度 (Volume charge density) $[\text{C} \cdot \text{m}^{-3}] $
* $\mathbf{J}$: 電流密度 (Current density) $[\text{A} \cdot \text{m}^{-2}] $
* $\mathbf{J}_f$: 自由電流密度 (Free current density)  $[\text{A} \cdot \text{m}^{-2}] $
* $\mu_0$: 真空磁導率 (Vacuum permeability)，精確值為 $4\pi \times 10^{-7}$ $[\text{T} \cdot \text{m} \cdot \text{A}^{-1}] $ 
* $\varepsilon_0$: 真空電容率 (Vacuum permittivity)，近似值為 $8.854 \times 10^{-12}$ $[\text{F} \cdot \text{m}^{-1}] $

> **【數學推導：如何從馬克士威方程組導出電磁波？】**
> 我們想證明電場 $\mathbf{E}$ 滿足波動方程式。假設在**無電荷且無電流的真空中**（即 $\rho = 0$ 且 $\mathbf{J} = 0$），方程式退化為：
> (1) $\nabla \cdot \mathbf{E} = 0$
> (2) $\nabla \cdot \mathbf{B} = 0$
> (3) $\nabla \times \mathbf{E} = -\frac{\partial \mathbf{B}}{\partial t}$
> (4) $\nabla \times \mathbf{B} = \mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t}$
> **Step 1:** 對法拉第定律 (3) 的兩邊同時取旋度 (curl)：
> 
> $$\nabla \times (\nabla \times \mathbf{E}) = \nabla \times \left(-\frac{\partial \mathbf{B}}{\partial t}\right)$$
> 
> 
> **Step 2:** 利用向量恆等式 $\nabla \times (\nabla \times \mathbf{A}) = \nabla(\nabla \cdot \mathbf{A}) - \nabla^2\mathbf{A}$ 展開左邊，並將右邊的空間微分與時間微分交換：
> 
> $$\nabla(\nabla \cdot \mathbf{E}) - \nabla^2\mathbf{E} = -\frac{\partial}{\partial t}(\nabla \times \mathbf{B})$$
> 
> 
> **Step 3:** 代入真空高斯定律 (1) $\nabla \cdot \mathbf{E} = 0$，以及真空安培定律 (4) $\nabla \times \mathbf{B} = \mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t}$：
> 
> $$0 - \nabla^2\mathbf{E} = -\frac{\partial}{\partial t}\left(\mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t}\right)$$
> 
> 
> **Step 4:** 整理後即可得到標準的三維波動方程式：
> 
> $$\nabla^2\mathbf{E} = \mu_0 \varepsilon_0 \frac{\partial^2 \mathbf{E}}{\partial t^2}$$
> 
> 
> 
> 與古典波動方程式 $\nabla^2 f = \frac{1}{v^2} \frac{\partial^2 f}{\partial t^2}$ 比較，可得知電磁波的傳播速度為 $v = \frac{1}{\sqrt{\mu_0 \varepsilon_0}} = c$。這證明了光本質上就是一種電磁波。





## 2. Blackbody Radiation (黑體輻射)

黑體 (Blackbody) 是一個理想化的物理模型：它能吸收所有入射的電磁輻射，不論頻率與入射角為何。既然沒有反射，我們觀測到黑體放出的電磁波就**只與物體的絕對溫度 $T$ 有關**。這排除了材料特性的干擾，讓觀測結果具有普適性。

### 2.1 Stefan-Boltzmann Law (斯特凡-波茲曼定律)

$$R = \sigma T^4$$

* $R$: 輻射出射度 (Power radiated per unit area)，即單位面積在單位時間內輻射出的總能量，$[\text{W} \cdot \text{m}^{-2}] $
* $T$: 黑體的絕對溫度，單位是 $[\text{K} ] $
* $\sigma$: 斯特凡-波茲曼常數 (Stefan-Boltzmann constant)，值約為 $5.67 \times 10^{-8}$ $[\text{W} \cdot \text{m}^{-2} \cdot \text{K}^{-4}] = [\text{J} \cdot \text{s}^{-1} \cdot \text{m}^{-2} \cdot \text{K}^{-4}]$

### 2.2 Wien's Displacement Law (維恩位移定律)

$$\lambda_{\text{max}} \propto \frac{1}{T} \implies \lambda_{\text{max}} T = 2.898 \times 10^{-3} \text{ m}\cdot\text{K}$$

* $\lambda_{\text{max}}$: 黑體輻射光譜中，能量強度達到最大值時所對應的波長，單位是 $[\text{m} ] $。
* **物理意義:** 溫度越高的物體，其輻射最強的波長越短（例如打鐵時，溫度升高，顏色從不可見的紅外線轉為暗紅，再轉為亮黃甚至偏藍）。

### 2.3 能量量子化的誕生

1. **Rayleigh-Jeans Law 的古典災難 (Ultraviolet Catastrophe):**
* 古典物理假設空腔內的電磁波是駐波，且根據 **能量均分定理** ，每一種頻率的駐波模式都分配到相同的平均能量 $k_B T$（$k_B$ 為波茲曼常數）。
* 結果導致：隨著頻率趨近無窮大（紫外區），輻射能量也會發散至無窮大，這與實驗結果完全不符。


2. **Planck's Postulate (普朗克假設):**
* 為了解決這個災難並擬合實驗數據，普朗克大膽假設：**振盪器吸收或發射的能量不是連續的，而是量子化的（離散的）。**
* 公式：$E_n = n h \nu$ ，其中 $n = 0, 1, 2, \dots$
* $h$: 普朗克常數，約為 $6.626 \times 10^{-34}$  $[\text{J} \cdot \text{s}]$
* $\nu$: 輻射頻率 $[ \text{s}^{-1}]$
* **老師的解答（為什麼要假設非連續？）：** 你可以想像能量是貨幣，古典物理認為能量是「液體」，要給多少都可以；但普朗克假設能量是「硬幣」，面額是 $h\nu$。對於高頻率（紫外光）的電磁波，它的單枚硬幣面額極大。在常溫下，系統擁有的總零錢（熱能 $k_B T$）根本湊不出一枚高頻硬幣，因此高頻輻射根本不會產生。這完美壓制了高頻區的能量發散，解決了紫外災難。




## 3. Einstein: Photoelectric Effect (光電效應)

光電效應證明了光不僅在發射時是量子化的，其在空間中傳播的本質也是一顆顆的粒子（光子）。

* **實驗裝置:** 真空玻璃管內有兩金屬極板。光源照射目標金屬（如鈉 Na），釋放出電子。透過外加逆向電壓（截止電壓 $V_s$）可測量出射電子的最大動能。必須抽真空以避免光電子與空氣分子碰撞而損失能量。

### 3.1 古典電磁理論 (EM Theory) 的三大崩潰

1. **沒有時間延遲 (Time Delay):**
* **實驗事實:** 光一照到金屬，電子在小於 $10^{-9} [\text{s}]$ 內立刻射出。
* **古典計算打臉:** 假設極弱光源。鈉原子密度約 $10^{19}$ $[\text{atoms} \cdot \text{m}^{-2}]$  。若入射光功率極低，平均每個原子每秒吸收的能量可能僅有 $10^{-25}$ $[\text{J} \cdot \text{s}^{-1}]$
* 要讓電子逃離金屬，需要克服功函數 (Work function) 約 $1 [\text{eV}] = 1.6 \times 10^{-19} [\text{J}]$
* 古典理論認為能量像波一樣均勻散布，因此單一電子累積足夠能量所需時間為：

$$t = \frac{1.6 \times 10^{-19} [\text{ J}]}{10^{-25} [\text{J}\cdot \text{s}^{-1}]} = 1.6 \times 10^6 [\text{s}]$$



這大約是 18 天！但實驗卻是瞬間發生。


1. **光強度不影響最大動能:**
* **實驗事實:** 同頻率下，越亮的光只會產生「越多」的電子（電流變大），但電子的最大動能固定。
* **古典理論:** 光越亮代表電磁波振幅越大，傳遞能量應該越多，電子動能應該變大。


3. **存在底線頻率 (Threshold Frequency):**
* **實驗事實:** 照射光頻率必須大於某個底線頻率 $\nu_0$ 才有電子射出。若頻率不夠，光再亮都沒用。射出電子的動能與頻率呈線性關係。
* **古典理論:** 只要光夠亮、照得夠久，任何頻率的光都應該能把電子打出來，無法解釋底線頻率的存在。



### 3.2 愛因斯坦的光子理論 (Photon Theory)

1905年，愛因斯坦將普朗克的概念延伸：光在空間中的傳播也是一個個能量包，稱為「光子」(Photon)。

* 單一光子能量：$E = h\nu$
* 電子游離能量、功函數：$IE = \phi = h\nu_0$
* **解釋光電效應 (能量守恆方程式):**
一顆電子只能「一次吞下一整顆」光子。電子吸收能量 $h\nu$ 後，一部分用來當作「過路費」克服金屬表面的束縛能（稱為功函數 Work function，記為 $\phi$ 或 $W$ 或 $\psi$），剩下的能量就變成電子飛出去的最大動能 $K_{\text{max}}$。

$$\begin{aligned}
h\nu &= K_{\text{max}} + \phi \\
K_{\text{max}} &= h\nu - \phi \\
K_{\text{max}} &= h(\nu - \nu_0)
\end{aligned}$$



這完美解釋了三大難題：

1. 一對一碰撞，不需累積時間。
2. 光變亮只代表光子「顆數」變多，單顆光子的能量不變。
3. 若 $h\nu < \phi$，單顆光子能量付不起過路費，電子出不來。

## 4. Stability of Atoms (原子的穩定性)

* **古典物理的致命傷:** 在拉塞福的行星模型中，電子繞著帶正電的原子核作圓周運動。由於電子具有向心加速度，根據古典電磁學（Larmor formula），加速運動的帶電體必須持續輻射出電磁波。
* **毀滅的倒數計時:** 一旦持續放出能量，電子的軌道半徑就會不斷縮小，最終以螺旋軌跡墜毀入原子核中。計算顯示，這個過程大約只需 $10^{-9}$ 秒。這意味著如果古典物理是對的，宇宙中所有的物質會在瞬間湮滅。
* **量子解方 (Bohr Model):** 波耳假設了角動量量子化（$L = n\hbar$），後來由德布羅意（de Broglie）的物質波理論給出物理圖像：電子在特定軌道上形成「駐波」（Standing wave），首尾相接不發生破壞性干涉。在這些特定的駐波軌道（定態）上，電子不會輻射能量。

## 5. Quantum Physics 諾獎歷史整理

近代物理的發展是一連串打破常規的接力賽：

* **Planck (1918):** Energy Quantization (能量量子化)
* **Einstein (1921):** Photoelectric Effect (光電效應 - 證明光的粒子性)
* **Bohr (1922):** Atomic Structure (原子結構 - 角動量量子化)
* **Compton (1926):** Particle Properties of Light (康普頓效應 - 再次印證光子帶有動量)
* **de Broglie (1927):** Wave-particle Duality (波粒二象性 - 粒子也具有波動性)
* **Schrödinger (1933):** Quantum Mechanics (量子力學 - 提出薛丁格方程式)

## 6. Governing Equations 總整理與重點回顧

### 🌟 核心觀念總結

1. **古典力學的基石：** 牛頓第一定律確立了「慣性坐標系」的存在，這不僅支撐了牛頓第二定律，在未來愛因斯坦的狹義相對論中更是至關重要的前提假設。
2. **光的電磁本質：** 透過馬克士威方程組，將法拉第定律取旋度並代入安培-馬克士威定律與高斯定律，可以嚴格導出電場 $\mathbf{E}$ 與磁場 $\mathbf{B}$ 均符合波動方程式，且波速等於光速，證明了光就是電磁波。
3. **黑體輻射的測量：** 根據 Stefan-Boltzmann 定律，只要測量黑體放出的電磁波總輻射強度，就能反推該物體的絕對溫度 $T$。
4. **能量的離散化：** 普朗克透過假設「能量量子化」（$E = nh\nu$），限制了高頻振盪器的激發機率，成功解釋了黑體輻射光譜在高頻區塊的能量分佈，解決了紫外災難。
5. **光的粒子性：** 光電效應證明了光能量具有空間局域性（光子）。因為電子的逸出時間極短，且存在截止頻率，這只能用光子與電子發生「一對一、全有或全無」的能量轉移來解釋。
6. **物質的波動性與原子穩定：** 氫原子的穩定存在（不會因為輻射電磁波而瞬間塌縮），必須依賴角動量量子化的假設。電子以「機率駐波」的形式存在於軌道中，打破了古典電磁學必定輻射能量的限制。
