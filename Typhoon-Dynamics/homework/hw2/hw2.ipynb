同學你好！這份講義探討的是大氣動力學（特別是針對颱風等軸對稱系統）中非常進階且核心的「次環流方程式（Eliassen Equation）」與「對稱不穩定性」。

既然你喜歡用比喻來學習，在我們進入數學之前，先想像一個**高速旋轉的陀螺**（或是花式溜冰選手）。
陀螺主要在原地高速旋轉，這股強大的旋轉氣流就是系統的**「主環流（Primary Circulation）」**。
但是，如果陀螺摩擦到地面（摩擦力），或是內部受到不均勻的加熱（潛熱釋放），它原本完美的平衡就會被破壞。為了不讓自己倒下，陀螺的內部結構會產生非常微小的「上下」與「向內/向外」的移動來重新找回平衡。這個用來「調整並維持系統平衡」的微小翻轉運動，就是**「次環流（Secondary Circulation）」** 。

## Question 1：解釋方程式代表意義與主次環流變數


$\frac{v^{2}}{r}+fv=\frac{\partial\phi}{\partial r}$

$\frac{\partial v}{\partial t}+u(f+\zeta)+w\frac{\partial v}{\partial z}=X$

$\frac{\partial\phi}{\partial z}=\frac{g}{\theta_{0}}\theta$ 

$\frac{\partial\rho ru}{r\partial r}+\frac{\partial\rho w}{\partial z}=0$ 

$\frac{\partial\theta}{\partial t}+u\frac{\partial\theta}{\partial r}+w\frac{\partial\theta}{\partial z}=\frac{\theta}{T}\frac{Q}{c_{p}}$ 

## Answer 1

我們用圓柱座標系統 $(r, \lambda, z)$ 來描述颱風，這些方程式描述了流體的基本物理守恆定律：

1. $\frac{v^{2}}{r}+fv=\frac{\partial\phi}{\partial r}$
  * **變數**   
    * $v$：切線風速 (Tangential wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $r$：徑向距離 (Radial distance) $[\text{m}]$
    * $f$：科氏參數 (Coriolis parameter) $\sim 10^{-5}[\text{s}^{-1}]$
    * $\phi$：重力位勢 / 動力氣壓 (Geopotential / Kinematic pressure) $[\text{m}^2 \cdot \text{s}^{-2}]$
  * **意義：** 這是徑向（向內/向外）的**梯度風平衡 (Gradient Wind Balance)**。它表示離心力 ($\frac{v^2}{r}$)、科氏力 ($fv$) 與氣壓梯度力 ($\frac{\partial\phi}{\partial r}$) 達到完美的動態平衡


2. $\frac{\partial v}{\partial t}+u(f+\zeta)+w\frac{\partial v}{\partial z}=X$ 
  * **變數**   
    * $v$：切線風速 (Tangential wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $t$：時間 (Time) $[\text{s}]$
    * $u$：徑向風速 (Radial wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $f$：科氏參數 (Coriolis parameter) $\sim 10^{-5}[\text{s}^{-1}]$
    * $\zeta$：相對渦度 (Relative vorticity) $[\text{s}^{-1}]$
    * $w$：垂直風速 (Vertical wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $z$：垂直高度 (Vertical height) $[\text{m}]$
  * **意義：** 這是切線方向的**動量方程式（角動量守恆）**。描述切線風速 $v$ 隨時間的變化，是由於徑向風 $u$ 與垂直風 $w$ 帶來了不同區域的角動量，加上外界摩擦力 $X$ 所導致 。
  

3. $\frac{\partial\phi}{\partial z}=\frac{g}{\theta_{0}}\theta$ 
  * **變數**
    * $\phi$：重力位勢 / 動力氣壓 (Geopotential / Kinematic pressure) $[\text{m}^2 \cdot \text{s}^{-2}]$
    * $z$：垂直高度 (Vertical height) $[\text{m}]$
    * $g$：重力加速度 (Gravitational acceleration) $9.81[\text{m} \cdot \text{s}^{-2}]$
    * $\theta_{0}$：參考態位溫 (Reference potential temperature) $\sim 300[\text{K}]$
    * $\theta$：位溫 (Potential temperature) $[\text{K}]$
  * **意義：** 這是垂直方向的**靜力平衡 (Hydrostatic Balance)**。表示垂直氣壓梯度力與重力（這裡用位溫 $\theta$ 形式表示的浮力）互相抵銷 
  
4. $\frac{\partial\rho ru}{r\partial r}+\frac{\partial\rho w}{\partial z}=0$ 
  * **變數**
    * $\rho$：空氣密度 (Air density) $\sim 1[\text{kg} \cdot \text{m}^{-3}]$
    * $r$：徑向距離 (Radial distance) $[\text{m}]$
    * $u$：徑向風速 (Radial wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $w$：垂直風速 (Vertical wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $z$：垂直高度 (Vertical height) $[\text{m}]$
    * **意義：** 這是**質量連續方程式 (Continuity Equation)**。確保流體質量守恆，有空氣輻合進來就必須有空氣上升或下降流出 
5. $\frac{\partial\theta}{\partial t}+u\frac{\partial\theta}{\partial r}+w\frac{\partial\theta}{\partial z}=\frac{\theta}{T}\frac{Q}{c_{p}}$ 
  * **變數**
    * $\theta$：位溫 (Potential temperature) $[\text{K}]$
    * $t$：時間 (Time) $[\text{s}]$
    * $u$：徑向風速 (Radial wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $w$：垂直風速 (Vertical wind speed) $[\text{m} \cdot \text{s}^{-1}]$
    * $r$：徑向距離 (Radial distance) $[\text{m}]$
    * $z$：垂直高度 (Vertical height) $[\text{m}]$
    * $T$：絕對溫度 (Absolute temperature) $[\text{K}]$
    * $Q$：非絕熱加熱率 (Diabatic heating rate) $[\text{W} \cdot \text{kg}^{-1}]$
    * $c_{p}$：等壓比熱 (Specific heat at constant pressure) $1004[\text{J} \cdot \text{kg}^{-1} \cdot \text{K}^{-1}]$
  * **意義：** 這是**熱力學方程式 (Thermodynamic Equation)**。描述溫度的變化是由於冷暖空氣被推來推去（平流），以及外界的非絕熱加熱 $Q$（例如水氣凝結放熱）所造成 。

* **主環流變數：** $v$ (切線風)、$\phi$ (重力位勢/氣壓)、$\theta$ (位溫)。它們是颱風的主要結構，通常處於梯度風與靜力平衡。 
* **次環流變數：** $u$ (徑向風)、$w$ (垂直風)。它們相對微弱，負責在主環流失去平衡時進行調整。

### 作業 2：A, B, C 參數代表意義與次環流形狀

這三個參數決定了流體抵抗變形的能力（恢復力），我們統稱為「穩定度參數」。你可以把這想像成空間中三個方向的彈簧，決定了次環流（翻轉橢圓）被擠壓的形狀。

* $\rho A=\frac{g}{\theta_{0}}\frac{\partial\theta}{\partial z}$ 
    * **意義：** $A$ 代表**靜力穩定度 (Static Stability)** 。它衡量大氣抵抗「垂直運動」的能力。
* $\rho B=-\frac{g}{\theta_{0}}\frac{\partial\theta}{\partial r}=-(f+\frac{2v}{r})\frac{\partial v}{\partial z}$ 
    * **意義：** $B$ 代表**斜壓性 (Baroclinicity)** 或熱力風垂直風切 。它連接著徑向的溫度梯度與垂直的風速變化，代表大氣的傾斜程度。
* $\rho C=(f+\frac{2v}{r})(f+\frac{\partial rv}{r\partial r})$ 
    * **意義：** $C$ 代表**慣性穩定度 (Inertial Stability)** 。它衡量流體抵抗「水平（徑向）運動」的能力。

**如何影響次環流形狀？**
因為連續方程式的關係，我們可以用一個流函數 $\psi$ 來代表次環流，它在空間中會形成一個封閉的橢圓形狀
1.  **當 $A \gg C$（靜力穩定度主宰）：** 垂直方向的「彈簧」很硬，很難上下走。所以次環流會被壓扁，形成**水平拉長、垂直範圍小**的橢圓
2.  **當 $C \gg A$（慣性穩定度主宰）：** 水平方向的「彈簧」很硬，很難往內外走。所以次環流會變成**垂直拉長、水平被限制**的橢圓
3.  **參數 $B$ 的影響：** 它相當於一個交叉項，會造成這個次環流橢圓發生**傾斜 (Tilt)**

### 作業 3：$AC - B^2$ 代表意義、Stiffness 與熱力風平衡調整

* **$AC - B^2$ 的意義：**
    在數學上，它是穩定度矩陣的行列式值；在物理上，它等同於系統的**位渦 (Potential Vorticity, PV)** 或是 $P$ 值
    要讓次環流方程式有穩定的解（數學上的橢圓型偏微分方程式），必須滿足 $AC - B^2 > 0$，這也就是系統呈現「對稱穩定 (Symmetric stable)」的必要條件。若 $AC - B^2 < 0$，系統就會發生對稱不穩定
* **Stiffness (剛性) 的特性：**
    講義中提到 "PV is a measure of the 'stiffness'" [cite: 143]。你可以把 PV 想像成這團流體的「綜合剛性」。當 PV 很大（大氣非常穩定）時，流體很硬，任何微小的擾動都會產生高頻的震盪（頻率 $\omega$ 很大）；當 PV 變小（例如因為潛熱釋放導致靜力穩定度下降），流體變軟，就容易被觸發不穩定現像
* **次環流如何調整主環流的熱力風平衡？**
    主環流原本維持著完美的「熱力風平衡」 。但是，當颱風底部發生**摩擦 (Friction, $X / F$)** 或是雲帶中有**潛熱釋放 (Diabatic heating, $Q$)** 時，這個平衡會瞬間被破壞 
    這時候，次環流 ($u, w$) 就會立刻啟動！上升氣流 $w$ 會透過絕熱冷卻來改變溫度場 $\theta$，而徑向氣流 $u$ 會帶來不同的角動量來改變風場 $v$。透過 $u$ 和 $w$ 的輸送，次環流能迅速把溫度和風場「喬」回符合熱力風平衡的狀態

### 作業 4：推導次環流方程式與交互作用概念

*(這題講義說不用交，但老師還是帶你看懂背後的魔術！)*

**推導概念 (簡單推導)：**
我們想要把 $u$ 和 $w$ 統整成一個方程式。
1.  **引入流函數 $\psi$：** 為了滿足質量守恆 $\frac{\partial\rho ru}{r\partial r}+\frac{\partial\rho w}{\partial z}=0$ ，我們定義 $\rho w=\frac{\partial r\psi}{r\partial r}$ 且 $\rho u=-\frac{\partial\psi}{\partial z}$ [cite: 56, 65]。這樣只要解出 $\psi$，就知道 $u$ 和 $w$ 了。
2.  **消去時間項：** 熱力風平衡告訴我們，溫度的徑向梯度與風的垂直切變是綁定的。如果我們把動量方程式對 $z$ 微分，把熱力學方程式對 $r$ 微分，然後把它們相加/相減來消去時間微分 $\frac{\partial}{\partial t}$。
3.  **整理成 Eliassen 方程式：** 把第一步定義的 $\psi$ 代入第二步整理好的方程式，並把參數 $A, B, C$ 替換進去，就會得到一個二階偏微分方程式：
    $$\mathcal{L}\psi=\frac{\partial J}{\partial r}-\frac{\partial F}{\partial z}$$ 
    其中 $\mathcal{L}(\cdot)=\frac{\partial}{\partial r}(A\frac{\partial r(\cdot)}{r\partial r}+B\frac{\partial(\cdot)}{\partial z})+\frac{\partial}{\partial z}(B\frac{\partial r(\cdot)}{r\partial r}+C\frac{\partial(\cdot)}{\partial z})$ 。

**意義與主次交互作用（準平衡）：**
這個方程式是一個「診斷方程式」（沒有時間項）。等號右邊是**外力強迫項**（熱源 $J$ 的徑向不均勻，以及摩擦力 $F$ 的垂直不均勻）
這告訴我們一個非常漂亮的物理圖像：我們為什麼要區分主、次環流？
因為大氣喜歡處於「準平衡 (Quasi-balance)」。主環流 ($v, \theta$) 負責提供穩定的舞台（決定了 $A, B, C$ 這些彈簧的硬度），而外力強迫一旦試圖打破平衡，系統不用等到整個垮掉，它會**瞬間診斷**出需要多少的次環流 $\psi$ 來抵銷這些外力，確保系統永遠踩在平衡的鋼索上 。
