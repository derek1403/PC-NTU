# **一、原始的完全可壓縮流體力學偏微分方程組** 轉成 **密度加權的線性化微擾方程組（且分離出非線性強迫項）** (From the original fully compressible Navier-Stokes equations to the density-weighted linearized perturbation equations (with isolated nonlinear forcing terms))



$$\begin{gather*}
\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}} u \right] + \frac{\partial }{\partial x} \left[ \bar{\rho}^{\frac{-1}{2}} p' \right]  &=& \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) u + \frac{\rho'}{\bar{\rho} \rho  } \frac{\partial p'}{\partial x} \right) \tag{2a}\\
\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}} v \right] + \frac{\partial }{\partial y} \left[ \bar{\rho}^{\frac{-1}{2}} p' \right]  &=& \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) v + \frac{\rho'}{\bar{\rho} \rho  }\frac{\partial p'}{\partial y} \right) \tag{2b}\\
\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}}  w \right]  + g\bar{\rho}^{\frac{-1}{2}}\left(\rho' - \frac{p'}{c^2}\right) + \left(  \frac{\partial}{\partial z} + \Gamma \right)[\bar{\rho}^{\frac{-1}{2}}p'] &=& - \bar{\rho}^{\frac{1}{2}} \left( (\mathbf{u} \cdot \nabla) w  +  \frac{\rho'^2}{\rho \bar{\rho} } g +  \frac{\rho'}{\rho \bar{\rho}} \frac{\partial p'}{\partial z}  \right) \tag{2c}\\
\frac{\partial ( \bar{\rho}^{\frac{-1}{2}} p') }{\partial t} + c^2 \left( \frac{\partial ( \bar{\rho}^{\frac{1}{2}} u) }{\partial x } +  \frac{\partial  (\bar{\rho}^{\frac{1}{2}} v)}{\partial y } + \left( \frac{\partial }{\partial z} -  \Gamma \right)\left[\bar{\rho}^{\frac{1}{2}} w \right]  \right)   &=&  -\bar{\rho}^{\frac{-1}{2}} \left( (\mathbf{u} \cdot \nabla ) p' + c^2  \rho'  \nabla \cdot \mathbf{u} \right) \tag{2d}\\
\frac{\partial }{\partial t} \left[ \bar{\rho}^{\frac{-1}{2}}  \left( \rho' - \frac{p'}{c^2}  \right)   \right] -  w \bar{\rho}^{\frac{1}{2}} \frac{N^2}{g}   &=& -\bar{\rho}^{\frac{-1}{2}} \left( \mathbf{u}\cdot\nabla \rho' -\frac{1}{c^2} \mathbf{u}\cdot\nabla p'  \right) \tag{2e}
\end{gather*}$$

* $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
* $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
* $\rho '$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$
* $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}]$
* $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
* $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
* $\Gamma $ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
* $N^{2}$ : 浮力頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
* $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$


## **為什麼要花這麼大功夫轉成這樣？（The Ultimate Goal）**

當方程式轉成具有「反厄米特特性（Skew-Hermitian）」的線性算子 $L$ 後，數學上會保證三件極度美好的事情（正如講義提到的）：
1. 特徵值純虛數：這代表方程式的解是純粹的「震盪波」，不會有數值上無端呈指數爆炸的解（保證穩定性）。
2. 特徵函數正交：你可以用這組解當作基底（Basis），像傅立葉轉換一樣，把大氣真實的狀態拆解成無數個獨立波的疊加。
3. 精確處理快波：在數值模式（如天氣預報）中，聲波跑得非常快，會嚴重限制時間步長（CFL condition）。把線性項抽出來並精確解析後，程式在時間積分時就可以跨大步長，只針對非線性強迫項 $N$ 做數值積分（例如講義 (14) 式的處理方式），大幅節省超級電腦的運算資源。


## 假設與已知 (Assumptions & Preliminaries)


* **【假設 1】 忽略黏滯力（Inviscid fluid assumption）  ：**
  
  $$ \mu = 0$$

  * $\mu $ : 動力黏滯係數 (Dynamic viscosity) $[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$


* **【已知 1】 [動量方程式 (Momentum Equation) - 納維-斯托克斯 (Navier-Stokes)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#momentum-equation-navier-stokes)  ：**
  
  * (a) 
   
  $$\begin{gather*}
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &=& \mathbf{g} - \frac{1}{\rho} \nabla p + \frac{\mu}{\rho} \nabla^2 \mathbf{u} \\
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &\overset{\text{假設 1}}{=}& \mathbf{g} - \frac{1}{\rho} \nabla p + \frac{0}{\rho} \nabla^2 \mathbf{u} \\
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &=& \mathbf{g} - \frac{1}{\rho} \nabla p  \\
  \end{gather*}$$

  * (b) x 方向
  
  $$\frac{\partial u}{\partial t } + (\mathbf{u} \cdot \nabla) u \overset{\text{已知 1(a)}}{=} - \frac{1}{\rho} \frac{\partial p}{\partial x} $$

  * (c) y 方向  
  
  $$\frac{\partial v}{\partial t } + (\mathbf{u} \cdot \nabla) v \overset{\text{已知 1(a)}}{=} - \frac{1}{\rho} \frac{\partial p}{\partial y} $$

  * (d) z 方向
  
  $$\frac{\partial w}{\partial t } + (\mathbf{u} \cdot \nabla) w \overset{\text{已知 1(a)}}{=} -g - \frac{1}{\rho} \frac{\partial p}{\partial z} $$


  * $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
  * $\mathbf{g}$: 重力加速度向量 $(0, 0, -g)$ $[\text{m} \cdot \text{s}^{-2}]$
  * $p$: 氣壓 $ [\text{Pa}],[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}] $
  * $\rho$: 流體密度 $[\text{kg} \cdot \text{m}^{-3}]$
  * $\mu$: 動力黏滯係數 (Dynamic viscosity) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}] $

* **【已知 2】 [連續方程式 (Continuity Equation)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#continuity-equation)  ：**
  
  $$\begin{gather*}
  \frac{\partial \rho}{\partial t } + \nabla \cdot (\rho \mathbf{u}) &=& 0 \\
  \frac{\partial \rho}{\partial t } + \frac{\partial (\rho u)}{\partial x }+ \frac{\partial (\rho v)}{\partial y }+ \frac{\partial (\rho w)}{\partial z } &=& 0 \\
  \frac{\partial \rho}{\partial t } + \left( \frac{\partial \rho }{\partial x } u + \rho \frac{\partial  u}{\partial x } \right)+ \left( \frac{\partial \rho }{\partial y } v + \rho \frac{\partial  v}{\partial y } \right) + \left( \frac{\partial \rho }{\partial z } w + \rho \frac{\partial  w}{\partial z } \right)&=& 0 \\
  \frac{\partial \rho}{\partial t } + u \frac{\partial \rho }{\partial x } + v\frac{\partial \rho }{\partial y }   + w \frac{\partial \rho }{\partial z } &=& - \rho \left( \frac{\partial  u}{\partial x } + \frac{\partial  v}{\partial y } +\frac{\partial  w}{\partial z }\right)\\
  \end{gather*}$$

  * $\rho$: 流體密度 $[\text{kg} \cdot \text{m}^{-3}]$
  * $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
  * $t$: 時間 $[\text{s}]$。

* **【假設 2】 等熵過程假設（Isentropic process assumption）：**
  
  $$\frac{Ds}{Dt} = 0$$

  * $s$ : 比熵／單位質量熵 (Specific entropy) $[\text{J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}]$



* **【假設 3】 線性微擾狀態假設（Linear perturbation state assumption） ：** 將流體的氣壓與密度分為隨高度變化的基本背景態（Basic state）與隨時空變化的微小擾動態（Perturbation）。

  * (a) 氣壓場拆解：
  
  $$p(x,y,z,t) = \bar{p}(z) + p'(x,y,z,t)$$

  * (b) 密度場拆解：
  
  $$\rho(x,y,z,t) = \bar{\rho}(z) + \rho'(x,y,z,t)$$

  * $\bar{p}(z)$ : 僅隨高度 $z$ 改變的基本背景態氣壓 (Basic state pressure)，$ [\text{Pa}],[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}] $
  * $\bar{\rho}(z)$ : 僅隨高度 $z$ 改變的基本背景態密度 (Basic state density)，$[\text{kg} \cdot \text{m}^{-3}]$
  * $p'$ : 隨時空變化的壓力微擾項 (Perturbation pressure)，滿足 $p' \ll \bar{p}$， $ [\text{Pa}],[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}] $
  * $\rho'$ : 隨時空變化的微擾項 (Perturbation density)，滿足  $\rho' \ll \bar{\rho}$，$[\text{kg} \cdot \text{m}^{-3}]$


* **【假設 4】 背景態靜力平衡（Hydrostatic equilibrium of the basic state） ：** 假設無任何外部干擾時，背景大氣在垂直方向上重力與氣壓梯度力完全抵消。

  $$\frac{d\bar{p}}{dz} = -\bar{\rho}g$$

  * $\bar{p}$ : 基本背景態氣壓 (Basic state pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $z$ : 垂直高度座標 (Vertical height coordinate) $[\text{m}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $\frac{d\bar{p}}{dz}$ : 背景氣壓隨高度變化的全導數（垂直氣壓梯度） $[\text{Pa}\cdot\text{m}^{-1}]$


* **【假設 5】等溫基本態假設（Isothermal basic state）：** 假設基本背景場的大氣溫度為不隨時間與空間變化的常數
 
 
  $$T(x,y,z,t) = T_0 $$

  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $T(x,y,z,t)$ : 瞬時大氣絕對溫度場 (Instantaneous absolute temperature field) $[\text{K}]$




* **【定義 1】音速的熱力學與大氣特徵定義（Thermodynamic and atmospheric definition of the speed of sound）  ：**
  
  * (a)  熱力學基本定義：
  
  $$\begin{gather*}
  c^2 &\overset{\text{def}}{=}& \left( \frac{\partial p}{\partial \rho} \right)_s \\
  \end{gather*}$$

  * (b) 理想氣體與背景態化簡：
 
  $$\begin{gather*}
  c^2 &\overset{\text{def}}{=}& \gamma R_d T \\
  &\overset{\text{假設 5}}{=}& \gamma R_d T_0
  \end{gather*}$$

  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $c^{2}$ : 音速的平方，代表氣壓隨密度變化的等熵彈性模量比值 $[\text{m}^2\cdot\text{s}^{-2}]$
  * $p$ : 氣壓 (Pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\rho $ : 流體密度 (Fluid density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $s$ : 比熵／單位質量熵 (Specific entropy) $[\text{J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}]$
  * $\left(\frac{\partial p}{\partial \rho }\right)_{s}$ : 在恆定比熵（等熵）條件下，氣壓對密度的偏微分偏導數 (Isentropic pressure derivative with respect to density) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $(\dots)_s$ : 表示在恆定比熵（等熵）條件下的偏微分
  * $\gamma $ : 絕熱指數／熱容量比 (Adiabatic index / Ratio of specific heats)，對於大氣雙原子主要氣體（如 $\text{N}_2, \text{O}_2$）而言 $\gamma = \frac{c_p}{c_v}\approx 1.4$ [無單位]
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1},[\text{m}^2\cdot\text{s}^{-2}\cdot\text{K}^{-1}]$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$

* **【定義 2】 等溫大氣標高（Scale height of an isothermal atmosphere） ：**  標高代表大氣壓力和密度隨高度垂直遞減的特徵尺度。在等溫大氣假設下，這個高度尺度為一個常數。
  
  $$\begin{gather*}
  H_s &\overset{\text{def}}{=}& \frac{R_d T}{g} \\
  &\overset{\text{假設 5}}{=}& \frac{R_d T_0}{g}
  \end{gather*}$$

  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$，代表大氣氣壓或密度遞減為地表的 $1/e$（約 36.8%）時所需上升的高度
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1},[\text{m}^2\cdot\text{s}^{-2}\cdot\text{K}^{-1}]$
  * $T$ : 瞬時絕對溫度 (Instantaneous absolute temperature) $[\text{K}]$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$

* **【定義 3】浮力頻率平方 / Brunt-Väisälä frequency 平方 ：** 浮力頻率代表大氣在垂直方向上的靜力穩定度指標。當流體微塊受到垂直擾動時，此頻率決定了它在原平衡位置上下振盪的特徵頻率。
  
  $$\begin{gather*}
  N^2 &\overset{\text{def}}{=}& \frac{\kappa g}{H_s} \\
  &=& \left(1 - \frac{1}{\gamma} \right) \frac{ g}{H_s} \\
  \end{gather*}$$

  * $N^{2}$ : 浮力頻率平方 / Brunt-Väisälä frequency squared $[\text{s}^{-2}]$
  * $\kappa $ : 乾空氣大氣熱力學常數 (Kappa constant for dry air) $\kappa = \frac{R_d}{C_p} = 1 - \frac{1}{\gamma}$，對於大氣雙原子氣體而言 $\kappa \approx 0.286$ [無單位]
  * $C_{p}$ : 乾空氣定壓比熱 (Specific heat of dry air at constant pressure) $C_p \approx 1005 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}$
  * $\gamma $ : 絕熱指數／熱容量比 (Adiabatic index)，大氣雙原子氣體 $\gamma \approx 1.4$ [無單位]
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$


* **【定義 4】  大氣密度層結修正因子 / 垂直衰減係數（Atmospheric density stratification correction factor / Eckart parameter $\Gamma $） ：** 該常數因子結合了聲學剪切（由音速 $c^{2}$ 決定）與浮力剪切（由浮力頻率 $N^{2}$ 決定），用於在波動方程中消除因大氣密度隨高度指數遞減所造成的振幅放大效應。
  
  $$\begin{gather*}
  \Gamma &\overset{\text{def}}{=}& \frac{1}{2}\left(\frac{g}{c^2} - \frac{N^2}{g}\right) \\
  &\overset{\text{定義 3}}{=}& \frac{1}{2}\left(\frac{g}{c^2} - \frac{\left(1 - \frac{1}{\gamma} \right) \frac{ g}{H_s}}{g}\right) \\
  &=& \frac{1}{2}\left(\frac{g}{c^2} - \left(1 - \frac{1}{\gamma} \right) \frac{ 1}{H_s}\right) \\
  &\overset{\text{定義 1(b)}}{=}& \frac{1}{2}\left(\frac{g}{\gamma R_d T_0} - \left(1 - \frac{1}{\gamma} \right) \frac{ 1}{H_s}\right) \\
  &\overset{\text{定義 2}}{=}& \frac{1}{2}\left(\frac{g}{\gamma R_d T_0} - \left(1 - \frac{1}{\gamma} \right) \frac{ 1}{\frac{R_d T_0}{g}}\right) \\
  &=& \frac{1}{2}\left(\frac{g}{\gamma R_d T_0} - \left(1 - \frac{1}{\gamma} \right) \frac{ g}{R_d T_0}\right) \\
  &=& \frac{1}{2}\frac{g}{R_d T_0}\left( \frac{1}{\gamma } - \left(1 - \frac{1}{\gamma} \right)\right) \\
  &=& \frac{1}{2}\frac{g}{R_d T_0}\left( \frac{2}{\gamma} - 1   \right) \\
  &=&  \frac{g}{\gamma R_d T_0} - \frac{1}{2}\frac{g}{R_d T_0}    \\
  &\overset{\text{定義 1(b),定義 2}}{=}& \frac{g}{c^2} - \frac{1}{2}\frac{1}{H_s}  \\
  &=&  \frac{g}{c^2} - \frac{1}{2H_s}    \\
  \end{gather*}$$

  * $\Gamma $ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c^{2}$ : 音速的平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $N^{2}$ : 浮力頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $\gamma $ : 絕熱指數／熱容量比 (Adiabatic index)，大氣雙原子氣體 $\gamma \approx 1.4$ [無單位]
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$



* **【已知 3】 一般熱力學狀態方程式之物質導數 (Material Derivative of the General Equation of State)  ：**
  
  $$\begin{gather*}
  \frac{Dp}{Dt} &=& \frac{D}{Dt} \left[ p(\rho, s) \right] \\
  \frac{Dp}{Dt}&=& \left( \frac{\partial p}{\partial \rho} \right)_s \frac{D\rho}{Dt} + \left( \frac{\partial p}{\partial s} \right)_\rho \frac{Ds}{Dt} \\
  \frac{Dp}{Dt}&\overset{\text{假設 2}}{=}& \left( \frac{\partial p}{\partial \rho} \right)_s \frac{D\rho}{Dt} + \left( \frac{\partial p}{\partial s} \right)_\rho \cdot 0 \\
  \frac{Dp}{Dt}&=& \left( \frac{\partial p}{\partial \rho} \right)_s \frac{D\rho}{Dt}  \\
  \frac{Dp}{Dt} &\overset{\text{定義 1}}{=}& c^2 \frac{D\rho}{Dt}  \\
  \frac{1}{c^2} \frac{Dp}{Dt} &=&  \frac{D\rho}{Dt}  \\
  \end{gather*}$$

  * $c^{2}$ : 音速的平方，代表氣壓隨密度變化的等熵彈性模量比值 $[\text{m}^2\cdot\text{s}^{-2}]$
  * $p$ : 氣壓 (Pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\rho $ : 流體密度 (Fluid density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $s$ : 比熵／單位質量熵 (Specific entropy) $[\text{J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}]$
  * $\left(\frac{\partial p}{\partial \rho }\right)_{s}$ : 在恆定比熵（等熵）條件下，氣壓對密度的偏微分偏導數 (Isentropic pressure derivative with respect to density) $[\text{m}^2\cdot\text{s}^{-2}]$



* **【已知 4】理想氣體狀態方程式 (Ideal Gas Equation of State)：** 

  $$\begin{gather*}
  p &=& \rho R_d T \\
  p &\overset{\text{假設 5}}{=}& \rho R_d T_0 \\
  \bar{p} &=&\overline{\rho R_d T_0 } \\
  \bar{p} &=&\bar{\rho } R_d T_0  \\
  \bar{p} &\overset{\text{定義 2}}{=}& \bar{\rho} H_s g \\ 
  \end{gather*}$$

  * $p$ : 瞬時總氣壓場 (Instantaneous total pressure field) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1},[\text{m}^2\cdot\text{s}^{-2}\cdot\text{K}^{-1}]$
  * $T$ : 瞬時絕對溫度場 (Instantaneous absolute temperature field) $[\text{K}]$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $\bar{p}$ : 基本背景態氣壓 (Basic state pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$

* **【推導 1】 倒數密度項的代數拆解（Algebraic decomposition of reciprocal density）：** 

  * (a) 計算基本態倒數與總場倒數之差
  
  $$\begin{gather*}
  \frac{1}{\bar{\rho}} - \frac{1}{\rho} &=&   \frac{\rho - \bar{\rho}}{\bar{\rho} \rho }   \\
  &\overset{\text{已知 1(b)}}{=}&  \frac{(\bar{\rho} + \rho') - \bar{\rho} }{\bar{\rho} \rho  } \\
  &=& \frac{\rho'}{\bar{\rho} \rho  } \\
  \end{gather*}$$

  * (b) 利用恆等式展開總場倒數密度

  $$\begin{gather*}
  \frac{1}{\rho} &=& \frac{1}{\rho} - \frac{1}{\bar{\rho}} + \frac{1}{\bar{\rho}}\\
  &=& - \left( \frac{1}{\bar{\rho}} - \frac{1}{\rho} \right) + \frac{1}{\bar{\rho}} \\
  &\overset{\text{推導 1(a)}}{=}& -\left( \frac{\rho'}{\rho \bar{\rho} }  \right) + \frac{1}{\bar{\rho}} \\
  \end{gather*}$$

  * $\rho$ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\rho'$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$

* **【推導 2】 背景態密度垂直梯度公式（Derivation of basic state density vertical gradient） ：** 

  $$\begin{gather*}
  \frac{d\bar{p}}{dz} &\overset{\text{假設 4}}{=}& -\bar{\rho}g \\
  \frac{d}{dz}\left[ \bar{\rho} H_s g \right] &\overset{\text{已知 4}}{=}& -\bar{\rho}g \\
  H_s g \frac{d \bar{\rho}}{dz} &=& -\bar{\rho}g \\
  \frac{d \bar{\rho}}{d z} &=& -\frac{\bar{\rho}}{H_s} \\
  \frac{\partial \bar{\rho}}{\partial z} &=& -\frac{\bar{\rho}}{H_s} \\
  \end{gather*}$$

  * $\bar{p}$ : 基本背景態氣壓 (Basic state pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $z$ : 垂直高度座標 (Vertical height coordinate) $[\text{m}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$



* **【推導 3】  氣壓微擾項與密度權重的垂直微分展開（Vertical derivative expansion of pressure perturbation with density weighting） ：** 

  $$\begin{gather*}
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] &=&  \frac{\partial (\bar{\rho}^{\frac{-1}{2}}) }{\partial z} p' + \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z} \\ 
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] &=&  \left(-\frac{1}{2}\bar{\rho}^{\frac{-3}{2}} \frac{\partial \bar{\rho}}{\partial z}   \right)  p' + \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z} \\ 
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] &\overset{\text{推導 2}}{=}& \left(-\frac{1}{2}\bar{\rho}^{\frac{-3}{2}} \left( -\frac{\bar{\rho}}{H_s} \right)   \right)  p' + \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z} \\ 
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p']  &=&  \frac{1}{2H_s}\bar{\rho}^{\frac{-1}{2}} p'+ \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z}\\ 
  -\bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z}   &=& - \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] + \frac{1}{2H_s}\bar{\rho}^{\frac{-1}{2}} p' \\ 
  \end{gather*}$$

  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$



* **【推導 4】垂直速度項與密度權重的垂直微分展開（Vertical derivative expansion of vertical velocity with density weighting） ：** 

  $$\begin{gather*} 
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &=&   \frac{d (\bar{\rho}^{\frac{1}{2}}) }{d z} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &=&   \frac{1}{2}  \bar{\rho}^{\frac{-1}{2}} \frac{d \bar{\rho} }{d z} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &\overset{\text{推導 2}}{=}&   \frac{1}{2}  \bar{\rho}^{\frac{-1}{2}} \left( - \frac{\bar{\rho}}{H_s} \right) w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &=&   -\frac{1}{2H_s}  \bar{\rho}^{\frac{1}{2}} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &\overset{\text{定義 4}}{=}&  \left( \Gamma - \frac{g}{c^2} \right) \bar{\rho}^{\frac{1}{2}} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} - \left( \Gamma - \frac{g}{c^2} \right) \bar{\rho}^{\frac{1}{2}} w &=&     \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \end{gather*}$$

  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $w$ : 垂直向速度場（此處主要考慮一階微擾項 $w'$） (Vertical velocity) $[\text{m}\cdot\text{s}^{-1}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
  * $\Gamma $ : 大氣密度層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$


* **【推導 5】氣壓物質導數 $\frac{Dp}{Dt}$ 的線性化拆解（Linearization and decomposition of pressure material derivative） ：** 

  $$\begin{gather*} 
  \frac{Dp}{Dt} &=&  \frac{\partial p}{\partial t} + u \frac{\partial p}{\partial x} + v \frac{\partial p}{\partial y} + w \frac{\partial p}{\partial z}  \\
  &\overset{\text{假設 3(a)}}{=}&  \frac{\partial (\bar{p}+p') }{\partial t} + u \frac{\partial (\bar{p}+p')}{\partial x} + v \frac{\partial (\bar{p}+p')}{\partial y} + w \frac{\partial (\bar{p}+p')}{\partial z}  \\
  &=&  \left( \frac{\partial \bar{p} }{\partial t} +  \frac{\partial p' }{\partial t}  \right)+ u \left( \frac{\partial \bar{p} }{\partial x} +  \frac{\partial p' }{\partial x}  \right) + v \left( \frac{\partial \bar{p} }{\partial y} +  \frac{\partial p' }{\partial y}  \right) + w \left( \frac{\partial \bar{p} }{\partial z} +  \frac{\partial p' }{\partial z}  \right) \\
  &=&  \left( 0 +  \frac{\partial p' }{\partial t}  \right)+ u \left( 0 +  \frac{\partial p' }{\partial x}  \right) + v \left( 0 +  \frac{\partial p' }{\partial y}  \right) + w \left( \frac{d \bar{p} }{d z} +  \frac{\partial p' }{\partial z}  \right) \\
  &=&   \frac{\partial p' }{\partial t} + u \frac{\partial p' }{\partial x}   + v\frac{\partial p' }{\partial y}  + w\frac{d \bar{p} }{d z} +  w\frac{\partial p' }{\partial z}\\
  &=&   \frac{\partial p' }{\partial t} +  (\mathbf{u} \cdot \nabla ) p'  + w\frac{d \bar{p} }{d z} \\
  &\overset{\text{假設 4}}{=}&   \frac{\partial p' }{\partial t} +  (\mathbf{u} \cdot \nabla ) p'  - w \bar{\rho} g  \\
  \end{gather*}$$

  * $p$ : 瞬時總氣壓場 (Instantaneous total pressure field) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{p}$ : 基本背景態氣壓，僅為高度 $z$ 的函數 (Basic state pressure) $[\text{Pa}]$
  * $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}]$
  * $\mathbf{u}$ : 三維速度向量 $(u, v, w)$ $[\text{m}\cdot\text{s}^{-1}]$
  * $u, v, w$ : 速度場在 $x, y, z$ 軸方向的分量 (Velocity components) $[\text{m}\cdot\text{s}^{-1}]$
  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $(\mathbf{u} \cdot \nabla ) p'$ : 擾動氣壓的平流項 (Advection of pressure perturbation) $[\text{Pa}\cdot\text{s}^{-1}]$

* **【推導 6】密度物質導數 $\frac{D \rho}{Dt}$ 的線性化拆解（Linearization and decomposition of density material derivative） ：** 

  $$\begin{gather*} 
  \frac{D\rho}{Dt} &=&  \frac{\partial \rho}{\partial t} + u \frac{\partial \rho}{\partial x} + v \frac{\partial \rho}{\partial y} + w \frac{\partial \rho}{\partial z}  \\
  &\overset{\text{假設 3(b)}}{=}&  \frac{\partial (\bar{\rho}+\rho') }{\partial t} + u \frac{\partial (\bar{\rho}+\rho')}{\partial x} + v \frac{\partial (\bar{\rho}+\rho')}{\partial y} + w \frac{\partial (\bar{\rho}+\rho')}{\partial z}  \\
  &=&  \left( \frac{\partial \bar{\rho} }{\partial t} +  \frac{\partial \rho' }{\partial t}  \right)+ u \left( \frac{\partial \bar{\rho} }{\partial x} +  \frac{\partial \rho' }{\partial x}  \right) + v \left( \frac{\partial \bar{\rho} }{\partial y} +  \frac{\partial \rho' }{\partial y}  \right) + w \left( \frac{\partial \bar{\rho} }{\partial z} +  \frac{\partial \rho' }{\partial z}  \right) \\
  &=&  \left( 0 +  \frac{\partial \rho' }{\partial t}  \right)+ u \left( 0 +  \frac{\partial \rho' }{\partial x}  \right) + v \left( 0 +  \frac{\partial \rho' }{\partial y}  \right) + w \left( \frac{d \bar{\rho} }{d z} +  \frac{\partial \rho' }{\partial z}  \right) \\
  &=&   \frac{\partial \rho' }{\partial t} + u \frac{\partial \rho' }{\partial x}   + v\frac{\partial \rho' }{\partial y}  + w\frac{d \bar{\rho} }{d z} +  w\frac{\partial \rho' }{\partial z}\\
  &=&   \frac{\partial \rho' }{\partial t} +  (\mathbf{u} \cdot \nabla ) \rho'  + w\frac{d \bar{\rho} }{d z} \\
  &\overset{\text{推導 2}}{=}&   \frac{\partial \rho' }{\partial t} +  (\mathbf{u} \cdot \nabla ) \rho'  + w \left( -\frac{\bar{\rho}}{H_s} \right)  \\
  &=&   \frac{\partial \rho' }{\partial t} +  (\mathbf{u} \cdot \nabla ) \rho'  - w \frac{\bar{\rho}}{H_s}   \\
  \end{gather*}$$

  * $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\bar{\rho}$ : 基本背景態密度，僅為高度 $z$ 的函數 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\rho '$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\mathbf{u}$ : 三維速度向量 $(u, v, w)$ $[\text{m}\cdot\text{s}^{-1}]$
  * $u, v, w$ : 速度場在 $x, y, z$ 軸方向的分量 (Velocity components) $[\text{m}\cdot\text{s}^{-1}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
  * $(\mathbf{u} \cdot \nabla ) \rho'$ : 擾動密度的平流項 (Advection of density perturbation) $[\text{kg}\cdot\text{m}^{-3}\cdot\text{s}^{-1}]$



* **【推導 7】浮力頻率 $N^2$ 與音速 $c$ 及標高 $H_s$ 的恆等式變形（Algebraic identity transformation of buoyancy frequency, speed of sound, and scale height） ：** 

  $$\begin{gather*} 
  \Gamma &=&  \Gamma  \\
  \frac{1}{2}\left(\frac{g}{c^2} - \frac{N^2}{g}\right) &\overset{\text{定義 4}}{=}&  \Gamma \\
  \frac{1}{2}\left(\frac{g}{c^2} - \frac{N^2}{g}\right) &\overset{\text{定義 4}}{=}&  \frac{g}{c^2} - \frac{1}{2H_s} \\
  \frac{1}{2} \frac{g}{c^2} - \frac{1}{2} \frac{N^2}{g} &=&  \frac{g}{c^2} - \frac{1}{2H_s} \\
  - \frac{1}{2} \frac{N^2}{g} &=&  \frac{1}{2}\frac{g}{c^2} - \frac{1}{2H_s} \\
  - \frac{N^2}{g} &=& \frac{g}{c^2} - \frac{1}{H_s} \\
  \end{gather*}$$

  * $\Gamma $ : 大氣密度層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
