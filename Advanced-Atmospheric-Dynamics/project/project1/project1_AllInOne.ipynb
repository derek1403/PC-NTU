### Eigenfunctions and eigenvalues of the fully compressible model.

We begin with a fully compressible nonhydrostatic model

$$\begin{gather*}
\frac{Du}{Dt} &=&  -\frac{1}{\rho} \frac{\partial p}{\partial x}, \tag{1a} \\
\frac{Dv}{Dt} &=&  -\frac{1}{\rho} \frac{\partial p}{\partial y}, \tag{1b} \\
\frac{Dw}{Dt} &=&  -\frac{1}{\rho} \frac{\partial p}{\partial z} - g, \tag{1c} \\
\frac{D\rho}{Dt} &=&  -\rho \left( \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} + \frac{\partial w}{\partial z} \right), \tag{1d} \\
\frac{D\rho}{Dt} &=&  \frac{1}{c^2} \frac{Dp}{Dt}, \tag{1e}
\end{gather*}$$

where $c^2$ is the square of sound speed and $D/Dt = \partial/\partial t + u\partial/\partial x + v\partial/\partial y + w\partial/\partial z$.

Let us divide the pressure and density into isothermal basic states and deviation parts, i.e.,

$$\begin{gather*}
P(x, y, z, t) &=&  \bar{p}(z) + p'(x, y, z, t), \\
\rho(x, y, z, t) &=&  \bar{\rho}(z) + \rho'(x, y, z, t),
\end{gather*}$$

where $\partial\bar{p}/\partial z = -\bar{\rho}g$.

With a resting basic state in an isothermal atmosphere (temperature $T_0$), (1) can be linearized in the form (with the nonlinear terms on the right hand side)



$$\begin{gather*}
\frac{\partial}{\partial t} \left(\bar{\rho}^{\frac{1}{2}}u\right) + \frac{\partial}{\partial x} \left( \frac{p'}{\bar{\rho}^{\frac{1}{2}}} \right) &=&  N_1, \tag{2a} \\
\frac{\partial}{\partial t} \left(\bar{\rho}^{\frac{1}{2}}v\right) + \frac{\partial}{\partial y} \left( \frac{p'}{\bar{\rho}^{\frac{1}{2}}} \right) &=&  N_2, \tag{2b} \\
\frac{\partial}{\partial t} \left(\bar{\rho}^{\frac{1}{2}}w\right) + \left( \frac{\partial}{\partial z} + \Gamma \right) \left( \frac{p'}{\bar{\rho}^{\frac{1}{2}}} \right) + g \left( \frac{1}{\bar{\rho}^{\frac{1}{2}}} \left(\rho' - \frac{p'}{c^2}\right) \right) &=&  N_3, \tag{2c} \\
\frac{\partial}{\partial t} \left( \frac{p'}{\bar{\rho}^{\frac{1}{2}}} \right) + c^2 \left( \frac{\partial}{\partial x} \left(\bar{\rho}^{\frac{1}{2}}u\right) + \frac{\partial}{\partial y} \left(\bar{\rho}^{\frac{1}{2}}v\right) + \left( \frac{\partial}{\partial z} - \Gamma \right) \left(\bar{\rho}^{\frac{1}{2}}w\right) \right) &=&  N_4, \tag{2d} \\
\frac{\partial}{\partial t} \left( \frac{1}{\bar{\rho}^{\frac{1}{2}}} \left(\rho' - \frac{p'}{c^2}\right) \right) - \frac{N^2}{g} \left(\bar{\rho}^{\frac{1}{2}}w\right) &=&  N_5, \tag{2e}
\end{gather*}$$

where the constant parameters in the isothermal atmosphere are

$$\begin{gather*}
c^2 &=&  \gamma R_d T_0, \tag{3a} \\
\Gamma &=&  \frac{1}{2} \left( \frac{g}{c^2} - \frac{N^2}{g} \right), \tag{3b} \\
N^2 &=&  \frac{\kappa g}{H_s}, \tag{3c}
\end{gather*}$$

and

$$\begin{gather*}
H_s &=&  R_d T_0 / g. \tag{3d}
\end{gather*}$$

The $\gamma$, $R_d$, and $\kappa$ take the usual meaning in the thermodynamics.

Now define the vectors

$$\tilde{u} = \begin{pmatrix}
\bar{\rho}^{\frac{1}{2}} u \\
\bar{\rho}^{\frac{1}{2}} v \\
\bar{\rho}^{\frac{1}{2}} w \\
p' / \bar{\rho}^{\frac{1}{2}} \\
(\rho' - p'/c^2) / \bar{\rho}^{\frac{1}{2}}
\end{pmatrix} = \begin{pmatrix}
U \\
V \\
W \\
P \\
\Pi
\end{pmatrix},$$

and

$$\tilde{N} = (N_1, N_2, N_3, N_4, N_5)^T,$$

we can write (2) as

$$\frac{\partial \tilde{u}}{\partial t} + L\tilde{u} = \tilde{N}, \tag{4}$$

Where

$$L = \begin{pmatrix}
0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\
0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\
0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\
c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\
0 & 0 & -\frac{N^2}{g} & 0 & 0
\end{pmatrix}.$$

Next we define the inner product of the vectors $\tilde{f}$ and $\tilde{g}$ as

$$(\tilde{f}, \tilde{g}) = \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( f_1 g_1^* + f_2 g_2^* + f_3 g_3^* + \frac{1}{c^2} f_4 g_4^* + \frac{g^2}{N^2} f_5 g_5^* \right) dxdydz, \tag{5}$$

where the $f^*$ is the complex conjugate of $f$. All the components of $\tilde{f}$ and $\tilde{g}$ are periodic in $x \text{ and } y$ and $f_3 = g_3 = 0$ at $z = 0, L_z$ (bottom and top). With the inner product (5) we can show that

$$(L\tilde{f}, \tilde{g}) = -(\tilde{f}, L\tilde{g}), \tag{6}$$

i.e., the operator $L$ is skew-Hermitian. Skew-Hermitian operator with appropriate boundary conditions have three properties that are of extreme importance in physics, both quantum and classical. The three properties are
* (1) the eigenvalues of a skew-Hermitian operator are imaginary,
* (2) the eigenfunctions of a skew-Hermitian operator are orthogonal,
* (3) the eigenfunctions of a skew-Hermitian operator form a complete set.

Finally, we note that (6) is a energy norm, it consists of the kinetic energy $\bar{\rho}/2(u^2 + v^2 + w^2)$, the elastic energy $1/(2\bar{\rho})(p'^2/c^2)$ and the thermobaric energy ( the internal energy)

$$I = \frac{\bar{\rho}}{2} \frac{g^2}{N^2} \left( \frac{\theta'}{\bar{\theta}} \right)^2,$$

or with the linearized equation of state, $(\rho - p'/c^2)\bar{\rho}^{-1/2} = -\bar{\rho}^{-1/2}(\theta'/\bar{\theta})$ ( Eckart 1960),

or with the linearized equation of state, $(\rho' - p'/c^2)\bar{\rho}^{-1/2} = -\bar{\rho}^{1/2}(\theta'/\bar{\theta})$ ( Eckart 1960),

$$I = \frac{g^2}{N^2} \frac{\left( \rho' - \frac{p'}{c^2} \right)^2}{2\bar{\rho}},$$

Thus, we can find the normalized eigenvectors $\tilde{K}_{klmn}$ of L, which satisfy

$$L\tilde{K}_{klmn} = i\nu_{klmn}\tilde{K}_{klmn} \tag{7}$$

where $\nu_{klmn}$ is real, and $k, l, m$ can be related to the wavenumbers $\bar{k}, \bar{l}, \bar{m}$ by

$$\bar{k} = \frac{2\pi k}{L_x}, \quad \bar{l} = \frac{2\pi l}{L_y}, \quad \bar{m} = \frac{2\pi m}{L_z}.$$

The $\tilde{K}_{klmn}$ form a complete space.

For a given $k, l, m$, there are six $n$’s which must be considered.

(1) The first gives zero frequency and corresponds to steady, horizontally non-divergent flow. This is really the degeneracy of the internal gravity waves with the wave number vector points in the vertical.

(2) The second frequency corresponds to Lamb waves. This solution was first discovered by Laplace (1778-1779) in discussing thermal oscillations in the atmosphere. He effectively stated that there is a mode with equivalent depth equal to the scale height of the atmosphere, and this will satisfy the tidal equations. This mode is now usually called Lamb wave, following a more complete discussion by Lamb (1910). A feature of this mode is that the velocity is everywhere parallel to the earth’s surface, i.e., $w = 0$. The propagation speed is the speed of sound and has a value of about $300\text{ m s}^{-1}$. The eigenvectors are found to be

$$\tilde{K}_{klmn} = A_{klmn} \begin{pmatrix}
\bar{k} \\
\bar{l} \\
0 \\
\nu_{klmn} \\
0
\end{pmatrix} e^{i(\bar{k}x + \bar{l}y)} e^{-\Gamma z}$$
$$ . \tag{8a}$$

with the dispersion relationship

$$\nu_{klmn}^2 = c^2 (\bar{k}^2 + \bar{l}^2), \tag{8b}$$

and the normalization factor

$$A_{klmn} = \left( \frac{(\bar{k}^2 + \bar{l}^2)(1 - e^{-2\Gamma L_z})}{\Gamma L_z} \right)^{-\frac{1}{2}}, \tag{8c}$$

(3) The remaining frequencies are the four roots (two acoustic waves and two buoyancy waves) of

$$\frac{\nu_{klmn}^4}{c^2} - \nu_{klmn}^2 \left( \bar{k}^2 + \bar{l}^2 + \bar{m}^2 + \Gamma^2 + \frac{N^2}{c^2} \right) + (\bar{k}^2 + \bar{l}^2) N^2 = 0, \tag{9a}$$

with the eigenvectors

$$\tilde{K}_{klmn} = A_{klmn} \begin{pmatrix}
\bar{k} (\bar{m} \cos \bar{m}z - \Gamma \sin \bar{m}z) \\
\bar{l} (\bar{m} \cos \bar{m}z - \Gamma \sin \bar{m}z) \\
-\frac{(\bar{m}^2 + \Gamma^2)\nu_{klmn}^2}{N^2 - \nu_{klmn}^2} i \sin \bar{m}z \\
\nu_{klmn} (\bar{m} \cos \bar{m}z - \Gamma \sin \bar{m}z) \\
\frac{N^2}{g} \frac{(\bar{m}^2 + \Gamma^2)}{N^2 - \nu_{klmn}^2} \nu_{klmn} \sin \bar{m}z
\end{pmatrix} e^{i(\bar{k}x + \bar{l}y)}$$
$$ . \tag{9b}$$

Where

$$A_{klmn} = \left( \left( \frac{\bar{m}^2 + \Gamma^2}{N^2 - \nu_{klmn}^2} \right) \left( N^2 (\bar{k}^2 + \bar{l}^2) - \frac{\nu_{klmn}^4}{c^2} \right) \right)^{-\frac{1}{2}}. \tag{9c}$$

If we take the inner product of (4) with $\tilde{K}_{klmn}$ to obtain

$$\left( \frac{\partial \tilde{u}}{\partial t}, \tilde{K}_{klmn} \right) + (L\tilde{u}, \tilde{K}_{klmn}) = (\tilde{N}, \tilde{K}_{klmn}). \tag{10}$$

With the skew-Hermitian property, we can write the second term of (10) as

$$(L\tilde{u}, \tilde{K}_{klmn}) = -(\tilde{u}, L\tilde{K}_{klmn}) = i\nu_{klmn} (\tilde{u}, \tilde{K}_{klmn}). \tag{11}$$

Now define the transform pairs of

$$\hat{u}_{klmn}(t) = (\tilde{u}, \tilde{K}_{klmn}), \tag{12a}$$

with inverse

$$\tilde{u} = \sum_{klmn} \hat{u}_{klmn}(t) \tilde{K}_{klmn}(x, y, z), \tag{12b}$$

$\tilde{N}$ and $\hat{N}_{klmn}$ can be defined similarly.

We can use (12) to write (10) as a set of scalar equations

$$\frac{d\hat{u}_{klmn}(t)}{dt} + i\nu_{klmn}\hat{u}_{klmn}(t) = \hat{N}_{klmn}, \tag{13}$$

which are the desired spectral equations. Equation (13) can be easily integrated as

$$\hat{u}_{klmn}(t + \Delta t) = \hat{u}_{klmn}(t)e^{-i\nu_{klmn}(\Delta t)} + \int_{t}^{t+\Delta t} \hat{N}_{klmn} e^{i\nu_{klmn}(t' - t - \Delta t)} dt'. \tag{14}$$

The computational procedures are:
* (1) Predict $\hat{u}_{klmn}$ in spectral space.
* (2) Transform $\hat{u}_{klmn}$ to physical space to get $\tilde{u}$.
* (3) Evaluate $\tilde{N}$ in physical space.
* (4) Transform $\tilde{N}$ to spectral space to get $\hat{N}_{klmn}$.
* (5) Go to step (1).

In step (3) and (4), the transform method of Orszag (1970) can be used. If $k$ is the truncated wave number, it is then possible to evaluate the $\hat{N}_{klmn}$ exactly using $3k + 1$ points trapezoidal quadrature, which prevents aliasing error in the calculation of the transformation of the nonlinear term. Note that the treatment of linear term in (14) is exact, which enlarge the time step of integration without the CFL constraint of the fast linear waves.

## 中文版 (In chinese)

### **一、原始的完全可壓縮流體力學偏微分方程組** 轉成 **密度加權的線性化微擾方程組（且分離出非線性強迫項）** (From the original fully compressible Navier-Stokes equations to the density-weighted linearized perturbation equations (with isolated nonlinear forcing terms))



$$\begin{gather*}
\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}} u \right] + \frac{\partial }{\partial x} \left[ \bar{\rho}^{\frac{-1}{2}} p' \right]  &=& \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) u + \frac{\rho'}{\bar{\rho} \rho  } \frac{\partial p'}{\partial x} \right) \tag{2a}\\
\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}} v \right] + \frac{\partial }{\partial y} \left[ \bar{\rho}^{\frac{-1}{2}} p' \right]  &=& \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) v + \frac{\rho'}{\bar{\rho} \rho  }\frac{\partial p'}{\partial y} \right) \tag{2b}\\
\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}}  w \right]  + g\bar{\rho}^{\frac{-1}{2}}\left(\rho' - \frac{p'}{c^2}\right) + \left(  \frac{\partial}{\partial z} + \Gamma \right)[\bar{\rho}^{\frac{-1}{2}}p'] &=& - \bar{\rho}^{\frac{1}{2}} \left( (\mathbf{u} \cdot \nabla) w  +  \frac{\rho'^2}{\rho \bar{\rho} } g +  \frac{\rho'}{\rho \bar{\rho}} \frac{\partial p'}{\partial z}  \right) \tag{2c}\\
\frac{\partial ( \bar{\rho}^{\frac{-1}{2}} p') }{\partial t} + c^2 \left( \frac{\partial ( \bar{\rho}^{\frac{1}{2}} u) }{\partial x } +  \frac{\partial  (\bar{\rho}^{\frac{1}{2}} v)}{\partial y } + \left( \frac{\partial }{\partial z} -  \Gamma \right)\left[\bar{\rho}^{\frac{1}{2}} w \right]  \right)   &=&  -\bar{\rho}^{\frac{-1}{2}} \left( (\mathbf{u} \cdot \nabla ) p' + c^2  \rho'  \nabla \cdot \mathbf{u} \right) \tag{2d}\\
\frac{\partial }{\partial t} \left[ \bar{\rho}^{\frac{-1}{2}}  \left( \rho' - \frac{p'}{c^2}  \right)   \right] -  w \bar{\rho}^{\frac{1}{2}} \frac{N^2}{g}   &=& -\bar{\rho}^{\frac{-1}{2}} \left( \mathbf{u}\cdot\nabla \rho' -\frac{1}{c^2} \mathbf{u}\cdot\nabla p'  \right) \tag{2e}
\end{gather*}$$

* $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
* $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
* $\rho '$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$
* $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}]$
* $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
* $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
* $\Gamma $ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
* $N^{2}$ : 浮力頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
* $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$


#### **為什麼要花這麼大功夫轉成這樣？（The Ultimate Goal）**

當方程式轉成具有「反厄米特特性（Skew-Hermitian）」的線性算子 $L$ 後，數學上會保證三件極度美好的事情（正如講義提到的）：
1. 特徵值純虛數：這代表方程式的解是純粹的「震盪波」，不會有數值上無端呈指數爆炸的解（保證穩定性）。
2. 特徵函數正交：你可以用這組解當作基底（Basis），像傅立葉轉換一樣，把大氣真實的狀態拆解成無數個獨立波的疊加。
3. 精確處理快波：在數值模式（如天氣預報）中，聲波跑得非常快，會嚴重限制時間步長（CFL condition）。把線性項抽出來並精確解析後，程式在時間積分時就可以跨大步長，只針對非線性強迫項 $N$ 做數值積分（例如講義 (14) 式的處理方式），大幅節省超級電腦的運算資源。


#### 假設與已知 (Assumptions & Preliminaries)


* **【假設 1】 忽略黏滯力（Inviscid fluid assumption）  ：**
  
  $$ \mu = 0$$

  * $\mu $ : 動力黏滯係數 (Dynamic viscosity) $[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$


* **【已知 1】 [動量方程式 (Momentum Equation) - 納維-斯托克斯 (Navier-Stokes)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#momentum-equation-navier-stokes)  ：**
  
  * (a) 
   
  $$\begin{gather*}
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &=& \mathbf{g} - \frac{1}{\rho} \nabla p + \frac{\mu}{\rho} \nabla^2 \mathbf{u} \\
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &\overset{\text{假設 1}}{=}& \mathbf{g} - \frac{1}{\rho} \nabla p + \frac{0}{\rho} \nabla^2 \mathbf{u} \\
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &=& \mathbf{g} - \frac{1}{\rho} \nabla p  \\
  \end{gather*}$$

  * (b) x 方向
  
  $$\frac{\partial u}{\partial t } + (\mathbf{u} \cdot \nabla) u \overset{\text{已知 1(a)}}{=} - \frac{1}{\rho} \frac{\partial p}{\partial x} $$

  * (c) y 方向  
  
  $$\frac{\partial v}{\partial t } + (\mathbf{u} \cdot \nabla) v \overset{\text{已知 1(a)}}{=} - \frac{1}{\rho} \frac{\partial p}{\partial y} $$

  * (d) z 方向
  
  $$\frac{\partial w}{\partial t } + (\mathbf{u} \cdot \nabla) w \overset{\text{已知 1(a)}}{=} -g - \frac{1}{\rho} \frac{\partial p}{\partial z} $$


  * $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
  * $\mathbf{g}$: 重力加速度向量 $(0, 0, -g)$ $[\text{m} \cdot \text{s}^{-2}]$
  * $p$: 氣壓 $ [\text{Pa}],[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}] $
  * $\rho$: 流體密度 $[\text{kg} \cdot \text{m}^{-3}]$
  * $\mu$: 動力黏滯係數 (Dynamic viscosity) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}] $

* **【已知 2】 [連續方程式 (Continuity Equation)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#continuity-equation)  ：**
  
  $$\begin{gather*}
  \frac{\partial \rho}{\partial t } + \nabla \cdot (\rho \mathbf{u}) &=& 0 \\
  \frac{\partial \rho}{\partial t } + \frac{\partial (\rho u)}{\partial x }+ \frac{\partial (\rho v)}{\partial y }+ \frac{\partial (\rho w)}{\partial z } &=& 0 \\
  \frac{\partial \rho}{\partial t } + \left( \frac{\partial \rho }{\partial x } u + \rho \frac{\partial  u}{\partial x } \right)+ \left( \frac{\partial \rho }{\partial y } v + \rho \frac{\partial  v}{\partial y } \right) + \left( \frac{\partial \rho }{\partial z } w + \rho \frac{\partial  w}{\partial z } \right)&=& 0 \\
  \frac{\partial \rho}{\partial t } + u \frac{\partial \rho }{\partial x } + v\frac{\partial \rho }{\partial y }   + w \frac{\partial \rho }{\partial z } &=& - \rho \left( \frac{\partial  u}{\partial x } + \frac{\partial  v}{\partial y } +\frac{\partial  w}{\partial z }\right)\\
  \end{gather*}$$

  * $\rho$: 流體密度 $[\text{kg} \cdot \text{m}^{-3}]$
  * $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
  * $t$: 時間 $[\text{s}]$。

* **【假設 2】 等熵過程假設（Isentropic process assumption）：**
  
  $$\frac{Ds}{Dt} = 0$$

  * $s$ : 比熵／單位質量熵 (Specific entropy) $[\text{J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}]$



* **【假設 3】 線性微擾狀態假設（Linear perturbation state assumption） ：** 將流體的氣壓與密度分為隨高度變化的基本背景態（Basic state）與隨時空變化的微小擾動態（Perturbation）。

  * (a) 氣壓場拆解：
  
  $$p(x,y,z,t) = \bar{p}(z) + p'(x,y,z,t)$$

  * (b) 密度場拆解：
  
  $$\rho(x,y,z,t) = \bar{\rho}(z) + \rho'(x,y,z,t)$$

  * $\bar{p}(z)$ : 僅隨高度 $z$ 改變的基本背景態氣壓 (Basic state pressure)，$ [\text{Pa}],[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}] $
  * $\bar{\rho}(z)$ : 僅隨高度 $z$ 改變的基本背景態密度 (Basic state density)，$[\text{kg} \cdot \text{m}^{-3}]$
  * $p'$ : 隨時空變化的壓力微擾項 (Perturbation pressure)，滿足 $p' \ll \bar{p}$， $ [\text{Pa}],[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}] $
  * $\rho'$ : 隨時空變化的微擾項 (Perturbation density)，滿足  $\rho' \ll \bar{\rho}$，$[\text{kg} \cdot \text{m}^{-3}]$


* **【假設 4】 背景態靜力平衡（Hydrostatic equilibrium of the basic state） ：** 假設無任何外部干擾時，背景大氣在垂直方向上重力與氣壓梯度力完全抵消。

  $$\frac{d\bar{p}}{dz} = -\bar{\rho}g$$

  * $\bar{P}$ : 基本背景態氣壓 (Basic state pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $z$ : 垂直高度座標 (Vertical height coordinate) $[\text{m}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $\frac{d\bar{P}}{dz}$ : 背景氣壓隨高度變化的全導數（垂直氣壓梯度） $[\text{Pa}\cdot\text{m}^{-1}]$


* **【假設 5】等溫基本態假設（Isothermal basic state）：** 假設基本背景場的大氣溫度為不隨時間與空間變化的常數
 
 
  $$T(x,y,z,t) = T_0 $$

  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $T(x,y,z,t)$ : 瞬時大氣絕對溫度場 (Instantaneous absolute temperature field) $[\text{K}]$




* **【定義 1】音速的熱力學與大氣特徵定義（Thermodynamic and atmospheric definition of the speed of sound）  ：**
  
  * (a)  熱力學基本定義：
  
  $$\begin{gather*}
  c^2 &\overset{\text{def}}{=}& \left( \frac{\partial p}{\partial \rho} \right)_s \\
  \end{gather*}$$

  * (b) 理想氣體與背景態化簡：
 
  $$\begin{gather*}
  c^2 &\overset{\text{def}}{=}& \gamma R_d T \\
  &\overset{\text{假設 5}}{=}& \gamma R_d T_0
  \end{gather*}$$

  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $c^{2}$ : 音速的平方，代表氣壓隨密度變化的等熵彈性模量比值 $[\text{m}^2\cdot\text{s}^{-2}]$
  * $p$ : 氣壓 (Pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\rho $ : 流體密度 (Fluid density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $s$ : 比熵／單位質量熵 (Specific entropy) $[\text{J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}]$
  * $\left(\frac{\partial P}{\partial \rho }\right)_{s}$ : 在恆定比熵（等熵）條件下，氣壓對密度的偏微分偏導數 (Isentropic pressure derivative with respect to density) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $(\dots)_s$ : 表示在恆定比熵（等熵）條件下的偏微分
  * $\gamma $ : 絕熱指數／熱容量比 (Adiabatic index / Ratio of specific heats)，對於大氣雙原子主要氣體（如 $\text{N}_2, \text{O}_2$）而言 $\gamma = \frac{c_p}{c_v}\approx 1.4$ [無單位]
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1},[\text{m}^2\cdot\text{s}^{-2}\cdot\text{K}^{-1}]$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$

* **【定義 2】 等溫大氣標高（Scale height of an isothermal atmosphere） ：**  標高代表大氣壓力和密度隨高度垂直遞減的特徵尺度。在等溫大氣假設下，這個高度尺度為一個常數。
  
  $$\begin{gather*}
  H_s &\overset{\text{def}}{=}& \frac{R_d T}{g} \\
  &\overset{\text{假設 5}}{=}& \frac{R_d T_0}{g}
  \end{gather*}$$

  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$，代表大氣氣壓或密度遞減為地表的 $1/e$（約 36.8%）時所需上升的高度
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1},[\text{m}^2\cdot\text{s}^{-2}\cdot\text{K}^{-1}]$
  * $T$ : 瞬時絕對溫度 (Instantaneous absolute temperature) $[\text{K}]$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$

* **【定義 3】浮力頻率平方 / Brunt-Väisälä frequency 平方 ：** 浮力頻率代表大氣在垂直方向上的靜力穩定度指標。當流體微塊受到垂直擾動時，此頻率決定了它在原平衡位置上下振盪的特徵頻率。
  
  $$\begin{gather*}
  N^2 &\overset{\text{def}}{=}& \frac{\kappa g}{H_s} \\
  &=& \left(1 - \frac{1}{\gamma} \right) \frac{ g}{H_s} \\
  \end{gather*}$$

  * $N^{2}$ : 浮力頻率平方 / Brunt-Väisälä frequency squared $[\text{s}^{-2}]$
  * $\kappa $ : 乾空氣大氣熱力學常數 (Kappa constant for dry air) $\kappa = \frac{R_d}{C_p} = 1 - \frac{1}{\gamma}$，對於大氣雙原子氣體而言 $\kappa \approx 0.286$ [無單位]
  * $C_{p}$ : 乾空氣定壓比熱 (Specific heat of dry air at constant pressure) $C_p \approx 1005 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}$
  * $\gamma $ : 絕熱指數／熱容量比 (Adiabatic index)，大氣雙原子氣體 $\gamma \approx 1.4$ [無單位]
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$


* **【定義 4】  大氣密度層結修正因子 / 垂直衰減係數（Atmospheric density stratification correction factor / Eckart parameter $\Gamma $） ：** 該常數因子結合了聲學剪切（由音速 $c^{2}$ 決定）與浮力剪切（由浮力頻率 $N^{2}$ 決定），用於在波動方程中消除因大氣密度隨高度指數遞減所造成的振幅放大效應。
  
  $$\begin{gather*}
  \Gamma &\overset{\text{def}}{=}& \frac{1}{2}\left(\frac{g}{c^2} - \frac{N^2}{g}\right) \\
  &\overset{\text{定義 3}}{=}& \frac{1}{2}\left(\frac{g}{c^2} - \frac{\left(1 - \frac{1}{\gamma} \right) \frac{ g}{H_s}}{g}\right) \\
  &=& \frac{1}{2}\left(\frac{g}{c^2} - \left(1 - \frac{1}{\gamma} \right) \frac{ 1}{H_s}\right) \\
  &\overset{\text{定義 1(b)}}{=}& \frac{1}{2}\left(\frac{g}{\gamma R_d T_0} - \left(1 - \frac{1}{\gamma} \right) \frac{ 1}{H_s}\right) \\
  &\overset{\text{定義 2}}{=}& \frac{1}{2}\left(\frac{g}{\gamma R_d T_0} - \left(1 - \frac{1}{\gamma} \right) \frac{ 1}{\frac{R_d T_0}{g}}\right) \\
  &=& \frac{1}{2}\left(\frac{g}{\gamma R_d T_0} - \left(1 - \frac{1}{\gamma} \right) \frac{ g}{R_d T_0}\right) \\
  &=& \frac{1}{2}\frac{g}{R_d T_0}\left( \frac{1}{\gamma } - \left(1 - \frac{1}{\gamma} \right)\right) \\
  &=& \frac{1}{2}\frac{g}{R_d T_0}\left( \frac{2}{\gamma} - 1   \right) \\
  &=&  \frac{g}{\gamma R_d T_0} - \frac{1}{2}\frac{g}{R_d T_0}    \\
  &\overset{\text{定義 1(b),定義 2}}{=}& \frac{g}{c^2} - \frac{1}{2}\frac{1}{H_s}  \\
  &=&  \frac{g}{c^2} - \frac{1}{2H_s}    \\
  \end{gather*}$$

  * $\Gamma $ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c^{2}$ : 音速的平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $N^{2}$ : 浮力頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $\gamma $ : 絕熱指數／熱容量比 (Adiabatic index)，大氣雙原子氣體 $\gamma \approx 1.4$ [無單位]
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$



* **【已知 3】 一般熱力學狀態方程式之物質導數 (Material Derivative of the General Equation of State)  ：**
  
  $$\begin{gather*}
  \frac{Dp}{Dt} &=& \frac{D}{Dt} \left[ p(\rho, s) \right] \\
  \frac{Dp}{Dt}&=& \left( \frac{\partial p}{\partial \rho} \right)_s \frac{D\rho}{Dt} + \left( \frac{\partial p}{\partial s} \right)_\rho \frac{Ds}{Dt} \\
  \frac{Dp}{Dt}&\overset{\text{假設 2}}{=}& \left( \frac{\partial p}{\partial \rho} \right)_s \frac{D\rho}{Dt} + \left( \frac{\partial p}{\partial s} \right)_\rho \cdot 0 \\
  \frac{Dp}{Dt}&=& \left( \frac{\partial p}{\partial \rho} \right)_s \frac{D\rho}{Dt}  \\
  \frac{Dp}{Dt} &\overset{\text{定義 1}}{=}& c^2 \frac{D\rho}{Dt}  \\
  \frac{1}{c^2} \frac{Dp}{Dt} &=&  \frac{D\rho}{Dt}  \\
  \end{gather*}$$

  * $c^{2}$ : 音速的平方，代表氣壓隨密度變化的等熵彈性模量比值 $[\text{m}^2\cdot\text{s}^{-2}]$
  * $p$ : 氣壓 (Pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\rho $ : 流體密度 (Fluid density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $s$ : 比熵／單位質量熵 (Specific entropy) $[\text{J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1}]$
  * $\left(\frac{\partial P}{\partial \rho }\right)_{s}$ : 在恆定比熵（等熵）條件下，氣壓對密度的偏微分偏導數 (Isentropic pressure derivative with respect to density) $[\text{m}^2\cdot\text{s}^{-2}]$



* **【已知 4】理想氣體狀態方程式 (Ideal Gas Equation of State)：** 

  $$\begin{gather*}
  p &=& \rho R_d T \\
  p &\overset{\text{假設 5}}{=}& \rho R_d T_0 \\
  \bar{p} &=&\overline{\rho R_d T_0 } \\
  \bar{p} &=&\bar{\rho } R_d T_0  \\
  \bar{p} &\overset{\text{定義 2}}{=}& \bar{\rho} H_s g \\ 
  \end{gather*}$$

  * $p$ : 瞬時總氣壓場 (Instantaneous total pressure field) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $R_{d}$ : 乾空氣氣體常數 (Gas constant for dry air) $R_d \approx 287.05 \text{ J}\cdot\text{kg}^{-1}\cdot\text{K}^{-1},[\text{m}^2\cdot\text{s}^{-2}\cdot\text{K}^{-1}]$
  * $T$ : 瞬時絕對溫度場 (Instantaneous absolute temperature field) $[\text{K}]$
  * $T_{0}$ : 基本背景態的常數絕對溫度 (Basic state constant absolute temperature) $[\text{K}]$
  * $\bar{p}$ : 基本背景態氣壓 (Basic state pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$






* **【推導 1】 倒數密度項的代數拆解（Algebraic decomposition of reciprocal density）：** 

  * (a) 計算基本態倒數與總場倒數之差
  
  $$\begin{gather*}
  \frac{1}{\bar{\rho}} - \frac{1}{\rho} &=&   \frac{\rho - \bar{\rho}}{\bar{\rho} \rho }   \\
  &\overset{\text{已知 1(b)}}{=}&  \frac{(\bar{\rho} + \rho') - \bar{\rho} }{\bar{\rho} \rho  } \\
  &=& \frac{\rho'}{\bar{\rho} \rho  } \\
  \end{gather*}$$

  * (b) 利用恆等式展開總場倒數密度

  $$\begin{gather*}
  \frac{1}{\rho} &=& \frac{1}{\rho} - \frac{1}{\bar{\rho}} + \frac{1}{\bar{\rho}}\\
  &=& - \left( \frac{1}{\bar{\rho}} - \frac{1}{\rho} \right) + \frac{1}{\bar{\rho}} \\
  &\overset{\text{推導 1(a)}}{=}& -\left( \frac{\rho'}{\rho \bar{\rho} }  \right) + \frac{1}{\bar{\rho}} \\
  \end{gather*}$$

  * $\rho$ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\rho'$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$

* **【推導 2】 背景態密度垂直梯度公式（Derivation of basic state density vertical gradient） ：** 

  $$\begin{gather*}
  \frac{d\bar{p}}{dz} &\overset{\text{假設 4}}{=}& -\bar{\rho}g \\
  \frac{d}{dz}\left[ \bar{\rho} H_s g \right] &\overset{\text{已知 4}}{=}& -\bar{\rho}g \\
  H_s g \frac{d \bar{\rho}}{dz} &=& -\bar{\rho}g \\
  \frac{d \bar{\rho}}{d z} &=& -\frac{\bar{\rho}}{H_s} \\
  \frac{\partial \bar{\rho}}{\partial z} &=& -\frac{\bar{\rho}}{H_s} \\
  \end{gather*}$$

  * $\bar{p}$ : 基本背景態氣壓 (Basic state pressure) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $z$ : 垂直高度座標 (Vertical height coordinate) $[\text{m}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.807 \text{ m}\cdot\text{s}^{-2}$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$



* **【推導 3】  氣壓微擾項與密度權重的垂直微分展開（Vertical derivative expansion of pressure perturbation with density weighting） ：** 

  $$\begin{gather*}
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] &=&  \frac{\partial (\bar{\rho}^{\frac{-1}{2}}) }{\partial z} p' + \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z} \\ 
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] &=&  \left(-\frac{1}{2}\bar{\rho}^{\frac{-3}{2}} \frac{\partial \bar{\rho}}{\partial z}   \right)  p' + \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z} \\ 
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] &\overset{\text{推導 2}}{=}& \left(-\frac{1}{2}\bar{\rho}^{\frac{-3}{2}} \left( -\frac{\bar{\rho}}{H_s} \right)   \right)  p' + \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z} \\ 
  \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p']  &=&  \frac{1}{2H_s}\bar{\rho}^{\frac{-1}{2}} p'+ \bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z}\\ 
  -\bar{\rho}^{\frac{-1}{2}} \frac{\partial p'}{\partial z}   &=& - \frac{\partial}{\partial z}[\bar{\rho}^{\frac{-1}{2}}p'] + \frac{1}{2H_s}\bar{\rho}^{\frac{-1}{2}} p' \\ 
  \end{gather*}$$

  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $P'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$



* **【推導 4】垂直速度項與密度權重的垂直微分展開（Vertical derivative expansion of vertical velocity with density weighting） ：** 

  $$\begin{gather*} 
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &=&   \frac{d (\bar{\rho}^{\frac{1}{2}}) }{d z} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &=&   \frac{1}{2}  \bar{\rho}^{\frac{-1}{2}} \frac{d \bar{\rho} }{d z} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &\overset{\text{推導 2}}{=}&   \frac{1}{2}  \bar{\rho}^{\frac{-1}{2}} \left( - \frac{\bar{\rho}}{H_s} \right) w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &=&   -\frac{1}{2H_s}  \bar{\rho}^{\frac{1}{2}} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} &\overset{\text{定義 4}}{=}&  \left( \Gamma - \frac{g}{c^2} \right) \bar{\rho}^{\frac{1}{2}} w  + \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \frac{\partial (\bar{\rho}^{\frac{1}{2}} w )}{\partial z} - \left( \Gamma - \frac{g}{c^2} \right) \bar{\rho}^{\frac{1}{2}} w &=&     \bar{\rho}^{\frac{1}{2}} \frac{\partial  w }{\partial z} \\
  \end{gather*}$$

  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $w$ : 垂直向速度場（此處主要考慮一階微擾項 $w'$） (Vertical velocity) $[\text{m}\cdot\text{s}^{-1}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
  * $\Gamma $ : 大氣密度層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.807 \text{ m}\cdot\text{s}^{-2}$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$


* **【推導 5】氣壓物質導數 $\frac{Dp}{Dt}$ 的線性化拆解（Linearization and decomposition of pressure material derivative） ：** 

  $$\begin{gather*} 
  \frac{Dp}{Dt} &=&  \frac{\partial p}{\partial t} + u \frac{\partial p}{\partial x} + v \frac{\partial p}{\partial y} + w \frac{\partial p}{\partial z}  \\
  &\overset{\text{假設 3(a)}}{=}&  \frac{\partial (\bar{p}+p') }{\partial t} + u \frac{\partial (\bar{p}+p')}{\partial x} + v \frac{\partial (\bar{p}+p')}{\partial y} + w \frac{\partial (\bar{p}+p')}{\partial z}  \\
  &=&  \left( \frac{\partial \bar{p} }{\partial t} +  \frac{\partial p' }{\partial t}  \right)+ u \left( \frac{\partial \bar{p} }{\partial x} +  \frac{\partial p' }{\partial x}  \right) + v \left( \frac{\partial \bar{p} }{\partial y} +  \frac{\partial p' }{\partial y}  \right) + w \left( \frac{\partial \bar{p} }{\partial z} +  \frac{\partial p' }{\partial z}  \right) \\
  &=&  \left( 0 +  \frac{\partial p' }{\partial t}  \right)+ u \left( 0 +  \frac{\partial p' }{\partial x}  \right) + v \left( 0 +  \frac{\partial p' }{\partial y}  \right) + w \left( \frac{d \bar{p} }{d z} +  \frac{\partial p' }{\partial z}  \right) \\
  &=&   \frac{\partial p' }{\partial t} + u \frac{\partial p' }{\partial x}   + v\frac{\partial p' }{\partial y}  + w\frac{d \bar{p} }{d z} +  w\frac{\partial p' }{\partial z}\\
  &=&   \frac{\partial p' }{\partial t} +  (\mathbf{u} \cdot \nabla ) p'  + w\frac{d \bar{p} }{d z} \\
  &\overset{\text{假設 4}}{=}&   \frac{\partial p' }{\partial t} +  (\mathbf{u} \cdot \nabla ) p'  - w \bar{\rho} g  \\
  \end{gather*}$$

  * $p$ : 瞬時總氣壓場 (Instantaneous total pressure field) $[\text{Pa}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{p}$ : 基本背景態氣壓，僅為高度 $z$ 的函數 (Basic state pressure) $[\text{Pa}]$
  * $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}]$
  * $\mathbf{u}$ : 三維速度向量 $(u, v, w)$ $[\text{m}\cdot\text{s}^{-1}]$
  * $u, v, w$ : 速度場在 $x, y, z$ 軸方向的分量 (Velocity components) $[\text{m}\cdot\text{s}^{-1}]$
  * $\bar{\rho}$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.807 \text{ m}\cdot\text{s}^{-2}$
  * $(\mathbf{u} \cdot \nabla ) p'$ : 擾動氣壓的平流項 (Advection of pressure perturbation) $[\text{Pa}\cdot\text{s}^{-1}]$

* **【推導 6】密度物質導數 $\frac{D \rho}{Dt}$ 的線性化拆解（Linearization and decomposition of density material derivative） ：** 

  $$\begin{gather*} 
  \frac{D\rho}{Dt} &=&  \frac{\partial \rho}{\partial t} + u \frac{\partial \rho}{\partial x} + v \frac{\partial \rho}{\partial y} + w \frac{\partial \rho}{\partial z}  \\
  &\overset{\text{假設 3(b)}}{=}&  \frac{\partial (\bar{\rho}+\rho') }{\partial t} + u \frac{\partial (\bar{\rho}+\rho')}{\partial x} + v \frac{\partial (\bar{\rho}+\rho')}{\partial y} + w \frac{\partial (\bar{\rho}+\rho')}{\partial z}  \\
  &=&  \left( \frac{\partial \bar{\rho} }{\partial t} +  \frac{\partial \rho' }{\partial t}  \right)+ u \left( \frac{\partial \bar{\rho} }{\partial x} +  \frac{\partial \rho' }{\partial x}  \right) + v \left( \frac{\partial \bar{\rho} }{\partial y} +  \frac{\partial \rho' }{\partial y}  \right) + w \left( \frac{\partial \bar{\rho} }{\partial z} +  \frac{\partial \rho' }{\partial z}  \right) \\
  &=&  \left( 0 +  \frac{\partial \rho' }{\partial t}  \right)+ u \left( 0 +  \frac{\partial \rho' }{\partial x}  \right) + v \left( 0 +  \frac{\partial \rho' }{\partial y}  \right) + w \left( \frac{d \bar{\rho} }{d z} +  \frac{\partial \rho' }{\partial z}  \right) \\
  &=&   \frac{\partial \rho' }{\partial t} + u \frac{\partial \rho' }{\partial x}   + v\frac{\partial \rho' }{\partial y}  + w\frac{d \bar{\rho} }{d z} +  w\frac{\partial \rho' }{\partial z}\\
  &=&   \frac{\partial \rho' }{\partial t} +  (\mathbf{u} \cdot \nabla ) \rho'  + w\frac{d \bar{\rho} }{d z} \\
  &\overset{\text{推導 2}}{=}&   \frac{\partial \rho' }{\partial t} +  (\mathbf{u} \cdot \nabla ) \rho'  + w \left( -\frac{\bar{\rho}}{H_s} \right)  \\
  &=&   \frac{\partial \rho' }{\partial t} +  (\mathbf{u} \cdot \nabla ) \rho'  - w \frac{\bar{\rho}}{H_s}   \\
  \end{gather*}$$

  * $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\bar{\rho}$ : 基本背景態密度，僅為高度 $z$ 的函數 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\rho '$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\mathbf{u}$ : 三維速度向量 $(u, v, w)$ $[\text{m}\cdot\text{s}^{-1}]$
  * $u, v, w$ : 速度場在 $x, y, z$ 軸方向的分量 (Velocity components) $[\text{m}\cdot\text{s}^{-1}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$
  * $(\mathbf{u} \cdot \nabla ) \rho'$ : 擾動密度的平流項 (Advection of density perturbation) $[\text{kg}\cdot\text{m}^{-3}\cdot\text{s}^{-1}]$



* **【推導 7】浮力頻率 $N^2$ 與音速 $c$ 及標高 $H_s$ 的恆等式變形（Algebraic identity transformation of buoyancy frequency, speed of sound, and scale height） ：** 

  $$\begin{gather*} 
  \Gamma &=&  \Gamma  \\
  \frac{1}{2}\left(\frac{g}{c^2} - \frac{N^2}{g}\right) &\overset{\text{定義 4}}{=}&  \Gamma \\
  \frac{1}{2}\left(\frac{g}{c^2} - \frac{N^2}{g}\right) &\overset{\text{定義 4}}{=}&  \frac{g}{c^2} - \frac{1}{2H_s} \\
  \frac{1}{2} \frac{g}{c^2} - \frac{1}{2} \frac{N^2}{g} &=&  \frac{g}{c^2} - \frac{1}{2H_s} \\
  - \frac{1}{2} \frac{N^2}{g} &=&  \frac{1}{2}\frac{g}{c^2} - \frac{1}{2H_s} \\
  - \frac{N^2}{g} &=& \frac{g}{c^2} - \frac{1}{H_s} \\
  \end{gather*}$$

  * $\Gamma $ : 大氣密度層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.807 \text{ m}\cdot\text{s}^{-2}$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $H_{s}$ : 大氣標高 (Atmospheric scale height) $[\text{m}]$

### **三、確認算子 $\mathbf{L}$ 的反厄米特特性**

$$(\mathbf{L}\tilde{f}, \tilde{g}) = -(\tilde{f}, L\tilde{g})$$

### **四、確認內積的範數正定性**

$$(\tilde{u}, \tilde{u}) \ge 0$$

在線性代數中，一個空間要定義 **長度** 或 **距離** ，就必須引入範數（Norm）。但一個合法的範數不能隨便定義，它必須滿足三個嚴格的數學條件，其中最關鍵的第一條就是「正定性」（Positive Definiteness）。 對於任何非零的向量 $\bar{u}$，它的自我內積（或範數平方）必須嚴格大於零；只有當向量本身是零向量時，其值才等於零。
1. $(\tilde{u}, \tilde{u}) > 0 \quad (\text{if } \tilde{u} \neq \mathbf{0})$
2. $(\tilde{u}, \tilde{u}) = 0 \quad (\text{iff } \tilde{u} = \mathbf{0})$

  * $c$ : 本地音速 $[\text{m}\cdot\text{s}^{-1}]$

* **【已知 2】 能量權重空間平均內積（Energy-weighted spatially averaged inner product）  ：**

  * $(\mathbf{\tilde{f}}, \mathbf{\tilde{g}})$ : 兩個波動狀態向量在全域空間上的內積值 (Spatially averaged inner product) [多元複合單位]
  * $f_{i}$ : 第一個測試函數向量 $\bar{\mathbf{f}}$ 的第 $i$ 個純量分量
  * $g_{i}^{*}$ : 第二個測試函數向量 $\bar{\mathbf{g}}$ 的第 $i$ 個純量分量的共軛複數 (Complex conjugate)
  * $x, y, z$ : 三維直角空間座標 $[\text{m}]$$L_x, L_y$ : 水平方向的長度與寬度 $[\text{m}]$
  * $L_{z}$ : 垂直方向的大氣總高度 $[\text{m}]$

&\ge& 0 \\
\end{gather*}$$


且 $(\tilde{u}, \tilde{u}) = 0$ 的唯一條件是系統完全處於靜止的背景平衡態（無任何速度、氣壓與位溫擾動），這完美滿足了線性代數中「能量範數 (Energy Norm)」的嚴格定義 。

### 五、大氣矩陣微分算符 $\mathbf{L}$ 的特徵解與色散關係推導

#### 假設與已知 (Assumptions & Preliminaries)


* **【已知 1】 空間微分波動算符矩陣（Spatial differential wave operator matrix） ：**
  
  $$ \mathbf{L} \overset{\text{def}}{=} \begin{bmatrix} 
  0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\[2ex]
  0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\[2ex]
  0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\[2ex]
  c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\[2ex]
  0 & 0 & -\frac{N^2}{g} & 0 & 0 \\[2ex]
  \end{bmatrix}  $$

  * $\mathbf{L}$ : 封裝空間偏微分與大氣層結常數的 $5 \times 5$ 矩陣微分算符 (Spatial differential operator matrix)
  * $\Gamma $ : 大氣層結修正因子 $[\text{m}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 $[\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $\frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z}$ : 各時空空間軸向的偏微分算符 $[\text{m}^{-1}]$
  
* **【定義 1】 空間結構向量 $\tilde{K}$ ：**

  $$\tilde{K} \overset{\text{def}}{=} \begin{bmatrix}  
  \hat{u}\\[2ex]
  \hat{v}\\[2ex]
  \hat{w}\\[2ex]
  \hat{p}\\[2ex]
  \hat{b}\\[2ex]
  \end{bmatrix}$$

  * $\bar{\mathbf{K}}$ : 大氣三維空間結構特徵向量 (Spatial structure eigenvector)
  * $\hat{u}, \hat{v}, \hat{w}$ : 經密度加權後，在波數空間中的 $x, y, z$ 三軸方向特徵速度結構函數 $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $\hat{p}$ : 經密度加權後，在波數空間中的特徵氣壓結構函數 $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $\hat{b}$ : 經密度加權後，在波數空間中的特徵熱力學／浮力結構函數 $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-3}]$



* **【已知 1】 特徵值分量方程式（Eigenvalue component equations）：** 將算子 $\mathbf{L}\tilde{K} = i\nu\tilde{K}$ 展開，尋找特徵頻率 $\nu$ 與空間結構向量 $\tilde{K} = (\hat{u}, \hat{v}, \hat{w}, \hat{p}, \hat{b})^T$ 的關係。

  * (a) 展開 $\mathbf{L}\tilde{K} = i\nu\tilde{K}$ 

  $$\begin{gather*}
  \mathbf{L}\tilde{K} &=& i\nu\tilde{K} \tag{7}\\
  \begin{bmatrix} 
  0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\[2ex]
  0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\[2ex]
  0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\[2ex]
  c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\[2ex]
  0 & 0 & -\frac{N^2}{g} & 0 & 0 \\[2ex]
  \end{bmatrix} \begin{bmatrix}   \hat{u}\\[2ex]  \hat{v}\\[2ex]  \hat{w}\\[2ex]  \hat{p}\\[2ex]  \hat{b}\\[2ex]  \end{bmatrix}
  &\overset{\text{已知 1, 定義 1}}{=}& i\nu \begin{bmatrix}   \hat{u}\\[2ex]  \hat{v}\\[2ex]  \hat{w}\\[2ex]  \hat{p}\\[2ex]  \hat{b}\\[2ex]  \end{bmatrix} \\
  \begin{bmatrix}   
  \frac{\partial \hat{p}}{\partial x}\\[2ex]  
  \frac{\partial \hat{p}}{\partial y}\\[2ex]  
  \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} + g \hat{b}\\[2ex]  
  c^2 \left( \frac{\partial \hat{u}}{\partial x} + \frac{\partial \hat{v}}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right)\\[2ex]  
  -\frac{N^2}{g} \hat{w}\\[2ex]   
  \end{bmatrix} &=& i\nu \begin{bmatrix}   \hat{u}\\[2ex]  \hat{v}\\[2ex]  \hat{w}\\[2ex]  \hat{p}\\[2ex]  \hat{b}\\[2ex]  \end{bmatrix} \\
  \end{gather*}$$


  * (b) x 方向動量：

  $$ \frac{\partial \hat{p}}{\partial x} \overset{\text{已知 1(a)}}{=} i\nu \hat{u}$$


  * (c) y 方向動量：

  $$ \frac{\partial \hat{p}}{\partial y} \overset{\text{已知 1(a)}}{=} i\nu \hat{v}$$


  * (d) 垂直方向動量：

  $$ \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} + g \hat{b} \overset{\text{已知 1(a)}}{=} i\nu \hat{w}$$


  * (e) 連續方程式（質量守恆）：

  $$c^2 \left( \frac{\partial \hat{u}}{\partial x} + \frac{\partial \hat{v}}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right) \overset{\text{已知 1(a)}}{=} i\nu \hat{p} $$


  * (f) 熱力學方程式（浮力頻率）：

  $$-\frac{N^2}{g} \hat{w} \overset{\text{已知 1(a)}}{=} i\nu \hat{b}$$


  * $\hat{u}, \hat{v}, \hat{w}, \hat{p}, \hat{b}$ : 三維空間上的微擾特徵振幅函數
  * $\nu$ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $i$ : 虛數單位 $(i = \sqrt{-1})$
  * $\Gamma$ : 大氣層結修正因子 $[\text{m}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 $[\text{s}^{-2}]$
  * $c^{2}$ : 本地音速平方 $[\text{m}^2\cdot\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$


* **【假設 1】 水平平面波解 (Horizontal plane wave solution)：** 假設大氣水平方向無邊界限制，擾動以平面波形式 $e^{i (\bar{k}+\bar{l})}$展開。
  
  * (a) x 方向空間導數：

  $$\begin{gather*}
  \frac{\partial}{\partial x} \left[ e^{i (\bar{k}+\bar{l})} \right] &=&  i\bar{k}  \left(e^{i (\bar{k}+\bar{l})} \right ) \\
  \frac{\partial}{\partial x} &	\Longleftrightarrow & i\bar{k}
  \end{gather*}$$


  * (b) y 方向空間導數：

  $$\begin{gather*}
  \frac{\partial}{\partial y} \left[ e^{i (\bar{k}+\bar{l})} \right] &=&  i\bar{l}  \left(e^{i (\bar{k}+\bar{l})} \right ) \\
  \frac{\partial}{\partial y} &	\Longleftrightarrow & i\bar{l}
  \end{gather*}$$



  * $\bar{k}$ : $x$ 方向的水平波數 (x-axis Horizontal wavenumbers) $[\text{m}^{-1}]$
  * $\bar{l}$ : $y$ 方向的水平波數 (y-axis Horizontal wavenumbers) $[\text{m}^{-1}]$


* **【定義 2】 水平波數平方 (Horizontal wavenumber squared)：**

  $$\bar{K}^2 \overset{\text{def}}{=} \bar{k}^2 + \bar{l}^2$$

  * $K^{2}$ : 水平總波數的平方 (Horizontal total wavenumber squared) $[\text{m}^{-2}]$
  * $k$ : $x$ 軸方向的水平波數 (Zonal wavenumber) $[\text{m}^{-1}]$
  * $l$ : $y$ 軸方向的水平波數 (Meridional wavenumber) $[\text{m}^{-1}]$

* **【假設 2】 垂直邊界與波結構 (Vertical boundary and wave structure)：** 對於一般三維波動（$\hat{w} \neq 0$），垂直方向滿足剛體邊界條件，對應的氣壓微擾垂直結構滿足二階特徵導數關係。

  $$\frac{\partial^2 \hat{p}}{\partial z^2} = -\bar{m}^2 \hat{p}$$


  * $\bar{m}$ : 垂直方向波數 (Vertical wavenumber) $[\text{m}^{-1}]$
  * $m$ : 垂直方向波數 (Vertical wavenumber) $[\text{m}^{-1}]$
  * $\hat{p}$ : 經密度加權後的特徵氣壓結構函數
  * $\frac{\partial ^{2}\hat{p}}{\partial z^{2}}$ : 特徵氣壓對高度的二階偏導數 $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{-\frac{3}{2}}\cdot\text{s}^{-2}]$


* **【推導 1】 水平速度與氣壓的代數關係：**

  * (a)
  
  $$\begin{gather*}
  i\nu \hat{u} &\overset{\text{已知 2(b)}}{=}& \frac{\partial \hat{p}}{\partial x} \\
  i\nu \hat{u} &\overset{\text{假設 1(a)}}{=}& i\bar{k}\hat{p} \\
  \hat{u} &=& \frac{\bar{k}}{\nu}\hat{p} \\
  \end{gather*}$$
  
  * (b)

  $$\begin{gather*}
  i\nu \hat{v} &\overset{\text{已知 2(c)}}{=}& \frac{\partial \hat{p}}{\partial y} \\
  i\nu \hat{v} &\overset{\text{假設 1(b)}}{=}& i\bar{l}\hat{p} \\
  \hat{v} &=& \frac{\bar{l}}{\nu}\hat{p} \\
  \end{gather*}$$

  * $\hat{u}, \hat{v}$ : 經密度加權後的水平特徵速度分量 $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $\hat{p}$ : 經密度加權後的特徵氣壓結構函數 $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $\nu$ : 系統的實數固有特徵角頻率 $[\text{s}^{-1}]$
  * $k, l$ : 水平方向各空間軸向的波數 $[\text{m}^{-1}]$


* **【推導 2】 垂直速度與氣壓的連續關係 （Continuity relation between vertical velocity and pressure）：**
  
  $$\begin{gather*}
  i\nu \hat{p} &\overset{\text{已知 2(e)}}{=}& c^2 \left( \frac{\partial \hat{u}}{\partial x} + \frac{\partial \hat{v}}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right) \\
  i\nu \hat{p} &\overset{\text{假設 1(a)(b)}}{=}& c^2 \left( i\bar{k}\hat{u} + i\bar{l}\hat{v} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right) \\
  i\nu \hat{p} &\overset{\text{推導 1(a)(b)}}{=}& c^2 \left( i\bar{k}\frac{\bar{k}}{\nu}\hat{p} + i\bar{l}\frac{\bar{l}}{\nu}\hat{p} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right) \\
  i\nu \hat{p} &=& c^2 \left( i\frac{\bar{k}^2 + \bar{l}^2}{\nu}\hat{p} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right) \\
  i\nu \hat{p} &\overset{\text{定義 2}}{=}& c^2 \left( i\frac{\bar{K}^2}{\nu}\hat{p} + \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \right) \\
  i\nu \hat{p} &=& i\frac{c^2 \bar{K}^2}{\nu}\hat{p} + c^2 \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} \\
  c^2 \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} &=& i\left( \nu - \frac{c^2 \bar{K}^2}{\nu} \right)\hat{p} \\
  \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} &=& i\left( \frac{\nu^2 - c^2 \bar{K}^2}{c^2 \nu} \right)\hat{p} \\
  \end{gather*}$$

  * $\nu$ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $\hat{p}$ : 經密度加權後的特徵氣壓結構函數 (Weighted pressure eigenvector) $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $\hat{u}$ : 經密度加權後的 $x$ 軸方向水平特徵速度分量 (Weighted zonal velocity eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $\hat{v}$ : 經密度加權後的 $y$ 軸方向水平特徵速度分量 (Weighted meridional velocity eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $\hat{w}$ : 經密度加權後的垂直特徵速度結構函數 (Weighted vertical velocity eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $\Gamma$ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $k$ : $x$ 軸方向的水平波數 (Zonal wavenumber) $[\text{m}^{-1}]$
  * $l$ : $y$ 軸方向的水平波數 (Meridional wavenumber) $[\text{m}^{-1}]$
  * $K^{2}$ : 水平總波數平方 (Horizontal total wavenumber squared) $[\text{m}^{-2}]$


* **【推導 3】 浮力與垂直速度的代數關係：**
  
  $$\begin{gather*}
  i\nu \hat{b} &\overset{\text{已知 2(f)}}{=}& -\frac{N^2}{g} \hat{w} \\
  \hat{b} &=& i\frac{N^2}{g\nu}\hat{w} \\
  \end{gather*}$$

  * $\nu$ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $\hat{b}$ : 經密度加權後的特徵浮力／熱力學結構函數 (Weighted buoyancy eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-3}]$
  * $\hat{w}$ : 經密度加權後的垂直特徵速度結構函數 (Weighted vertical velocity eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$


* **【推導 4】 垂直氣壓梯度與垂直速度的動力關係：**

  $$\begin{gather*}
  i\nu \hat{w} &\overset{\text{已知 2(d)}}{=}& \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} + g \hat{b} \\
  i\nu \hat{w} &\overset{\text{推導 3}}{=}& \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} + g \left( i\frac{N^2}{g\nu}\hat{w} \right) \\
  i\nu \hat{w} &=& \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} + i\frac{N^2}{\nu}\hat{w}  \\
  \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} &=& i\nu \hat{w} - i\frac{N^2}{\nu}\hat{w} \\
  \left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p} &=& i\left( \frac{\nu^2 - N^2}{\nu} \right)\hat{w} \\
  \end{gather*}$$

  * $\nu$ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $\hat{w}$ : 經密度加權後的垂直特徵速度結構函數 (Weighted vertical velocity eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $\hat{p}$ : 經密度加權後的特徵氣壓結構函數 (Weighted pressure eigenvector) $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $\hat{b}$ : 經密度加權後的特徵浮力／熱力學結構函數 (Weighted buoyancy eigenvector) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-3}]$
  * $\Gamma$ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$


* **【推導 5】 氣壓微擾的核心垂直結構常微分方程式 (Core vertical structure ODE)：**
  
  $$\begin{gather*}
  \left(\frac{\partial}{\partial z} - \Gamma\right) \left[\left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p}\right] &=& \left(\frac{\partial}{\partial z} - \Gamma\right) \left[\left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p}\right] \\
  \left(\frac{\partial^2}{\partial z^2} - \Gamma^2\right)\hat{p} &=& \left(\frac{\partial}{\partial z} - \Gamma\right) \left[\left(\frac{\partial}{\partial z} + \Gamma\right)\hat{p}\right] \\
  \left(\frac{\partial^2}{\partial z^2} - \Gamma^2\right)\hat{p} &\overset{\text{推導 4}}{=}& \left(\frac{\partial}{\partial z} - \Gamma\right) \left[ i\left( \frac{\nu^2 - N^2}{\nu} \right)\hat{w} \right] \\
  \left(\frac{\partial^2}{\partial z^2} - \Gamma^2\right)\hat{p} &=& i\left( \frac{\nu^2 - N^2}{\nu} \right)  \left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w}  \\
  \left(\frac{\partial^2}{\partial z^2} - \Gamma^2\right)\hat{p} &\overset{\text{推導 2}}{=}& i\left( \frac{\nu^2 - N^2}{\nu} \right) \left( i\left( \frac{\nu^2 - c^2 \bar{K}^2}{c^2 \nu} \right)\hat{p} \right)  \\
  \left(\frac{\partial^2}{\partial z^2} - \Gamma^2\right)\hat{p} &=& -\left( \frac{(\nu^2 - N^2)(\nu^2 - c^2 \bar{K}^2)}{c^2 \nu^2} \right)\hat{p} \\
  \end{gather*}$$

  * $\Gamma$ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $\nu$ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $\hat{p}$ : 經密度加權後，在波數空間中的特徵氣壓結構函數 $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $K^{2}$ : 水平總波數平方 (Horizontal total wavenumber squared) $[\text{m}^{-2}]$



#### 證明

##### 1. 藍姆波色散關係式 (Lamb waves dispersion relationship, 當 $\hat{w}=0$)

當垂直擾動速度為零（$\hat{w}=0$），且氣壓微擾場不為零（$\hat{p} \neq 0$）時：

$$\begin{gather*}
\left(\frac{\partial}{\partial z} - \Gamma\right)\hat{w} &\overset{\text{推導 2}}{=}& i\left( \frac{\nu^2 - c^2 \bar{K}^2}{c^2 \nu} \right)\hat{p} \\
\left(\frac{\partial}{\partial z} - \Gamma\right)(0) &=& i\left( \frac{\nu^2 - c^2 \bar{K}^2}{c^2 \nu} \right)\hat{p} \\
0 &=& \frac{\nu^2 - c^2 \bar{K}^2}{c^2 \nu} \\
\nu^2 - c^2 \bar{K}^2 &=& 0 \\
\nu^2 &=& c^2 \bar{K}^2 \\
\nu^2 &\overset{\text{定義 1}}{=}& c^2(\bar{k}^2 + \bar{l}^2) \tag{8b}\\
\end{gather*}$$

##### 2. 三維聲波與浮力波色散關係式 (Acoustic and Buoyancy waves dispersion relationship, 當 $\hat{w} \neq 0$)

對於一般的三維波動，特徵方程式必須滿足垂直波數的二階空間導數特徵關係：

$$\begin{gather*}
\left(\frac{\partial^2}{\partial z^2} - \Gamma^2\right)\hat{p} &\overset{\text{推導 5}}{=}& -\left( \frac{(\nu^2 - N^2)(\nu^2 - c^2 \bar{K}^2)}{c^2 \nu^2} \right)\hat{p} \\
\left(-\bar{m}^2 - \Gamma^2\right)\hat{p} &\overset{\text{假設 2}}{=}& -\left( \frac{(\nu^2 - N^2)(\nu^2 - c^2 \bar{K}^2)}{c^2 \nu^2} \right)\hat{p} \\
\bar{m}^2 + \Gamma^2 &=& \frac{(\nu^2 - N^2)(\nu^2 - c^2 \bar{K}^2)}{c^2 \nu^2} \\
c^2 \nu^2 \left( \bar{m}^2 + \Gamma^2 \right) &=& \left( \nu^2 - N^2 \right)\left( \nu^2 - c^2 \bar{K}^2 \right) \\
c^2 \nu^2 \left( \bar{m}^2 + \Gamma^2 \right) &=& \nu^4 - \nu^2 c^2 \bar{K}^2 - \nu^2 N^2 + N^2 c^2 \bar{K}^2 \\
\nu^4 - \nu^2 c^2 \bar{K}^2 - \nu^2 N^2 - \nu^2 c^2 \left( \bar{m}^2 + \Gamma^2 \right) + N^2 c^2 \bar{K}^2 &=& 0 \\
\nu^4 - \nu^2 \left[ c^2 \bar{K}^2 + c^2 \bar{m}^2 + c^2 \Gamma^2 + N^2 \right] + c^2 \bar{K}^2 N^2 &=& 0 \\
\frac{\nu^4}{c^2} - \nu^2 \left( \bar{K}^2 + \bar{m}^2 + \Gamma^2 + \frac{N^2}{c^2} \right) + \bar{K}^2 N^2 &=& 0 \\
\frac{\nu^4}{c^2} - \nu^2 \left( \bar{k}^2 + \bar{l}^2 + \bar{m}^2 + \Gamma^2 + \frac{N^2}{c^2} \right) + \left( \bar{k}^2 + \bar{l}^2 \right)N^2 &\overset{\text{定義 2}}{=}& 0 \tag{9a}\\
\end{gather*}$$


**特徵方程式的根**


$$\begin{gather*}
c^2 \nu^2 \left( \bar{m}^2 + \Gamma^2 \right) &=& \left( \nu^2 - N^2 \right)\left( \nu^2 - c^2 \bar{K}^2 \right) \\
0 &=& (\nu ^{2})^{2}-\left[N^{2}+c^{2}\bar{K}^{2}+c^{2}\left(\bar{m}^{2}+\Gamma^{2}\right)\right]\nu ^{2}+c^{2}N^{2}\bar{K}^{2} \\
\nu ^{2} &=& \frac{c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\pm \sqrt{\left[c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\bar{K}^{2}}}{2}  \\
\nu ^{2} &=& \frac{c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2}\pm \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]}}{2} 
\end{gather*}$$


#### 四種流場與波動的動力學特徵討論(Discussion on Dynamic Characteristics of Four Flow Fields and Waves)



##### 1. 非輻散流 (Non-divergent flow)

* **動力學特徵**：這對應於特徵頻率為零的解 $\nu_{klmn}^2 = 0$，代表一種穩定、不隨時間震盪的狀態 。
* **物理意義**：在流場上表現為水平非輻散流（horizontally non-divergent flow） 。從波動動力學的角度來看，這實際上是內重力波（internal gravity waves）的一種退化狀態（degeneracy），發生在波數向量（wave number vector）完全指向垂直方向的時候 。在旋轉大氣中，這種低頻/零頻的解通常對應於滿足地轉平衡的大尺度渦旋天氣系統（即氣象學中最關心的氣流）。



##### 2. 藍姆波 (Lamb waves)

* **動力學特徵**：其頻率滿足色散關係式 $\nu_{klmn}^2 = c^2(\bar{k}^2 + \bar{l}^2)$ 。其傳播速度即為音速，大約為 $300 \text{ m}\cdot\text{s}^{-1}$ 。
* **物理意義**：藍姆波是一種非常特殊的波動模式 。它最大的特徵是流場的速度在任何地方都平行於地球表面，也就是垂直速度 $w=0$ 。這意味著它相當於一種僅在水平方向上傳播的二維純聲波。由於沒有垂直運動，它的能量被重力完全束縛在地表附近，不會向高空頻散。



##### 3. 聲波 (Acoustic waves)

* **動力學特徵**：屬於特徵方程式解出的 **高頻正負根** $\frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} + \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]}\right)$ 。受重力修正後的聲波，其頻率存在一個下限，亦即頻率必須大於聲學截斷頻率（Acoustic cutoff frequency，$N_A \approx 1.11N$） 。
* **物理意義**：聲波的主要恢復力是流體的可壓縮性（即彈性氣壓梯度力）。在完全可壓縮模型中，聲波在水平與垂直方向皆可迅速傳播。然而，因為聲波的移動速度極快，在數值天氣預報模式中會嚴格限制計算的時間步長（CFL 條件限制），且聲波本身幾乎不攜帶具有氣象意義的能量，因此在許多大尺度動力模型中，常會透過假設「不可壓縮流」或「滯彈性近似（Anelastic approximation）」將聲波人為濾除。



##### 4. 浮力波 / 內重力波 (Buoyancy waves / Internal gravity waves)

* **動力學特徵**：屬於特徵方程式解出的 **低頻正負根** $\frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} - \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]}\right)$ 。與聲波相反，浮力波的頻率存在一個上限，其頻率必須小於浮力頻率（Brunt-Väisälä frequency，$N$） 。
* **物理意義**：浮力波的恢復力來自於穩定層結大氣中的重力與浮力。當空氣塊被垂直位移時，由於周遭環境密度的差異而被拉回原位，進而在浮力與重力的交替作用下產生震盪。浮力波在垂直與水平方向上皆可色散能量，對於大氣中的動量傳輸（如地形產生的重力波拖曳）以及晴空亂流的生成具有至關重要的影響。









#### 三種流場頻率範圍 (Frequency Ranges)


我們定義水平波數平方為 $\bar{K}^2 \overset{\text{def}}{=} \bar{k}^2 + \bar{l}^2$ ，波數的物理範圍是：
* $0 \le \bar{K} < \infty$ 
* $0 \le \bar{m} < \infty$

##### 1. 藍姆波 (Lamb Waves)


* **下限：**

$$\begin{gather*}
\lim\limits_{\bar{K}^2 \to 0^+}  \nu^2 &\overset{\text{(8b)}}{=}& \lim\limits_{\bar{K}^2 \to 0^+} c^2(\bar{k}^2 + \bar{l}^2) \\
&\overset{\text{定義 2}}{=}& \lim\limits_{\bar{K}^2 \to 0^+} c^2 \bar{K}^2 \\
&=& 0^+
\end{gather*}$$

* **上限：**
  
$$\begin{gather*}
\lim\limits_{\bar{K}^2 \to \infty}  \nu^2 &\overset{\text{(8b)}}{=}& \lim\limits_{\bar{K}^2 \to \infty} c^2(\bar{k}^2 + \bar{l}^2) \\
&\overset{\text{定義 2}}{=}& \lim\limits_{\bar{K}^2 \to \infty} c^2 \bar{K}^2 \\
&\rightarrow& \infty
\end{gather*}$$

* **頻率範圍：** $\nu \in (0, \infty)$ 



#### 2. 聲波 (Acoustic waves)  




* **下限（聲學截斷頻率 Acoustic Cutoff Frequency）：** 聲波的頻率極小值發生在三維波數皆為零（$\bar{K} = 0, \bar{m} = 0$）的時候 。這個最小值被定義為聲學截斷頻率 $N_A$。如講義中 Fig 6.18 的說明所述，對於大氣參數，$N_A \approx 1.11N$ 。

$$\begin{gather*}
\nu_{\text{A}}^{2} &=& \frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} +  \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]} \right) \\
\nu_{\text{A},\text{min}}^{2} &=& \frac{1}{2}\left( c^{2}(0^{2}+0^{2}+\Gamma^{2})+N^{2} +  \sqrt{\left(c^{2}\cdot 0^{2}-N^{2}\right)^{2}+c^{2}(0^{2}+\Gamma^{2})\left[2(c^{2}\cdot 0^{2}+N^{2})+c^{2}(0^{2}+\Gamma^{2})\right]} \right) \\
\nu_{\text{A},\text{min}}^{2} &=& \frac{1}{2}\left( c^{2}\Gamma^{2}+N^{2} +  \sqrt{N^{4}+c^{2}\Gamma^{2} (2N^{2}+c^{2}\Gamma^{2})} \right) \\
\nu_{\text{A},\text{min}}^{2} &=& \frac{1}{2}\left( c^{2}\Gamma^{2}+N^{2} +  \sqrt{N^{4}+2c^{2}\Gamma^{2}N^{2}+(c^{2}\Gamma^{2})^{2}} \right) \\
\nu_{\text{A},\text{min}}^{2} &=& \frac{1}{2}\left( c^{2}\Gamma^{2}+N^{2} +  \sqrt{(N^{2}+c^{2}\Gamma^{2})^{2}} \right) \\
\nu_{\text{A},\text{min}}^{2} &=& \frac{1}{2}\left( c^{2}\Gamma^{2}+N^{2} +  N^{2}+c^{2}\Gamma^{2} \right) \\
\nu_{\text{A},\text{min}}^{2} &=& c^{2}\Gamma^{2}+N^{2}  \\
\end{gather*}$$


* **上限：** 當波數趨近於無限大時，聲波頻率自然趨近無限大。

$$\begin{gather*}
\nu_{\text{A}}^{2} &=& \frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} +  \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]} \right) \\
\nu_{\text{A},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty} \lim\limits_{\bar{m}^2 \to \infty} \frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} +  \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]} \right) \\
&\rightarrow& \infty
\end{gather*}$$

* **頻率範圍：** $\nu_{\text{A}} \in [1.11N, \infty)$ 或寫為 $[N_A, \infty)$



#### 2. 浮力波 (Buoyancy waves)



* **下限：** 當水平波數 $\bar{K} \to 0$（超長波）時，$C \to 0$，此時 $\nu_B^2 \to 0$。

$$\begin{gather*}
\nu_{\text{B}}^{2} &=& \frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} -  \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]} \right) \\
\nu_{\text{B}}^{2} &=& \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\bar{K}^{2}}\right) \\
\nu_{\text{B},\text{min}}^{2} &=& \frac{1}{2}\left(c^{2}\left(0^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(0^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\cdot 0^{2}}\right) \\
\nu_{\text{B},\text{min}}^{2} &=& \frac{1}{2}\left(c^{2}\left(\bar{m}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}}\right) \\
\nu_{\text{B},\text{min}}^{2} &=& \frac{1}{2}\left(c^{2}\left(\bar{m}^{2}+\Gamma^{2}\right)+N^{2} - \left( c^{2}\left(\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right) \right)\\
\nu_{\text{B},\text{min}}^{2} &=& 0^+\\
\end{gather*}$$



* **上限：** 當 $\bar{m} \to 0$ 且 $\bar{K} \to \infty$（極扁平的擾動）時，可以證明頻率會漸近趨近於浮力頻率 $N$ 。也就是說，環境的穩定度 $N$ 是內部重力波震盪的最快極限。

$$\begin{gather*}
\nu_{\text{B}}^{2} &=& \frac{1}{2}\left( c^{2}(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2})+N^{2} -  \sqrt{\left(c^{2}\bar{K}^{2}-N^{2}\right)^{2}+c^{2}(\bar{m}^{2}+\Gamma^{2})\left[2(c^{2}\bar{K}^{2}+N^{2})+c^{2}(\bar{m}^{2}+\Gamma^{2})\right]} \right) \\
\nu_{\text{B}}^{2} &=& \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\bar{K}^{2}}\right) \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty} \lim\limits_{\bar{m}^2 \to 0} \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(\bar{K}^{2}+\bar{m}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\bar{K}^{2}}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+0^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(\bar{K}^{2}+0^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\bar{K}^{2}}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2}\right]^{2}-4c^{2}N^{2}\bar{K}^{2}}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[\left( c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)\right)^2+2 c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)  N^{2}+N^{4}\right]-4c^{2}N^{2}\bar{K}^{2}}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{\left[ c^{4}\left(\bar{K}^{4}+2\bar{K}^{2}\Gamma^{2}+\Gamma^{4}\right)+2 c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)  N^{2}+N^{4}\right]-4c^{2}N^{2}\bar{K}^{2}}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} - \sqrt{(c^{2}\bar{K}^{2})^{2}+2c^{2}(c^{2}\Gamma^{2}-N^{2})\bar{K}^{2}+(N^{4}+2c^{2}\Gamma^{2}N^{2}+c^{4}\Gamma^{4})}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} - c^{2}\bar{K}^{2} \sqrt{1+\frac{2(c^{2}\Gamma^{2}-N^{2})}{c^{2}\bar{K}^{2}}+\frac{(N^{4}+2c^{2}\Gamma^{2}N^{2}+c^{4}\Gamma^{4})}{c^{4}\bar{K}^{4}}}\right)  \\
\nu_{\text{B},\text{Max}}^{2} &\approx& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} - c^{2}\bar{K}^{2} \left(1+\frac{1}{2}\frac{2(c^{2}\Gamma^{2}-N^{2})}{c^{2}\bar{K}^{2}}\right) \right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\left(\bar{K}^{2}+\Gamma^{2}\right)+N^{2} -  \left(c^{2}\bar{K}^{2}+(c^{2}\Gamma^{2}-N^{2})\right) \right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \lim\limits_{\bar{K}^2 \to \infty}  \frac{1}{2}\left(c^{2}\bar{K}^{2}+c^{2}\Gamma^{2}+N^{2} -  \left(c^{2}\bar{K}^{2}+c^{2}\Gamma^{2}-N^{2}\right) \right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& \frac{1}{2}\left(N^{2} +N^{2} \right)  \\
\nu_{\text{B},\text{Max}}^{2} &=& N^{2}  \\
\end{gather*}$$


* **頻率範圍：** $\nu_{\text{B}} \in (0, N]$ 





### 六、從物理空間轉換至頻譜空間：突破聲波 CFL 條件限制

$$\hat{u}_{klmn}(t+\Delta t) = \hat{u}_{klmn}(t)e^{-i\nu_{klmn}(\Delta t)} + \int_{t}^{t+\Delta t}\hat{N}_{klmn}(t')e^{i\nu_{klmn}(t' - t - \Delta t)}dt' \tag{14}$$



方程式 (13) 看起來跟方程式 (4) $\frac{\partial \tilde{u}}{\partial t} + L\tilde{u} = \tilde{N}$ 長得很像，這**絕對不是巧合**。方程式 (4) 是在「三維物理空間」中的偏微分方程式（PDE），而方程式 (13) 是在「頻譜空間（Spectral Space）」中的常微分方程式（ODE）。

我們前面之所以要千辛萬苦證明 $\mathbf{L}$ 是反厄米特算子，就是要保證它解出來的特徵向量 $\tilde{K}_{klmn}$ 是**互相正交的（Orthogonal）**。有了這組正交基底，我們就可以利用「Galerkin 投影法（Galerkin Projection）」，把物理空間的複雜 PDE，拆解成無限多條獨立的 ODE。


#### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 物理空間微擾矩陣偏微分方程式 (Matrix PDE in physical space)：** 
  
  $$\frac{\partial \tilde{u}}{\partial t} + \mathbf{L}\tilde{u} = \tilde{N}$$

  * $\tilde{u}(x,y,z,t)$ : 物理空間中的三維狀態向量
  * $\tilde{N}(x,y,z,t)$ : 物理空間中的非線性強迫項向量
  * $\mathbf{L}$ : 空間微分線性算子


* **【已知 2】 算子 $\mathbf{L}$ 的特徵值問題與正交性 (Eigenvalue problem and Orthogonality)：** 因 $L$ 為反厄米特算子，其特徵向量形成一組完備且正交的基底。為簡化標示，將四維索引 $(k,l,m,n)$ 簡寫為單一索引 $\alpha$ 與 $\beta$
  
  * (a) 特徵值方程式：

  $$\mathbf{L}\tilde{K}_\alpha = i\nu_\alpha \tilde{K}_\alpha$$


  * (b) 正交歸一性：

  $$(\tilde{K}_\alpha, \tilde{K}_\beta) = \delta_{\alpha\beta}$$


  * $\alpha, \beta$ : 波態的四維索引簡寫，代表特定的波數與頻率組合
  * $\tilde{K}_\alpha(x,y,z)$ : 第 $\alpha$ 個波態的空間特徵向量
  * $i\nu_\alpha$ : 第 $\alpha$ 個波態的純虛數特徵值
  * $\delta_{\alpha\beta}$ : 克羅內克函數 (Kronecker delta)，當 $\alpha=\beta$ 時為 1，否則為 0


* **【定義 1】 頻譜展開 (Spectral Expansion)：** 將物理空間的狀態向量與非線性強迫項，展開為特徵基底向量 $\tilde{K}_\alpha$ 的線性組合。

  * (a) 狀態向量展開：

  $$\tilde{u}(x,y,z,t) = \sum_{\alpha} \hat{u}_\alpha(t) \tilde{K}_\alpha(x,y,z)$$


  * (b) 非線性項展開：

  $$\tilde{N}(x,y,z,t) = \sum_{\alpha} \hat{N}_\alpha(t) \tilde{K}_\alpha(x,y,z)$$


  * $\hat{u}_\alpha(t)$ : 隨時間變化的波態振幅係數 (Wave amplitude coefficient)
  * $\hat{N}_\alpha(t)$ : 非線性強迫項在特定波態上的投影係數

* **【定義 2】 波數索引：** $k, l, m$ 決定了波的 **長相與大小（空間結構）** ，而 $n$ 決定了波的 **靈魂與物理性質（波動種類）**

  * (a) 空間波數索引：$k, l, m$

  $$\bar{k} = \frac{2\pi}{L_x}k$$

  $$\bar{l} = \frac{2\pi}{L_y}l$$

  $$\bar{m} = \frac{\pi}{L_z}m$$


  * (b) 波態 / 頻率索引：$n$

    * 【特徵根 1】$n = 1$：非輻散流 (Non-divergent flow)
      * 對應關係：這是解出 $\nu_{klm1} = 0$ 的那個根。
      * 物理意義：因為頻率為 0，這代表流場完全不隨時間震盪（Steady state）。它對應的是滿足地轉平衡的緩慢天氣系統（如高低壓渦旋）。
    * 【特徵根 2 & 3】$n = 2, 3$：聲波 (Acoustic waves)
      * 對應關係：這是由色散方程式 (9a) 解出來的一對「高頻正負根」
        * $\nu_{klm2} = +\nu_A$ （向東傳播的三維聲波）
        * $\nu_{klm3} = -\nu_A$ （向西傳播的三維聲波）
      * 物理意義：即使空間波長 $(k, l, m)$ 跟天氣系統一樣大，只要它按照這個極高的頻率震盪，它就是聲波。它的恢復力是空氣的可壓縮性（彈性）。
    * 【特徵根 4 & 5】$n = 4, 5$：浮力波 / 內重力波 (Buoyancy waves)
      * 對應關係：這是由色散方程式 (9a) 解出來的一對「低頻正負根」
        * $\nu_{klm4} = +\nu_B$ （向東傳播的浮力波）
        * $\nu_{klm5} = -\nu_B$ （向西傳播的浮力波）
      * 物理意義：在同一個空間網格 $(k, l, m)$ 上，如果空氣是以低於浮力頻率 $N$ 的速度上下擺盪，它就是浮力波。




  * $k$：對應 $x$ 方向（東西向）的波數。決定了水平特徵值 
  * $l$：對應 $y$ 方向（南北向）的波數。決定了水平特徵值 
  * $m$：對應 $z$ 方向（垂直向）的波數。決定了垂直特徵值 
  * $n$：代表了 5 個不同的特徵根 $n = 1, 2, 3, 4, 5$


* **【推導 1】純時間的一階常微分方程式：**

  $$\begin{gather*}
  \frac{\partial \tilde{u}}{\partial t} + \mathbf{L}\tilde{u} &\overset{\text{已知 1}}{=}& \tilde{N} \\
  \frac{\partial }{\partial t} \left( \sum_{\alpha} \hat{u}_\alpha(t) \tilde{K}_\alpha \right) + \mathbf{L} \left( \sum_{\alpha} \hat{u}_\alpha(t) \tilde{K}_\alpha \right) &\overset{\text{定義 1(a)(b)}}{=}& \sum_{\alpha} \hat{N}_\alpha(t) \tilde{K}_\alpha \\

  \sum_{\alpha} \left( \frac{d\hat{u}_\alpha(t)}{dt} + i\nu_\alpha \hat{u}_\alpha(t) \right) \tilde{K}_\alpha &=& \sum_{\alpha} \hat{N}_\alpha(t) \tilde{K}_\alpha \\

  \sum_{\alpha} \left( \frac{d\hat{u}_\alpha(t)}{dt} + i\nu_\alpha \hat{u}_\alpha(t) \right) (\tilde{K}_\alpha, \tilde{K}_\beta) &=& \sum_{\alpha} \hat{N}_\alpha(t) (\tilde{K}_\alpha, \tilde{K}_\beta) \\
  \sum_{\alpha} \left( \frac{d\hat{u}_\alpha(t)}{dt} + i\nu_\alpha \hat{u}_\alpha(t) \right) \delta_{\alpha\beta} &\overset{\text{已知 2(b)}}{=}& \sum_{\alpha} \hat{N}_\alpha(t) \delta_{\alpha\beta} \\
  \frac{d\hat{u}_\beta(t)}{dt} + i\nu_\beta \hat{u}_\beta(t) &=& \hat{N}_\beta(t) \\
  \frac{d\hat{u}_{klmn}(t)}{dt} + i\nu_{klmn}\hat{u}_{klmn}(t)  &=&  \hat{N}_{klmn}(t)
  \end{gather*}$$

  * $i\nu_\alpha$ : 第 $\alpha$ 個波態的純虛數特徵值

e^{i\nu_{klmn}t} \frac{d\hat{u}_{klmn}(t)}{dt} + i\nu_{klmn} e^{i\nu_{klmn}t} \hat{u}_{klmn}(t)  &=&  \hat{N}_{klmn}(t) e^{i\nu_{klmn}t} \\

\int_{t}^{t+\Delta t} \frac{d}{dt'} \left[ \hat{u}_{klmn}(t') e^{i\nu_{klmn}t'} \right] dt'  &=&  \int_{t}^{t+\Delta t} \hat{N}_{klmn}(t') e^{i\nu_{klmn}t'} dt' \\
\hat{u}_{klmn}(t+\Delta t) e^{i\nu_{klmn}(t+\Delta t)} - \hat{u}_{klmn}(t) e^{i\nu_{klmn}t}  &=& \int_{t}^{t+\Delta t} \hat{N}_{klmn}(t') e^{i\nu_{klmn}t'} dt' \\
\hat{u}_{klmn}(t+\Delta t) e^{i\nu_{klmn}(t+\Delta t)} &=&  \hat{u}_{klmn}(t) e^{i\nu_{klmn}t} + \int_{t}^{t+\Delta t} \hat{N}_{klmn}(t') e^{i\nu_{klmn}t'} dt' \\
\hat{u}_{klmn}(t+\Delta t)  &=&  \hat{u}_{klmn}(t) e^{i\nu_{klmn}t} e^{-i\nu_{klmn}(t+\Delta t)} + \int_{t}^{t+\Delta t} \hat{N}_{klmn}(t') e^{i\nu_{klmn}t'} e^{-i\nu_{klmn}(t+\Delta t)} dt' \\
\hat{u}_{klmn}(t+\Delta t)  &=&  \hat{u}_{klmn}(t)e^{-i\nu_{klmn}(\Delta t)} + \int_{t}^{t+\Delta t}\hat{N}_{klmn}(t')e^{i\nu_{klmn}(t' - t - \Delta t)}dt'
\end{gather*}$$


#### 突破音速 CFL 條件限制？

在常規的大氣數值模式開發中，決定時間步長 $\Delta t$ 大小的關鍵是 **CFL 條件（Courant-Friedrichs-Lewy condition）**。CFL 條件要求 $\Delta t \le \frac{\Delta x}{C_{max}}$，其中 $C_{max}$ 是系統中「跑得最快的波」的波速。

**方程式 (14) 的突破性意義：**

* **線性項的「解析解 (Exact Solution)」：** 觀察 (14) 式的第一項 $\hat{u}_{klmn}(t)e^{-i\nu_{klmn}(\Delta t)}$。我們並沒有用任何數值差分法去逼近線性算子造成的波動演進，而是直接用微積分算出了它的**精確解析解**。因為是解析的旋轉相位 $e^{-i\nu_{klmn}\Delta t}$，無論你的 $\Delta t$ 設多大，線性快波都絕對不會數值發散，它們只會在複數平面上完美地旋轉。
* **計算重心的轉移：** 透過這個積分因子轉換，數值誤差與不穩定性現在**只來自於對非線性項 $\hat{N}_{klmn}$ 的數值積分**（即等式右邊第二項的積分式） 。而非線性項代表的是慢速的平流作用與低頻交互作用。
* **結論：** 因為會引發 CFL 不穩定的線性快波已經被數學公式「完美吸收」了，所以新的 CFL 條件不再由音速 $c$ 決定，而是由流體本身的運動速度 $U$ 決定。這使得數值模式的時間步長 $\Delta t$ 可以放大 10 倍甚至更多，大幅提升了數值積分的效率，卻又不會受到線性快波 CFL 限制的阻礙 。


#### 實務計算的完整演算法步驟 (Detailed Computational Algorithm)

##### **【預備階段：初始化】**

**1. 初始化物理空間網格與參數**

* 定義三維網格 $(x, y, z)$ 與時間步長 $\Delta t$。
* 給定環境背景態參數：$\bar{\rho}(z), \Gamma, N^2, c^2$。

**2. 初始化物理空間狀態向量 (Physical State)**

* 在 $t=0$ 時刻，取得大氣的初始擾動場：$u_0, v_0, w_0, p'_0, \rho'_0$。
* 轉換為密度加權狀態向量 $\tilde{u}(x,y,z,0)$：
$\tilde{u} = \left( \bar{\rho}^{1/2}u, \bar{\rho}^{1/2}v, \bar{\rho}^{1/2}w, \bar{\rho}^{-1/2}p', \bar{\rho}^{-1/2}(\rho' - \frac{p'}{c^2}) \right)^T$

**3. 預先計算頻譜基底與特徵值 (Pre-compute Eigen-structure)**

* 給定截斷波數上限。對於每一個波數組合 $(k, l, m)$ 與 每一個物理波態 $(n = 1 \dots 5)$：
* 透過色散方程式 **(8b) 或 (9a)** 解出特徵頻率 $\nu_{klmn}$。
* 算出對應的空間特徵向量 $\tilde{K}_{klmn}(x,y,z)$。



**4. 初始頻譜轉換 (Initial Forward Transform)**

* 將初始狀態 $\tilde{u}(x,y,z,0)$ 投影到頻譜空間。透過快速傅立葉轉換 (FFT) 與正交內積，算出 $t=0$ 時刻的所有波態振幅：
$\hat{u}_{klmn}(0) = \iiint \tilde{u}(x,y,z,0) \cdot \tilde{K}_{klmn}^* \, dx dy dz$

##### **【主迴圈：時間積分 ($t \to t+\Delta t$)】**

以下步驟開始進入迴圈，假設目前時間為 $t$，我們已知頻譜空間的振幅 $\hat{u}_{klmn}(t)$：

**5. 轉換至物理空間取得 $\tilde{u}(t)$ (Spectral to Physical)**

* *對應講義 Step (2)*
* 利用反向傅立葉轉換 (Inverse FFT) 疊加所有的波，算出目前的物理場：

$$\tilde{u}(x,y,z,t) = \sum_{k,l,m} \sum_{n=1}^5 \hat{u}_{klmn}(t) \tilde{K}_{klmn}(x,y,z)$$



**6. 在物理空間計算非線性強迫項 $\tilde{N}(t)$ (Evaluate Nonlinearity)**

* *對應講義 Step (3)*
* 在網格點上，利用算出來的 $\tilde{u}(t)$ 計算空間梯度與乘積。例如計算平流項 $(u\cdot\nabla)u$ 以及密度的非線性修正。
* 彙整出等號右邊的非線性向量 $\tilde{N}(x,y,z,t) = (N_1, N_2, N_3, N_4, N_5)^T$。
*(註：這一步是整個模式運算量最大、最耗時的地方)*

**7. 將非線性項轉換至頻譜空間取得 $\hat{N}_{klmn}(t)$ (Forward Transform)**

* *對應講義 Step (4)*
* 將算好的非線性網格資料，再用 FFT 轉回頻譜空間，得到每個波態分配到的非線性強迫力：

$$\hat{N}_{klmn}(t) = \iiint \tilde{N}(x,y,z,t) \cdot \tilde{K}_{klmn}^* \, dx dy dz$$



**8. 在頻譜空間進行精確時間積分 (Predict in Spectral Space)**

* *對應講義 Step (1)*
* 利用積分因子方程式 **(14)**，將狀態推進到 $t+\Delta t$。
* 實務上，積分號 $\int_{t}^{t+\Delta t} \dots$ 需要數值離散化。若採用最簡單的指數時間差分歐拉法 (ETD1, 假設 $\hat{N}$ 在這 $\Delta t$ 內為常數)，計算公式為：

$$\hat{u}_{klmn}(t+\Delta t) = \hat{u}_{klmn}(t) \cdot e^{-i\nu_{klmn}\Delta t} + \hat{N}_{klmn}(t) \left[ \frac{1 - e^{-i\nu_{klmn}\Delta t}}{i\nu_{klmn}} \right]$$

*(注意：在這裡，無論 $n=2,3$ 聲波的 $\nu$ 有多大，指數函數 $e^{-i\nu\Delta t}$ 的絕對值永遠是 1，只是在複數平面上瘋狂旋轉，完全不會讓數值炸掉！這就是解開 CFL 封印的魔法！)*

**9. 更新時間，進入下一輪**

* 將計算出的 $\hat{u}_{klmn}(t+\Delta t)$ 設為下一輪的已知值。
* 時間 $t = t + \Delta t$。
* 回到 **步驟 5** 繼續執行。



### 七、轉換法中消除混疊之網格點數條件證明

**證明目標：** 

$$N \ge 3K + 1$$

#### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 物理空間中截斷最大波數為 $K$ 的頻譜展開式：**

  $$\begin{gather*}
  u(x) &=& \sum_{k_1=-K}^{K} \hat{u}_{k_1} e^{ik_1 x} \\
  v(x) &=& \sum_{k_2=-K}^{K} \hat{v}_{k_2} e^{ik_2 x}
  \end{gather*}$$



  * $u(x)$ : 物理空間上的變數分量／東西向速度微擾場 (Physical zonal velocity component) $[\text{m}\cdot\text{s}^{-1}]$
  * $v(x)$ : 物理空間上的變數分量／南北向速度微擾場 (Physical meridional velocity component) $[\text{m}\cdot\text{s}^{-1}]$
  * $K$ : 頻譜截斷最大波數 (Truncated maximum wavenumber) $[\text{m}^{-1}]$，代表數值模式中能解析的最高波數限制
  * $k_{1}$ : 頻譜空間中對應 $u$ 場的整數波數索引 (Wavenumber spectrum index for u) $[\text{m}^{-1}]$
  * $k_{2}$ : 頻譜空間中對應 $v$ 場的整數波數索引 (Wavenumber spectrum index for v) $[\text{m}^{-1}]$
  * $\hat{u}_{k_{1}}$ : 頻譜空間中對應波數 $k_{1}$ 的 $u$ 場傅立葉振幅係數 (Fourier amplitude coefficient of u) $[\text{m}\cdot\text{s}^{-1}]$
  * $\hat{v}_{k_{2}}$ : 頻譜空間中對應波數 $k_{2}$ 的 $v$ 場傅立葉振幅係數 (Fourier amplitude coefficient of v) $[\text{m}\cdot\text{s}^{-1}]$

  * $e^{ik_{1}x}$ : 對應波數 $k_{1}$ 的空間諧振基底函數 (Spatial harmonic basis function) [無單位]
  * $e^{ik_{2}x}$ : 對應波數 $k_{2}$ 的空間諧振基底函數 (Spatial harmonic basis function) [無單位]

  * $N$ : 物理空間的總網格點數 (Total number of grid points in physical space) [無單位]，代表空間離散化的解析度限制

  * $m$ : 進入積分函數內部求和的淨總和波數 (Net combined wavenumber) $[\text{m}^{-1}]$，在頻譜非線性乘積展開中通常由多個交互作用波數相加減構成（如 $m = k_1 + k_2 - k$）

  $$|k| \le K$$

  * $|k|$ : 目標頻譜波數的絕對值 (Absolute value of the target wavenumber) $[\text{m}^{-1}]$
  * $K$ : 頻譜截斷最大波數 (Truncated maximum wavenumber) $[\text{m}^{-1}]$，代表數值模式中所定義並保留的最高物理有效波數上限

  * $k_{2}$ : 頻譜空間中第二個物理變數微擾場（如 $v$ 場）的傅立葉整數波數索引 (Fourier wavenumber index for the second variable) $[\text{m}^{-1}]$
  * $k$ : 目標計算與保留的頻譜波數索引 (Target spectral wavenumber index) $[\text{m}^{-1}]$



* **【推導 2】 離散積分函數內最高波動頻率 $\max(|m|)$ 之極值：**

  $$\begin{gather*}
  \max(|m|) &\overset{\text{定義 2}}{=}& \max(|k_1 + k_2 - k|) \\
  \max(|m|) &\le& \max(|k_1|) + \max(|k_2|) + \max(|-k|) \\
  \max(|m|) &\overset{\text{已知 1, 定義 1}}{\le}& K + K + K \\
  \max(|m|) &=& 3K
  \end{gather*}$$

  * $\max(|m|)$ : 進入離散積分時可能產生的最高淨總和波數絕對值 (Maximum absolute net wavenumber) $[\text{m}^{-1}]$
  * $k_{1}$ : 頻譜空間中第一個物理變數微擾場（如 $u$ 場）的傅立葉整數波數索引 (Fourier wavenumber index for the first variable) $[\text{m}^{-1}]$
  * $k_{2}$ : 頻譜空間中第二個物理變數微擾場（如 $v$ 場）的傅立葉整數波數索引 (Fourier wavenumber index for the second variable) $[\text{m}^{-1}]$
  * $k$ : 目標計算與保留的頻譜波數索引 (Target spectral wavenumber index) $[\text{m}^{-1}]$

#### 證明

求取目標波數 $k$ 的離散傅立葉轉換係數 $\hat{F}_{k,\text{discrete}}$：

$$\begin{gather*}
\hat{F}_{k,\text{discrete}} &=& \frac{1}{N} \sum_{j=0}^{N-1} F(x_j) e^{-ik x_j} \\
\hat{F}_{k,\text{discrete}} &\overset{\text{推導 1}}{=}& \frac{1}{N} \sum_{j=0}^{N-1} \left( \sum_{k_1=-K}^{K} \sum_{k_2=-K}^{K} \hat{u}_{k_1} \hat{v}_{k_2} e^{i(k_1+k_2)x_j} \right) e^{-ik x_j} \\
\hat{F}_{k,\text{discrete}} &=& \frac{1}{N} \sum_{j=0}^{N-1} \sum_{k_1=-K}^{K} \sum_{k_2=-K}^{K} \hat{u}_{k_1} \hat{v}_{k_2} e^{i(k_1+k_2-k)x_j} \\
\hat{F}_{k,\text{discrete}} &\overset{\text{定義 2}}{=}& \sum_{k_1=-K}^{K} \sum_{k_2=-K}^{K} \hat{u}_{k_1} \hat{v}_{k_2} \left( \frac{1}{N} \sum_{j=0}^{N-1} e^{i m x_j} \right)
\end{gather*}$$

為確保精確萃取係數且無混疊誤差，必須保證上式中括號內的離散積分項，絕對不會觸發 $p \neq 0$ 的非零投影條件：

$$\begin{gather*}
\max(|m|) &<& |p|N \quad (p \in \mathbb{Z}, p \neq 0) \\
\max(|m|) &<& |1|N  \\
\max(|m|) &<& N \\
3K &\overset{\text{推導 2}}{<}& N \\
N &\ge& 3K + 1
\end{gather*}$$

### 八、三維波動色散方程式之無因次化轉換證明

**證明目標：** 

$$Y^4 - Y^2 \left[ S(X^2 + Z^2 + G^2) + 1 \right] + S X^2 = 0$$

#### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 三維一般波動色散方程式 (3D wave dispersion relation) ：**

  $$\frac{\nu^4}{c^2} - \nu^2 \left( \bar{k}^2 + \bar{l}^2 + \bar{m}^2 + \Gamma^2 + \frac{N^2}{c^2} \right) + \left( \bar{k}^2 + \bar{l}^2 \right)N^2 = 0$$


  * $\nu $ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $\bar{k}$ : $x$ 軸方向的水平波數／東西向波數 (Zonal wavenumber) $[\text{m}^{-1}]$
  * $\bar{l}$ : $y$ 軸方向的水平波數／南北向波數 (Meridional wavenumber) $[\text{m}^{-1}]$
  * $\bar{m}$ : $z$ 軸方向的垂直波數 (Vertical wavenumber) $[\text{m}^{-1}]$
  * $\Gamma$ : 大氣層結修正因子 $[\text{m}^{-1}]$
  * $N$ : 浮力頻率 (Brunt-Väisälä frequency) $[\text{s}^{-1}]$


* **【定義 1】 無因次頻率 (Normalized frequency) ：**

  $$\begin{gather*}
  Y &\overset{\text{def}}{=}& \frac{\nu}{N} \\
  \nu &=& YN
  \end{gather*}$$

  * $Y$ : 歸一化無因次波頻 (Normalized dimensionless frequency) [無單位]，代表波動頻率相對於浮力頻率的比值
  * $\nu $ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $N$ : 浮力頻率 (Brunt-Väisälä frequency) $[\text{s}^{-1}]$


* **【定義 2】 無因次水平波數 (Normalized horizontal wavenumber) ：**

  $$\begin{gather*}
  X &\overset{\text{def}}{=}& \sqrt{\bar{k}^2 + \bar{l}^2} H_s \\
  \bar{k}^2 + \bar{l}^2 &=& \frac{X^2}{H_s^2}
  \end{gather*}$$

  * $X$ : 歸一化無因次水平波數 (Normalized dimensionless horizontal wavenumber) [無單位]，代表水平波數與大氣特徵高度尺度的乘積
  * $\bar{k}$ : $x$ 軸方向的水平波數／東西向波數 (Zonal wavenumber) $[\text{m}^{-1}]$
  * $\bar{l}$ : $y$ 軸方向的水平波數／南北向波數 (Meridional wavenumber) $[\text{m}^{-1}]$
  * $H_{s}$ : 等溫大氣標高 (Atmospheric scale height) $[\text{m}]$

* **【定義 3】 無因次垂直波數 (Normalized vertical wavenumber) ：**

  $$\begin{gather*}
  Z &\overset{\text{def}}{=}& \bar{m}H_s \\
  \bar{m}^2 &=& \frac{Z^2}{H_s^2}
  \end{gather*}$$

  * $Z$ : 歸一化無因次垂直波數 (Normalized dimensionless vertical wavenumber) [無單位]，代表垂直波數與大氣垂直尺度的比例關係
  * $\bar{m}$ : $z$ 軸方向的垂直波數 (Vertical wavenumber) $[\text{m}^{-1}]$
  * $H_{s}$ : 等溫大氣標高 (Atmospheric scale height) $[\text{m}]$

* **【定義 4】 無因次大氣層結參數 (Normalized stratification parameter) ：**

  $$\begin{gather*}
  G &\overset{\text{def}}{=}& \Gamma H_s \\
  \Gamma^2 &=& \frac{G^2}{H_s^2}
  \end{gather*}$$

  * $G$ : 歸一化無因次大氣層結參數 (Normalized dimensionless stratification parameter) [無單位]，代表密度非均勻性對垂直特徵尺度的修正量
  * $\Gamma $ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $H_{s}$ : 等溫大氣標高 (Atmospheric scale height) $[\text{m}]$


* **【定義 5】 無因次音速參數 (Normalized sound speed parameter) ：**

  $$S \overset{\text{def}}{=} \frac{c^2}{N^2 H_s^2}$$

  * $S$ : 歸一化無因次音速參數 (Normalized dimensionless sound speed parameter) [無單位]，物理上代表聲波傳播速度與浮力波特徵尺度的比值
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $N^{2}$ : 浮力頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $H_{s}^{2}$ : 大氣標高的平方 (Atmospheric scale height squared) $[\text{m}^2]$


* **【推導 1】 頻率高次方項代換 ：**

  * (a) 特徵頻率平方代換：

  $$\begin{gather*}
  \nu &\overset{\text{定義 1}}{=}& Y N \\
  \nu^2 &=& Y^2 N^2
  \end{gather*}$$

  * (b) 特徵頻率四次方代換：

  $$\begin{gather*}
  \nu &\overset{\text{定義 1}}{=}& Y N \\
  \nu^4 &=& Y^4 N^4
  \end{gather*}$$

  * $\nu $ : 系統的實數固有特徵角頻率 (Real eigenfrequency) $[\text{s}^{-1}]$
  * $Y$ : 歸一化無因次波頻 (Normalized dimensionless frequency) [無單位]
  * $N$ : 浮力頻率 (Brunt-Väisälä frequency) $[\text{s}^{-1}]$


#### 證明

$$\begin{gather*}
\frac{\nu^4}{c^2} - \nu^2 \left( \bar{k}^2 + \bar{l}^2 + \bar{m}^2 + \Gamma^2 + \frac{N^2}{c^2} \right) + \left( \bar{k}^2 + \bar{l}^2 \right)N^2 &\overset{\text{已知 1}}{=}& 0 \\
\frac{Y^4 N^4}{c^2} - Y^2 N^2 \left( \frac{X^2}{H_s^2} + \frac{Z^2}{H_s^2} + \frac{G^2}{H_s^2} + \frac{N^2}{c^2} \right) + \left( \frac{X^2}{H_s^2} \right)N^2 &\overset{\text{定義 2,3,4, 推導 1(a)(b)}}{=}& 0 \\
\frac{Y^4 N^4}{c^2} - Y^2 N^2 \left( \frac{X^2 + Z^2 + G^2}{H_s^2} + \frac{N^2}{c^2} \right) + \frac{X^2 N^2}{H_s^2} &=& 0 \\
\frac{c^2}{N^4} \left[ \frac{Y^4 N^4}{c^2} - Y^2 N^2 \left( \frac{X^2 + Z^2 + G^2}{H_s^2} + \frac{N^2}{c^2} \right) + \frac{X^2 N^2}{H_s^2} \right] &=& 0 \cdot \frac{c^2}{N^4} \\
Y^4 - Y^2 \frac{c^2}{N^2} \left( \frac{X^2 + Z^2 + G^2}{H_s^2} + \frac{N^2}{c^2} \right) + X^2 \frac{c^2 N^2}{N^4 H_s^2} &=& 0 \\
Y^4 - Y^2 \left( \frac{c^2}{N^2 H_s^2}(X^2 + Z^2 + G^2) + \frac{c^2}{N^2}\frac{N^2}{c^2} \right) + \frac{c^2}{N^2 H_s^2} X^2 &=& 0 \\
Y^4 - Y^2 \left( \frac{c^2}{N^2 H_s^2}(X^2 + Z^2 + G^2) + 1 \right) + \frac{c^2}{N^2 H_s^2} X^2 &=& 0 \\
Y^4 - Y^2 \left[ S(X^2 + Z^2 + G^2) + 1 \right] + S X^2 &\overset{\text{定義 5}}{=}& 0 \\
\end{gather*}$$



這是一道非常經典的氣象動力學繪圖題。Gill 的 *Atmosphere-Ocean Dynamics* 書中 Figure 6-17 展示了這四種波在頻譜空間上的漸近行為與界線。為了用 Python 重現這張圖，我們需要先將你剛才推導出的色散方程式「無因次化（Non-dimensionalize）」。

#### 數學轉換（無因次化）

我們定義三個無因次變數，作為繪圖的 X 軸、Y 軸與不同曲線的參數：


1. **X 軸（無因次水平波數）：** $X = \bar{k}H_s$
2. **Y 軸（無因次頻率）：** $Y = \frac{\nu}{N}$
3. **曲線參數（無因次垂直波數）：** $Z = \bar{m}H_s$

將這些變數代入三維波動的色散方程式 $(9a)$，並引入地球大氣的常數：

* 取雙原子氣體絕熱指數 $\gamma = 1.4$
* 聲速平方與浮力頻率的比例常數：$S = \frac{c^2}{N^2 H_s^2} = \frac{\gamma^2}{\gamma - 1} \approx 4.9$
* 大氣層結參數：$G = \Gamma H_s = \frac{1}{\gamma} - 0.5 \approx 0.214$

方程式將化簡為一個漂亮的關於 $Y^2$ 的一元二次方程式：


$$\begin{gather*}
Y^4 - Y^2 \left[ S(X^2 + Z^2 + G^2) + 1 \right] + S X^2 &=& 0 \\
\end{gather*}$$


而藍姆波 (Lamb wave) 的色散方程式 $(8b)$ 則化簡為：


$$Y = \sqrt{S} X$$

這張圖是探討大氣波動極為核心的**色散關係圖（Dispersion Diagram）**。它視覺化了流體力學方程式在特定邊界條件下，所允許存在的波動狀態。

##### 觀察圖表的核心物理意義

當您執行程式碼後，會看到雙對數圖上呈現非常清晰的界線：

1. **右下角的浮力波 (Gravity Waves)**：不管水平波數 $\bar{k}H_s$ 多大，所有低頻點狀虛線最終都會被**灰色虛線 ($\nu/N = 1$)** 壓制。這完美印證了我們在第 5 題推導的：環境層結穩定度 $N$ 是內重力波震盪的最快極限。
2. **左上角的聲波 (Acoustic Waves)**：所有的短虛線在向左延伸（極長波）時，都不會掉到 0，而是被**紫色虛線 ($\nu/N \approx 1.11$)** 托住，這就是聲學截斷頻率 $N_A$ 的視覺化證明。
3. **貫穿中央的藍姆波 (Lamb Wave)**：這是一條斜率為 1 的直線，代表它是非色散的（波速為常數音速），且其頻率不受 $N$ 或是 $N_A$ 的上下限約束，橫跨了整個頻譜。


#####  坐標軸的物理意義

* **X 軸（$\bar{k}H_s$）：無因次水平波數**
這代表波在水平方向上的「空間擁擠程度」。數值越小（靠左），代表水平波長越長（例如跨越數千公里的大尺度天氣系統）；數值越大（靠右），代表水平波長越短（例如幾公里寬的積雨雲或微尺度擾動）。
* **Y 軸（$\omega/N$）：無因次頻率**
這代表波隨時間「震盪的快慢」。它被環境的浮力頻率 $N$ 給標準化了。數值越大（靠上），波震盪越劇烈；數值越小（靠下），波的變化越緩慢。


#####  每一條線的意義與「量子化」疑問

* **這些線代表什麼？** 每條虛線或點虛線，都代表一個具有特定垂直結構（特定的垂直波數 $\bar{m}$）的波，其頻率與水平波長之間的對應關係。
* **是特舉幾條線，還是只有這幾條？** 圖中畫出的 $mH_s = 0, 0.4, 0.8, 1.2$ 只是「舉例」**。但在封閉的大氣模型中（例如有堅硬的地表與模式頂部），垂直波數 $\bar{m}$ 確實是**被量子化（Quantized）的！為了滿足上下邊界垂直速度 $w=0$ 的條件，波的垂直結構只能是半波長的整數倍（即 $\bar{m} = \frac{n\pi}{L_z}$）。
* 因此，這張圖實際上存在著無數條離散的曲線（對應 $n=1, 2, 3...$），這些密集的離散曲線填滿了整塊灰色的允許區域 。




#####  點與點之間的相對關係

* **同一條線上不同的點：**
代表具有「相同垂直層數（垂直波長固定）」，但「水平尺度不斷縮放」的同族波動。當水平波數 $\bar{k}$ 改變時，頻率 $\omega$ 必須沿著這條線嚴格跟著改變。此外，曲線上任意一點的切線斜率，在物理上精確等於該波的**水平群速度（Horizontal Group Velocity）**，也就是波傳遞能量的速度 。


* **同一個 X 座標不同點：**
代表我們在看「水平寬度一模一樣」的擾動。沿著垂直線往上下看，會發現它能以多種不同的頻率震盪。高頻對應的是垂直波長不同的聲波，低頻對應的是垂直波長不同的浮力波。
* **同一個 Y 座標不同點：**
代表我們用一個「固定頻率的節拍器」去擾動大氣。這條水平線會切過許多不同的曲線，意味著同一個震盪頻率，可以激發出多種水平波長與垂直波長組合不同的波，它們會在空間中同時向外傳播並發生干涉。
 
#####  白色的區域與不存在的波

圖中沒有被灰色網點覆蓋的白色區域（包含聲波下限與浮力波上限之間的空白帶），被稱為「衰減區（Evanescent Zone）」或「禁帶（Forbidden Zone）」。

* **絕對不存在這種 wave 嗎？**
這取決於您對「存在」的定義。在數學與物理上，**「能夠自由、無損耗向遠方傳遞能量的自由行進波（Freely propagating waves）」在這些白色區域是絕對不存在的** 。

* **物理本質：**
如果您強行以白色區域內的頻率（例如 $\omega/N = 1.05$，介於浮力極限與聲學截斷之間）去震盪空氣，數學解依然存在，但其垂直波數 $\bar{m}$ 會變成**虛數**。這代表擾動無法形成波動，其振幅會隨空間距離呈「指數型快速衰減（Exponential decay）」。擾動的能量會被困在震源附近，無法像水波一樣輻射出去。
* **特例：**
您會注意到有一條黑色的實線（藍姆波，Lamb wave）大搖大擺地切過了這塊白色禁區 。這是因為藍姆波是極度特殊的**二維邊界波**（垂直速度為零），它不受三維內部重力波與聲波的頻率極限約束，因此能存在於內部波的禁帶之中。
