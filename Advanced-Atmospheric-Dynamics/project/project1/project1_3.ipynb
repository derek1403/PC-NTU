# **三、確認算子 $\mathbf{L}$ 的反厄米特特性**

$$(\mathbf{L}\tilde{f}, \tilde{g}) = -(\tilde{f}, L\tilde{g})$$

算子 $\mathbf{L}$ 在指定內積 $(\cdot , \cdot)$ 下滿足反厄米特（Anti-Hermitian）特性，意即 $(\mathbf{L}\tilde{f}, \tilde{g}) = -(\tilde{f}, \mathbf{L}\tilde{g})$。這在物理上意味著該系統不存在能量的憑空耗散或產生，是波動得以穩定傳播的數學保證。


## 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 空間微分波動算符矩陣（Spatial differential wave operator matrix） ：**
  
  $$ \mathbf{L} \overset{\text{def}}{=} \begin{bmatrix} 
  0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\[2ex]
  0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\[2ex]
  0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\[2ex]
  c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\[2ex]
  0 & 0 & -\frac{N^2}{g} & 0 & 0 \\[2ex]
  \end{bmatrix}  $$

  * $\mathbf{L}$ : 封裝空間偏微分與大氣層結常數的 $5 \times 5$ 矩陣微分算符 (Spatial differential operator matrix)
  * $\Gamma $ : 大氣層結修正因子 $[\text{m}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 $[\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $\frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z}$ : 各時空空間軸向的偏微分算符 $[\text{m}^{-1}]$
  

* **【定義 1】 測試函數向量與大氣波動邊界條件（Test function vectors and atmospheric wave boundary conditions）  ：**

  * (a) 五維測試函數向量定義：
  
  $$
  \tilde{f} \overset{\text{let}}{=} \begin{bmatrix}  f_1  \\[2ex] f_2 \\[2ex] f_3 \\[2ex] f_4 \\[2ex] f_5 \\[2ex] \end{bmatrix}  \quad , \quad
  \tilde{g} \overset{\text{let}}{=} \begin{bmatrix}  g_1  \\[2ex] g_2 \\[2ex] g_3 \\[2ex] g_4 \\[2ex] g_5 \\[2ex] \end{bmatrix}
  $$

  * (b) 水平週期性與垂直剛性壁面邊界條件：

  $$\left\{\begin{matrix}
  f_i(0, y, z, t) &=& f_i(L_x, y, z, t) &\quad , \quad i \in \{1, 2, 3, 4, 5\} \\
  f_i(x, 0, z, t) &=& f_i(x, L_y, z, t) &\quad , \quad i \in \{1, 2, 3, 4, 5\} \\
  f_3(x, y, 0, t) &=& 0 &\quad (\text{底部 } z=0) \\
  f_3(x, y, L_z, t) &=& 0 &\quad (\text{頂部 } z=L_z) \\
  \end{matrix}\right.$$

  $$\left\{\begin{matrix}
  g_i(0, y, z, t) &=& g_i(L_x, y, z, t) &\quad , \quad i \in \{1, 2, 3, 4, 5\} \\
  g_i(x, 0, z, t) &=& g_i(x, L_y, z, t) &\quad , \quad i \in \{1, 2, 3, 4, 5\} \\
  g_3(x, y, 0, t) &=& 0 &\quad (\text{底部 } z=0) \\
  g_3(x, y, L_z, t) &=& 0 &\quad (\text{頂部 } z=L_z) \\
  \end{matrix}\right.$$

  * $f_i, g_i$ : 向量中的各個純量分量元素 (Vector component elements)，其物理量綱與【定義 1 (前章)】的大氣狀態向量各元件一一對應
  * $x, y, z$ : 三維空間位置座標 $[\text{m}]$
  * $t$ : 時間 $[\text{s}]$
  * $L_{x}$ : 系統在 $x$ 軸方向（東西向）的週期長度 (Domain length in x-direction) $[\text{m}]$
  * $L_{y}$ : 系統在 $y$ 軸方向（南北向）的週期長度 (Domain length in y-direction) $[\text{m}]$
  * $L_{z}$ : 大氣層垂直方向的總厚度／頂部高度 (Domain height in z-direction) $[\text{m}]$
  * $f_3, g_3$ : 對應垂直速度 $w$ 的函數分量；在頂部與底部為零（即滿足剛性無穿透條件 Rigid non-penetration boundary condition）
  

* **【定義 2】 能量權重空間平均內積（Energy-weighted spatially averaged inner product）  ：**

  $$ (\tilde{f}, \tilde{g}) \overset{\text{def}}{=} \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( f_1 g_1^* + f_2 g_2^* + f_3 g_3^* + \frac{1}{c^2} f_4 g_4^* + \frac{g^2}{N^2} f_5 g_5^* \right) dxdydz $$

  * $(\mathbf{\tilde{f}}, \mathbf{\tilde{g}})$ : 兩個波動狀態向量在全域空間上的內積值 (Spatially averaged inner product) [多元複合單位]
  * $f_{i}$ : 第一個測試函數向量 $\tilde{\mathbf{f}}$ 的第 $i$ 個純量分量
  * $g_{i}^{*}$ : 第二個測試函數向量 $\tilde{\mathbf{g}}$ 的第 $i$ 個純量分量的共軛複數 (Complex conjugate)
  * $x, y, z$ : 三維直角空間座標 $[\text{m}]$
  * $L_x, L_y$ : 水平方向的長度與寬度 $[\text{m}]$
  * $L_{z}$ : 垂直方向的大氣總高度 $[\text{m}]$
  * $\frac{1}{L_{x}L_{y}L_{z}}$ : 空間體積歸一化因子 (Volume normalization factor) $[\text{m}^{-3}]$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $\frac{1}{c^2} f_4 g_4^*$ : 經彈性音速權重調整後的彈性可壓縮位能項 (Compressible potential energy component)
  * $\frac{g^2}{N^2} f_5 g_5^*$ : 經靜力穩定度權重調整後的重力浮力熱力位能項 (Gravitational potential energy component)

* **【推導 1】  算符矩陣作用於狀態向量的各分量展開（Component expansion of operator matrix acting on state vectors） ：**

  * (a) 算符矩陣 $\mathbf{L}$ 作用於測試向量 $\tilde{\mathbf{f}}$：

  $$\begin{gather*} 
  \mathbf{L}\tilde{f} &\overset{\text{已知 1,定義 1(a)}}{=}&  \begin{bmatrix} 
  0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\[2ex]
  0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\[2ex]
  0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\[2ex]
  c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\[2ex]
  0 & 0 & -\frac{N^2}{g} & 0 & 0 \\[2ex]
  \end{bmatrix}  \begin{bmatrix}  f_1  \\[2ex] f_2 \\[2ex] f_3 \\[2ex] f_4 \\[2ex] f_5 \\[2ex] \end{bmatrix} \\
  &=& \begin{bmatrix}  \frac{\partial f_4}{\partial x}  \\[2ex]
  \frac{\partial f_4}{\partial y} \\[2ex] 
  \left(\frac{\partial}{\partial z} + \Gamma\right)f_4 + g f_5 \\[2ex] 
  c^2 \left[ \frac{\partial f_1}{\partial x} + \frac{\partial f_2}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)f_3 \right] \\[2ex] 
  -\frac{N^2}{g} f_3 \\[2ex] \end{bmatrix}
  \end{gather*}$$

  * (b) 算符矩陣 $\mathbf{L}$ 作用於測試向量 $\tilde{\mathbf{g}}$：

  $$\begin{gather*} 
  \mathbf{L}\tilde{g} &\overset{\text{已知 1,定義 1(a)}}{=}&  \begin{bmatrix} 
  0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\[2ex]
  0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\[2ex]
  0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\[2ex]
  c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\[2ex]
  0 & 0 & -\frac{N^2}{g} & 0 & 0 \\[2ex]
  \end{bmatrix}  \begin{bmatrix}  g_1  \\[2ex] g_2 \\[2ex] g_3 \\[2ex] g_4 \\[2ex] g_5 \\[2ex] \end{bmatrix} \\
  &=& \begin{bmatrix}  \frac{\partial g_4}{\partial x}  \\[2ex]
  \frac{\partial g_4}{\partial y} \\[2ex] 
  \left(\frac{\partial}{\partial z} + \Gamma\right)g_4 + g g_5 \\[2ex] 
  c^2 \left[ \frac{\partial g_1}{\partial x} + \frac{\partial g_2}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)g_3 \right] \\[2ex] 
  -\frac{N^2}{g} g_3 \\[2ex] \end{bmatrix}
  \end{gather*}$$

  * $\mathbf{L}$ : 空間微分波動算符矩陣
  * $\mathbf{\tilde{f}}, \mathbf{\tilde{g}}$ : 五維測試函數向量
  * $f_i , g_i$ : 各向量對應之純量分量元素
  * $\Gamma $ : 大氣層結修正因子 $[\text{m}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 $[\text{s}^{-2}]$
  * $c^{2}$ : 本地音速平方 $[\text{m}^2\cdot\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$

* **【推導 2】  單維度分部積分與邊界項消去（Single-dimensional integration by parts and boundary term elimination） ：**

  * (a) $x$ 方向之 $f_{4}$ 微分項分部積分：
  
  $$\begin{gather*} 
  \int_{0}^{L_x} \frac{\partial f_4}{\partial x}g_1^* \ dx &\overset{\text{IbP}}{=}& \left[ f_4 \cdot g_1^* \right]_{0}^{L_x} -  \int_{0}^{L_x}  \frac{\partial g_1^*}{\partial x} f_4 \ dx \\
  &=& \left( f_4(L_x) \cdot g_1^*(L_x)  -  f_4(0) \cdot g_1^*(0) \right) -  \int_{0}^{L_x}  \frac{\partial g_1^*}{\partial x} f_4 \ dx \\
  &\overset{\text{定義 1(b)}}{=}& \left( f_4(0) \cdot g_1^*(0)  -  f_4(0) \cdot g_1^*(0) \right) -  \int_{0}^{L_x}  \frac{\partial g_1^*}{\partial x} f_4 \ dx \\
  &=& -  \int_{0}^{L_x}  \frac{\partial g_1^*}{\partial x} f_4 \ dx \\
  \end{gather*}$$

  * (b) $y$ 方向之 $f_{4}$ 微分項分部積分：
  
  $$\begin{gather*} 
  \int_{0}^{L_y} \frac{\partial f_4}{\partial y}g_2^* \ dy &\overset{\text{IbP}}{=}& \left[ f_4 \cdot g_2^* \right]_{0}^{L_y} -  \int_{0}^{L_y}  \frac{\partial g_2^*}{\partial y} f_4 \ dy \\
  &=& \left( f_4(L_y) \cdot g_2^*(L_y)  -  f_4(0) \cdot g_2^*(0) \right) -  \int_{0}^{L_y}  \frac{\partial g_2^*}{\partial y} f_4 \ dy \\
  &\overset{\text{定義 1(b)}}{=}& \left( f_4(0) \cdot g_2^*(0)  -  f_4(0) \cdot g_2^*(0) \right) -  \int_{0}^{L_y}  \frac{\partial g_2^*}{\partial y} f_4 \ dy \\
  &=& -  \int_{0}^{L_y}  \frac{\partial g_2^*}{\partial y} f_4 \ dy \\
  \end{gather*}$$

  * (c) $z$ 方向之 $f_{4}$ 微分項分部積分：

  $$\begin{gather*} 
  \int_{0}^{L_z} \frac{\partial f_4}{\partial z}g_3^* \ dz &\overset{\text{IbP}}{=}& \left[ f_4 \cdot g_3^* \right]_{0}^{L_z} -  \int_{0}^{L_z}  \frac{\partial g_3^*}{\partial z} f_4 \ dz \\
  &=& \left( f_4(L_z) \cdot g_3^*(L_z)  -  f_4(0) \cdot g_3^*(0) \right) -  \int_{0}^{L_z}  \frac{\partial g_3^*}{\partial z} f_4 \ dz \\
  &\overset{\text{定義 1(b)}}{=}& \left( 0 \cdot 0  -  0 \cdot 0 \right) -  \int_{0}^{L_z}  \frac{\partial g_3^*}{\partial z} f_4 \ dz \\
  &=& -  \int_{0}^{L_z}  \frac{\partial g_3^*}{\partial z} f_4 \ dz \\
  \end{gather*}$$

  * (d) $x$ 方向之 $f_{1}$ 微分項分部積分：

  $$\begin{gather*} 
  \int_{0}^{L_x} \frac{\partial f_1}{\partial x}g_4^* \ dx &\overset{\text{IbP}}{=}& \left[ f_1 \cdot g_4^* \right]_{0}^{L_x} -  \int_{0}^{L_x}  \frac{\partial g_4^*}{\partial x} f_1 \ dx \\
  &=& \left( f_1(L_x) \cdot g_4^*(L_x)  -  f_1(0) \cdot g_4^*(0) \right) -  \int_{0}^{L_x}  \frac{\partial g_4^*}{\partial x} f_1 \ dx \\
  &\overset{\text{定義 1(b)}}{=}& \left( f_1(0) \cdot g_4^*(0)  -  f_1(0) \cdot g_4^*(0) \right) -  \int_{0}^{L_x}  \frac{\partial g_4^*}{\partial x} f_1 \ dx \\
  &=& -  \int_{0}^{L_x}  \frac{\partial g_4^*}{\partial x} f_1 \ dx \\
  \end{gather*}$$

  * (e) $y$ 方向之 $f_{2}$ 微分項分部積分：

  $$\begin{gather*} 
  \int_{0}^{L_y} \frac{\partial f_2}{\partial y}g_4^* \ dy &\overset{\text{IbP}}{=}& \left[ f_2 \cdot g_4^* \right]_{0}^{L_y} -  \int_{0}^{L_y}  \frac{\partial g_4^*}{\partial y} f_2 \ dy \\
  &=& \left( f_2(L_y) \cdot g_4^*(L_y)  -  f_2(0) \cdot g_4^*(0) \right) -  \int_{0}^{L_y}  \frac{\partial g_4^*}{\partial y} f_2 \ dy \\
  &\overset{\text{定義 1(b)}}{=}& \left( f_2(0) \cdot g_4^*(0)  -  f_2(0) \cdot g_4^*(0) \right) -  \int_{0}^{L_y}  \frac{\partial g_4^*}{\partial y} f_2 \ dy \\
  &=& -  \int_{0}^{L_y}  \frac{\partial g_4^*}{\partial y} f_2 \ dy \\
  \end{gather*}$$

  * (f) $z$ 方向之 $f_{3}$ 微分項分部積分： 

  $$\begin{gather*} 
  \int_{0}^{L_z} \frac{\partial f_3}{\partial z}g_4^* \ dz &\overset{\text{IbP}}{=}& \left[ f_3 \cdot g_4^* \right]_{0}^{L_z} -  \int_{0}^{L_z}  \frac{\partial g_4^*}{\partial z} f_3 \ dz \\
  &=& \left( f_3(L_z) \cdot g_4^*(L_z)  -  f_3(0) \cdot g_4^*(0) \right) -  \int_{0}^{L_z}  \frac{\partial g_4^*}{\partial z} f_3 \ dz \\
  &\overset{\text{定義 1(b)}}{=}& \left( 0 \cdot 0  -  0 \cdot 0 \right) -  \int_{0}^{L_z}  \frac{\partial g_4^*}{\partial z} f_3 \ dz \\
  &=& -  \int_{0}^{L_z}  \frac{\partial g_4^*}{\partial z} f_3 \ dz \\
  \end{gather*}$$

  * $\text{IbP}$ : 分部積分法 (Integration by Parts) 算符法則
  * $f_i, g_i$ : 測試函數向量的空間連續分量元素
  * $g_{i}^{*}$ : 測試函數分量的共軛複數 (Complex conjugate)
  * $x, y, z$ : 水平與垂直空間位置座標 $[\text{m}]$
  * $L_x, L_y$ : 水平邊界邊長週期特徵長度 $[\text{m}]$
  * $L_{z}$ : 垂直域頂部高度 $[\text{m}]$

## 證明


$$\begin{gather*} 
(\mathbf{L}\tilde{f}, \tilde{g})  &\overset{\text{定義 2}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( (L\tilde{f})_1 g_1^* + (L\tilde{f})_2 g_2^* + (L\tilde{f})_3 g_3^* + \frac{1}{c^2} (L\tilde{f})_4 g_4^* + \frac{g^2}{N^2} (L\tilde{f})_5 g_5^* \right) dxdydz \\
&\overset{\text{推導 1(a)}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( \frac{\partial f_4}{\partial x} g_1^* + \frac{\partial f_4}{\partial y} g_2^* + \left(\left(\frac{\partial}{\partial z} + \Gamma\right)f_4 + g f_5 \right) g_3^* + \frac{1}{c^2} c^2 \left[ \frac{\partial f_1}{\partial x} + \frac{\partial f_2}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)f_3 \right] g_4^* + \frac{g^2}{N^2} (-\frac{N^2}{g} f_3) g_5^* \right) dxdydz \\
&=& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( \frac{\partial f_4}{\partial x}g_1^* + \frac{\partial f_4}{\partial y}g_2^* + \frac{\partial f_4}{\partial z}g_3^* + \Gamma f_4 g_3^* + g f_5 g_3^* + \frac{\partial f_1}{\partial x}g_4^* + \frac{\partial f_2}{\partial y}g_4^* + \frac{\partial f_3}{\partial z}g_4^* - \Gamma f_3 g_4^* - g f_3 g_5^* \right) dxdydz \\
&\overset{\text{推導 2(a)(b)(c)(d)(e)(f)}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( -\frac{\partial g_1^*}{\partial x}f_4 - \frac{\partial g_2^*}{\partial y}f_4 - \frac{\partial g_3^*}{\partial z}f_4 + \Gamma f_4 g_3^* + g f_5 g_3^* - \frac{\partial g_4^*}{\partial x}f_1 - \frac{\partial g_4^*}{\partial y}f_2 - \frac{\partial g_4^*}{\partial z}f_3 - \Gamma f_3 g_4^* - g f_3 g_5^*\right) dxdydz \\
&=& - \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x}  \left( \frac{\partial g_1^*}{\partial x}f_4 + \frac{\partial g_2^*}{\partial y}f_4 + \frac{\partial g_3^*}{\partial z}f_4 - \Gamma f_4 g_3^* - g f_5 g_3^* + \frac{\partial g_4^*}{\partial x}f_1 + \frac{\partial g_4^*}{\partial y}f_2 + \frac{\partial g_4^*}{\partial z}f_3 + \Gamma f_3 g_4^* + g f_3 g_5^*\right) dxdydz \\
&=& -\frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x}  \left( \frac{\partial g_4^*}{\partial x}f_1 + \frac{\partial g_4^*}{\partial y}f_2 + \left( \frac{\partial g_4^*}{\partial z} + \Gamma g_4^* + g g_5^* \right) f_3 + \left( \frac{\partial g_1^*}{\partial x} + \frac{\partial g_2^*}{\partial y} + \frac{\partial g_3^*}{\partial z} - \Gamma g_3^* \right) f_4 - g g_3^* f_5 \right) dxdydz \\
&=& - \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x}  \left( \frac{\partial g_4^*}{\partial x}f_1 + \frac{\partial g_4^*}{\partial y}f_2 + \left( \left(\frac{\partial}{\partial z} + \Gamma\right)g_4^* + g g_5^* \right) f_3 + \left[ \frac{\partial g_1^*}{\partial x} + \frac{\partial g_2^*}{\partial y} + \left(\frac{\partial}{\partial z} - \Gamma\right)g_3^* \right] f_4 + \frac{g^2}{N^2} (-\frac{N^2}{g} g_3^*) f_5 \right) dxdydz \\
&\overset{\text{定義 2}}{=}& -(\tilde{f}, \mathbf{L}\tilde{g})
\end{gather*}$$

證明完畢 ！算子 $\mathbf{L}$ 在此內積定義下確實是反厄米特（Skew-Hermitian）算子。
