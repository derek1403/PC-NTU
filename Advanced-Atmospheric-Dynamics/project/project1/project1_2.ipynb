# **二、流體力學偏微分方程式** 轉成 **線性代數矩陣方程式** (From fluid dynamics PDEs to linear algebraic matrix equations.)

$$\frac{\partial \tilde{u}}{\partial t} + \mathbf{L} \tilde{u} = \tilde{N}$$





## 假設與已知 (Assumptions & Preliminaries)


* **【已知 1】  密度加權的線性化微擾方程組與非線性強迫項分離（Density-weighted linearized perturbation equations with nonlinear forcing separation） ：**

  * (a) $x$ 方向動量方程式：

  $$\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}} u \right] + \frac{\partial }{\partial x} \left[ \bar{\rho}^{\frac{-1}{2}} p' \right]  = \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) u + \frac{\rho'}{\bar{\rho} \rho  } \frac{\partial p'}{\partial x} \right) $$

  * (b) $y$ 方向動量方程式：
  
  $$\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}} v \right] + \frac{\partial }{\partial y} \left[ \bar{\rho}^{\frac{-1}{2}} p' \right]  = \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) v + \frac{\rho'}{\bar{\rho} \rho  }\frac{\partial p'}{\partial y} \right) $$
  
  * (c) $z$ 方向動量方程式：
  
  $$\frac{\partial }{\partial t } \left[ \bar{\rho}^{\frac{1}{2}}  w \right]  + g\bar{\rho}^{\frac{-1}{2}}\left(\rho' - \frac{p'}{c^2}\right) + \left(  \frac{\partial}{\partial z} + \Gamma \right)[\bar{\rho}^{\frac{-1}{2}}p'] = - \bar{\rho}^{\frac{1}{2}} \left( (\mathbf{u} \cdot \nabla) w  +  \frac{\rho'^2}{\rho \bar{\rho} } g +  \frac{\rho'}{\rho \bar{\rho}} \frac{\partial p'}{\partial z}  \right)$$

  * (d) 結合狀態方程之連續（質量守恆）方程式：

  $$\frac{\partial ( \bar{\rho}^{\frac{-1}{2}} p') }{\partial t} + c^2 \left( \frac{\partial ( \bar{\rho}^{\frac{1}{2}} u) }{\partial x } +  \frac{\partial  (\bar{\rho}^{\frac{1}{2}} v)}{\partial y } + \left( \frac{\partial }{\partial z} -  \Gamma \right)\left[\bar{\rho}^{\frac{1}{2}} w \right]  \right)   =  -\bar{\rho}^{\frac{-1}{2}} \left( (\mathbf{u} \cdot \nabla ) p' + c^2  \rho'  \nabla \cdot \mathbf{u} \right) $$

  * (e) 內能／熱力學能量（等熵）方程式：

  $$\frac{\partial }{\partial t} \left[ \bar{\rho}^{\frac{-1}{2}}  \left( \rho' - \frac{p'}{c^2}  \right)   \right] -  w \bar{\rho}^{\frac{1}{2}} \frac{N^2}{g}   = -\bar{\rho}^{\frac{-1}{2}} \left( \mathbf{u}\cdot\nabla \rho' -\frac{1}{c^2} \mathbf{u}\cdot\nabla p'  \right)$$

  * $\rho $ : 瞬時總密度場 (Instantaneous total density field) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $\rho '$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$
  * $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}]$
  * $\mathbf{u}$: 三維速度向量 $(u, v, w)$  $[\text{m} \cdot \text{s}^{-1}]$
  * $g$ : 地表重力加速度常數 (Standard acceleration of gravity) $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $\Gamma $ : 大氣層結修正因子 (Stratification correction factor) $[\text{m}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$


* **【定義 1】 經密度加權的大氣狀態向量（Density-weighted atmospheric state vector） ：**

  $$\begin{gather*} 
  \tilde{u} &\overset{\text{def}}{=}& \begin{bmatrix} U \\[1.5ex] V \\[1.5ex] W \\[1.5ex] P \\[1.5ex] \Pi \\[1.5ex] \end{bmatrix} 
  &\overset{\text{let}}{=}&  \begin{bmatrix} \bar{\rho}^{\frac{1}{2}} u \\[1.5ex] \bar{\rho}^{\frac{1}{2}} v \\[1.5ex] \bar{\rho}^{\frac{1}{2}} w \\[1.5ex] \bar{\rho}^{\frac{-1}{2}} p' \\[1.5ex] \bar{\rho}^{\frac{-1}{2}}  \left( \rho' - \frac{p'}{c^2}  \right)\\[1.5ex] \end{bmatrix}
  \end{gather*}$$

  * $\tilde{\mathbf{u}}$ : 大氣波動狀態向量 (Atmospheric wave state vector)
  * $U, V, W$ : 經密度平方根加權後的 $x, y, z$ 三軸方向動量型速度分量 $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $P$ : 經密度倒數平方根加權後的縮放氣壓微擾項 $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $\Pi $ : 經密度倒數平方根加權後的非等熵度量／勢度微擾項 (Non-isentropic perturbation representative) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-3}]$
  * $u, v, w, p', \rho'$ : 原始微擾場的速度、氣壓與密度分量
  * $\bar{\rho }$ : 基本背景態密度 $[\text{kg}\cdot\text{m}^{-3}]$
  * $c$ : 本地音速 $[\text{m}\cdot\text{s}^{-1}]$


* **【定義 2】 非線性強迫力向量（Nonlinear forcing vector） ：**

  $$\tilde{N} \overset{\text{def}}{=} \begin{bmatrix} 
  \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) u + \frac{\rho'}{\bar{\rho} \rho  } \frac{\partial p'}{\partial x} \right) \\[2ex] 
  \bar{\rho}^{\frac{1}{2}} \left(-  (\mathbf{u} \cdot \nabla) v + \frac{\rho'}{\bar{\rho} \rho  }\frac{\partial p'}{\partial y} \right) \\[2ex] 
  - \bar{\rho}^{\frac{1}{2}} \left( (\mathbf{u} \cdot \nabla) w  +  \frac{\rho'^2}{\rho \bar{\rho} } g +  \frac{\rho'}{\rho \bar{\rho}} \frac{\partial p'}{\partial z}  \right) \\[2ex] 
  -\bar{\rho}^{\frac{-1}{2}} \left( (\mathbf{u} \cdot \nabla ) p' + c^2  \rho'  \nabla \cdot \mathbf{u} \right) \\[2ex] 
  -\bar{\rho}^{\frac{-1}{2}} \left( \mathbf{u}\cdot\nabla \rho' -\frac{1}{c^2} \mathbf{u}\cdot\nabla p'  \right) \\[2ex]
  \end{bmatrix} $$

  * $\tilde{\mathbf{N}}$ : 包含平流與微擾乘積項的五維非線性強迫力向量 (Nonlinear forcing column vector)
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $\rho $ : 瞬時總密度場 $[\text{kg}\cdot\text{m}^{-3}]$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $(\mathbf{u} \cdot \nabla)$ : 三維平流算符 $[\text{s}^{-1}]$

* **【定義 3】 空間微分波動算符矩陣（Spatial differential wave operator matrix） ：**
  
  $$ \mathbf{L} \overset{\text{def}}{=} \begin{bmatrix} 
  0 & 0 & 0 & \frac{\partial}{\partial x} & 0 \\[2ex]
  0 & 0 & 0 & \frac{\partial}{\partial y} & 0 \\[2ex]
  0 & 0 & 0 & \left(\frac{\partial}{\partial z} + \Gamma\right) & g \\[2ex]
  c^2 \frac{\partial}{\partial x} & c^2 \frac{\partial}{\partial y} & c^2 \left(\frac{\partial}{\partial z} - \Gamma\right) & 0 & 0 \\[2ex]
  0 & 0 & -\frac{N^2}{g} & 0 & 0 \\[2ex]
  \end{bmatrix}  $$

  * $\mathbf{L}$ : 封裝空間偏微分與大氣層結常數的 $5 \times 5$ 矩陣微分算符 (Spatial differential operator matrix)
  * $\Gamma $ : 大氣層結修正因子 $[\text{m}^{-1}]$
  * $N^{2}$ : 浮力頻率平方 $[\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$
  * $\frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z}$ : 各時空空間軸向的偏微分算符 $[\text{m}^{-1}]$
