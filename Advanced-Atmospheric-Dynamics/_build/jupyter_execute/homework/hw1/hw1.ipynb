# Question 1

氧氣是維持生命機能的要素。在理想狀態下，我們以管道層流來模擬生物循環系統的氧氣輸送，若 $[O_2]$ 為流體內氧氣的濃度，且流體的體積積流量為 $Q = \frac{\pi r^4 \Delta p}{8 \mu l}$，則單位時間內的氧氣輸送量公式為：
   
$$J_{O_2} = [O_2] \frac{\pi r^4 \Delta p}{8 \mu l}$$

* $J_{O_2}$ 是氧氣輸送速率，單位是 $[\text{mol} \cdot \text{s}^{-1}]$
* $[O_2]$ 是氧氣濃度，單位是 $[\text{mol} \cdot \text{m}^{-3}]$
* $Q$ 流體的體積積流量，單位是 $[\text{m}^3 \cdot \text{s}^{-1}]$
* $r$ 為管徑，單位是 $[\text{m}]$
* $\Delta p$ 為管道兩端壓力差，單位是 $[\text{Pa}]$
* $\mu$ 為流體黏滯係數，單位是 $[\text{Pa} \cdot \text{s}]=[\text{Nt} \cdot \text{s}\cdot \text{m}^{-2}]$
* $l$ 為管道長度，單位是 $[\text{m}]$

1. 以人類為例，若血管發生病變導致半徑縮小 20%（即剩餘 80% 半徑），則心臟需增加多少比例的「血壓差」才能維持原有的氧氣輸送量？

2. 在溫度 20°C 時，水的黏滯係數 $\mu_{H_2O}$ 約為空氣 $\mu_{air}$ 的 55倍，且水中的氧氣濃度 $[O_2]$ 比空氣少約 40倍。若要維持相同的氧氣輸送量與管道幾何（半徑、長度），以空氣或水輸送，兩者需要的壓力差倍數關係為何？哪一種流體輸送較省力？

3. 承上題，若兩系統的「壓力差」與「氧氣輸送量」皆相同，兩個流體系統的管道半徑倍數差異為何？哪一個流體輸送需要較粗的管道？

4. 請討論：生物系統為何多數使用「液體管道」作為生命循環系統？（請先論述為何基於前述計算，這個問題是成立/值得探討的，再給出 3 個可能理由）

## 1.1

設原半徑為 $r$，原壓力差為 $\Delta p$，阻塞後的半徑 $r' = 0.8r$，新壓力差為 $\Delta p'$，由於氧氣輸送量 $J_{O_2}$ 維持不變，且流體與 $l$ 不變。

$$\begin{gather*}
[O_2] \frac{\pi r^4 \Delta p}{8 \mu l} &= J_{O_2} =& [O_2] \frac{\pi (r')^4 \Delta p'}{8 \mu l} \\
r^4 \Delta p &=& (r')^4 \Delta p' \\
\Delta p' &=& \left(\frac{r}{0.8r}\right)^4 \Delta p \\
&=& \left(\frac{1}{0.8}\right)^4 \Delta p \\
&\approx&  2.441 \Delta p
\end{gather*}$$

心臟需要將血壓差提高到原來的 2.441倍，因此需增加的血壓差比例為 $(2.441 - 1) \times 100\% = \mathbf{144.1\%}$。



## 1.2

在 $J_{O_2}, l, r$ 皆相同的情況下，壓力差與 $\frac{\mu}{[O_2]}$ 成正比。

$$\begin{gather*}
\frac{\Delta p_{\text{water}}}{\Delta p_{\text{air}}} &=& \frac{J_{O_2} \frac{8 \mu_{\text{water}} l}{\pi r^4 [O_2]_{\text{water}}}}{J_{O_2} \frac{8 \mu_{\text{air}} l}{\pi r^4 [O_2]_{\text{air}}}}\\
&=&\left(\frac{\mu_{\text{water}}}{\mu_{\text{air}}}\right) \times \left(\frac{[O_2]_{\text{air}}}{[O_2]_{\text{water}}}\right) \\
&=& 55 \times 40 \\
&=&\mathbf{2200}
\end{gather*}$$


以水為基底的液體輸送，需要的壓力差是空氣的 2200倍。因此，以空氣輸送較為省力。

## 1.3

在 $J_{O_2}, l, \Delta p$ 皆相同的情況下，$r^4$ 與 $\frac{\mu}{[O_2]}$ 成正比。

$$\begin{gather*}
\frac{ r^4_{\text{water}}}{ r^4_{\text{air}}} &=& \frac{J_{O_2} \frac{8 \mu_{\text{water}} l}{\pi \Delta p [O_2]_{\text{water}}}}{J_{O_2} \frac{8 \mu_{\text{air}} l}{\pi \Delta p [O_2]_{\text{air}}}}\\
&=&\left(\frac{\mu_{\text{water}}}{\mu_{\text{air}}}\right) \times \left(\frac{[O_2]_{\text{air}}}{[O_2]_{\text{water}}}\right) \\
&=& 55 \times 40 \\
&=&\mathbf{2200} \\
\frac{r_{\text{water}}}{r_{\text{air}}} &=& (2200)^{1/4} \\
&\approx& \mathbf{6.85}
\end{gather*}$$

液體系統需要的管道半徑約為空氣系統的 6.85倍。液體輸送需要較粗的管道

## 1.4

1. 演化起源與內部環境恆定： **生命起源於海洋** ，早期細胞適應了在水溶液中進行生化反應。液體循環系統在生物體內創造了一個微型的「內部海洋」， **維持了細胞所需的滲透壓與離子平衡**。

2. 多功能溶劑與運輸載體：液體（水）是極佳的極性溶劑，血液不僅輸送氧氣（可利用紅血球中的血紅素來大幅突破純水的溶氧限制），還能同時溶解並輸送養分（葡萄糖、氨基酸）、荷爾蒙、免疫細胞，並帶走細胞代謝產生的水溶性廢物（尿素、乳酸），這是純氣體管道無法做到的。

3. 不可壓縮性與結構支撐：液體幾乎不可壓縮。在具有血壓的封閉血管系統中，液體可以提供流體靜力骨架（Hydrostatic skeleton）的 **支撐作用** ，並且在生物體受到外界壓力或重力變化時， **液體管路較不易像氣體管路那樣被輕易壓扁而導致循環中斷** 。

# Question 2

1. 請簡述 Navier-Stokes 方程式背後的物理核心概念，並概述其推導步驟。

2. 在推導過程中，為何「本構方程式 (Constitutive equation)」是不可或缺的？

3. 在數值流體力學中，為何質量或能量的守恆方程式多以「通量形式 (Flux form)」出現？

4. 在可壓縮流體（如大氣）中，若探討跟隨流體運動的古典守恆變數，應使用「密度」還是「混合比」？為什麼？

## 2.1

**物理核心：牛頓第二運動定律 ($F = ma$) 在流體上的應用。**

Navier-Stokes equation 本質上就是流體的 $\mathbf{F}=m\mathbf{a}$ ，只是流體會變形、會流動，所以我們不能只看一個固定的質點，而是看一個arbitrary volume

**推導步驟**

* **Step 1: 描述加速度 ($\mathbf{a}$)** 
加速度有兩種來源：一是 **隨時間的局部變化** $\frac{\partial \mathbf{v}}{\partial t}$，二是 **因為位置改變而帶來的變化**，即平流項 $(\mathbf{v} \cdot \nabla)\mathbf{v}$。

  * $\mathbf{a}$ 是3維加速度向量
  * $\mathbf{v}$ 是3維速度向量

  對流體而言，跟隨氣塊的總加速度寫為：

$$\frac{D\mathbf{v}}{Dt} = \frac{\partial \mathbf{v}}{\partial t} + (\mathbf{v} \cdot \nabla)\mathbf{v}$$




* **Step 2: 盤點考慮的作用力 ($\mathbf{F}$)**
作用在流體微小控制體積上的力有兩種：
1. **體積力 (Body forces)**：超距力，通常只考慮有重力 $\rho \mathbf{g}$ 有時考慮科氏力
   * $\rho \mathbf{g}$ 是3維重力向量
2. **表面力 (Surface forces)**：相鄰流體層互相推擠或摩擦的力，這需要用**應力張量 (Stress Tensor) $\mathbf{\sigma}_{3 \times 3}$** 來描述，表面的淨作用力即為應力張量的散度 $\nabla \cdot \sigma$。
   * $\mathbf{\sigma}_{3 \times 3}$ 是外力對表面作用的應力張量 $(3 \times 3)$矩陣
   * $\nabla \cdot \boldsymbol{\sigma} = \begin{bmatrix} \frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \end{bmatrix} \begin{bmatrix} \sigma_{xx} & \sigma_{xy} & \sigma_{xz} \\ \sigma_{yx} & \sigma_{yy} & \sigma_{yz} \\ \sigma_{zx} & \sigma_{zy} & \sigma_{zz} \end{bmatrix} = \begin{bmatrix} f_x & f_y & f_z \end{bmatrix}$


* **Step 3: 組合出 柯西動量方程式（Cauchy momentum equation）**
把質量（其實是密度 $\rho$）、加速度和力組合起來：

$$\rho \frac{D\mathbf{v}}{Dt} = \underbrace{\nabla \cdot \mathbf{\sigma}_{3 \times 3}}_{\text{表面力}} + \underbrace{\rho \mathbf{g}}_{\text{體積力}}$$


## 2.2

**簡單來說，本構方程式 (Constitutive Equation)是確定我們想考慮的表面力，並且提供一些額外的假設來達到數學上closure的條件**

而大氣科學所使用的 **本構方程式** 考慮的
* **表面力:** 氣壓梯度力、黏滯力
* **額外的假設** 【正向應力 (Normal Stresses) —— 矩陣的對角線】 和 【剪應力 (Shear Stresses) —— 矩陣的非對角線】
  * $\sigma_{xx} = -p + 2\mu \frac{\partial u}{\partial x} - \frac{2}{3}\mu (\nabla \cdot \mathbf{v})$
  * $\sigma_{yy} = -p + 2\mu \frac{\partial v}{\partial y} - \frac{2}{3}\mu (\nabla \cdot \mathbf{v})$
  * $\sigma_{zz} = -p + 2\mu \frac{\partial w}{\partial z} - \frac{2}{3}\mu (\nabla \cdot \mathbf{v})$
  * $\sigma_{xy} = \sigma_{yx} = \mu \left( \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \right)$
  * $\sigma_{xz} = \sigma_{zx} = \mu \left( \frac{\partial u}{\partial z} + \frac{\partial w}{\partial x} \right)$
  * $\sigma_{yz} = \sigma_{zy} = \mu \left( \frac{\partial v}{\partial z} + \frac{\partial w}{\partial y} \right)$


也就是說

$$\mathbf{\sigma}_{3 \times 3} = -p \mathbf{I} + \mu \left( \nabla \mathbf{v} + (\nabla \mathbf{v})^T - \frac{2}{3}(\nabla \cdot \mathbf{v})\mathbf{I} \right)$$



把這個本構方程式代入 Cauchy 方程式，把 $\sigma$ 換掉，才終於用出 **Navier-Stokes Equation**，沒有本構方程式系統就無法封閉、無法進行數值求解。



### 2.3 為何守恆定律的方程式以「特定形式（通量形式）」出現？

無論是質量、動量還是能量的守恆在理論與數值運算中，我們極度偏好將它們寫成 Eulerian 觀點下的**通量形式 (Flux form)** 或**守恆形式 (Conservative form)**：

$$\frac{\partial \phi}{\partial t} + \nabla \cdot (\phi \mathbf{v}) = S$$


* $\phi$ 是某個物理量
* $\nabla \cdot (\phi \mathbf{v})$ 是該物理量的通量散度
* $S$ 是源/匯

會這樣寫是因為 **[雷諾傳輸定理 (Reynolds Transport Theorem)](https://github.com/derek1403/PC-NTU/blob/main/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#leibniz-integral-rule)** 。方程式寫成散度形式的最大好處是：當我們在網格上對整個空間進行積分時，根據高斯散度定理 (Divergence Theorem)，$\iiint \nabla \cdot (\phi \mathbf{v}) dV$ 會轉化為邊界上的通量 $\iint (\phi \mathbf{v}) \cdot d\vec{A}$。
這意味著，只要沒有東西流出系統邊界，系統內部的總量（無論網格怎麼切割、內部怎麼平流）在數值計算上都會**絕對守恆**，這對於確保數值天氣預報模式跑了好幾天後不會憑空創造出大氣質量或能量來說是至關重要的數學結構。



### 4. 古典守恆變數會是「密度」還是「混合比」？

這取決於你的流體環境。如果是針對大氣科學這樣的**可壓縮流體**，答案絕對是**混合比 (Mixing ratio)**。

* **密度 ($\rho$) 為什麼不守恆？**
密度定義為質量除以體積。想像一個氣塊從地面被抬升到高空，外界氣壓變低，氣塊會膨脹（體積變大）。即便氣塊裡面的分子數量（質量）完全沒變，它的密度還是下降了。也就是說，跟隨氣塊運動時 $\frac{D\rho}{Dt} \neq 0$。
* **混合比為什麼守恆？**
混合比的定義通常是「某物質（如水氣）的質量」除以「乾空氣的質量」。因為這兩個質量在氣塊單純經歷絕熱膨脹或壓縮的過程中，都不會隨體積變化而改變，所以只要沒有發生相變（如凝結下雨），跟隨氣塊運動時，混合比是守恆的 $\frac{D(Mixing\_Ratio)}{Dt} = 0$。
