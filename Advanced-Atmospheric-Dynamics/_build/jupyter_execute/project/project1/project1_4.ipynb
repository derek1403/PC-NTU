# **四、確認內積的範數正定性**

$$(\tilde{u}, \tilde{u}) \ge 0$$

在線性代數中，一個空間要定義 **長度** 或 **距離** ，就必須引入範數（Norm）。但一個合法的範數不能隨便定義，它必須滿足三個嚴格的數學條件，其中最關鍵的第一條就是「正定性」（Positive Definiteness）。 對於任何非零的向量 $\bar{u}$，它的自我內積（或範數平方）必須嚴格大於零；只有當向量本身是零向量時，其值才等於零。
1. $(\tilde{u}, \tilde{u}) > 0 \quad (\text{if } \tilde{u} \neq \mathbf{0})$
2. $(\tilde{u}, \tilde{u}) = 0 \quad (\text{iff } \tilde{u} = \mathbf{0})$

## 假設與已知 (Assumptions & Preliminaries)



* **【已知 1】 經密度加權的大氣狀態向量（Density-weighted atmospheric state vector） ：**

  * (a)
  $$\begin{gather*} 
  \tilde{u} &\overset{\text{def}}{=}& \begin{bmatrix} \bar{\rho}^{\frac{1}{2}} u \\[1.5ex] \bar{\rho}^{\frac{1}{2}} v \\[1.5ex] \bar{\rho}^{\frac{1}{2}} w \\[1.5ex] \bar{\rho}^{\frac{-1}{2}} p' \\[1.5ex] \bar{\rho}^{\frac{-1}{2}}  \left( \rho' - \frac{p'}{c^2}  \right)\\[1.5ex] \end{bmatrix}
  \end{gather*}$$

  * (b) 實數物理場 

  $$ \tilde{u} = \tilde{u}^{*}$$

  * $\~{\mathbf{u}}$ : 大氣波動狀態向量 (Atmospheric wave state vector)
  * $U, V, W$ : 經密度平方根加權後的 $x, y, z$ 三軸方向動量型速度分量 $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-1}\cdot\text{s}^{-1}]$
  * $P$ : 經密度倒數平方根加權後的縮放氣壓微擾項 $[\text{kg}^{-\frac{1}{2}}\cdot\text{m}^{\frac{1}{2}}\cdot\text{s}^{-2}]$
  * $\Pi $ : 經密度倒數平方根加權後的非等熵度量／勢度微擾項 (Non-isentropic perturbation representative) $[\text{kg}^{\frac{1}{2}}\cdot\text{m}^{-3}]$
  * $u, v, w, p', \rho'$ : 原始微擾場的速度、氣壓與密度分量
  * $\bar{\rho }$ : 基本背景態密度 $[\text{kg}\cdot\text{m}^{-3}]$
  * $c$ : 本地音速 $[\text{m}\cdot\text{s}^{-1}]$

* **【已知 2】 能量權重空間平均內積（Energy-weighted spatially averaged inner product）  ：**

  $$ (\tilde{f}, \tilde{g}) \overset{\text{def}}{=} \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( f_1 g_1^* + f_2 g_2^* + f_3 g_3^* + \frac{1}{c^2} f_4 g_4^* + \frac{g^2}{N^2} f_5 g_5^* \right) dxdydz $$

  * $(\mathbf{\tilde{f}}, \mathbf{\tilde{g}})$ : 兩個波動狀態向量在全域空間上的內積值 (Spatially averaged inner product) [多元複合單位]
  * $f_{i}$ : 第一個測試函數向量 $\bar{\mathbf{f}}$ 的第 $i$ 個純量分量
  * $g_{i}^{*}$ : 第二個測試函數向量 $\bar{\mathbf{g}}$ 的第 $i$ 個純量分量的共軛複數 (Complex conjugate)
  * $x, y, z$ : 三維直角空間座標 $[\text{m}]$$L_x, L_y$ : 水平方向的長度與寬度 $[\text{m}]$
  * $L_{z}$ : 垂直方向的大氣總高度 $[\text{m}]$
  * $\frac{1}{L_{x}L_{y}L_{z}}$ : 空間體積歸一化因子 (Volume normalization factor) $[\text{m}^{-3}]$
  * $c^{2}$ : 本地音速平方 (Speed of sound squared) $[\text{m}^2\cdot\text{s}^{-2}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$
  * $\frac{1}{c^2} f_4 g_4^*$ : 經彈性音速權重調整後的彈性可壓縮位能項 (Compressible potential energy component)
  * $\frac{g^2}{N^2} f_5 g_5^*$ : 經靜力穩定度權重調整後的重力浮力熱力位能項 (Gravitational potential energy component)


* **【已知 3】 微擾動能 (Kinetic Energy, $E_k$)  ：** 描述氣流微擾運動所蘊含的機械動能密度
 
  * (a) 動能公式 

  $$\begin{gather*} 
  E_k &=& \frac{1}{2} \bar{\rho}(u^2 + v^2 + w^2) \\
  2 E_k&=& \bar{\rho}(u^2 + v^2 + w^2) 
  \end{gather*}$$

  * (b) 正定性（能量恆大於或等於零）

  $$E_k \ge 0$$

  * $E_{k}$ : 大氣微擾動能密度 (Perturbation kinetic energy density) $[\text{J}\cdot\text{m}^{-3}],[\text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-2}]$
  * $\bar{\rho }$ : 基本背景態密度 (Basic state density) $[\text{kg}\cdot\text{m}^{-3}]$
  * $u, v, w$ : 速度場在三軸方向的微擾分量 (Perturbation velocity components) $[\text{m}\cdot\text{s}^{-1}]$


* **【已知 4】 彈性能 / 聲學位能 （Elastic Energy / Acoustic Potential Energy, $E_{e}$  ：** 大氣受到聲波壓縮或膨脹時，因流體彈性所儲存的位能密度

  * (a) 彈性能公式

  $$\begin{gather*} 
  E_e &=& \frac{1}{2\bar{\rho}}\frac{p'^2}{c^2} \\
  2 E_e &=& \frac{1}{\bar{\rho}}\frac{p'^2}{c^2}
  \end{gather*}$$

  * (b) 正定性
  
  $$E_e \ge 0$$

  * $E_{e}$ : 大氣微擾彈性能密度 (Perturbation elastic energy density) $[\text{J}\cdot\text{m}^{-3}]$
  * $p'$ : 氣壓的時空微擾項 (Pressure perturbation variable) $[\text{Pa}]$
  * $c$ : 本地音速 (Local speed of sound) $[\text{m}\cdot\text{s}^{-1}]$


* **【已知 5】 熱壓能 / 浮力位能 (Thermobaric / Internal Energy, $I$)  ：** 大氣微塊因偏離等熵狀態或受到重力浮力效應（層結擾動）時所蘊含的熱力學內部能與浮力位能密度 

  * (a) 熱壓能公式 
  
  $$\begin{gather*} 
  I &=& \frac{g^2 \left(\rho' - \frac{p'}{c^2}\right)^2}{2\bar{\rho} N^2} \\
  2 I &=& \frac{g^2 \left(\rho' - \frac{p'}{c^2}\right)^2}{\bar{\rho} N^2}
  \end{gather*}$$

  * (b) 正定性（當系統靜力穩定，即 $N^2 > 0$ 時恆成立）
  
  $$I \ge 0$$

  * $I$ : 大氣微擾熱壓能密度 (Perturbation thermobaric energy density) $[\text{J}\cdot\text{m}^{-3}]$
  * $\rho'$ : 密度的時空微擾項 (Density perturbation variable) $[\text{kg}\cdot\text{m}^{-3}]$
  * $g$ : 地表重力加速度常數 $g \approx 9.80665 \text{ m}\cdot\text{s}^{-2}$
  * $N^{2}$ : 浮力頻率平方 / 班特-外薩拉頻率平方 (Brunt-Väisälä frequency squared) $[\text{s}^{-2}]$

## 證明

$$\begin{gather*} 
(\tilde{u}, \tilde{u}) &\overset{\text{已知 2}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( \tilde{u}_1 \tilde{u}_1^* + \tilde{u}_2 \tilde{u}_2^* + \tilde{u}_3 \tilde{u}_3^* + \frac{1}{c^2}\tilde{u}_4 \tilde{u}_4^* + \frac{g^2}{N^2}\tilde{u}_5 \tilde{u}_5^* \right) dxdydz \\
&\overset{\text{已知 1(b)}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( \tilde{u}_1^2 + \tilde{u}_2^2  + \tilde{u}_3^2 + \frac{1}{c^2}\tilde{u}_4^2 + \frac{g^2}{N^2}\tilde{u}_5^2 \right) dxdydz \\
&\overset{\text{已知 1(a)}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( \left(\bar{\rho}^{1/2}u\right)^2 + \left(\bar{\rho}^{1/2}v\right)^2 + \left(\bar{\rho}^{1/2}w\right)^2 + \frac{1}{c^2}\left(\bar{\rho}^{-1/2}p'\right)^2 + \frac{g^2}{N^2}\left(\bar{\rho}^{-1/2}\left(\rho' - \frac{p'}{c^2}\right)\right)^2 \right) dxdydz \\
&=& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( \bar{\rho}(u^2 + v^2 + w^2) + \frac{1}{\bar{\rho}}\frac{p'^2}{c^2} + \frac{g^2}{N^2 \bar{\rho}}\left(\rho' - \frac{p'}{c^2}\right)^2  \right) dxdydz \\
&\overset{\text{已知 3(a),已知 4(a),已知 5(a)}}{=}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( E_k + E_e + I  \right) dxdydz \\
&\overset{\text{已知 3(b),已知 4(b),已知 5(b)}}{\ge}& \frac{1}{L_x L_y L_z} \int_{0}^{L_z} \int_{0}^{L_y} \int_{0}^{L_x} \left( 0 + 0 + 0  \right) dxdydz \\
&\ge& 0 \\
\end{gather*}$$


且 $(\tilde{u}, \tilde{u}) = 0$ 的唯一條件是系統完全處於靜止的背景平衡態（無任何速度、氣壓與位溫擾動），這完美滿足了線性代數中「能量範數 (Energy Norm)」的嚴格定義 。
