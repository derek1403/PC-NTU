# Matsuno (1966) 學習筆記
## Hermite 函數與赤道波的數學物理連結


## 核心思想

- Matsuno 發現**赤道大氣動力學** ≡ **量子諧振子 (Harmonic Oscillator)** — 數學結構完全等價
- 科氏力隨緯度變化 → 形成天然「勢阱」→ 波動能量被困在赤道 → **赤道波導 (Equatorial Waveguide)**


## 1. Weber 方程式

$$\frac{d^2\hat{v}}{dy^2} + \left( \lambda - y^2 \right)\hat{v} = 0$$

- $\lambda = \omega^2 - k^2 + k/\omega$（特徵值）
- 與[**一維量子諧振子薛丁格方程式**](https://derek1403.github.io/PC-NTU/Modern-Physics/_build/html/lecture/05_Quantum_Mechanics/quantum_mechanics.html#schrodingers-equation-steady-state-form)形式相同
- $-y^2$ 項的物理意義對照：

| 量子力學 | 赤道波 |
|---|---|
| 位能勢阱，困住電子 | 科氏參數 $\beta y$，困住波動能量 |


## 2. 邊界條件 → 高斯衰減包絡

- 物理要求：$y \to \pm\infty$ 時，$\hat{v} \to 0$（能量不向兩極發散）
- 大 $y$ 極限下方程式退化：$\hat{v}'' \approx y^2 \hat{v}$，解為 $e^{\pm y^2/2}$
- 只能取 $e^{-y^2/2}$（發散解捨棄）
- Matsuno 的關鍵代換：

$$\hat{v}(y) = V(y)\, e^{-y^2/2}$$

- $e^{-y^2/2}$ = **高斯衰減包絡線**
  - 決定赤道波能量南北遞減的空間尺度
  - 對應真實物理長度 = **赤道羅斯貝變形半徑**


## 3. Hermite 微分方程式 & 量子化條件

代入後 $V(y)$ 滿足：

$$V'' - 2yV' + (\lambda - 1)V = 0$$

### 為什麼 $\lambda$ 不能任意取值？

- 任意 $\lambda$ → $V(y)$ 展開為無窮級數 → 增長速度 $\sim e^{y^2}$ → 抵消高斯衰減 → 物理崩潰
- 解救條件：無窮級數必須**截斷**為有限次多項式

### 量子化條件

$$\lambda - 1 = 2n, \quad n = 0, 1, 2, \dots$$

- $\lambda$ 只能取**離散**數值
- 這正是**色散關係式的數學起源**
- 赤道波的頻率 $\omega$ 與波數 $k$ 被「量子化」，如同量子態

## 4. Hermite Polynomial vs. Hermite Function

### Hermite Polynomial $H_n(y)$（埃爾米特多項式）

- 即 $V(y)$ 的解
- 前幾個：$H_0 = 1,\quad H_1 = 2y,\quad H_2 = 4y^2 - 2, \dots$
- **物理意義：**
  - $n$ = 多項式最高次方 = 南北方向**穿越零點（節點）的次數**
  - $n=0$：南北風全程同號
  - $n=1$：南北風在赤道兩側反號

### Hermite Function $\psi_n(y)$（埃爾米特函數）

$$\psi_n(y) = e^{-y^2/2} H_n(y)$$

- 多項式 × 高斯衰減 = **真實的空間特徵解**
- 關鍵數學性質：在 $(-\infty, +\infty)$ 上**完備且正交**：

$$\int_{-\infty}^{\infty} \psi_m(y)\, \psi_n(y)\, dy = 0 \quad (m \neq n)$$

- 構成赤道 $\beta$ 平面上的正交基底


## 5. 正交完備性的物理意義（最終結論）

- 任何複雜的**熱帶天氣擾動 / 對流加熱**，都可以被完美**投影分解**成無限多個獨立赤道波的線性疊加：
  - 羅斯貝波 (Rossby Wave)
  - 重力波 (Gravity Wave)
  - 克耳文波 (Kelvin Wave)
- 各模態互相獨立、不干擾
- 奠定**現代熱帶氣象學**的理論基礎


## 一句話總結

> 赤道科氏力的 $\beta y$ 結構 → 數學上等同量子勢阱 → 邊界條件強迫波動「量子化」→ 色散關係離散化 → Hermite 函數成為熱帶大氣的自然正交基底。
