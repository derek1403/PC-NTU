# Matsuno (1966) vs. Gill (1980)
## 長波近似的核心差異

## 0. 色散 (Dispersion) 快速複習

- **色散波：** $\omega$ 與 $k$ 非線性正比 → 相速度 $c = \omega/k$ 隨波長改變 → 波包「散開」
- **非色散波：** $c$ 與 $k$ 無關（如淺水重力波 $c = \sqrt{gH}$）→ 波形不變形
- 關鍵量：**群速度** $c_g = \partial\omega/\partial k$（能量傳遞速度），色散波中 $c \neq c_g$

> **類比：** 池塘丟石頭 → 長波快、短波慢 → 波紋「散開」= 色散。真空中的光 = 非色散。


## 1. 兩者目標根本不同

| | Matsuno (1966) | Gill (1980) |
|---|---|---|
| **想問的問題** | 赤道大氣裡有哪些**自由波**？ | 持續加熱後大氣**穩態響應**長什麼樣？ |
| **研究對象** | 瞬變波動 (Transient waves) | 定態環流 (Steady-state response) |
| **代表現象** | 颱風生成、渦旋羅斯貝波 | 沃克環流、ENSO、季風 |

## 2. 數學近似的差異

### 完整色散關係（出發點相同）
$$\omega^2 - k^2 - \frac{k}{\omega} - 1 = 2n$$

### Matsuno：只砍 $\omega^2$（低頻近似）
- 假設：$\omega \ll 1$（低頻運動）→ 忽略 $\omega^2$
- 結果：

$$\omega = \frac{-k}{k^2 + 2n + 1}$$

- **$k^2$ 保留** → 羅斯貝波仍為**色散波**
  - 長波（小 $k$）：相速度快
  - 短波（大 $k$）：相速度極慢，近乎停滯

### Gill：同時砍 $\omega^2$ 與 $k^2$（嚴格長波近似）
- 假設：$L_x \gg L_y$（東西尺度遠大於南北）→ $k \to 0$
- 動力學操作：南北向動量方程式中 $\partial v/\partial t$ 直接砍掉 → 強迫地轉平衡
  
$$yu + \frac{\partial \phi}{\partial y} = 0$$

- 結果：

$$\omega = \frac{-k}{2n + 1}$$

- **$k^2$ 消失** → $\omega \propto k$ → 羅斯貝波變為**非色散波**
  - 給定模態 $n$，所有波長一律以 $c = -1/(2n+1)$ 向西傳播


## 3. 核心概念：$k^2$ 項的命運

> **"The fate of the $k^2$ term"** 是理解兩者差異的鑰匙。

- **Matsuno 保留 $k^2$** → 物理完整，但數學複雜，難以求解析解
- **Gill 消滅 $k^2$** → 物理略失真，但換來數學優雅：
  - 可用**特徵線法 (Method of Characteristics)** 求解
  - 畫出著名的 **Gill Pattern**：
    - 加熱源**東側**：赤道克耳文波下沉區
    - 加熱源**西側**：一對羅斯貝波氣旋式環流（雙胞胎氣旋）


## 4. 總結比較表

| 比較項目 | Matsuno (1966) | Gill (1980) |
|---|---|---|
| 動力學假設 | 僅忽略 $\omega^2$ | 忽略 $\omega^2$ 與 $\partial v/\partial t$ |
| 保留 / 忽略 $k^2$ | **保留** | **忽略** |
| 色散關係 | $\omega = \dfrac{-k}{k^2+2n+1}$ | $\omega = \dfrac{-k}{2n+1}$ |
| 波的特性 | **色散波**（短波慢） | **非色散波**（同速） |
| 適用場景 | 瞬變天氣、颱風 | 大尺度氣候、ENSO |


## 一句話總結

> Matsuno 問「波怎麼走」，所以必須保留色散性；Gill 問「加熱後穩態場長什麼樣」，所以用長波近似換取解析解的可能——兩人各自犧牲了不同的物理細節，萃取出不同的真理。
