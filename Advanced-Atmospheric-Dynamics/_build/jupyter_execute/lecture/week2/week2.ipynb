## 1 連續方程式與擴散守恆 (Continuity Equation & Diffusion)

### 1.1 基本物理量與假設

在推導連續方程式前，我們必須確立幾個核心概念：
* **物質體積 (Material Volume, $V_m$)**：隨著流體運動的封閉體積，其內部質量守恆。
* **連續體假設**：流體密度 $\rho$ 必須能在宏觀上被平滑定義。這要求觀測的物理尺度 $L$ 必須遠大於分子的平均自由徑 $l$（即 $l \ll L$）。
* **任意體積 (Arbitrary Volume)**：方程式對空間中任何一個固定大小的控制體積皆成立。

> **💡 物理史類比：富蘭克林的池塘實驗**
> 班傑明·富蘭克林曾將一茶匙的油滴入池塘，觀察到油膜擴散至約半英畝大就不再擴散。假設油膜最終厚度為單一分子層，利用巨觀的體積與面積關係 ($V = A \cdot h$)，即可估算出油分子的直徑。這是一個利用巨觀連續體現象去推測微觀尺度的經典例子，與我們定義 $\rho$ 的概念相呼應。

### 1.2 定義「一般化」的連續方程式

$$\frac{\partial C}{\partial t} + \nabla \cdot \vec{J}_C = 0$$

* $C$ (Concentration / Content)：代表我們想討論的「某種物理量的空間分佈」，它可以是質量密度 ($\rho$)、熱能密度 ($e$)、電荷密度、甚至是機率密度假設 $C$ 的單位是 $[C]$
* $\vec{J}_C$ (Flux density)：代表這個物理量 $C$ 在空間中流動的「通量密度」， $\vec{J}_C$ 的單位是 $[C] \cdot \frac{m}{s}$
  * $\vec{J}_C$ 的普遍物理意義永遠是：「每單位時間 ($s$)、穿過每單位截面積 ($m^2$) 的總物理量 ($[C] \cdot m^3$)」，這就是「通量 (Flux)」的本質


#### **應用一：流體連續方程式 (Continuity Equation for Fluid)**

給定 $C$ 是流體密度 $\rho$、$\vec{J}_C = \rho \mathbf{u}$，即得質量守恆的連續方程式。此式**已於 week1 完整證明，此處直接引用不再重證**：

* **【已知】[連續方程式 (Continuity Equation)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#continuity-equation)：**

  $$\frac{\partial \rho}{\partial t } + \nabla \cdot (\rho \mathbf{u}) = 0$$

  * $\rho$ : 流體密度 (Density) $[\text{kg} \cdot \text{m}^{-3}]$
  * $\mathbf{u}$ : 三維速度向量 (Velocity vector) $(u, v, w)$ $[\text{m} \cdot \text{s}^{-1}]$
  * $t$ : 時間 (Time) $[\text{s}]$

#### **應用二：熱傳導方程式 (Heat Conduction Equation)**

##### 目標 (Goal)

證明在無巨觀流體運動、無內部熱源的均質介質中，溫度場滿足熱傳導方程式（擴散方程式）：

$$\frac{\partial T}{\partial t} = \alpha \nabla^2 T$$

* $T$ : 絕對溫度 (Absolute temperature) $[\text{K}]$
* $t$ : 時間 (Time) $[\text{s}]$
* $\alpha$ : 熱擴散係數 (Thermal diffusivity) $[\text{m}^{2}\cdot \text{s}^{-1}]$
* $\nabla^2$ : 拉普拉斯算子 (Laplacian) $[\text{m}^{-2}]$

![](./pic/Fourier's_law_of_heat_conduction.png)

##### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】一般化連續方程式 (Generalized Continuity Equation)：** 即本章 1.2 所定義的守恆律骨架。

  $$\frac{\partial C}{\partial t} + \nabla \cdot \vec{J}_C = 0$$

  * $C$ : 廣義守恆量密度 (Conserved content density) $[C \cdot \text{m}^{-3}]$
  * $\vec{J}_C$ : 廣義通量密度 (Flux density) $[C \cdot \text{m}^{-2} \cdot \text{s}^{-1}]$

* **【已知 2】傅立葉熱傳導定律 (Fourier's Law of Heat Conduction)：** 熱量從高溫流向低溫，大小與溫度梯度成正比；負號代表熱流方向與溫度梯度相反。

  $$\vec{J}_q \overset{\text{def}}{=} -k \nabla T$$

  * $\vec{J}_q$ : 熱通量密度 (Heat flux density) $[\text{J} \cdot \text{m}^{-2} \cdot \text{s}^{-1}] = [\text{W} \cdot \text{m}^{-2}]$
  * $k$ : 熱傳導係數 (Thermal conductivity) $[\text{W} \cdot \text{m}^{-1} \cdot \text{K}^{-1}]$
  * $\nabla T$ : 溫度梯度 (Temperature gradient) $[\text{K} \cdot \text{m}^{-1}]$

* **【定義 1】體積熱能密度 (Volumetric Thermal Energy Density)：** 把廣義守恆量 $C$ 具體化為「單位體積內所含的熱能」。

  $$e \overset{\text{def}}{=} \rho c_p T$$

  * $e$ : 體積熱能密度 (Volumetric thermal energy density) $[\text{J} \cdot \text{m}^{-3}]$
  * $\rho$ : 密度 (Density) $[\text{kg} \cdot \text{m}^{-3}]$
  * $c_p$ : 等壓比熱 (Specific heat at constant pressure) $[\text{J} \cdot \text{kg}^{-1} \cdot \text{K}^{-1}]$
  * $T$ : 絕對溫度 (Absolute temperature) $[\text{K}]$

* **【定義 2】熱擴散係數 (Thermal Diffusivity)：** 「導熱能力」除以「儲熱能力」，量度溫度訊號在介質中擴散的快慢。

  $$\alpha \overset{\text{def}}{=} \frac{k}{\rho c_p}$$

  * $\alpha$ : 熱擴散係數 (Thermal diffusivity) $[\text{m}^{2}\cdot \text{s}^{-1}]$
  * $k$（熱傳導係數）：代表材料「導熱的能力」。金屬的 $k$ 很大，木頭的 $k$ 很小。
    單位 $[\text{W} \cdot \text{m}^{-1}\cdot \text{K}^{-1}]$ 的白話：材料厚度為 $1$ 公尺、兩端溫差為 $1$ 度時，每秒鐘有多少焦耳的熱能穿過 $1$ 平方公尺的面積。
  * $\rho c_p$（體積熱容量）：代表材料「儲存熱能的能力」。水的值很大，空氣的值很小。
    單位 $[\text{kg} \cdot \text{m}^{-3} \cdot \text{J} \cdot \text{kg}^{-1}\cdot \text{K}^{-1}] = [\text{J}\cdot \text{m}^{-3}\cdot \text{K}^{-1}]$ 的白話：要讓 $1$ 立方公尺的物體升溫 $1$ 度，需要吃掉多少熱量。

* **【假設 1】無巨觀流體運動 (No Macroscopic Motion)：** 探討的是固體，或完全靜止沒有對流 (Advection) 的流體，熱能只能靠分子碰撞（傳導）傳遞。故廣義通量只剩傳導項：

  $$\vec{J}_C = \vec{J}_q$$

* **【假設 2】均質且不可壓縮 (Homogeneous & Incompressible)：** 材料均勻，$\rho$、$c_p$、$k$ 在時空中皆為常數、不隨溫度變化，故可自微分算子中提出。

  * (a) 三個材料參數皆為常數：

    $$\frac{\partial \rho}{\partial t} = \frac{\partial c_p}{\partial t} = \frac{\partial k}{\partial t} = 0 , \quad \nabla \rho = \nabla c_p = \nabla k = 0$$

  * (b) 時間項提出常數：

    $$\frac{\partial (\rho c_p T)}{\partial t} = \rho c_p \frac{\partial T}{\partial t}$$

  * (c) 空間項提出常數：

    $$\nabla \cdot (-k \nabla T) = -k \nabla \cdot (\nabla T) = -k \nabla^2 T$$

* **【假設 3】無內部熱源 (No Internal Heat Source)：** 系統內部沒有化學反應或輻射加熱（無核反應、無微波加熱），故內部生熱率為零，守恆律右端維持 $0$。

  $$\begin{gather*}
  \dot{Q}_{\text{source}} &=& 0 \\
  \frac{\partial e}{\partial t} + \nabla \cdot \vec{J}_q &=& \dot{Q}_{\text{source}} = 0
  \end{gather*}$$

  * $\dot{Q}_{\text{source}}$ : 內部生熱率 (Internal heat source rate) $[\text{W} \cdot \text{m}^{-3}]$

##### proof

$$\begin{gather*}
0 &\overset{\text{已知 1}}{=}& \frac{\partial C}{\partial t} + \nabla \cdot \vec{J}_C \\
0 &\overset{\text{假設 1,假設 3}}{=}& \frac{\partial e}{\partial t} + \nabla \cdot \vec{J}_q \\
0 &\overset{\text{定義 1,已知 2}}{=}& \frac{\partial (\rho c_p T)}{\partial t} + \nabla \cdot (-k \nabla T) \\
0 &\overset{\text{假設 2(b)(c)}}{=}& \rho c_p \frac{\partial T}{\partial t} - k \nabla^2 T \\
\rho c_p \frac{\partial T}{\partial t} &=& k \nabla^2 T \\
\frac{\partial T}{\partial t} &=& \frac{k}{\rho c_p} \nabla^2 T \\
\frac{\partial T}{\partial t} &\overset{\text{定義 2}}{=}& \alpha \nabla^2 T
\end{gather*}$$

**📐 數學與物理的現實落差：**
這個偏微分方程是建立在泰勒展開式中極限 $\Delta x \to 0$ 的數學基礎上。但在實際大氣觀測中，我們儀器的解析度無法做到真正的 $\Delta x \to 0$。此外，大氣中的亂流（Turbulence）不僅存在大尺度能量傳遞至小尺度（Energy cascade），也存在小尺度擾動發展成大尺度結構的現象（如斜壓不穩定、正壓不穩定），這涉及到從線性行為過渡到非線性行為的複雜動力學。

## 2. 動量方程式與黏滯耗散 (Momentum Equation & Viscous Dissipation)

### 2.1 積分形式的動量方程式 (Integral Form)

納維-斯托克斯方程式的積分形式描述控制體積 $V_m$ 內動量的變化率等於所受外力總和。此式**已於 week1 由雷諾傳輸定理完整證明，此處直接引用不再重證**：

* **【已知】[動量方程式 (Momentum Equation, Navier-Stokes)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#momentum-equation-navier-stokes)：**

  $$\begin{gather*}
  \frac{d}{dt} \int_{V_m} \rho \mathbf{u} \, dV &=& \int_{V_m} \rho \mathbf{g} \, dV - \oint_{\partial V_m} P \, d\mathbf{A} + \int_{V_m} \mu \nabla^2 \mathbf{u} \, dV \\
  \frac{\partial \mathbf{u}}{\partial t } + (\mathbf{u} \cdot \nabla) \mathbf{u} &=& \mathbf{g} - \frac{1}{\rho} \nabla P + \frac{\mu}{\rho} \nabla^2 \mathbf{u}
  \end{gather*}$$

  * $\mathbf{u}$ : 三維速度向量 (Velocity vector) $(u, v, w)$ $[\text{m} \cdot \text{s}^{-1}]$
  * $\mathbf{g}$ : 重力加速度向量 (Gravitational acceleration) $(0, 0, -g)$ $[\text{m} \cdot \text{s}^{-2}]$
  * $P$ : 氣壓 (Pressure) $[\text{Pa}] = [\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}]$
  * $\rho$ : 流體密度 (Density) $[\text{kg} \cdot \text{m}^{-3}]$
  * $\mu$ : 動力黏滯係數 (Dynamic viscosity) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$
  * $V_m$ : 物質體積 (Material volume) $[\text{m}^{3}]$


### 2.2 黏滯力與拉普拉斯算子 ($\nabla^2$)
方程式中的最後一項($\int_{V_m} \mu \nabla^2 \mathbf{u} \, dV$)是由流體內摩擦引起的黏滯力，在物理上 $\mu \frac{\partial u_i}{\partial x_j}$ 代表 $u_i$ 動量在 $x_j$ 方向上的傳送（動量通量），其散度即為 $\mu \nabla^2 \mathbf{u}$。

**這在做什麼：** 拉普拉斯算子 $\nabla^2$ 在物理方程式中通常扮演「平滑化」（Smoothing）的角色，它是一個「消耗項」（Sink）。
* **數學推導視角**：若我們將速度場轉換到傅立葉空間（Fourier space），對空間微分兩次會產生 $-k^2$（$k$ 為波數），這意味著高頻擾動（波數 $k$ 很大，即劇烈的空間變化）會被快速衰減。它會將空間中的極大值壓低、極小值拉高。
* **類比**：如果把世界看作一個動力系統，$\frac{\partial \mathbf{u}}{\partial t} = \nu \nabla^2 \mathbf{u}$ 就是一個「抹平一切差異」的方程式（黑色玩笑地說，這是某種物理學上的「$\mathbb{C}\mathbb{N}-\dot{8}\dot{9}\mathit{64}$大同世界」方程式）。

## 3. 亂流與渦流擴散 (Turbulence & Eddy Diffusivity)

### 3.1 擾動變異數的耗散 (Dissipation of Scalar Variance)

#### 目標 (Goal)

在多數的流體力學基礎推導中，$q$ 代表「任意一種具備擴散性質的純量」（例如：溫度 $T$、比濕 $r$、甚至是某種化學濃度），積分 $\int \frac{1}{2} {q'}^2 \, dV$ 在統計數學上真正的名字叫做 **「擾動變異數 (Scalar Variance)」**，它量度的是這個純量場「不均勻的程度」（例如一塊冷空氣與一塊熱空氣的溫差起伏）。[想要看嚴格擾動動能的耗散證明請到這裡](123)。

證明擾動變異數（代表亂流動能 Kinetic Energy, KE）的耗散率正比於梯度平方，且必然為負（單調衰減）：

$$\frac{d}{dt} \int \frac{{q'}^2}{2} \, dV = - \nu \int |\nabla q'|^2 \, dV$$

* $q$ : 具擴散性質的純量 (Diffusive scalar) $[q]$
* $\bar{q}$ : 純量的平均態 (Mean state) $[q]$
* $q'$ : 純量的擾動量 (Perturbation) $[q]$
* $\mathbf{u}$ : 三維速度向量 (Velocity vector) $[\text{m} \cdot \text{s}^{-1}]$
* $\nu$ : 分子擴散係數 (Molecular diffusivity) $[\text{m}^{2}\cdot \text{s}^{-1}]$
* $V$ : 積分體積 (Integration volume) $[\text{m}^{3}]$
* $t$ : 時間 (Time) $[\text{s}]$

#### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】平流擴散方程式 (Advection-Diffusion Equation)：** 純量 $q$ 的守恆律。

  $$\frac{\partial q}{\partial t} + \mathbf{u} \cdot \nabla q = \nu \nabla^2 q$$

  * $\frac{\partial q}{\partial t}$ : 局地隨時間的變化 (Local tendency) $[q \cdot \text{s}^{-1}]$
  * $\mathbf{u} \cdot \nabla q$ : 被流體速度 $\mathbf{u}$ 帶著走的平流項 (Advection) $[q \cdot \text{s}^{-1}]$
  * $\nu \nabla^2 q$ : 分子擴散項，抹平差異的力量 (Molecular diffusion) $[q \cdot \text{s}^{-1}]$

* **【已知 2】散度定理 (Divergence Theorem)：**

  $$\int_{V} \nabla \cdot \mathbf{F} \, dV = \oint_{\partial V} \mathbf{F} \cdot d\mathbf{A}$$

  * $\mathbf{F}$ : 任意可微向量場 (Vector field)
  * $d\mathbf{A}$ : 向外的法向量面積元素 (Outward area element) $[\text{m}^{2}]$

* **【已知 3】乘積律 (Product Rule)：** 把 $q' \nabla^2 q'$ 拆成一個全微分項與一個正定項。

  $$q' \nabla^2 q' = \nabla \cdot (q' \nabla q') - \nabla q' \cdot \nabla q'$$

* **【假設 1】雷諾分解 (Reynolds Decomposition)：** 在亂流中把物理量分解為平均態與擾動態，並假設擾動量純粹受擴散影響。

  $$q \overset{\text{def}}{=} \bar{q} + q'$$

* **【假設 2】背景場與線性化 (Background State & Linearization)：**

  * (a) 沒有背景風（或跟著背景風走）：

    $$\bar{\mathbf{u}} = 0$$

  * (b) 背景場很均勻（沒有梯度）：

    $$\nabla \bar{q} = 0$$

  * (c) 線性化（忽略極小項）：擾動非常微弱，兩個微小的擾動相乘可忽略：

    $$\mathbf{u}' \cdot \nabla q' \approx 0$$

* **【假設 3】封閉邊界條件 (Closed Boundary Condition)：** 積分域取週期邊界，或擾動量在無窮遠處衰減為零，故穿過邊界的擾動通量淨值為零。

  $$\oint_{\partial V} (q' \nabla q') \cdot d\mathbf{A} = 0$$

* **【推導 1】只考慮耗散的擾動方程式 (Dissipation-Only Perturbation Equation)：** 把【假設 1,2】代入【已知 1】，平流項整條消失，只剩分子擴散。

  $$\begin{gather*}
  \frac{\partial q'}{\partial t} &\overset{\text{已知 1,假設 1}}{=}& \nu \nabla^2 q' - (\bar{\mathbf{u}} + \mathbf{u}') \cdot \nabla (\bar{q} + q') \\
  &\overset{\text{假設 2(a)(b)}}{=}& \nu \nabla^2 q' - \mathbf{u}' \cdot \nabla q' \\
  &\overset{\text{假設 2(c)}}{=}& \nu \nabla^2 q'
  \end{gather*}$$

#### proof

$$\begin{gather*}
\frac{d}{dt} \int \frac{{q'}^2}{2} \, dV &=& \int q' \frac{\partial q'}{\partial t} \, dV \\
&\overset{\text{推導 1}}{=}& \nu \int q' (\nabla^2 q') \, dV \\
&\overset{\text{已知 3}}{=}& \nu \int \nabla \cdot (q' \nabla q') \, dV - \nu \int \nabla q' \cdot \nabla q' \, dV \\
&\overset{\text{已知 2}}{=}& \nu \oint_{\partial V} (q' \nabla q') \cdot d\mathbf{A} - \nu \int \nabla q' \cdot \nabla q' \, dV \\
&\overset{\text{假設 3}}{=}& - \nu \int \nabla q' \cdot \nabla q' \, dV \\
&=& - \nu \int |\nabla q'|^2 \, dV
\end{gather*}$$




**物理意義：**
這證明了**耗散是與梯度的平方 ($|\nabla q'|^2$) 成正比**的（負號代表能量正在減少），在 2D 亂流的 Energy cascade（能量串級）中系統會傾向讓梯度極小化，如果我們尋找使耗散最小化的穩態函數，就會得到拉普拉斯方程式（Laplace equation）$\nabla^2 q' = 0$。

> **💡 為什麼「變異數的耗散」可以拿來解釋「擾動動能的耗散」？（奶精攪拌類比）**
>
> 關鍵在於耗散率的被積函數是**梯度的平方** $|\nabla q'|^2$，而不是擾動本身的大小 $|q'|^2$。
>
> 把一滴奶精滴進咖啡裡：**攪拌（亂流）本身並不會直接把奶精「抹掉」**，它做的事情是把奶精拉成越來越細的絲，
> 讓奶精與咖啡的**介面面積暴增**。介面越多、越薄，介面兩側的濃度落差就被擠在越短的距離內，
> 也就是梯度 $|\nabla q'|$ 被放大到極大 ─ 於是 $\int |\nabla q'|^2 \, dV$ 幾乎要趨向無限大。
>
> 而耗散率正比於這個積分，所以**攪拌得越厲害，耗散就越猛**：
> 亂流（大尺度）負責「製造梯度」，分子擴散 $\nu$（小尺度）負責「吃掉梯度」。
> 這就是為什麼即使 $\nu$ 小得可憐，只要亂流把梯度撐得夠大，耗散依然可觀。
>
> 結果就是：攪拌 → 梯度暴增 → 耗散暴增 → 過一小段時間後 $q'$ 被抹平，咖啡變得**均勻**（$\nabla q' \to 0$，耗散停止）。
> 擾動動能 (KE) 的耗散走的是完全相同的劇本，只是把純量 $q'$ 換成速度擾動 $\mathbf{u}'$、
> 把「濃度介面」換成「速度剪切層」：亂流把大渦拉成小渦（Energy cascade），速度梯度 $|\nabla \mathbf{u}'|$ 被放大，
> 黏滯力就在最小的尺度上把動能吃成熱能。**這也是為什麼耗散總是發生在最小尺度上。**

> **💡 巨觀與微觀的類比：渦流擴散係數 $K$**
> 在地球大氣中，分子黏滯係數 $\nu$ 非常小（幾乎要趴在貼近地板的邊界層才感受得到），在大尺度氣象學中我們經常使用「渦流擴散係數 (Eddy diffusivity, $K$)」，它在數學形式上與 $\nu$ 一模一樣但物理機制不同：$\nu$ 是分子間的碰撞交換，而 $K$ 是一大塊熱空氣和一大塊冷空氣在大氣中互相翻攪（亂流）。即使自由大氣不考慮分子黏滯力 $\mu$，我們依舊能用相同的數學框架來理解大氣亂流。

## 4. 白金漢 $\pi$ 定理與雷諾數 (Buckingham $\pi$ Theorem & Reynolds Number)

### 4.1 因次分析 (Dimensional Analysis)

根據白金漢 $\pi$ 定理（Buckingham $\pi$ Theorem）：

$$\text{無因次參數數量} = \text{物理變數數量} - \text{基本單位數量}$$

假設流體所受外力 $F$ 與動力黏滯係數 $\mu$、特徵速度 $v$、特徵長度 $l$、流體密度 $\rho$ 有關。
函數關係：$F(\mu, v, l, \rho, F_{force}) = 0$
這裡有 5 個變數，基本單位有 3 個（質量 M、長度 L、時間 T）。
因此，我們可以得出 $5 - 3 = 2$ 個無因次參數，組合後得到：

$$\frac{F}{\rho v^2 l^2} = f\left(\frac{\rho v l}{\mu}\right)$$

$\frac{\rho v l}{\mu}$ 就是著名的 **[雷諾數 (Reynolds Number, $Re$)](https://derek1403.github.io/Theory_Playground/_build/html/02_Concepts/Fluid_Dynamics/Reynolds_number.html)**
* $Re = \frac{\text{慣性力}}{\text{黏滯力}} = \frac{\text{擴散時間尺度}}{\text{慣性時間尺度}}$
* 當 $Re$ 很大時，擴散時間遠大於慣性時間，意味著在短暫的慣性運動時間內，擴散效應可忽略不計

### 4.2 阻力係數圖 的物理拆解

[![](./pic/Drag_coefficient_Reynolds_number_for_sphere.png)](https://commons.wikimedia.org/wiki/File%3ASmooth_sphere_drag_coefficient.svg)

觀察摩擦力係數與雷諾數的關係圖：
* **A區 (層流，Low $Re$ domain, $Re<10^2$)：** 
  * 圖上斜率為 $y \propto 1/x$，即

    $$\begin{gather*}
    \frac{F}{\rho v^2 l^2} &\propto& \frac{\mu}{\rho v l} \\
    F &\propto& \mu v l
    \end{gather*}$$

  * 這代表在低雷諾數下（如微小顆粒或黏稠流體），阻力主要由黏滯力 $\mu$ 貢獻，密度 $\rho$ 幾乎沒有發揮作用（對應斯托克斯阻力定律 Stokes' drag）
  
* **B區 (慣性主導，$10^2 < Re < 10^5$)：**
  * 圖上為水平線 $y \sim \text{const}$，即

    $$\begin{gather*}
    \frac{F}{\rho v^2 l^2} &\sim& 1 \\
    F &\propto& \rho v^2 l^2
    \end{gather*}$$

  * 此時慣性力主導
  * *類比*：同樣射速的水槍與空氣槍，水槍造成的打擊力 ($F$) 會比空氣槍大約 1000 倍，因為 $\rho_{\text{water}} \approx 1000 \rho_{\text{air}}$，此時處於 B 區，力與密度成正比

* **C區 (亂流，High $Re$ domain, $Re > 10^5$)：**
  * 沒有明確的函數關係


### 4.3 PM2.5 在大氣中的沉降 (Settling of PM2.5)

#### 目標 (Goal)

估算 PM2.5 顆粒在大氣中的終端速度 $v_{\text{terminal}}$，並據此評估其自邊界層與對流層頂完全沉降所需的時間尺度。

* $v_{\text{terminal}}$ : 終端速度 (Terminal velocity) $[\text{m} \cdot \text{s}^{-1}]$
* $r$ : 顆粒半徑 (Particle radius) $[\text{m}]$
* $\rho_p$ : 顆粒密度 (Particle density) $[\text{kg} \cdot \text{m}^{-3}]$
* $g$ : 重力加速度 (Gravitational acceleration) $[\text{m} \cdot \text{s}^{-2}]$
* $\mu$ : 空氣動力黏滯係數 (Dynamic viscosity of air) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$
* $H$ : 沉降高度 (Settling height) $[\text{m}]$
* $\tau$ : 沉降時間尺度 (Settling timescale) $[\text{s}]$

#### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】[低雷諾數終端速度公式 (Terminal Velocity at Low $Re$)](https://derek1403.github.io/Theory_Playground/_build/html/02_Concepts/Fluid_Dynamics/Terminal_velocity.html#terminal-velocity-of-air)：** 由斯托克斯阻力與重力平衡導出，已於 Theory_Playground 證明，此處直接引用。

  $$v_{\text{terminal}} = \frac{2}{9}\frac{r^2 \rho_p g}{\mu}$$

* **【假設 1】低雷諾數區間 (Low $Re$ Regime)：** PM2.5 顆粒極小（$r$ 小）且沉降速度慢（$v$ 小），故落在 §4.2 的 A 區，必須用【已知 1】而非 $v \propto \sqrt{r}$ 的高雷諾數公式。

  $$\begin{gather*}
  Re &=& \frac{\rho v l}{\mu} \\
  &\ll& 1
  \end{gather*}$$

* **【假設 2】參數量級 (Order of Magnitude of Parameters)：**

  * (a) 顆粒半徑：PM2.5 的 $2.5 \ \mu\text{m}$ 指的是**直徑**，故半徑要取其一半：

    $$r \sim \frac{1}{2} \cdot 2.5 \times 10^{-6} \ \text{m}$$

  * (b) 顆粒密度（假設類似水滴或固體微粒）：

    $$\rho_p \sim 10^{3} \ \text{kg} \cdot \text{m}^{-3}$$

  * (c) 重力加速度：

    $$g \sim 10 \ \text{m} \cdot \text{s}^{-2}$$

  * (d) 空氣動力黏滯係數：

    $$\mu \sim 10^{-5} \ \text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}$$

* **【假設 3】沉降高度 (Settling Height)：**

  * (a) 邊界層厚度：

    $$H_{\text{PBL}} \sim 10^{3} \ \text{m}$$

  * (b) 對流層頂高度：

    $$H_{\text{trop}} \sim 10^{4} \ \text{m}$$

#### solve 終端速度

$$\begin{gather*}
v_{\text{terminal}} &\overset{\text{已知 1,假設 1}}{=}& \frac{2}{9}\frac{r^2 \rho_p g}{\mu} \\
&\overset{\text{假設 2}}{\approx}& \frac{2}{9} \cdot \frac{\left( \frac{1}{2} \cdot 2.5 \times 10^{-6} \ \text{m} \right)^2 \left( 10^{3} \ \text{kg}\cdot\text{m}^{-3} \right) \left( 10 \ \text{m}\cdot\text{s}^{-2} \right)}{10^{-5} \ \text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-1}} \\
&=& \frac{2}{9} \cdot \frac{\left( 1.25 \times 10^{-6} \ \text{m} \right)^2 \left( 10^{4} \ \text{kg}\cdot\text{m}^{-3}\cdot\text{s}^{-2} \right)}{10^{-5} \ \text{kg}\cdot\text{m}^{-1}\cdot\text{s}^{-1}} \\
&\approx& 3.5 \times 10^{-4} \ \text{m} \cdot \text{s}^{-1} \\
&\sim& 10^{-4} \ \text{m} \cdot \text{s}^{-1}
\end{gather*}$$

顆粒半徑 $r$ 越大，終端速度越快（$v_{\text{terminal}} \propto r^2$），這就是為什麼粗顆粒很快落地、而細懸浮微粒能久懸不下。

#### solve 沉降時間尺度

1. 若 PM2.5 處於**邊界層 (Planetary Boundary Layer)**：

$$\begin{gather*}
\tau_{\text{PBL}} &\overset{\text{def}}{=}& \frac{H_{\text{PBL}}}{v_{\text{terminal}}} \\
&\overset{\text{假設 3(a)}}{\approx}& \frac{10^{3} \ \text{m}}{3.5 \times 10^{-4} \ \text{m}\cdot\text{s}^{-1}} \\
&\approx& 2.9 \times 10^{6} \ \text{s} \\
&\approx& 1 \ \text{個月}
\end{gather*}$$

2. 若 PM2.5 進入**對流層頂 (Tropopause)**：

$$\begin{gather*}
\tau_{\text{trop}} &\overset{\text{def}}{=}& \frac{H_{\text{trop}}}{v_{\text{terminal}}} \\
&\overset{\text{假設 3(b)}}{\approx}& \frac{10^{4} \ \text{m}}{3.5 \times 10^{-4} \ \text{m}\cdot\text{s}^{-1}} \\
&\approx& 2.9 \times 10^{7} \ \text{s} \\
&\approx& 1 \ \text{年}
\end{gather*}$$

*結論*：一旦污染物（如火山灰、核爆塵）被送入對流層頂或平流層，將會造成「年尺度」的長期全球氣候效應。

## 5. 圓管流體力學：波以耳定律與電路類比 (Poiseuille Flow)



### 5.1 目標 (Goal)

在圓管中的穩態層流下，證明三件事：

* (a) **速度剖面 (Velocity Profile)**：管內徑向位置 $r$ 處的速度呈拋物線分佈

  $$u(r) = \frac{\Delta P}{4\ell\mu} (r_0^2 - r^2)$$

* (b) **總流量 (Flow Rate)，即 Poiseuille's Law**：流量與管徑四次方成正比

  $$I = \frac{\pi \Delta P}{8 \ell \mu} r_0^4$$

* (c) **流量對管徑的敏感度 (Sensitivity)**：管徑的微小變化會被放大四倍

  $$\frac{\Delta I}{I} \approx 4 \frac{\Delta r_0}{r_0}$$

符號說明：

* $u(r)$ : 徑向位置 $r$ 處的軸向流速 (Axial velocity) $[\text{m} \cdot \text{s}^{-1}]$
* $r$ : 距管軸的徑向距離 (Radial distance) $[\text{m}]$
* $r_0$ : 圓管半徑 (Pipe radius) $[\text{m}]$
* $\ell$ : 圓管長度 (Pipe length) $[\text{m}]$
* $\Delta P$ : 管子兩端的壓力差 (Pressure difference) $[\text{Pa}] = [\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}]$
* $\mu$ : 動力黏滯係數 (Dynamic viscosity) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$
* $I$ : 總體積流量 (Volumetric flow rate) $[\text{m}^{3} \cdot \text{s}^{-1}]$

![](./pic/steady_state_flow_pipe.png)

### 5.2 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】牛頓黏滯定律 (Newton's Law of Viscosity)：** 流體內部的剪應力正比於速度梯度。

  $$\tau \overset{\text{def}}{=} \mu \frac{du}{dr}$$

  * $\tau$ : 剪應力 (Shear stress) $[\text{Pa}] = [\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-2}]$
  * $\mu$ : 動力黏滯係數 (Dynamic viscosity) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$
  * $u$ : 軸向流速 (Axial velocity) $[\text{m} \cdot \text{s}^{-1}]$
  * $r$ : 距管軸的徑向距離 (Radial distance) $[\text{m}]$
  * $\frac{du}{dr}$ : 徑向的速度梯度 (Velocity shear) $[\text{s}^{-1}]$

* **【已知 2】無滑動邊界條件 (No-Slip Boundary Condition)：** 在管壁邊緣，流體黏附於管壁，速度為零。

  $$u(r_0) = 0$$

  * $u(r_0)$ : 管壁處（$r = r_0$）的軸向流速 (Axial velocity at the wall) $[\text{m} \cdot \text{s}^{-1}]$
  * $r_0$ : 圓管半徑 (Pipe radius) $[\text{m}]$

* **【假設 1】穩態層流力平衡 (Steady-State Laminar Force Balance)：** 液體在管子裡等速前進（Steady state），故流體的內部摩擦阻力必須等於管子兩端的推力。

  * (a) **管子兩端的推力**：壓力差 $\Delta P$ 乘上截面積 $\pi r^2$。

    $$F_{\text{pressure}} = -\Delta P (\pi r^2)$$

  * (b) **流體的內部摩擦阻力**：剪應力乘上該半徑處的圓柱側面積 $2\pi r \ell$。

    $$F_{\text{viscous}} \overset{\text{已知 1}}{=} \mu \frac{du}{dr} (2\pi r \ell)$$

  * (c) 兩者平衡：

    $$\mu \frac{du}{dr} (2\pi r \ell) = -\Delta P (\pi r^2)$$

* **【定義 1】流阻 (Fluid Resistance)：** 仿照歐姆定律 $\Delta V = I \cdot R_{\text{electric}}$，把 $\Delta P = I \cdot R_{\text{fluid}}$ 中的比例係數**定義**為流阻。此處**只做定義**，其具體數值留待 5.3(d) 由 5.3(b) 代入導出。

  $$R_{\text{fluid}} \overset{\text{def}}{=} \frac{\Delta P}{I}$$

  * $R_{\text{fluid}}$ : 流阻 (Fluid resistance) $[\text{kg} \cdot \text{m}^{-4} \cdot \text{s}^{-1}]$
  * $\Delta P$ : 管子兩端的壓力差 (Pressure difference) $[\text{Pa}]$
  * $I$ : 總體積流量 (Volumetric flow rate) $[\text{m}^{3} \cdot \text{s}^{-1}]$

### 5.3 證明 (Proof)

#### (a) proof 速度剖面

$$\begin{gather*}
\mu \frac{du}{dr} (2\pi r \ell) &\overset{\text{假設 1(c)}}{=}& -\Delta P (\pi r^2) \\
du &=& \frac{-\Delta P}{2\ell\mu} r \, dr \\
\int_{u(r)}^{0} du &\overset{\text{已知 2}}{=}& \int_{r}^{r_0} \frac{-\Delta P}{2\ell\mu} r \, dr \\
0 - u(r) &=& \frac{-\Delta P}{2\ell\mu} \left[ \frac{1}{2}r^2 \right]_{r}^{r_0} \\
u(r) &=& \frac{\Delta P}{4\ell\mu} (r_0^2 - r^2)
\end{gather*}$$

這是一個標準的拋物線分佈：管子中心（$r=0$）速度最快，管壁處（$r = r_0$）速度為零。

#### (b) proof 總流量 (Poiseuille's Law)

將整個圓管切成無數個面積為 $2\pi r \, dr$ 的同心圓環，把每一圈的流量積分加總：

$$\begin{gather*}
I &\overset{\text{def}}{=}& \int_0^{r_0} u(r) \, 2\pi r \, dr \\
&\overset{\text{5.3(a)}}{=}& \int_0^{r_0} \frac{\Delta P}{4\ell\mu} 2\pi r (r_0^2 - r^2) \, dr \\
&=& \frac{\pi \Delta P}{2\ell\mu} \int_0^{r_0} (r_0^2 r - r^3) \, dr \\
&=& \frac{\pi \Delta P}{2\ell\mu} \left( \frac{1}{2}r_0^4 - \frac{1}{4}r_0^4 \right) \\
&=& \frac{\pi \Delta P}{8 \ell \mu} r_0^4
\end{gather*}$$

#### (c) proof 流量對管徑的敏感度

$$\begin{gather*}
I &\overset{\text{5.3(b)}}{=}& \frac{\pi \Delta P}{8 \ell \mu}  r_0^4 \\
dI &=& 4 \frac{\pi \Delta P}{8 \ell \mu}  r_0^3 \, dr_0 \\
\frac{dI}{I} &=& \frac{4 \frac{\pi \Delta P}{8 \ell \mu}  r_0^3 \, dr_0}{I } \\
\frac{dI}{I} &\overset{\text{5.3(b)}}{=}& \frac{4 \frac{\pi \Delta P}{8 \ell \mu}  r_0^3 \, dr_0}{\frac{\pi \Delta P}{8 \ell \mu} r_0^4} \\
\frac{\Delta I}{I} &\approx& 4 \frac{\Delta r_0}{r_0}
\end{gather*}$$

總流量 $I$ 與管徑的「四次方」成正比，$I \propto r_0^4$ 表示管徑的微小變化會對流量造成巨大影響：
若 $\frac{\Delta r_0}{r_0} \sim 5\%$，則 $\frac{\Delta I}{I} \sim 20\%$！

#### (d) proof 流阻

$$\begin{gather*}
R_{\text{fluid}} &\overset{\text{定義 1}}{=}& \frac{\Delta P}{I} \\
&\overset{\text{5.3(b)}}{=}& \frac{\Delta P}{\frac{\pi \Delta P}{8 \ell \mu} r_0^4} \\
&=& \frac{8 \ell \mu}{\pi r_0^4}
\end{gather*}$$

### 5.4 跨領域類比：流體與電路 (Circuit Analogy)
我們可以將複雜的流體公式改寫成國中學過的歐姆定律形式：

$$\Delta V = I \cdot R_{\text{electric}}$$

$$\Delta P = I \cdot R_{\text{fluid}}$$

其中的「流阻 (Fluid Resistance)」即【定義 1】，其值已於 5.3(d) 導出：

$$R_{\text{fluid}} \overset{\text{5.3(d)}}{=} \frac{8 \ell \mu}{\pi r_0^4}$$

這與電阻定律 $R_{\text{electric}} = \rho_e \frac{\ell}{A} \propto \frac{\ell}{r^2}$ 概念如出一轍。
> **💡 生物醫學類比：**
> * 人類吃高油食物導致血管稍微阻塞（$r_0$ 下降 5%），流阻 $R_{\text{fluid}}$ 會因為四次方關係急遽上升。為維持相同的血流量 $I$，心臟必須加壓，導致 $\Delta P$ 大幅上升（高血壓）。
> * 吃高鹽食物導致血液變稠（動力黏滯係數 $\mu$ 上升），同樣會讓流阻上升，引發高血壓。這正是心血管疾病危險的物理學根源。

## 6. 分支管路系統的最佳化：穆雷定律 (Murray's Law)



自然界中的血管、氣管，甚至植物的導管，為何會以特定的角度分叉？流體力學提供了一個尋求「總阻力最小」的極值問題解答。

### 6.1 目標 (Goal)

證明使總流阻最小的最佳分叉角度 $\theta$ 滿足：

$$\cos\theta = \frac{r_2^4}{r_1^4}$$

![](./pic/Three-Way_pipe.png)

* $\theta$ : 支管與主管的分叉角度 (Branching angle) $[\text{rad}]$
* $r_1$ : 主管半徑 (Radius of the main pipe) $[\text{m}]$
* $r_2$ : 支管半徑 (Radius of the branch pipe) $[\text{m}]$
* $a$ : 起點到末端的水平總距離 (Horizontal distance) $[\text{m}]$
* $b$ : 起點到末端的垂直距離 (Vertical distance) $[\text{m}]$
* $\mu$ : 動力黏滯係數 (Dynamic viscosity) $[\text{kg} \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$
* $R_{\text{總}}$ : 主管與支管串聯的總流阻 (Total fluid resistance) $[\text{kg} \cdot \text{m}^{-4} \cdot \text{s}^{-1}]$

物理背景：在固定壓力差下，把流體從主管（半徑 $r_1$）透過支管（半徑 $r_2$）送到末端，調整分叉角度 $\theta$ 以達到最大傳送量（即流阻最小化）。

### 6.2 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】流阻 (Fluid Resistance)：** 引用 §5 5.3(d) 已導出的 Poiseuille 流阻，不再重證。

  $$R_{\text{fluid}} \overset{\text{§5 5.3(d)}}{=} \frac{8 \ell \mu}{\pi r_0^4}$$

  * $R_{\text{fluid}}$ : 流阻 (Fluid resistance) $[\text{kg} \cdot \text{m}^{-4} \cdot \text{s}^{-1}]$
  * $\ell$ : 管子長度 (Pipe length) $[\text{m}]$
  * $r_0$ : 管子半徑 (Pipe radius) $[\text{m}]$

* **【已知 2】三角函數微分基本公式 (Derivatives of Trigonometric Functions)：**

  * (a) 餘切函數的導數：

    $$\frac{d}{d\theta}\left[ \cot\theta \right] = -\csc^2\theta$$

  * (b) 餘割函數的導數：

    $$\frac{d}{d\theta}\left[ \csc\theta \right] = -\csc\theta \cot\theta$$

* **【已知 3】三角函數基本恆等式 (Trigonometric Identities)：**

  * (a) 餘割函數：

    $$\csc\theta = \frac{1}{\sin\theta}$$

  * (b) 餘切函數：

    $$\cot\theta = \frac{\cos\theta}{\sin\theta}$$

* **【假設 1】幾何設定 (Geometric Setup)：** 末端座標為水平總距離 $a$、垂直距離 $b$。

  * (a) **支管的長度**：利用三角函數，為斜邊

    $$\ell_2 = b \csc\theta$$

  * (b) **主管的長度**：總水平長度減去支管的水平投影 $b \cot\theta$

    $$\ell_1 = a - b \cot\theta$$

* **【假設 2】流阻最小化的極值條件 (Minimization Condition)：** 最佳角度發生在總流阻對 $\theta$ 的一階導數為零之處。

  $$\frac{dR_{\text{總}}}{d\theta} = 0$$

* **【推導 1】總流阻方程式 (Total Resistance)：** 兩段管子串聯，流阻直接相加 $R_{\text{總}} = R_1 + R_2$。

  $$\begin{gather*}
  R_{\text{總}} &\overset{\text{已知 1}}{=}& \frac{8 \ell_1 \mu}{\pi r_1^4} + \frac{8 \ell_2 \mu}{\pi r_2^4} \\
  &\overset{\text{假設 1(a)(b)}}{=}& \frac{8\mu}{\pi} \left( \frac{a - b \cot\theta}{r_1^4} + \frac{b \csc\theta}{r_2^4} \right)
  \end{gather*}$$

### 6.3 proof 最佳分叉角度

$$\begin{gather*}
0 &\overset{\text{假設 2}}{=}& \frac{dR_{\text{總}}}{d\theta} \\
0 &\overset{\text{推導 1}}{=}& \frac{8\mu}{\pi} \frac{d}{d\theta} \left[ \frac{a - b \cot\theta}{r_1^4} + \frac{b \csc\theta}{r_2^4} \right] \\
0 &\overset{\text{已知 2(a)(b)}}{=}& \frac{8\mu}{\pi} \left[ \frac{b \csc^2\theta}{r_1^4} - \frac{b \csc\theta \cot\theta}{r_2^4} \right] \\
\frac{b \csc^2\theta}{r_1^4} &=& \frac{b \csc\theta \cot\theta}{r_2^4} \\
\frac{\csc^2\theta}{r_1^4} &=& \frac{\csc\theta \cot\theta}{r_2^4} \\
\frac{1}{\sin^2\theta} \cdot \frac{1}{r_1^4} &\overset{\text{已知 3(a)(b)}}{=}& \frac{\cos\theta}{\sin^2\theta} \cdot \frac{1}{r_2^4} \\
\frac{1}{r_1^4} &=& \frac{\cos\theta}{r_2^4} \\
\cos\theta &=& \frac{r_2^4}{r_1^4}
\end{gather*}$$

### 6.4 物理與生物的印證 (Physical & Biological Verification)
* **破水管類比 ($\theta \rightarrow \frac{\pi}{2}$)**：若主管破了一個小洞噴水，支管半徑 $r_2 \ll r_1$。此時 $\frac{r_2^4}{r_1^4} \approx 0$，即 $\cos\theta = 0$，解得 $\theta = 90^\circ$。這完美吻合了我們在日常生活中看到破水管的水會「垂直」噴出的現象。
* **樹枝的生長模型修正**：雖然樹木在演化中也服從流動阻力最小的力學原則，但正如筆記所洞見的，生物體並非純粹的管線。植物的**向光性 (Phototropism)** 會為了爭取更多陽光而強行改變枝條的生長角度，因此在建立真實的生物物理模型時，這項因素必須被視為額外的權重加入修正。

# 7. 旋轉座標系中的動量方程式 (Momentum Equations in a Rotating Frame)

## 7.1 目標 (Goal)

把[動量方程式](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#momentum-equation-navier-stokes)搬到旋轉座標系上，並證明：

* (a) **颱風的絕對角動量守恆 (Absolute Angular Momentum Conservation)**：在軸對稱、無摩擦的圓柱座標系下

  $$\frac{D}{Dt}\left[ v r + \frac{1}{2} f r^2 \right] = 0$$

* (b) **原始方程式 (Primitive Equations)**：在球座標系下套用傳統近似後，得到全世界數值天氣模式（GFS、WRF，以及 Pangu-Weather、GraphCast 等 AI 模型）底層所遵循的方程組

  $$\begin{aligned}
  \frac{Du}{Dt} - f v - \frac{u v \tan\phi}{a} &= -\frac{1}{\rho a \cos\phi} \frac{\partial P}{\partial \lambda} + F_\lambda \\
  \frac{Dv}{Dt} + f u + \frac{u^2 \tan\phi}{a} &= -\frac{1}{\rho a} \frac{\partial P}{\partial \phi} + F_\phi \\
  0 &= -\frac{1}{\rho} \frac{\partial P}{\partial z} - g
  \end{aligned}$$

符號說明（**全章統一用 $(u, v, w)$，與 week1 及 Theory_Playground 的推導一致**）：

* $u$ : 第一分量風速 (First component) $[\text{m} \cdot \text{s}^{-1}]$ ─ 球座標為緯向風（Zonal，向東為正）；圓柱座標為徑向風（Radial，向外為正）
* $v$ : 第二分量風速 (Second component) $[\text{m} \cdot \text{s}^{-1}]$ ─ 球座標為經向風（Meridional，向北為正）；圓柱座標為切向風（Azimuthal，逆時針為正）
* $w$ : 垂直風速 (Vertical velocity，向上為正) $[\text{m} \cdot \text{s}^{-1}]$
* $\lambda$ : 經度／方位角 (Longitude / Azimuth) $[\text{rad}]$
* $\phi$ : 緯度 (Latitude) $[\text{rad}]$
* $r$ : 距颱風中心的半徑 (Radius) $[\text{m}]$
* $R$ : 距地心的半徑 (Radial distance from Earth's center) $[\text{m}]$
* $a$ : 地球半徑 (Earth's radius) $[\text{m}]$，$a \approx 6371 \ \text{km}$
* $\Omega$ : 地球自轉角速度 (Earth's angular velocity) $[\text{rad} \cdot \text{s}^{-1}]$，$\Omega \approx 7.29 \times 10^{-5} \ \text{rad} \cdot \text{s}^{-1}$
* $f$ : 科氏參數 (Coriolis parameter) $[\text{s}^{-1}]$
* $\rho$ : 空氣密度 (Air density) $[\text{kg} \cdot \text{m}^{-3}]$
* $P$ : 氣壓 (Pressure) $[\text{Pa}]$
* $g$ : 有效重力加速度 (Effective gravity) $[\text{m} \cdot \text{s}^{-2}]$
* $F$ : 摩擦力／黏滯力 (Friction) $[\text{m} \cdot \text{s}^{-2}]$
* $\frac{D}{Dt}$ : 全質導數 (Material derivative) $[\text{s}^{-1}]$

## 7.2 座標系命名空間與 RHS 總定義 (Namespace & RHS)

在寫下任何方程式前，先把變數的「命名空間」定義清楚。大氣動力學中的 RHS（方程式右側）代表的是**絕對加速度以外的所有外力**，通常包含三項：氣壓梯度力 (Pressure Gradient Force)、重力 (Gravity，僅垂直方向)、摩擦力／黏滯力 (Friction，記為 $F$)。

* **【定義 1】科氏參數 (Coriolis Parameter)：**

  $$f \overset{\text{def}}{=} 2\Omega \sin\phi$$

* **【定義 2】三套座標系的 RHS 展開 (RHS in Three Coordinate Systems)：** 差別只在於「座標微分」與「實際物理距離」之間的度量換算（見 7.6）。

  * (a) **直角座標系 (Cartesian Coordinates)** $(x, y, z)$，速度分量 $(u, v, w)$：

    $$\begin{aligned}
    \text{RHS}_x &= -\frac{1}{\rho} \frac{\partial P}{\partial x} + F_x \\
    \text{RHS}_y &= -\frac{1}{\rho} \frac{\partial P}{\partial y} + F_y \\
    \text{RHS}_z &= -\frac{1}{\rho} \frac{\partial P}{\partial z} - g + F_z
    \end{aligned}$$

  * (b) **圓柱座標系 (Cylindrical Coordinates)** $(r, \lambda, z)$ ─ 適用於颱風中心分析，速度分量 $(u, v, w)$ 為徑向、切向、垂直：

    $$\begin{aligned}
    \text{RHS}_r &= -\frac{1}{\rho} \frac{\partial P}{\partial r} + F_r \\
    \text{RHS}_\lambda &= -\frac{1}{\rho r} \frac{\partial P}{\partial \lambda} + F_\lambda \\
    \text{RHS}_z &= -\frac{1}{\rho} \frac{\partial P}{\partial z} - g + F_z
    \end{aligned}$$

  * (c) **球面座標系 (Spherical Coordinates)** $(\lambda, \phi, R)$ ─ 適用於全球環流模式，速度分量 $(u, v, w)$ 為緯向、經向、垂直：

    $$\begin{aligned}
    \text{RHS}_\lambda &= -\frac{1}{\rho R \cos\phi} \frac{\partial P}{\partial \lambda} + F_\lambda \\
    \text{RHS}_\phi &= -\frac{1}{\rho R} \frac{\partial P}{\partial \phi} + F_\phi \\
    \text{RHS}_z &= -\frac{1}{\rho} \frac{\partial P}{\partial R} - g + F_z
    \end{aligned}$$

## 7.3 假設與已知 (Assumptions & Preliminaries)

  $$\frac{D \vec{V}}{Dt} + f\hat{k}\times \vec{V} = -\frac{1}{\rho}\nabla P - g\hat{k} + \text{Friction}$$

  *（本卡片是**根源**：下面的【已知 2】【已知 3】都是由它經座標轉換導出的，因此證明段不會直接引用它，而是從【已知 2】【已知 3】起手。）*

* **【已知 2】[圓柱座標系動量方程式 (Momentum Equation in Cylindrical Coordinates)](https://derek1403.github.io/Theory_Playground/_build/html/01_Derivations/Coordinate_System/cylindrical_coordinate_transformation_momentum_equation.html)：** 由【已知 1】經座標轉換與尺度分析導出（Theory_Playground 已證）。

  $$\begin{aligned}
  \text{[} u \text{ 徑向]} \quad &\frac{Du}{Dt} - \frac{v^2}{r} - f v = \text{RHS}_r \\
  \text{[} v \text{ 切向]} \quad &\frac{Dv}{Dt} + \frac{u v}{r} + f u = \text{RHS}_\lambda \\
  \text{[} w \text{ 垂直]} \quad &\frac{Dw}{Dt} = \text{RHS}_z
  \end{aligned}$$

  * $\frac{v^2}{r}$ : 離心力項 (Centrifugal force) $[\text{m} \cdot \text{s}^{-2}]$ ─ 當颱風風速 $v$ 極大、眼牆半徑 $r$ 很小時，這一項會變得無比巨大，這就是颱風中心氣壓極低的原因（梯度風平衡 Gradient wind balance）
  * $\frac{u v}{r}$ : 曲率項 (Metric/Curvature term) $[\text{m} \cdot \text{s}^{-2}]$

* **【已知 3】[球面座標系動量方程式 (Momentum Equation in Spherical Coordinates)](https://derek1403.github.io/Theory_Playground/_build/html/01_Derivations/Coordinate_System/spherical_coordinate_transformation_momentum_equation.html)：** 由【已知 1】經座標轉換導出（Theory_Playground 已證）。其中 $\left(2\Omega + \frac{u}{R \cos\phi}\right)$ 是把「地球本身的自轉」與「空氣相對地球的旋轉」合併而成的**絕對角速度係數**。

  $$\begin{aligned}
  \text{[} u \text{ 緯向]} \quad &\frac{Du}{Dt} + \left(2\Omega + \frac{u}{R \cos\phi}\right)(w \cos\phi - v \sin\phi) = \text{RHS}_\lambda \\
  \text{[} v \text{ 經向]} \quad &\frac{Dv}{Dt} + \left(2\Omega + \frac{u}{R \cos\phi}\right) u \sin\phi + \frac{v w}{R} = \text{RHS}_\phi \\
  \text{[} w \text{ 垂直]} \quad &\frac{Dw}{Dt} - \left(2\Omega + \frac{u}{R \cos\phi}\right) u \cos\phi - \frac{v^2}{R} = \text{RHS}_z
  \end{aligned}$$

* **【已知 4】徑向速度的定義 (Definition of Radial Velocity)：** 在圓柱座標系中，質點的半徑變化率即徑向風速。

  $$\frac{Dr}{Dt} = u$$

* **【假設 1】軸對稱無摩擦的切向條件 (Axisymmetric & Frictionless)：** 用於 7.4 的角動量守恆。等壓線完全同心（沒有切向氣壓梯度）、且忽略摩擦。

  $$\text{RHS}_\lambda \overset{\text{定義 2(b)}}{=} -\frac{1}{\rho r} \frac{\partial P}{\partial \lambda} + F_\lambda = 0$$

* **【假設 2】$f$-plane 近似 ($f$-plane Approximation)：** 颱風的水平尺度遠小於地球半徑，質點移動時所感受的科氏參數視為常數。

  $$\frac{Df}{Dt} = 0$$

* **【假設 3】淺層／傳統近似 (Shallow Fluid / Traditional Approximation)：** 地球半徑 $a \approx 6371 \ \text{km}$，而對流層厚度只有約 $10 \ \text{km}$ ─ 相對於地球，大氣層薄得像蘋果皮。

## 7.4 颱風系統：圓柱座標的應用 (Application to Tropical Cyclones)

把【已知 2】的 RHS 用【定義 2(b)】展開，就是颱風動力學的出發點：

$$\begin{aligned}
\text{[} u \text{ 徑向]} \quad &\frac{Du}{Dt} - \frac{v^2}{r} - f v = -\frac{1}{\rho} \frac{\partial P}{\partial r} + F_r \\
\text{[} v \text{ 切向]} \quad &\frac{Dv}{Dt} + \frac{u v}{r} + f u = -\frac{1}{\rho r} \frac{\partial P}{\partial \lambda} + F_\lambda
\end{aligned}$$

* **物理翻譯（徑向）**：質點向外的加速度 $\frac{Du}{Dt}$，扣掉離心力 $\frac{v^2}{r}$ 與科氏力 $f v$ 後，必須由氣壓梯度力與摩擦力來平衡。
* **物理翻譯（切向）**：若沒有切向氣壓差與摩擦力，當颱風把外圍空氣往中心吸入時（$u < 0$），為了守恆角動量，切向風速 $v$ 就會瘋狂加速。
* **慣性圓 (Inertial Circle)**：當 $\text{RHS}_r = 0$，徑向方程式退化為 $\frac{Du}{Dt} - \left(f + \frac{v}{r}\right)v = 0$。

### (a) proof 絕對角動量守恆

$$\begin{gather*}
0 &\overset{\text{假設 1}}{=}& \text{RHS}_\lambda \\
0 &\overset{\text{已知 2}}{=}& \frac{Dv}{Dt} + \frac{u v}{r} + f u \\
0 &=& r \frac{Dv}{Dt} + u v + f u r \\
0 &\overset{\text{已知 4}}{=}& r \frac{Dv}{Dt} + v \frac{Dr}{Dt} + f r \frac{Dr}{Dt} \\
0 &=& \frac{D \left[ v r \right]}{Dt} + \frac{1}{2} f \frac{D \left[ r^2 \right]}{Dt} \\
0 &\overset{\text{假設 2}}{=}& \frac{D}{Dt}\left[ v r + \frac{1}{2} f r^2 \right]
\end{gather*}$$

括號內的 $v r + \frac{1}{2} f r^2$ 就是**絕對角動量 (Absolute Angular Momentum)**。這解釋了為什麼颱風周圍的空氣被吸入中心（$u < 0$，半徑 $r$ 縮小）時，旋轉速度 $v$ 必須狂飆 ─ 因為絕對角動量是守恆的。

## 7.5 全球模式：原始方程式 (Primitive Equations)

數學很美，但大自然給了我們簡化的權利。套用【假設 3】的傳統近似與【假設 4】的尺度分析，把【已知 3】的球座標方程式化簡。

### (b-1) proof 緯向方程式

$$\begin{gather*}
\text{RHS}_\lambda &\overset{\text{已知 3}}{=}& \frac{Du}{Dt} + \left(2\Omega + \frac{u}{R \cos\phi}\right)(w \cos\phi - v \sin\phi) \\
&=& \frac{Du}{Dt} + 2\Omega w \cos\phi - 2\Omega v \sin\phi + \frac{u w}{R} - \frac{u v \tan\phi}{R} \\
&\overset{\text{假設 3(a)}}{=}& \frac{Du}{Dt} + 2\Omega w \cos\phi - 2\Omega v \sin\phi + \frac{u w}{a} - \frac{u v \tan\phi}{a} \\
&\overset{\text{假設 3(d-1)(e)}}{\approx}& \frac{Du}{Dt} - 2\Omega v \sin\phi - \frac{u v \tan\phi}{a} \\
&\overset{\text{定義 1}}{=}& \frac{Du}{Dt} - f v - \frac{u v \tan\phi}{a}
\end{gather*}$$

### (b-2) proof 經向方程式

$$\begin{gather*}
\text{RHS}_\phi &\overset{\text{已知 3}}{=}& \frac{Dv}{Dt} + \left(2\Omega + \frac{u}{R \cos\phi}\right) u \sin\phi + \frac{v w}{R} \\
&=& \frac{Dv}{Dt} + 2\Omega u \sin\phi + \frac{u^2 \tan\phi}{R} + \frac{v w}{R} \\
&\overset{\text{假設 3(a)}}{=}& \frac{Dv}{Dt} + 2\Omega u \sin\phi + \frac{u^2 \tan\phi}{a} + \frac{v w}{a} \\
&\overset{\text{假設 3(d-2)}}{\approx}& \frac{Dv}{Dt} + 2\Omega u \sin\phi + \frac{u^2 \tan\phi}{a} \\
&\overset{\text{定義 1}}{=}& \frac{Dv}{Dt} + f u + \frac{u^2 \tan\phi}{a}
\end{gather*}$$

### (b-3) proof 垂直方程式（靜力平衡 Hydrostatic Balance）

$$\begin{gather*}
\text{RHS}_z &\overset{\text{已知 3}}{=}& \frac{Dw}{Dt} - \left(2\Omega + \frac{u}{R \cos\phi}\right) u \cos\phi - \frac{v^2}{R} \\
&\overset{\text{假設 3(a)}}{=}& \frac{Dw}{Dt} - 2\Omega u \cos\phi - \frac{u^2 + v^2}{a} \\
&\overset{\text{假設 4(c)(d)(e)}}{\approx}& 0 \\
0 &\overset{\text{定義 2(c)}}{=}& -\frac{1}{\rho} \frac{\partial P}{\partial z} - g \\
\frac{\partial P}{\partial z} &=& -\rho g
\end{gather*}$$

### 大氣數值模式的基石

把【定義 2(c)】的 RHS 代回 (b-1)、(b-2)、(b-3)，就得到全世界所有大型天氣模式（從傳統的 GFS、WRF 到 Pangu-Weather、GraphCast 等 AI 模型）底層所遵循的**原始方程式 (Primitive Equations)**：

$$\begin{aligned}
\text{[} u \text{ 緯向]} \quad &\frac{Du}{Dt} - f v - \frac{u v \tan\phi}{a} = -\frac{1}{\rho a \cos\phi} \frac{\partial P}{\partial \lambda} + F_\lambda \\
\text{[} v \text{ 經向]} \quad &\frac{Dv}{Dt} + f u + \frac{u^2 \tan\phi}{a} = -\frac{1}{\rho a} \frac{\partial P}{\partial \phi} + F_\phi \\
\text{[} w \text{ 垂直]} \quad &0 = -\frac{1}{\rho} \frac{\partial P}{\partial z} - g \quad \text{(靜力平衡 Hydrostatic balance)}
\end{aligned}$$

這組方程式完美平衡了物理的嚴謹性與計算的可行性。

## 7.6 補充：座標系的選擇 (Supplement: Choosing a Coordinate System)

> 本節是**概念討論，不含證明**，用來回答「為什麼 $\nabla P$ 在不同座標長得不一樣」與「什麼時候必須用球座標」。

### 一、破解向量的幻象：為什麼 $\nabla P$ 長得不一樣？

「寫成向量 $\nabla P$ 在哪個座標都是對的。」這句話的意思是，**向量方程式是一種「抽象介面 (Abstract Interface)」**，它描述純粹的物理與幾何事實，還沒有被綁定到任何特定的「資料結構 (Grid System)」上。

* **物理事實**：氣壓梯度力永遠指向氣壓降得最快的方向。這個「最快傾斜的陡峭程度」，數學上就叫梯度算子 $\nabla$ (Nabla)。

但當我們要把抽象的 $\nabla$ 具體「實作 (Implement)」到特定座標系時，必須考慮那個座標系的**「度量張量 (Metric / Scale factor)」**：你的網格線在真實空間中代表多長的物理距離？

1. **在直角座標系中**：網格是正方形的。你在 $x$ 軸上前進一個單位的座標 $dx$，真實世界你就前進了 $1$ 公尺。

   $$\nabla P = \hat{i} \frac{\partial P}{\partial x} + \hat{j} \frac{\partial P}{\partial y} + \hat{k} \frac{\partial P}{\partial z}$$

2. **在球面座標系中**：網格是經緯度。當你在經度上移動一個微小角度 $d\lambda$ 時，實際在地球表面走過的**物理距離**並不是 $d\lambda$，而是會隨緯度收縮的弧長（赤道上的 $1^\circ$ 經度很寬，靠近極地時很窄），$dx = R \cos\phi \, d\lambda$。因此把「對物理距離微分」轉換成「對經緯度微分」時，分母就會自動吐出修正項：

   $$\frac{\partial P}{\partial (\text{物理距離 } x)} = \frac{1}{R \cos\phi} \frac{\partial P}{\partial \lambda}$$

**結論**：$-\frac{1}{\rho}\nabla P$ 永遠是對的，它只是把所有座標系底層的複雜度量（$R \cos\phi$ 等曲率項）打包封裝進了 $\nabla$ 這個符號裡。當你把它解壓縮（展開成純量分量）時，不同座標系自然會長得不一樣 ─ 這正是【定義 2】三套 RHS 的來源。

### 二、為什麼課堂上都用直角座標系 $(x, y, z)$？

因為掛上了大氣科學中最佛心的外掛：**「局部直角座標系近似 (Local Cartesian Approximation)」**，這包含了著名的 $f$-plane 與 $\beta$-plane 近似（即【假設 2】的推廣）。

想像你拿一張平坦的 A4 畫紙，貼在地球表面的某個緯度 $\phi_0$ 上（例如貼在台灣上空）：定義 $x$ 是向東的實體距離、$y$ 是向北的實體距離。只要討論的天氣系統（例如颱風，尺度約 1000 公里）相對於地球半徑（6371 公里）還算小，就可以假裝這張平坦的 A4 紙就是地球表面，球面曲率造成的項（例如 $\frac{u v \tan\phi}{a}$）小到可以忽略。

這就是為什麼在探討「颱風動力學」、「斜壓不穩定」、「羅斯貝波」的基礎理論時，一律用 $x, y, z$：否則方程式裡會塞滿 $\sin\phi$、$\cos\phi$ 與非線性項，人類根本無法在黑板上求出解析解 (Analytical solution)。

### 三、限制在哪？什麼時候「必須」用球座標？

**當天氣系統的尺度大到可以「感受到地球是圓的」時，直角座標系就會崩潰。**

* **局部尺度 (Mesoscale / Synoptic scale, $L < 2000 \ \text{km}$)**：單一颱風、鋒面、雷暴雲。用局部直角座標系 $x, y, z$ 非常完美，誤差極小。
* **行星尺度 (Planetary scale, $L \sim 10000 \ \text{km}$)**：超長波、全球季風環流、跨越半個地球的遙相關機制。這時地球曲率 $R \cos\phi$ 有劇烈變化，若繼續用 $x, y, z$，算出來的能量傳遞與波速會完全錯掉。

**必須用 3D 球座標系的時機：運行全球數值天氣模式 (Global NWP Models)。**

* 這些模型的網格資料格式（例如 NetCDF 檔案、用 Xarray 讀取），座標維度通常是 `(time, level, latitude, longitude)`。
* 計算某點的氣壓梯度力時，絕對不能只拿相鄰格點的值相減除以常數 $dx$，必須在程式碼中寫入 $\frac{1}{a \cos\phi} \frac{\Delta P}{\Delta \lambda}$（即【定義 2(c)】），否則極地的氣壓梯度力會被嚴重低估。
