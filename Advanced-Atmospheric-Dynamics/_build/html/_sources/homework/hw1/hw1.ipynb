# Question 1

氧氣是維持生命機能的要素。在理想狀態下，我們以管道層流來模擬生物循環系統的氧氣輸送，若 $[O_2]$ 為流體內氧氣的濃度，且流體的體積積流量為 $Q = \frac{\pi r^4 \Delta p}{8 \mu l}$，則單位時間內的氧氣輸送量公式為：
   
$$J_{O_2} = [O_2] \frac{\pi r^4 \Delta p}{8 \mu l}$$

* $J_{O_2}$ 是氧氣輸送速率，單位是 $[\text{mol} \cdot \text{s}^{-1}]$
* $[O_2]$ 是氧氣濃度，單位是 $[\text{mol} \cdot \text{m}^{-3}]$
* $Q$ 流體的體積積流量，單位是 $[\text{m}^3 \cdot \text{s}^{-1}]$
* $r$ 為管徑，單位是 $[\text{m}]$
* $\Delta p$ 為管道兩端壓力差，單位是 $[\text{Pa}]$
* $\mu$ 為流體黏滯係數，單位是 $[\text{Pa} \cdot \text{s}]=[\text{Nt} \cdot \text{s}\cdot \text{m}^{-2}]$
* $l$ 為管道長度，單位是 $[\text{m}]$

1. 以人類為例，若血管發生病變導致半徑縮小 20%（即剩餘 80% 半徑），則心臟需增加多少比例的「血壓差」才能維持原有的氧氣輸送量？

2. 在溫度 20°C 時，水的黏滯係數 $\mu_{H_2O}$ 約為空氣 $\mu_{air}$ 的 55倍，且水中的氧氣濃度 $[O_2]$ 比空氣少約 40倍。若要維持相同的氧氣輸送量與管道幾何（半徑、長度），以空氣或水輸送，兩者需要的壓力差倍數關係為何？哪一種流體輸送較省力？

3. 承上題，若兩系統的「壓力差」與「氧氣輸送量」皆相同，兩個流體系統的管道半徑倍數差異為何？哪一個流體輸送需要較粗的管道？

4. 請討論：生物系統為何多數使用「液體管道」作為生命循環系統？（請先論述為何基於前述計算，這個問題是成立/值得探討的，再給出 3 個可能理由）

## 1.1

設原半徑為 $r$，原壓力差為 $\Delta p$，阻塞後的半徑 $r' = 0.8r$，新壓力差為 $\Delta p'$，由於氧氣輸送量 $J_{O_2}$ 維持不變，且流體與 $l$ 不變。

$$\begin{gather*}
[O_2] \frac{\pi r^4 \Delta p}{8 \mu l} &= J_{O_2} =& [O_2] \frac{\pi (r')^4 \Delta p'}{8 \mu l} \\
r^4 \Delta p &=& (r')^4 \Delta p' \\
\Delta p' &=& \left(\frac{r}{0.8r}\right)^4 \Delta p \\
&=& \left(\frac{1}{0.8}\right)^4 \Delta p \\
&\approx&  2.441 \Delta p
\end{gather*}$$

心臟需要將血壓差提高到原來的 2.441倍，因此需增加的血壓差比例為 $(2.441 - 1) \times 100\% = \mathbf{144.1\%}$。



## 1.2

在 $J_{O_2}, l, r$ 皆相同的情況下，壓力差與 $\frac{\mu}{[O_2]}$ 成正比。

$$\begin{gather*}
\frac{\Delta p_{\text{water}}}{\Delta p_{\text{air}}} &=& \frac{J_{O_2} \frac{8 \mu_{\text{water}} l}{\pi r^4 [O_2]_{\text{water}}}}{J_{O_2} \frac{8 \mu_{\text{air}} l}{\pi r^4 [O_2]_{\text{air}}}}\\
&=&\left(\frac{\mu_{\text{water}}}{\mu_{\text{air}}}\right) \times \left(\frac{[O_2]_{\text{air}}}{[O_2]_{\text{water}}}\right) \\
&=& 55 \times 40 \\
&=&\mathbf{2200}
\end{gather*}$$


以水為基底的液體輸送，需要的壓力差是空氣的 2200倍。因此，以空氣輸送較為省力。

## 1.3

在 $J_{O_2}, l, \Delta p$ 皆相同的情況下，$r^4$ 與 $\frac{\mu}{[O_2]}$ 成正比。

$$\begin{gather*}
\frac{ r^4_{\text{water}}}{ r^4_{\text{air}}} &=& \frac{J_{O_2} \frac{8 \mu_{\text{water}} l}{\pi \Delta p [O_2]_{\text{water}}}}{J_{O_2} \frac{8 \mu_{\text{air}} l}{\pi \Delta p [O_2]_{\text{air}}}}\\
&=&\left(\frac{\mu_{\text{water}}}{\mu_{\text{air}}}\right) \times \left(\frac{[O_2]_{\text{air}}}{[O_2]_{\text{water}}}\right) \\
&=& 55 \times 40 \\
&=&\mathbf{2200} \\
\frac{r_{\text{water}}}{r_{\text{air}}} &=& (2200)^{1/4} \\
&\approx& \mathbf{6.85}
\end{gather*}$$

液體系統需要的管道半徑約為空氣系統的 6.85倍。液體輸送需要較粗的管道

## 1.4

1. 演化起源與內部環境恆定： **生命起源於海洋** ，早期細胞適應了在水溶液中進行生化反應。液體循環系統在生物體內創造了一個微型的「內部海洋」， **維持了細胞所需的滲透壓與離子平衡**。

2. 多功能溶劑與運輸載體：液體（水）是極佳的極性溶劑，血液不僅輸送氧氣（可利用紅血球中的血紅素來大幅突破純水的溶氧限制），還能同時溶解並輸送養分（葡萄糖、氨基酸）、荷爾蒙、免疫細胞，並帶走細胞代謝產生的水溶性廢物（尿素、乳酸），這是純氣體管道無法做到的。

3. 不可壓縮性與結構支撐：液體幾乎不可壓縮。在具有血壓的封閉血管系統中，液體可以提供流體靜力骨架（Hydrostatic skeleton）的 **支撐作用** ，並且在生物體受到外界壓力或重力變化時， **液體管路較不易像氣體管路那樣被輕易壓扁而導致循環中斷** 。

## 2.1 Briefly describe the core physical concept behind the Navier–Stokes equations and outline the main steps in their derivation.

**物理核心：牛頓第二運動定律 ($F = ma$) 在流體上的應用。**

Navier-Stokes equation 本質上就是流體的 $\mathbf{F}=m\mathbf{a}$ ，只是流體會變形、會流動，所以我們不能只看一個固定的質點，而是看一個arbitrary volume

**推導步驟**

* **Step 1: 描述加速度 ($\mathbf{a}$)** 
加速度有兩種來源：一是 **隨時間的局部變化** $\frac{\partial \mathbf{u}}{\partial t}$，二是 **因為位置改變而帶來的變化**，即平流項 $(\mathbf{u} \cdot \nabla)\mathbf{u}$。

  * $\mathbf{a}$ 是3維加速度向量
  * $\mathbf{u}$ 是3維速度向量

  對流體而言，跟隨氣塊的總加速度寫為：

$$\frac{D\mathbf{u}}{Dt} = \frac{\partial \mathbf{u}}{\partial t} + (\mathbf{u} \cdot \nabla)\mathbf{u}$$




* **Step 2: 盤點考慮的作用力 ($\mathbf{F}$)**
作用在流體微小控制體積上的力有兩種：
1. **體積力 (Body forces)**：超距力，通常只考慮有重力 $\rho \mathbf{g}$ 有時考慮科氏力
   * $\rho \mathbf{g}$ 是3維重力向量
2. **表面力 (Surface forces)**：相鄰流體層互相推擠或摩擦的力，這需要用**應力張量 (Stress Tensor) $\mathbf{\sigma}_{3 \times 3}$** 來描述，表面的淨作用力即為應力張量的散度 $\nabla \cdot \sigma$。
   * $\mathbf{\sigma}_{3 \times 3}$ 是外力對表面作用的應力張量 $(3 \times 3)$矩陣
   * $\nabla \cdot \boldsymbol{\sigma} = \begin{bmatrix} \frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \end{bmatrix} \begin{bmatrix} \sigma_{xx} & \sigma_{xy} & \sigma_{xz} \\ \sigma_{yx} & \sigma_{yy} & \sigma_{yz} \\ \sigma_{zx} & \sigma_{zy} & \sigma_{zz} \end{bmatrix} = \begin{bmatrix} f_x & f_y & f_z \end{bmatrix}$


* **Step 3: 組合出 柯西動量方程式（Cauchy momentum equation）**
把質量（其實是密度 $\rho$）、加速度和力組合起來：

$$\rho \frac{D\mathbf{u}}{Dt} = \underbrace{\nabla \cdot \mathbf{\sigma}_{3 \times 3}}_{\text{表面力}} + \underbrace{\rho \mathbf{g}}_{\text{體積力}}$$


## 2.2 Why is the constitutive equation indispensable in the derivation process?

**簡單來說，本構方程式 (Constitutive Equation)是確定我們想考慮的表面力，並且提供一些額外的假設來達到數學上closure的條件**

而大氣科學所使用的 **本構方程式** 考慮的
* **表面力:** 氣壓梯度力、黏滯力
* **額外的假設** 【正向應力 (Normal Stresses) —— 矩陣的對角線】 和 【剪應力 (Shear Stresses) —— 矩陣的非對角線】
  * $\sigma_{xx} = -p + 2\mu \frac{\partial u}{\partial x} - \frac{2}{3}\mu (\nabla \cdot \mathbf{u})$
  * $\sigma_{yy} = -p + 2\mu \frac{\partial v}{\partial y} - \frac{2}{3}\mu (\nabla \cdot \mathbf{u})$
  * $\sigma_{zz} = -p + 2\mu \frac{\partial w}{\partial z} - \frac{2}{3}\mu (\nabla \cdot \mathbf{u})$
  * $\sigma_{xy} = \sigma_{yx} = \mu \left( \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \right)$
  * $\sigma_{xz} = \sigma_{zx} = \mu \left( \frac{\partial u}{\partial z} + \frac{\partial w}{\partial x} \right)$
  * $\sigma_{yz} = \sigma_{zy} = \mu \left( \frac{\partial v}{\partial z} + \frac{\partial w}{\partial y} \right)$


也就是說

$$\mathbf{\sigma}_{3 \times 3} = -p \mathbf{I} + \mu \left( \nabla \mathbf{u} + (\nabla \mathbf{u})^T - \frac{2}{3}(\nabla \cdot \mathbf{u})\mathbf{I} \right)$$



把這個本構方程式代入 Cauchy 方程式，把 $\sigma$ 換掉，才終於用出 **Navier-Stokes Equation**，沒有本構方程式系統就無法封閉、無法進行數值求解。



## 2.3 In computational fluid dynamics, why are conservation equations for mass or energy often written in flux form?

無論是質量、動量還是能量的守恆在理論與數值運算中，我們極度偏好將它們寫成 Eulerian 觀點下的**通量形式 (Flux form)** 或**守恆形式 (Conservative form)**：

$$\frac{\partial \phi}{\partial t} + \nabla \cdot (\phi \mathbf{u}) = S$$


* $\phi$ 是某個物理量
* $\nabla \cdot (\phi \mathbf{u})$ 是該物理量的通量散度
* $S$ 是源/匯

會這樣寫是因為 **[雷諾傳輸定理 (Reynolds Transport Theorem)](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#leibniz-integral-rule)** 

方程式寫成散度形式的最大好處是：當我們在網格上對整個空間進行積分時，根據高斯散度定理 (Divergence Theorem)，$\iiint \nabla \cdot (\phi \mathbf{u}) dV$ 會轉化為邊界上的通量 $\iint (\phi \mathbf{u}) \cdot d\mathbf{A}$。

這意味著，只要沒有東西流出系統邊界，系統內部的總量（無論網格怎麼切割、內部怎麼平流）在數值計算上都會**絕對守恆**，這對於確保數值天氣預報模式跑了好幾天後不會憑空創造出大氣質量或能量來說是至關重要的數學結構。



## 2.4 In compressible fluids (such as the atmosphere), when considering classical conserved variables that follow the motion of the fluid, should one use density or mixing ratio? Why?

**這取決於流體環境**。如果是針對大氣科學這樣的**可壓縮流體**，答案絕對是**混合比 (Mixing ratio)**。

* **密度 ($\rho$) 為什麼不守恆？**
密度定義為質量除以體積。想像一個氣塊從地面被抬升到高空，外界氣壓變低、氣塊會膨脹（體積變大）。即便氣塊裡面的分子數量（質量）完全沒變，它的密度還是下降了，跟隨氣塊運動時 $\frac{D\rho}{Dt} \neq 0$。

* **混合比為什麼守恆？**
混合比的定義通常是「某物質（如水氣）的質量」除以「乾空氣的質量」，因為這兩個質量在氣塊單純經歷絕熱膨脹或壓縮的過程中都不會隨體積變化而改變，所以只要沒有發生相變（如凝結下雨），跟隨氣塊運動時，混合比是守恆的 $\frac{D(Mixing\_Ratio)}{Dt} = 0$。

### 所有守恆定律都會以「連續方程式 (Continuity Equation)」的形式出現

$$\frac{\partial \rho}{\partial t} + \nabla \cdot \mathbf{J} = 0$$

* 在電磁學中：$\rho$ 是電荷密度（單位體積內的電荷量），$\mathbf{J}$ 是電流密度
* 在流體力學中：$\rho$ 是質量密度（單位體積內的質量），$\mathbf{J} = \rho \mathbf{v}$ 是質量通量。

**在電磁學都用「密度」**
因為在一般的電磁學問題中，電荷是在固定的導線或真空中移動，空間本身（或導線的體積）並不會因為電子的流動而劇烈膨脹或收縮。既然「體積」是固定的，我們理所當然可以直接用「單位體積內的量（密度）」來定義守恆。

### [大氣科學從「質量守恆」出發推導連續方程式](https://derek1403.github.io/PC-NTU/Advanced-Atmospheric-Dynamics/_build/html/lecture/week1/week1.html#continuity-equation)

#### 已知與假設
1. **【已知 A】質量守恆 (continuity equation):**
   1. **【已知 1】散度定理 (Divergence Theorem):**
   2. **【已知 2】質量守恆:** 
   3. **【已知 3】變分學基本引理 (微積分中的任意體積引理):** 
   4. **【已知 4】雷諾傳輸定理 (Reynolds Transport Theorem, RTT):** 
2. **【假設 1】不可壓縮流體:** 
   
   $$\nabla \cdot \mathbf{u} = 0$$

   * $\nabla \cdot \mathbf{u}$ 的物理意義就是「體積的相對膨脹率」
   * 如果 $\nabla \cdot \mathbf{u} > 0$（速度往外發散），代表這個水球正在被拉扯、體積膨脹。
   * 質量不變、體積膨脹，Arbitrary Volumes 裡面的空氣 **密度** 勢必會下降！
  

$$\begin{gather*}
0 &\overset{\text{已知 A}}{=}& \frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{u})&& \\
&=&  \frac{\partial \rho}{\partial t} + \mathbf{u} \cdot \nabla \rho + \rho \nabla \cdot \mathbf{u} &&\\
&=& \frac{D\rho}{Dt} + \rho \nabla \cdot \mathbf{u} &&\\
\frac{D\rho}{Dt} &=&  -\rho \nabla \cdot \mathbf{u} &\overset{\text{假設 1}}{=}&  0 \\
\end{gather*}$$

使用了不可壓縮流體假設後才能夠得到密度的守恆 $\frac{D\rho}{Dt} = 0$ ，而 **混和比在不相變的情況下不論可否壓縮都守恆**

海水中嗜甲烷菌 (Methanotrophs) 扮演著地球的過濾器，能消耗海底噴發的甲烷 ($CH_4$)，減少其進入大氣加劇溫室效應。

2. 請建構一個無因次參數（Dimensionless number）來量化此生化與物理的競爭過程，並說明該參數取決於哪些環境與幾何因素？

## 已知與假設

### 一、 氣泡幾何與物理動力學假設

1. **【假設 1】完美的球體假設 (Perfectly Spherical Bubble)**
* **假設內容**：氣泡的體積為 $\frac{4}{3}\pi R^3$，表面積為 $4\pi R^2$。
* **忽略了什麼**：實際上，當氣泡變大或上浮速度變快時，流體阻力會將氣泡壓扁成橢球狀（Ellipsoidal）甚至蘑菇狀（Spherical cap）。這會大幅增加表面積/體積比。


2. **【假設 2】瞬時終端速度 (Instantaneous Terminal Velocity)**
* **假設內容**：氣泡一離開海床，就以恆定的終端速度 $V(R)$ 往上飄，且 $\tau_{rise} = H / V(R)$。
* **忽略了什麼**：忽略了氣泡剛噴發時的加速期，也忽略了因為氣泡大小改變而導致的 $V(R)$ 隨深度變化的動態過程。


3. **【假設 3】孤立氣泡假設 (Isolated Single Bubble)**
* **假設內容**：我們只追蹤「單一顆」氣泡的命運，忽略周遭環境。
* **忽略了什麼**：在大規模爆發（Blowout）時，大量氣泡會互相碰撞、合併（Coalescence）變大，或是被流體剪力撕裂變小。群體氣泡也會產生向上的尾流（Wake），把周圍的水帶上來，使上浮速度比單顆氣泡更快。



### 二、 熱力學與海洋環境假設

4. **【假設 4】不可壓縮氣體 / 恆定密度 (Constant Gas Density)**
* **假設內容**：甲烷氣體的密度 $\rho_{CH_4}$ 是一個常數。
* **忽略了什麼**：這是物理上偏差最大的一個假設。海底水壓極大，當氣泡往上浮，靜水壓力隨深度驟降，根據理想氣體方程式（或波以耳定律 $P \cdot V = \text{常數}$ ），氣泡應該會**劇烈膨脹**（$R$ 變大、$\rho_{CH_4}$ 變小）。


5. **【假設 5】靜止的均勻海洋 (Stagnant and Homogeneous Ocean)**
* **假設內容**：海水是完全靜止的，且物理性質（溫度、鹽度）從海底到海面都一樣。
* **忽略了什麼**：忽略了強烈的水平洋流（會把氣泡吹離垂直路徑，增加滯留時間），以及海洋的溫度躍層（Thermocline），水溫的改變會嚴重影響氣體溶解度與細菌活性。


6. **【假設 6】純甲烷氣泡與單向傳輸 (Pure Methane and One-way Transfer)**
* **假設內容**：氣泡裡 100% 是甲烷，且只有甲烷往外跑。
* **忽略了什麼**：海水中溶解了大量的氮氣和氧氣。在上浮過程中，海水中的 $N_2$ 和 $O_2$ 會擴散**進入**氣泡內，改變氣泡的成分和浮力。



### 三、 生物化學機制假設

7. **【假設 7】零階表面反應 / 恆定消耗率 (Constant Surface Reaction Rate)**
* **假設內容**：細菌吃甲烷的速率 $k_{bio}$ 是一個常數，直接與氣泡表面積成正比。
* **忽略了什麼**：忽略了細菌的族群增長動態（例如越吃數量越多，速率應該要隨時間上升）。也忽略了周遭海水中氧氣或營養鹽可能耗盡的問題（如果氧氣不夠，細菌也無法代謝甲烷）。


8. **【假設 8】直接表面吞噬 (Direct Surface Consumption)**
* **假設內容**：我們把「物理溶解」和「生物消耗」合併成一個動作，假設細菌直接貼在氣泡表面把氣體吃掉。
* **忽略了什麼**：真實的微觀物理過程是「兩階段」的：甲烷氣體必須先**溶解 (Dissolve)** 進入氣泡周圍的液體邊界層，然後懸浮在水中的細菌再去吃掉這些**已溶解的甲烷 (Dissolved methane)**。我們將複雜的質傳系數 (Mass transfer coefficient) 簡化成了單一的 $k_{bio}$。




## 3.1

把整個過程拆解成兩件事： **「氣泡往上跑」** 與 **「細菌吃氣泡」** ，這兩件事各自需要花費一定的時間。

###  Step 1: 定義「物理傳輸特徵時間」 ($\tau_{\text{rise}}$)

假設海底深度為 $H$，氣泡半徑為 $R$，氣泡的終端上浮速度為 $V(R)$ 和【假設 1~6】，氣泡從海底浮到海面所需的特徵時間為：

$$\tau_{rise} = \frac{H}{V(R)}$$

較大的氣泡通常擁有較大的浮力、阻力，但在經驗法則上 **大氣泡的上浮速度 $V(R)$ 會比小氣泡快**（對於中等雷諾數以上的氣泡，速度大致與 $\sqrt{R}$ 成正比）。

### Step 2: 定義「生物消耗特徵時間」 ($\tau_{\text{bio}}$)

假設「細菌只能在表面積作用，但氣泡總量看的是體積」，【假設 7~8】

* **氣泡質量 ($M$)**：與體積成正比，$M = \rho_{CH_4} \frac{4}{3}\pi R^3$
* **消耗速率 ($\frac{dM}{dt}$)**：細菌在氣泡表面進行反應，所以消耗速率與表面積成正比。設單位面積的生物消耗率為常數 $k_{\text{bio}}$，則 $\frac{dM}{dt} = -k_{\text{bio}} 4\pi R^2$

定義「生物消耗特徵時間」為完全消耗掉整顆氣泡所需的時間量級：

$$\tau_{\text{bio}} \sim \frac{M}{|\frac{dM}{dt}|} = \frac{\rho_{CH_4} \frac{4}{3}\pi R^3}{k_{\text{bio}} 4\pi R^2} = \frac{\rho_{CH_4}}{3 k_{\text{bio}}} R$$


**注意到：** $\tau_{\text{bio}} \propto R$。這在物理上意味著，**氣泡半徑 $R$ 越大，細菌需要花費越長的時間才能把它完全吃完！**


## 3.2 Dimensionless number

### Step 3: 建構無因次參數 ($Da$)

將兩個特徵時間相除，定義出一個類似丹克勒數 ($Da$) 的無因次參數，用來比較「細菌消耗的速度」與「氣泡逃逸的速度」：


$$Da = \frac{\text{氣泡滯留時間}}{\text{完全消耗所需時間}} = \frac{\tau_{\text{rise}}}{\tau_{\text{bio}}} = \frac{3 k_{\text{bio}} H}{\rho_{CH_4} R \cdot V(R)}$$


利用我們推導出的 $Da$ ，就可以完美解釋大規模爆發與小氣泡的差異：

1. **小氣泡排放 (Small Bubble Emission)**：
* 半徑 $R$ 很小 $\rightarrow$ 上浮速度 $V(R)$ 慢 $\rightarrow$ 滯留時間 $\tau_{\text{rise}}$ 極長。
* 半徑 $R$ 很小 $\rightarrow$ 表面積/體積比例大 $\rightarrow$ 完全消耗時間 $\tau_{\text{bio}}$ 極短。
* 結果：**$Da \gg 1$**。代表氣泡還沒浮到水面，就在海水中被細菌吃得一乾二淨，不會影響大氣。


2. **大規模爆發 / 大氣泡排放 (Large Bubble Burst)**：
* 半徑 $R$ 很大 $\rightarrow$ 上浮速度 $V(R)$ 極快 $\rightarrow$ 滯留時間 $\tau_{\text{rise}}$ 極短。
* 半徑 $R$ 很大 $\rightarrow$ 表面積/體積比例小 $\rightarrow$ 細菌要花很久的時間才能吃完 ($\tau_{\text{bio}}$ 極長)。
* 結果：**$Da \ll 1$**。生物消耗的機制完全來不及作用，氣泡就如同搭特快車一樣衝出海面，將大量甲烷排放到大氣中。



3. **取決因素總結**：
根據公式 $Da = \frac{3 k_{\text{bio}} H}{\rho_{CH_4} R \cdot V(R)}$，此過程取決於：海底深度 ($H$)、嗜甲烷菌的生化反應速率常數 ($k_{\text{bio}}$)、甲烷氣體密度 ($\rho_{CH_4}$，受深海壓力影響)、以及氣泡的初始半徑 ($R$)。
