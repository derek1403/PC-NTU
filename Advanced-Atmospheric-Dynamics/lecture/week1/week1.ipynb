### leibniz integral rule (1D邊界會變化的例子)

$$
\frac{d}{dt} \int_{a(t)}^{b(t)} f(x,t) \, dx = \int_{a(t)}^{b(t)} \frac{\partial f(x,t)}{\partial t} \, dx+ f(x,b(x)) \frac{d b(x)}{dt} - f(x,a(x)) \frac{d a(x)}{dt}
$$

證明:

透過假設 $\frac{\partial F(x,t)}{\partial t } \overset{\text{let}}{=} f(x,t)$ ，所以也知道 $\left{ \frac{\partial F(a,t)}{\partial t } \overset{\text{let}}{=} f(a,t) \\ \frac{\partial F(b,t)}{\partial t } \overset{\text{let}}{=} f(b,t) \right . $

$$ 
\begin{gather*}
\frac{d}{dt} \int_{a(t)}^{b(t)} f(x,t) \, dx  &\overset{\text{let}}{=} & \frac{d}{dt} \int_{a(t)}^{b(t)} \frac{\partial F(x,t)}{\partial t } \, dx  \\
&\overset{\text{ftoc-1}}{=} & \frac{d}{dt}[F(b,t)-F(a,t)] \\
&\overset{\text{偏微分?}}{=} & \frac{\partial }{\partial t}[F(b,t)-F(a,t)]\frac{dt}{dt} + \frac{\partial }{\partial a}[F(b,t)-F(a,t)]\frac{da}{dt} + \frac{\partial }{\partial b}[F(b,t)-F(a,t)]\frac{db}{dt} \\
&\overset{\text{}}{=} & \int_{a(t)}^{b(t)} \frac{\partial f(x,t)}{\partial t} \, dx+ f(x,b(x)) \frac{d b(x)}{dt} - f(x,a(x)) \frac{d a(x)}{dt}
\end{gather*}
$$

可以分為
* 內部變化 : 
  * 邊界固定 $\int_{a(t)}^{b(t)} $
  * 內部變化 $\frac{\partial f(x,t)}{\partial t}$
* 邊界變化:
  * 通量:邊界值*速度
  * 邊界值 $f(x,b(x))$ 
  * 速度 $ \frac{d b(x)}{dt}$
