### 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 基礎組成關係 ：**

  $$A = B + C$$

* **【已知 2】 $C$ 與 $D$ 的關係 ：**

  $$C = 2D$$

* **【已知 3】 $E$ 的組成（本題未使用）：**

  $$E = F \cdot G$$

* **【已知 4】 系統的總能量常數 ：**

  $$U = 100$$

* **【假設 1】 忽略 $B$ ：**

  $$B = 0$$

* **【假設 2】 $D$ 的設定值 ：**

  $$D = 5$$

* **【定義 1】 能量函數 $H$ ：**

  $$H \overset{\text{def}}{=} A^2$$

* **【推導 1】 將 $A$ 化簡為 $C$ ：**

  $$\begin{gather*}
  A &\overset{\text{已知 1}}{=}& B + C \\
  &\overset{\text{假設 1}}{=}& 0 + C \\
  &=& C \\
  \end{gather*}$$

* **【推導 2】 以 $D$ 表示 $A$ ：**

  $$\begin{gather*}
  A &\overset{\text{推導 1}}{=}& C \\
  &\overset{\text{已知 2}}{=}& 2D \\
  \end{gather*}$$

* **【推導 3】 $A$ 的數值 ：**

  $$\begin{gather*}
  A &\overset{\text{推導 2}}{=}& 2D \\
  &\overset{\text{假設 2}}{=}& 2 \times 5 \\
  &=& 10 \\
  \end{gather*}$$

### 證明

驗證能量函數 $H$ 等於系統的總能量常數 $U$。

$$\begin{gather*}
H &\overset{\text{定義 1}}{=}& A^2 \\
&\overset{\text{推導 3}}{=}& 10^2 \\
&=& 100 \\
&\overset{\text{已知 4}}{=}& U \\
\end{gather*}$$
