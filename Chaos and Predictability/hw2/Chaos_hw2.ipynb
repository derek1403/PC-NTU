### (a) Rayleigh-Benard convection 在不同 Rayleigh number $(R_a)$ 之間的影響

#### 定義穩定:系統的物理量在時間序列上的統計特性不再發生顯著的長期變化
    
這被我精巧的分為兩種情況:

* **數值穩定:**

    數值上說這個系統穩定應該可以從以下3點看出。

    1. **相對平均差異小** 

        $$
        \text{相對差異} = \frac{|\text{後半段平均值} - \text{前半段平均值}|}{\text{前半段平均值}} <10\% 
        $$

        這代表系統在長時間運行後其平均狀態已收斂，不再有明顯的長期趨勢

    2. **數值落在穩定變異度範圍**

        $$
        \text{數值} \in [\text{標準差} - \text{平均值},\text{標準差} + \text{平均值}] 
        $$

        波動很大但只要平均值穩定，系統仍然被視為處於統計平衡，只是其平衡態是一個動態的、高度變化的狀態(湍流、混沌)。

    3. **累計平均值收斂**

        $$
        \text{數值}_{\text{cum}}(N) = \frac{1}{N} \sum_{i=1}^{N} \text{數值}_i
        $$

        如果一個數值的累計平均值曲線趨於水平，則表示該數值已經達到了統計上的穩定狀態。

* **圖看起來穩定:**
    
    如果系統的時空演化圖能夠收斂並穩定地保持一個或多個主導空間模態 (Dominant Spatial Modes) 的形式，則該系統被認為達到了動力平衡 (Dynamical Equilibrium) 或定態 (Steady State)

    $$
    \begin{equation}
    \left\{
    \begin{aligned}
    \zeta &= X(t) \sin\left(\frac{\pi x}{L}\right) \sin\left(\frac{\pi z}{H}\right) \\
    T &= Y(t) \cos\left(\frac{\pi x}{L}\right) \sin\left(\frac{\pi z}{H}\right) - Z(t) \sin\left(\frac{2\pi z}{H}\right)
    \end{aligned}
    \right.
    \tag{1}
    \end{equation}
    $$

### (c) 在 $(x,y,z)=(0,0,0)$ 的線性化與穩定性判定

#### Revisit the Lorenz 63 equations from Homework 1

**Lorenz63 系統:**

$$
\begin{equation}
\left\{
\begin{aligned}
\frac{dx}{dt} &= \sigma (y - x) \\
\frac{dy}{dt} &= x(\rho - z) - y \\
\frac{dz}{dt} &= xy - \beta z
\end{aligned}
\right.
\end{equation} \tag{hw1-1}
$$


**假設**在平衡點附近線性化 $ \left\{ \begin{aligned}  x&= x_{ref} +\delta x \\ y &= y_{ref} +\delta y \\ z &= z_{ref} +\delta z \end{aligned} \right.$ 則可得到線性化的矩陣形式和 $J(x,y,z)$ 和 **Tangent linear model (TLM)** 的形式 $\frac{d \delta x}{dt}= \text{TLM} \ \delta x$

$$
\frac{d}{dt} \begin{bmatrix}\delta x\\ \delta y\\ \delta z\\ \end{bmatrix} =
\underbrace{\begin{bmatrix}
-\sigma & \sigma & 0\\
\rho - z & -1 & -x\\
y & x & -\beta
\end{bmatrix}}_{J(x,y,z)=\text{TLM}(x,y,z)}
\begin{bmatrix}\delta x\\ \delta y\\ \delta z\\ \end{bmatrix}
$$

計算 $\text{TLM}(0,0,0)$ 的三個特徵值並根據題目給定的參數（$\sigma=10,\ \beta=\tfrac{8}{3},\ \rho=24.74$）判斷穩定性

$$
\begin{aligned}
\text{TLM}(0,0,0) \mathbf{v} &= \lambda \mathbf{v} \\
\det(\text{TLM}(0,0,0) - \lambda I) &= 0 \\
-(\beta +\lambda)((\sigma +\lambda)(1+\lambda)-\rho \sigma)&= 0
\end{aligned}
$$

$$
\lambda = ( -\beta,\quad \frac{-(\sigma+1)}{2}+\frac{1}{2}\sqrt{(\sigma+1)^2-4\sigma(1-\rho)},\quad \frac{-(\sigma+1)}{2}-\frac{1}{2}\sqrt{(\sigma+1)^2-4\sigma(1-\rho)} \quad ) \\
\lambda \approx (-2.66666667,\; 10.86001222,\; -21.86001222 ) \\
\mathbf{v} =(\begin{bmatrix}0 \\0 \\1\end{bmatrix}, \begin{bmatrix}1 \\ \frac{\sigma+\lambda}{\sigma} \\0\end{bmatrix} ,\begin{bmatrix}1 \\\frac{\sigma+\lambda}{\sigma} \\0\end{bmatrix}) \overset{\text{normalize}}{\approx}  (\begin{bmatrix}0 \\0 \\1\end{bmatrix},\underbrace{\begin{bmatrix} 0.4322811 \\0.9017389 \\0\end{bmatrix}}_{\text{unstable mode}}, \begin{bmatrix}-0.64461164 \\ 0.76451019 \\0\end{bmatrix} )
$$
  

**解釋與結論：**

* 有一個特徵值 $\lambda\approx 10.86001222>0$ :表示存在一個沿該特徵向量的**指數發散的mode**，系統在原點是不穩定的
* 其它兩個特徵值為負則表示在對應方向上是收斂的，但由於存在正實部的特徵值所以整體還是不穩定

#### **動力解釋為什麼這個mode會不穩定**

從 $Y(t)$ 的mode(1,1)可以看到左熱右冷，這樣的溫度分布會導致**順時鐘的氣塊流動**。從 $X(t)$ 的mode(1,1)可以看到流函數、渦度場的分布會導致**逆時鐘的氣塊流動**。當兩種不同方向的氣塊流相互撞擊容易引發**抵銷**，而在解析這種抵銷作用往往需要**小尺度、高精度的解析**，對於這樣細緻要求的預報會**容易產生分歧**所以引發很大的誤差，也就是**更容易混沌**。而且這種引起"不同方向的氣塊流相互撞擊抵銷"的溫度場、流場(渦度場)需要**特殊的比例才能夠耦合**，這個比例由數學上解出 $(Y \approx 2X)$ 。

若是兩種相同方向(溫度場:逆時針,渦度場:逆時針)的氣塊流相互撞擊合併加強逆時針的情況，從**我的看法**來說會加強該mode也就會**不容易產生分歧、更容易預報，不會導致該mode不穩定**。但是從**諸多家機器人**表達的想法我認為部分的他們想說這種加強是**正反饋指數級的相互加強**，因此會導致這個**mode不穩定**。

\begin{aligned}
\frac{dE}{dt} &= \frac{1}{2} \frac{d}{dt} \left[ \sum_{k}^{K} X_k^2  + \sum_{j,k}^{J,K} Y_{j,k}^2  + \sum_{\ell,j,k}^{L,J,K} Z_{\ell,j,k}^2 \right] \\
&=\frac{\mathbf{2}}{2} \left(  \sum_{k}^{K} X_k  \frac{d}{dt}\left[X_k\right]  + \sum_{j,k}^{J,K}  Y_{j,k} \frac{d}{dt}\left[Y_{j,k}\right]  + \sum_{\ell,j,k}^{L,J,K} Z_{\ell,j,k} \frac{d}{dt}\left[Z_{\ell,j,k}\right]  \right) \\
&=\frac{\mathbf{2}}{2} \left(  \sum_{k}^{K} X_k  \left[-X_{k-1} (X_{k-2} - X_{k+1}) - X_k + F - \left( \frac{hc}{b} \right) \sum_{j=0}^{J-1} Y_{j,k}\right] 
+ \sum_{j,k}^{J,K}  Y_{j,k} \left[-cb\, Y_{j+1,k} (Y_{j+2,k} - Y_{j-1,k}) - c\, Y_{j,k} + \frac{hc}{b} X_k - \left( \frac{\hat{h} c}{\hat{b}} \right) \sum_{\ell=0}^{L-1} Z_{\ell,j,k}\right]  
+ \sum_{\ell,j,k}^{L,J,K} Z_{\ell,j,k}\left[-\hat{c}b \cdot cb\, Z_{\ell+1,j,k} (Z_{\ell+2,j,k} - Z_{\ell-1,j,k}) - \zeta Z_{\ell,j,k} + \frac{\hat{h} c}{\hat{b}} Y_{j,k}\right]  \right)  \\
&\overset{\text{A,B,C}}{=}-\frac{\mathbf{2}}{2} \left(  \sum_{k}^{K} X_k^2  + c\sum_{j,k}^{J,K}  Y_{j,k}^2  + \zeta \sum_{\ell,j,k}^{L,J,K} Z_{\ell,j,k}^2 \right) \underset{\text{decaying}}{\overset{\text{enery}}{<}}0 \ \ \  ,\left(\forall c>0 \ , \forall \zeta>0 \right) \\
\end{aligned}
$$

* $\text{A}$ 單一尺度的週期性邊界非線性平流項加總不改變單一尺度的系統能量: 

    - 對於 $X_k$ 假設週期性邊界網格數量 $num(k)=4$ ，可以簡單使用瞪眼法證明平流項總合為0

1.  **尺度與量級的關係（能量分佈）：**
    * 系統中變量的振幅量級與其所屬尺度**呈正相關**，尺度越大振幅量級也越大；尺度越小振幅量級越小，這表明**系統的總能量或方差主要集中在較大的尺度上**

2.  **能量傳播的特性：**
    * **大、中尺度的左向傳播：** 在大尺度和中尺度上，可以清晰地觀察到**波狀擾動向左側傳遞**

3.  **小尺度的表現（雜訊與耗散）：**
    * **缺乏結構：** 在小尺度上，**缺乏可辨識的波傳結構或規律性**呈現為**雜訊**，可能是該尺度上的能量已經被填滿，或者這些小尺度的動態遠小於測量雜訊水平。

### (c) 改變 h1 h2 看可預報度的改變

| 耦合情境 | 可預測性 | 物理解釋 |
|---------|---------|----------|
| **h1↑, h2 固定** | ↓↓ | 中尺度強烈擾動大尺度，就像湍流破壞行星波 |
| **h2↑, h1 固定** | ↓ | 能量快速耗散到小尺度，但對 X 的直接影響較弱 |
| **h1, h2 都大** | ↓↓↓ | 能量在尺度間轉移的很快，最混沌的情況 |
| **h1, h2 都小** | ↑↑ | 尺度間部耦合，大尺度相對獨立、可預測 |
